## 4. Test 1: Podstawowe Pytanie o Supply Chain

```python
pytanie = "Co to jest optymalizacja ≈Ça≈Ñcucha dostaw?"

print(f"‚ùì Pytanie: {pytanie}\n")
print("‚è≥ Generowanie odpowiedzi...\n")

odpowiedz = engine.generate_explanation(pytanie)

print(f"ü§ñ Odpowied≈∫:\n{odpowiedz}")
```

```python
dane_kontekst = """
Dane: Surowiec X - zu≈ºycie w ostatnich 4 tygodniach:
- Tydzie≈Ñ 1: 120 kg
- Tydzie≈Ñ 2: 95 kg
- Tydzie≈Ñ 3: 150 kg
- Tydzie≈Ñ 4: 80 kg

≈örednie zu≈ºycie: 111.25 kg/tydzie≈Ñ
Odchylenie standardowe: 30.5 kg
"""

pytanie = f"{dane_kontekst}\n\nJakie wnioski mo≈ºesz wyciƒÖgnƒÖƒá z tych danych? Co polecasz dzia≈Çowi zakup√≥w?"

print("‚ùì Analiza danych zakupowych\n")
print("‚è≥ Generowanie analizy...\n")

analiza = engine.generate_explanation(pytanie)

print(f"ü§ñ Analiza AI:\n{analiza}")
```
