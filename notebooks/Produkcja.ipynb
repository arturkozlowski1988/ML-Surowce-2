# [CDN].[ZasobyView]

```sql
/****** Object:  View [CDN].[ZasobyView]    Script Date: 5.01.2026 21:12:40 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [CDN].[ZasobyView]
As
Select
P.RpT2_Rpt2ID    As TwZV_RpT2Id,
A.Twr_TwrId      As TwZV_TwrId,
A.Twr_Typ        As TwZV_TwrTyp,
A.Twr_Produkt    As TwZV_Produkt,
A.Twr_Kod        As TwZV_Kod,
A.Twr_NumerKat   As TwZV_NumerKat,
A.Twr_Nazwa      As TwZV_Nazwa,
E.Sciezka        As TwZV_Grupa,
A.Twr_KatId      As TwZV_KatId,
A.Twr_Kategoria  As TwZV_Kategoria,
A.Twr_KatZakId   As TwZV_KatZakId,
A.Twr_Opis       As TwZV_Opis,
A.Twr_JM         As TwZV_Jm,
A.Twr_JMZ        As TwZV_JmZ,
A.Twr_TwCNumer   As TwZV_TwCNumer,
B.TwI_MagId      As TwZV_MagId,
B.TwI_Data       As TwZV_Data,
B.TwI_Ilosc      As TwZV_Ilosc,
B.TwI_Braki      As TwZV_Braki,
B.TwI_Rezerwacje As TwZV_Rezerwacje,
B.TwI_Zamowienia As TwZV_Zamowienia,
B.TwI_Wartosc    As TwZV_Wartosc,
D.TwC_Wartosc    As TwZV_Cena,
D.TwC_Waluta     As TwZV_Waluta,
B.TwI_TwIId      As TwZV_TwIId
From CDN.RptTemp2 P join CDN.Towary A On 1=1
Left Outer Join CDN.TwrIlosci B
  On A.Twr_TwrID = B.TwI_TwrID
 And ( P.RpT2_TmpId Is Not Null Or B.TwI_MagId IS NULL )
 And ( P.RpT2_TmpId Is Null Or B.TwI_MagId = P.RpT2_TmpId )
 And B.TwI_Data = (
    Select Top 1 H.TwI_Data From CDN.TwrIlosci H Where H.TwI_TwrId = A.Twr_TwrId
    And ( P.RpT2_TmpId Is Not Null Or H.TwI_MagId Is Null )
    And ( P.RpT2_TmpId Is Null Or H.TwI_MagId = P.RpT2_TmpId )
    And H.TwI_Data <= P.RpT2_Data
    Order By H.TwI_Data Desc )
Left Outer Join CDN.Magazyny C ON B.TwI_MagId= C.Mag_MagId
Left Outer Join CDN.TwrCeny D ON A.Twr_TwrId= D.TwC_TwrID AND A.Twr_TwCNumer = D.TwC_TwCNumer
Left Outer Join CDN.TwrGrupyView E ON A.Twr_TwGGIDNumer = E.GIDNumer

GO
```

# [dbo].[CTIProdukcjaAtrybutyElemAddEdit]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaAtrybutyElemAddEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaAtrybutyElemAddEdit]
 @CAETypDok int,
 @CAEIdDok  int,
 @CAECANLp  int,
 @CAEWartosc varchar(250)
AS
BEGIN
	--jeżeli jest zapis to tylko aktualizuj
	if exists (select * from dbo.CtiAtrybutyElem 
	where CAE_TypDok = @CAETypDok and CAE_IdDok = @CAEIdDok and  CAE_CANLp = @CAECANLp)
	begin
		update dbo.CtiAtrybutyElem set CAE_Wartosc = @CAEWartosc
		where CAE_TypDok = @CAETypDok and CAE_IdDok = @CAEIdDok and  CAE_CANLp = @CAECANLp
	end
	else--jezeli nie ma to pobierz z konfiguracji (tylko przy TH) i zapisz wartość
	begin
		if @CAEWartosc <> ''
		begin

			declare @CAHID int = 0
			SELECT top 1 @CAHID = CAH_ID FROM  dbo.CtiAtrybutyElemHist INNER JOIN CtiAtrybutyNag
				ON CAH_Nazwa = CAN_Nazwa and CAH_Domysla = CAN_Domysla and CAH_TypMaski = CAN_TypMaski 
				and CAH_Maska =  CAN_Maska
			where  CAN_Lp =@CAECANLp

			if @CAHID = 0
			begin
				insert into dbo.CtiAtrybutyElemHist (  CAH_Nazwa, CAH_Domysla, CAH_TypMaski, CAH_Maska)
				SELECT distinct CAN_Nazwa,  CAN_Domysla, CAN_TypMaski, CAN_Maska
				FROM dbo.CtiAtrybutyNag where CAN_Lp = @CAECANLp
				Select @CAHID = SCOPE_IDENTITY() 
			end

			insert into dbo.CtiAtrybutyElem (CAE_TypDok, CAE_IdDok, CAE_CANLp, CAE_Typ,CAE_Wartosc,CAE_CAHid)
			select @CAETypDok, @CAEIdDok, CAN_Lp, CAN_Typ,@CAEWartosc, @CAHID
			FROM dbo.CtiAtrybutyNag where CAN_Lp = @CAECANLp
		end
	end
END
GO
```

# [dbo].[CTIProdukcjaBrakiAddEdit]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaBrakiAddEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaBrakiAddEdit]
@id int,
@Ope int,
@Status int,
@opis nvarchar(254),
@Typ int,
@DataRealizacji datetime = null
AS
BEGIN
declare @numer int = 0
declare @miesiac int =month(GETDATE())
declare @rok int =  year(GETDATE())
declare @nrpelny varchar(20) =''
declare @Seria varchar(5) = ''

select @numer = isnull(Max(CBN_Numer),0)+1  from dbo.CtiBrakiNag
where CBN_Rok = @rok --and isnull(CBN_Seria,'') = '' and CBN_Miesiac = @miesiac --tylko rok
select  @nrpelny='BR/'+cast(replace(str(@numer,5),' ','0')as char(5)) +'/'+cast(@rok as varchar(4))

	if @id = 0
	begin
	INSERT INTO dbo.CtiBrakiNag
		(CBN_Numer,CBN_Miesiac,CBN_Rok,CBN_Seria,  CBN_OpeWystaw, CBN_DataWystaw, CBN_OperModyf, CBN_DataModyf, CBN_Status,CBN_NrPelny,CBN_Opis,CBN_Typ, CBN_DataRealizacji)
		VALUES (@numer,@miesiac,@rok,@Seria,@Ope,GETDATE(),@Ope,GETDATE(),@Status,@nrpelny,@opis,@Typ,@DataRealizacji)
		Select SCOPE_IDENTITY() as ID 
	end
if @id > 0
	begin
		UPDATE dbo.CtiBrakiNag SET	
		CBN_OperModyf =@Ope,
		CBN_Status = @Status,
		CBN_DataModyf = GETDATE(),
		CBN_DataRealizacji = @DataRealizacji,
		CBN_Opis=@opis WHERE CBN_ID =@id
	end

END
GO
```

# [dbo].[CTIProdukcjaBrakiSurowca]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaBrakiSurowca]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE  [dbo].[CTIProdukcjaBrakiSurowca]   
@status int , @status2 int,@magazyn varchar(20), @ZPid int, @SurowiecId int, @ZPZid int,@typ int, @data varchar(20)
--typ 0
--konkretny surowiec @SurowiecId, @status = 0 , @status2 = 1,@magazyn = mag surowców operatora 
--typ 1
--zlecenie złozone
--typ 2
--zlecenie złożone braki - suma towarów bez info który numer zlecenia
--typ 3
--suma towarów bez info który numer zlecenia - braki
--typ 4
--zlecenie produkcyjne @ZPid, @status = 0 , @status2 = 1,@magazyn = mag surowców operatora
AS 
declare @magId int = 0 select @magId =mag_magid from cdn.magazyny where mag_symbol = @magazyn

declare @status3 int = 3
if (@typ = 1 or @typ = 4)
begin	set  @status3  = 3 end
else
begin	set  @status3  = @status end 

declare @statusGlowny int = 4
if (@typ = 1 or @typ = 4)
begin	set  @statusGlowny  = 4  end-- mniejsze od anulowanego
else
begin	set  @statusGlowny  = 3  end-- mniejsze od wydanego, zakończonego
 


declare @dataD datetime =convert(datetime, @data,120)

CREATE TABLE #WydaniaList (CZNId int, TwrId Int, Ilosc decimal(15,4), IloscZrealizowana decimal(15,4), IloscNaMagazynie decimal(15,4))

	--lista towarow z technologii dodanie do tabeli
		insert 	into #WydaniaList (CZNId, TwrId , Ilosc,IloscZrealizowana,IloscNaMagazynie)			
			select CZNId, TwrId, Ilosc,0,0 from
			(	select CZN_Id as CZNId, CZE_TwrId As TwrId,sum(CZE_ilosc) as Ilosc  
				from dbo.CtiZlecenieNag 
				INNER JOIN  dbo.CtiZlecenieElem on CZE_CZNId = CZN_ID and CZE_Typ in (1,2) --tylko surowce z cechą i surowce
				where
				1 = 1
				and (dbo.CTIProdukcjaStatusy(23, CZN_ID)<@statusGlowny)
				
				and (
					replace( dbo.CTIProdukcjaStatusy(2, CZN_ID) ,'2','1') in (@status,@status2,@status3)
				)
				and ((@ZPid = 0 and CZN_Id = CZN_Id) or  (@ZPid <> 0 and CZN_Id = @ZPid))--zlecenie
				and ((@SurowiecId = 0 and CZE_TwrId=CZE_TwrId) or (@SurowiecId <> 0 and CZE_TwrId = @SurowiecId))--surowiec
				group by CZE_TwrId,CZN_Id
			) e



--	aktualizacjia ilosci, po tym co juz zostało rozchodowane w ramach zlecenia (dok RW)		 --RW nie anulowane minus korekta 

		update #WydaniaList set IloscZrealizowana = Ilosc2  
		from
		(
				select  CZD_CZNId,TwrId2, cast(SUM(Ilosc )as decimal(15,4)) as Ilosc2
				from
				(
					select 
						CZD_CZNId as CZD_CZNId
						, e.TrE_TwrId as TwrId2
						, sum(isnull(e.tre_ilosc,0)+isnull(e2.tre_ilosc,0)) as Ilosc
					from CtiZlecenieDok 
						INNER JOIN CtiZlecenieNag ON CZD_CZNId = CZN_ID 
						INNER JOIN CDN.TraNag a ON CZD_TrnId = a.TrN_TrNID 
						INNER JOIN cdn.TraElem e on a.trn_trnid = e.TrE_TrNId
						LEFT JOIN CDN.TraNag a2 ON a.TrN_TrNID= isnull(a2.TrN_ZwrId,0) and isnull(a2.TrN_Rodzaj,0) = 304001
						LEFT JOIN cdn.TraElem e2 on isnull(a2.trn_trnid,0) = isnull(e2.TrE_TrNId,0) and e.TrE_TreId = isnull(e2.TrE_ZwrId,0) 
					where czd_typ = 5 and a.TrN_TypDokumentu = 304  and a.trn_anulowany = 0 
						--zlecenia potwierdzone i nie potwierdzone ale nie zamknete i nie wydane
						and dbo.CTIProdukcjaStatusy(23, CZD_CZNId)<@statusGlowny
						--and dbo.CTIProdukcjaStatusy(2, CZD_CZNId) < 3
						and	
						( replace( dbo.CTIProdukcjaStatusy(2, CZN_ID) ,'2','1') in (@status,@status2,@status3)
						)
						and ((@ZPid = 0 and CZD_CZNId = CZD_CZNId) or  (@ZPid <> 0 and CZD_CZNId = @ZPid))
					group by CZD_CZNId, e.TrE_TwrId
					having sum(isnull(e.tre_ilosc,0)+isnull(e2.tre_ilosc,0)) > 0

				) a
				group by TwrId2,CZD_CZNId
		)c
		where c.TwrId2 = TwrId
		and CZNId = c.CZD_CZNId




	--jakie towary z potrzebnych są na magazynie w jakich ilościach
	update #WydaniaList set IloscNaMagazynie = Ilosc2
		from
		(
			--SELECT  isnull(sum(isnull(c.ilosc,0)),0) as Ilosc2,b.TwrId as TwrId2,b.CZNId as CZNId2
			--FROM 
			--#WydaniaList b
			--cross apply  [CDN].[fnTwrDost](b.TwrId) c where isnull(c.ilosc,0) > 0
			--and (@magazyn = '' and Mag1 = Mag1) or  (@magazyn <> '' and Mag1 = @magazyn)
			--group by b.TwrId,b.CZNId

			
			SELECT  isnull(sum(isnull(isnull(b.TwI_Ilosc, 0),0)),0) as Ilosc2,a.TwrId as TwrId2,a.CZNId as CZNId2
			FROM #WydaniaList a
				INNER JOIN CDN.TwrIlosci B ON B.TwI_TwIId = 
					(SELECT TOP 1 IL.TwI_TwIId From CDN.TwrIlosci IL Where IL.TwI_TwrId = a.TwrId and ((@magId = '' and TwI_MagId = TwI_MagId) or  (@magId <> '' and TwI_MagId = @magId)) And IL.TwI_Data <= Convert(DATETIME,@dataD,120) ORDER BY IL.TwI_Data DESC)  
			group by a.TwrId,a.CZNId
		)g
		where g.TwrId2 = TwrId and g.CZNId2 = CZNId



if (@typ = 0)
begin		
	--select tabeli tymczasowej z id towaru suma ilości kodem nazwą i jednostką
	SELECT
		TwrId as ID, twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, --towar
		CZN_Id as CZNId,CZN_NrPelny as CZNNrPelny, --zlecenie
		Ilosc,IloscZrealizowana,
		case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end  as IloscDoRealizacji,
		IloscNaMagazynie, 
		--case when (Ilosc-IloscZrealizowana) < 0 then 0 else 
		--case when (Ilosc-IloscZrealizowana)-IloscNaMagazynie <0 then 0 else(Ilosc-IloscZrealizowana)-IloscNaMagazynie end
		--end  as braki
		case when (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0)) < 0 then 0
		else (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0)) end as braki
		,ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) AS IloscNaZleceniach
		,@magazyn as Magazyn, dbo.CTIProdukcjaStatusy(2, CZN_ID) AS Status,
		Twr_JM as Jm, isnull(CZZE_CZZNID,0) as CZZNID, isnull(CZZN_NrPelny,'') as CZZNNrPelny
	 FROM #WydaniaList
		INNER JOIN cdn.towary on TwrId = twr_twrid 
		INNER JOIN dbo.CtiZlecenieNag on CZN_Id = CZNId
		LEFT JOIN dbo.CtiZlecenieZElem on CZN_Id = CZZE_CZNID
		LEFT JOIN dbo.CtiZlecenieZNag on CZZE_CZZNID = CZZN_ID
	where (Ilosc-IloscZrealizowana) > 0
	order by CZN_Id,twr_kod
end
else if (@typ = 1)
begin
	SELECT
		TwrId as ID, twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, --towar
		CZN_Id as CZNId,CZN_NrPelny as CZNNrPelny, --zlecenie
		Ilosc,IloscZrealizowana,
		case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end  as IloscDoRealizacji,
		IloscNaMagazynie, 
		--case when (Ilosc-IloscZrealizowana) < 0 then 0 else 
		--case when (Ilosc-IloscZrealizowana)-IloscNaMagazynie <0 then 0 else(Ilosc-IloscZrealizowana)-IloscNaMagazynie end
		--end  as braki
		case when (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0)) < 0 then 0
		else (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0)) end as braki
		,ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) AS IloscNaZleceniach
		,@magazyn as Magazyn, dbo.CTIProdukcjaStatusy(2, CZN_ID) AS Status,
		Twr_JM as Jm, isnull(CZZE_CZZNID,0) as CZZNID, isnull(CZZN_NrPelny,'') as CZZNNrPelny
	FROM #WydaniaList
		INNER JOIN cdn.towary on TwrId = twr_twrid --where Ilosc > 0
		INNER JOIN dbo.CtiZlecenieNag on CZN_Id = CZNId
		INNER JOIN dbo.CtiZlecenieZElem on CZN_Id = CZZE_CZNID
		INNER JOIN dbo.CtiZlecenieZNag on CZZE_CZZNID = CZZN_ID
	where CZZE_CZZNID = @ZPZid
	order by CZN_Id,twr_kod
end
else if (@typ = 3)
begin
	SELECT
		TwrId as ID, twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, --towar
		0 as CZNId,'' as CZNNrPelny, --zlecenie
		sum(Ilosc) as Ilosc,sum(IloscZrealizowana) as IloscZrealizowana,
		sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  as IloscDoRealizacji,
		IloscNaMagazynie, 
		--sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) as braki,
		sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) -ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) as braki,
		ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) AS IloscNaZleceniach

		,@magazyn as Magazyn, 0 AS Status,
		Twr_JM as Jm, 0 as CZZNID, '' as CZZNNrPelny
		--,isnull(TwI_Ilosc, 0) AS [TwI_Ilosc], isnull(TwI_Braki, 0) AS TwIBraki 
		,isnull(TwI_Rezerwacje, 0) AS TwIRezerwacje, isnull(TwI_Zamowienia, 0) AS TwIZamowienia 
	FROM #WydaniaList
		INNER JOIN cdn.towary A on TwrId = twr_twrid
		INNER JOIN dbo.CtiZlecenieNag on CZN_Id = CZNId
		LEFT JOIN CDN.TwrIlosci B ON B.TwI_TwIId = 
		(SELECT TOP 1 IL.TwI_TwIId From CDN.TwrIlosci IL Where IL.TwI_TwrId = A.Twr_TwrID and ((@magId = '' and TwI_MagId = TwI_MagId) or  (@magId <> '' and TwI_MagId = @magId))
		And  IL.TwI_Data <= Convert(DATETIME,CAST(FLOOR(CAST(@dataD AS FLOAT))AS DATETIME),120) ORDER BY IL.TwI_Data DESC)  
	group by TwrId,twr_kod,twr_nazwa,IloscNaMagazynie,Twr_JM
		,isnull(TwI_Rezerwacje, 0), isnull(TwI_Zamowienia, 0)
	having sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) -ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) >0
	order by twr_kod
end
else if (@typ = 2)
begin
	SELECT
		TwrId as ID, twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, --towar
		0 as CZNId,'' as CZNNrPelny, --zlecenie
		sum(Ilosc) as Ilosc,sum(IloscZrealizowana) as IloscZrealizowana,
		sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  as IloscDoRealizacji,
		IloscNaMagazynie, 
		--sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else 
		--case when (Ilosc-IloscZrealizowana)-IloscNaMagazynie <0 then 0 else(Ilosc-IloscZrealizowana)-IloscNaMagazynie end
		--end)-isnull(TwI_Zamowienia, 0)  as braki
		--sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) as braki,
		
		sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) -ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) as braki,
		ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) AS IloscNaZleceniach
		,@magazyn as Magazyn, 0 AS Status,
		Twr_JM as Jm, 0 as CZZNID, '' as CZZNNrPelny
		--,isnull(TwI_Ilosc, 0) AS [TwI_Ilosc], isnull(TwI_Braki, 0) AS TwIBraki 
		,isnull(TwI_Rezerwacje, 0) AS TwIRezerwacje, isnull(TwI_Zamowienia, 0) AS TwIZamowienia 
	FROM #WydaniaList
		INNER JOIN cdn.towary A on TwrId = twr_twrid 
		INNER JOIN dbo.CtiZlecenieNag on CZN_Id = CZNId
		INNER JOIN dbo.CtiZlecenieZElem on CZN_Id = CZZE_CZNID
		INNER JOIN dbo.CtiZlecenieZNag on CZZE_CZZNID = CZZN_ID
		LEFT JOIN CDN.TwrIlosci B ON B.TwI_TwIId = 
		(SELECT TOP 1 IL.TwI_TwIId From CDN.TwrIlosci IL Where IL.TwI_TwrId = A.Twr_TwrID and ((@magId = '' and TwI_MagId = TwI_MagId) or  (@magId <> '' and TwI_MagId = @magId))
		And  IL.TwI_Data <= Convert(DATETIME,CAST(FLOOR(CAST(@dataD AS FLOAT))AS DATETIME),120) ORDER BY IL.TwI_Data DESC)  
	where CZZE_CZZNID = @ZPZid
	group by TwrId,twr_kod,twr_nazwa,IloscNaMagazynie,Twr_JM
		,isnull(TwI_Rezerwacje, 0), isnull(TwI_Zamowienia, 0)
	having sum(case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end)  -isnull(IloscNaMagazynie, 0) -isnull(TwI_Zamowienia, 0) -ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) >0
	order by twr_kod
end
else if (@typ = 4)
begin		
	--select tabeli tymczasowej z id towaru suma ilości kodem nazwą i jednostką
	SELECT
		TwrId as ID, twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, --towar
		CZN_Id as CZNId,CZN_NrPelny as CZNNrPelny, --zlecenie
		Ilosc,IloscZrealizowana,
		case when (Ilosc-IloscZrealizowana) < 0 then 0 else Ilosc-IloscZrealizowana end  as IloscDoRealizacji,
		IloscNaMagazynie, 
		--case when (Ilosc-IloscZrealizowana) < 0 then 0 else 
		--case when (Ilosc-IloscZrealizowana) - IloscNaMagazynie < 0 then 0 else (Ilosc-IloscZrealizowana)-IloscNaMagazynie end
		case when (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0)) < 0 then 0 
		else (Ilosc - IloscZrealizowana - IloscNaMagazynie - ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0))
		end as braki,
		ISNULL((SELECT SUM(CZN_Ilosc) FROM dbo.CtiZlecenieNag WHERE CZN_TwrId = TwrId AND CZN_Status IN (1,2)), 0) AS IloscNaZleceniach,
		@magazyn as Magazyn
		, dbo.CTIProdukcjaStatusy(2, CZN_ID) AS Status,
		Twr_JM as Jm, isnull(CZZE_CZZNID,0) as CZZNID, isnull(CZZN_NrPelny,'') as CZZNNrPelny
	
	 FROM #WydaniaList
		INNER JOIN cdn.towary on TwrId = twr_twrid 
		INNER JOIN dbo.CtiZlecenieNag on CZN_Id = CZNId
		LEFT JOIN dbo.CtiZlecenieZElem on CZN_Id = CZZE_CZNID
		LEFT JOIN dbo.CtiZlecenieZNag on CZZE_CZZNID = CZZN_ID
	
	group by TwrId, twr_kod, twr_nazwa, --towar
		CZN_Id,CZN_NrPelny, --zlecenie
		Ilosc,IloscZrealizowana,
		IloscNaMagazynie, 
		Twr_JM,CZZE_CZZNID, isnull(CZZN_NrPelny,'')
	order by CZN_Id,twr_kod
end

	DROP TABLE #WydaniaList
GO
```

# [dbo].[CTIProdukcjaCechyDlaDostawyList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaCechyDlaDostawyList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaCechyDlaDostawyList] 
AS
--atrybut musi być dodany przynajmniej na jednym towarze
	select DISTINCT TwA_DeAId,DeA_Kod,DeA_Format
	from cdn.twratrybuty
		inner join cdn.defatrybuty on TwA_DeAId = DeA_DeAId
	where  TwA_CzyPrzenosicNaDostawy = 1 and TwA_CzyObowiazkowyDostawy = 1
		and DeA_Format <> 5 and DeA_Format in (1,2,3,4)
	order by DeA_Kod asc

--Format pola
--1 - liczbowe 
--2 - tekstowe 
--3 - lista 
--4 - data 
--5 - binaria

GO
```

# [dbo].[CTIProdukcjaCofnijPotwierdzenie]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaCofnijPotwierdzenie]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaCofnijPotwierdzenie]
    @zpzID int
AS
BEGIN
    BEGIN TRY
        BEGIN TRAN T1

        DECLARE @FinaleZPId int 
        DECLARE @ID int

         SELECT @FinaleZPId = CZZE_CZNID 
         FROM dbo.CtiZlecenieZElem 
         WHERE CZZE_CZZNID = @zpzID AND CZZE_FinalneZP = 1

        DELETE FROM dbo.CtiZlecenieZElem 
        WHERE CZZE_ID = @zpzID AND CZZE_FinalneZP = 1

        EXEC dbo.CTIProdukcjaZlecenieDel @FinaleZPId, 1

        DECLARE Elementy_ZPZ CURSOR FOR
        SELECT CZZE_CZNID 
        FROM dbo.CtiZlecenieZElem
        WHERE CZZE_CZZNID = @zpzID AND CZZE_Typ = 22

        OPEN Elementy_ZPZ
        FETCH NEXT FROM Elementy_ZPZ INTO @ID

        WHILE @@FETCH_STATUS = 0
        BEGIN
            DELETE FROM dbo.CtiZlecenieZElem 
            WHERE CZZE_CZNID = @ID AND CZZE_CZZNID = @zpzID

            UPDATE dbo.CtiZlecenieZNag
            SET 
                CZZN_CZZNIdGlowne = dbo.CtiProdukcjaZpzGlowne(@ID, 22), 
                CZZN_CZZNNrPelnyGlowne = dbo.CtiProdukcjaZpzGlowneNumer(@ID, 22),
                CZZN_CZZNIdNadrzedne = dbo.CtiProdukcjaZpzNadrzedne(@ID, 22),
                CZZN_CZZNNrPelnyNadrzedne = dbo.CtiProdukcjaZpzNadrzedneNumer(@ID, 22)
            WHERE CZZN_ID = @ID

            FETCH NEXT FROM Elementy_ZPZ INTO @ID
        END

        CLOSE Elementy_ZPZ 
        DEALLOCATE Elementy_ZPZ

        DECLARE Elementy_ZP CURSOR FOR
        SELECT CZZE_CZNID 
        FROM dbo.CtiZlecenieZElem
        WHERE CZZE_CZZNID = @zpzID AND CZZE_Typ = 2

        OPEN Elementy_ZP
        FETCH NEXT FROM Elementy_ZP INTO @ID

        WHILE @@FETCH_STATUS = 0
        BEGIN
            DELETE FROM dbo.CtiZlecenieZElem 
            WHERE CZZE_CZNID = @ID AND CZZE_CZZNID = @zpzID

            UPDATE dbo.CtiZlecenieNag
            SET 
                CZN_CZZNIdGlowne = dbo.CtiProdukcjaZpzGlowne(@ID, 2), 
                CZN_CZZNNrPelnyGlowne = dbo.CtiProdukcjaZpzGlowneNumer(@ID, 2),
                CZN_CZZNIdNadrzedne = dbo.CtiProdukcjaZpzNadrzedne(@ID, 2),
                CZN_CZZNNrPelnyNadrzedne = dbo.CtiProdukcjaZpzNadrzedneNumer(@ID, 2)
            WHERE CZN_ID = @ID
            
            FETCH NEXT FROM Elementy_ZP INTO @ID
        END

        CLOSE Elementy_ZP 
        DEALLOCATE Elementy_ZP

        UPDATE dbo.CtiZlecenieZNag 
        SET CZZN_Status = 0 
        WHERE CZZN_ID = @zpzID

        COMMIT TRAN T1

    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRAN T1

        DECLARE @ErrorMessage NVARCHAR(4000)
        DECLARE @ErrorSeverity INT
        DECLARE @ErrorState INT

        SELECT 
            @ErrorMessage = ERROR_MESSAGE(),
            @ErrorSeverity = ERROR_SEVERITY(),
            @ErrorState = ERROR_STATE()

        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState)

        RETURN
    END CATCH
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaCofnijPotwierdzenieZUsuwaniem]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaCofnijPotwierdzenieZUsuwaniem]
 @zpzID int

AS
BEGIN
    BEGIN TRY
        -- Rozpoczęcie transakcji
        BEGIN TRANSACTION;

        DECLARE @idDoUsuniecia int;

        -- Przetwarzanie kursora dla typu 22
        DECLARE Elementy_ZPZ CURSOR FOR
        SELECT CZZE_CZNID FROM dbo.CtiZlecenieZElem
        WHERE CZZE_CZZNID = @zpzID AND CZZE_Typ = 22;

        OPEN Elementy_ZPZ;
        FETCH NEXT FROM Elementy_ZPZ INTO @idDoUsuniecia;

        WHILE @@FETCH_STATUS = 0
        BEGIN
            EXEC dbo.CTIProdukcjaZlecenieZDel @idDoUsuniecia, 0;
            FETCH NEXT FROM Elementy_ZPZ INTO @idDoUsuniecia;
        END

        CLOSE Elementy_ZPZ;
        DEALLOCATE Elementy_ZPZ;

        -- Przetwarzanie kursora dla typu 2
        DECLARE Elementy_ZP CURSOR FOR
        SELECT CZZE_CZNID FROM dbo.CtiZlecenieZElem
        WHERE CZZE_CZZNID = @zpzID AND CZZE_Typ = 2;

        OPEN Elementy_ZP;
        FETCH NEXT FROM Elementy_ZP INTO @idDoUsuniecia;

        WHILE @@FETCH_STATUS = 0
        BEGIN
            EXEC dbo.CTIProdukcjaZlecenieDel @idDoUsuniecia, 0;
            FETCH NEXT FROM Elementy_ZP INTO @idDoUsuniecia;
        END

        CLOSE Elementy_ZP;
        DEALLOCATE Elementy_ZP;

        -- Aktualizacja statusu
        UPDATE dbo.CtiZlecenieZNag SET CZZN_Status = 0 WHERE CZZN_ID = @zpzID;

        -- Zakończenie transakcji
        COMMIT TRAN T3;
    END TRY
    BEGIN CATCH
        -- Wycofanie transakcji w przypadku błędu
        IF @@TRANCOUNT > 0
        BEGIN
            ROLLBACK TRANSACTION;
        END

        -- Przekazanie informacji o błędzie
        DECLARE @ErrorMessage NVARCHAR(4000);
        DECLARE @ErrorSeverity INT;
        DECLARE @ErrorState INT;

        SELECT 
            @ErrorMessage = ERROR_MESSAGE(),
            @ErrorSeverity = ERROR_SEVERITY(),
            @ErrorState = ERROR_STATE();

        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH
END
GO
```

# [dbo].[CTIProdukcjaCountShortageMMWyd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaCountShortageMMWyd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaCountShortageMMWyd]
    @ZPid INT,
    @resultStatus int OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    DECLARE @shortage INT = 0;
    SELECT 
        a.tre_twrid AS corrTwrId,
        SUM(
            CASE 
                WHEN ISNULL(czd_korekta, 0) = 0 THEN a.tre_ilosc
                ELSE -1 * ISNULL(a.tre_ilosc, 0)
            END
        ) AS Ilosc2
    INTO #TempMM
    FROM dbo.ctizleceniedok
    INNER JOIN cdn.tranag ON trn_trnid = czd_trnid
    INNER JOIN cdn.traelem a ON czd_trnid = a.tre_trnid
    WHERE trn_anulowany = 0 
      AND czd_typ = 3
      AND czd_cznid = @ZPid
    GROUP BY a.tre_twrid;

    IF EXISTS (SELECT 1 FROM #TempMM WHERE Ilosc2 < 0)
    BEGIN
        SET @resultStatus = 0;
        DROP TABLE #TempMM;
        RETURN;
    END
    ELSE
    BEGIN
        SELECT 
            SUM(cze_ilosc) AS Ilosc1,
            cze_twrid
        INTO #BasicDoc
        FROM dbo.ctizlecenienag
        INNER JOIN dbo.ctizlecenieelem 
            ON cze_cznid = czn_id AND cze_typ IN (1, 2)
        WHERE czn_id = @ZPid
        GROUP BY cze_twrid;

        SELECT @shortage = COUNT(*)
        FROM #BasicDoc bd
        INNER JOIN #TempMM tm ON bd.cze_twrid = tm.corrTwrId
        WHERE bd.Ilosc1 > ISNULL(tm.Ilosc2, 0);
        IF (@shortage > 0)
			SET @resultStatus = 1
		ELSE 
			SET @resultStatus = 2
        DROP TABLE #BasicDoc;
    END
    DROP TABLE #TempMM;
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaCzyZlecenieMoznaRealizowac]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaCzyZlecenieMoznaRealizowac]
@zlecenieId INT

AS
BEGIN
SELECT CASE WHEN EXISTS (
    SELECT CZN_ID
	FROM(
		SELECT CZN_NrPelny,CZN_ID
		,ISNULL(e2.CZZE_CZNID,e.CZZE_CZNID) [ChildCZNID]
		,CASE WHEN ISNULL(e2.CZZE_CZNID,e.CZZE_CZNID) IS NULL THEN 0 ELSE e.CZZE_Ilosc - dbo.CTIProdukcjaZlecenieIloscZrealizowana(ISNULL(e2.CZZE_CZNID,e.CZZE_CZNID), 1, 0) END[IloscDoRealizacjiChild]
		FROM CtiZlecenieNag
		LEFT JOIN CtiZlecenieZElem t ON t.CZZE_CZNID=CZN_ID and t.CZZE_FinalneZP=1 and t.CZZE_Typ=2
		LEFT JOIN CtiZlecenieZElem e ON e.CZZE_FinalneZP=0 and t.CZZE_FinalneZP=1 and t.CZZE_CZZNID=e.CZZE_CZZNID and e.CZZE_Typ in (2,22)
		LEFT JOIN CtiZlecenieZElem e2 ON e2.CZZE_FinalneZP=1 and e2.CZZE_CZZNID=e.CZZE_CZNID and e2.CZZE_Typ = 2 and e.CZZE_Typ = 22
		where CZN_Status <> 0
		)t
		WHERE CZN_ID = @zlecenieId
		GROUP BY CZN_NrPelny,CZN_ID
		HAVING sum([IloscDoRealizacjiChild])=0
)
THEN CAST(1 AS BIT)
ELSE CAST(0 AS BIT) END
END
GO
```

# [dbo].[CTIProdukcjaDokElementy]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaDokElementy]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaDokElementy]   @id INT
AS
BEGIN

	DECLARE @KodCechy1 VARCHAR(max) = (SELECT TOP 1 knf_wartosc FROM dbo.CtiKonfiguracja WHERE knf_numer = 5)
	DECLARE @KodCechy2 VARCHAR(max) = (SELECT TOP 1 knf_wartosc FROM dbo.CtiKonfiguracja WHERE knf_numer = 19)
	DECLARE @idcechy1 INT;
	DECLARE @idcechy2 INT;

	SELECT DISTINCT @idcechy1 = TwA_DeAId FROM cdn.twratrybuty 
		INNER JOIN cdn.defatrybuty ON TwA_DeAId = DeA_DeAId 
		INNER JOIN dbo.CtiKonfiguracja ON KNF_Numer = 5 AND DeA_Kod = KNF_Wartosc 
		WHERE TwA_CzyPrzenosicNaDostawy = 1 AND TwA_DeAId <> 0

	SELECT DISTINCT @idcechy2 = TwA_DeAId FROM cdn.twratrybuty 
		INNER JOIN cdn.defatrybuty ON TwA_DeAId = DeA_DeAId 
		INNER JOIN dbo.CtiKonfiguracja ON KNF_Numer = 19 AND DeA_Kod = KNF_Wartosc 
		where TwA_CzyPrzenosicNaDostawy = 1 AND TwA_DeAId <> 0

	SELECT TrEID AS ID,TwrId,TreLp, TwrKod, REPLACE(TwrNazwa,CHAR(10), ' ') AS TwrNazwa, DostId, NumerPelny, SUM(iloscDostawa) AS ilosc, Cena
		, Cecha1DeAId, Cecha1Kod, CechaWartosc, Cecha2DeAId, Cecha2Kod, Cecha2Wartosc
		, JM, Thdomyslna, idTh,ZPilosc, status, StatusNazwa, IloscPW, Thzdomyslna, idThz
	FROM 
	(
		SELECT b.*
		, COALESCE(thWyb.CTN_Nazwa, thDom.CTN_Nazwa,'') AS  Thdomyslna
		, COALESCE(thWyb.CTN_ID, thDom.CTN_ID, 0) AS idTh

		, COALESCE(thzWyb.CTZN_Nazwa, thzDom.CTZN_Nazwa, '') AS  Thzdomyslna
		, COALESCE(thzWyb.CTZN_ID, thzDom.CTZN_ID, 0) AS idThz

		FROM
		(
			SELECT 
				tre.TrE_TypDokumentu Trn_Typ,
				tre.TrE_lp AS TreLp, 
				tre.TrE_TrEID AS TrEID, 
				tre.TrE_TwrId AS TwrId, 
				tre.TrE_TwrKod AS TwrKod, 
				tre.TrE_twrNazwa AS TwrNazwa, 
				tre.TrE_JMZ AS JM
				, ISNULL(TsD_TrSIdDost,0) AS DostId, ISNULL(TsD_NumerPelny,'') AS NumerPelny
				, CAST(ISNULL(TsD_Ilosc,tre.TrE_ilosc) AS DECIMAL(15,4)) AS iloscDostawa, 
				tre.TrE_cenaW AS Cena
				, ISNULL(@idcechy1, 0) AS Cecha1DeAId
				, ISNULL(@KodCechy1, '') AS Cecha1Kod
				, dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1) AS CechaWartosc
				, ISNULL(@idcechy2, 0) AS Cecha2DeAId
				, ISNULL(@KodCechy2, '') AS Cecha2Kod
				, dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 2) AS Cecha2Wartosc
				, COALESCE(
					(SELECT SUM(czzn_ilosc) FROM dbo.CtiZlecenieZNag WHERE ISNULL(czzn_DokZwiazTreId,0) = tre.TrE_TrEID AND  CZZN_DokZwiazTyp = 308)
					,(SELECT SUM(czn_ilosc) FROM dbo.CtiZlecenieNag WHERE ISNULL(czn_DokZwiazTreId,0) = tre.TrE_TrEID AND  CZN_DokZwiazTyp = 308)
					,0
				) AS ZPilosc
				, 0 AS status
				, '' AS StatusNazwa
				, ISNULL((SELECT SUM(e.TrE_ilosc) FROM dbo.CtiZlecenieNag 
						INNER JOIN CtiZlecenieDok ON CZD_CZNId = CZN_ID 
						INNER JOIN CDN.TraNag AS n ON CZD_TrnId = n.TrN_TrNID  
						INNER JOIN CDN.TraElem AS e ON n.TrN_TrNID  = e.TrE_TrNID AND e.TrE_TwrId = CZN_TwrId
						WHERE n.TrN_TypDokumentu = 303  AND n.trn_anulowany = 0 
							AND ISNULL(czn_DokZwiazTreId,0) = tre.TrE_TrEID AND CZN_DokZwiazTyp = 308 
					),0) AS IloscPW

			FROM cdn.traelem AS tre 
			LEFT JOIN cdn.TraSElemDost ON tre.TrE_TrEID = ISNULL(TsD_TrEID,0)
			WHERE tre.TrE_trnid = @id
		) b
		LEFT JOIN dbo.CtiTechnolDomyslna ON ISNULL(CTD_TwrId,0) = b.TwrId
		LEFT JOIN dbo.CtiTraElem ON CTre_TreId = b.TrEID

		LEFT JOIN dbo.CtiTechnolNag thDom ON CTD_ID = thDom.CTN_ID
		LEFT JOIN dbo.CtiTechnolNag thWyb ON CTre_ThId = thWyb.CTN_ID

		LEFT JOIN dbo.CtiTechnolZNag thzDom ON CTZN_TwrId = b.TwrId AND thzDom.CTZN_Domyslna = 1
		LEFT JOIN dbo.CtiTechnolZNag thzWyb ON CTre_ThzId = thzWyb.CTZN_ID
	) a
	GROUP BY 
		TrEID,TwrId,TreLp,TwrKod,TwrNazwa,DostId,NumerPelny,Cena,Cecha1DeAId, Cecha1Kod,CechaWartosc,Cecha2DeAId,Cecha2Kod,Cecha2Wartosc,JM
		,Thdomyslna,idTh,ZPilosc,status,StatusNazwa,IloscPW, Thzdomyslna, idThz,Trn_Typ
	HAVING CASE Trn_Typ WHEN 312 THEN sum(iloscDostawa) ELSE 1 END >0
	ORDER BY TreLp

END
GO
```

# [dbo].[CTIProdukcjaDostepneDostawyList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaDostepneDostawyList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaDostepneDostawyList]  @id INT, @PrcId INT
AS
-- lista zasobów dla MM

declare @Korekta INT
declare @MagId INT
declare @zpId INT
declare @typMM INT --6 kontrola -- 3 wydanie 
declare @cecha1 varchar(50)
declare @cecha2 varchar(50)

select @MagId = TrN_MagZrdId from cdn.TraNag INNER JOIN cdn.Magazyny on TrN_MagZrdId = Mag_MagId where  TrN_TrNId = @id
select @zpId = ISNULL(CZD_CZNId,0),@typMM = ISNULL(czd_typ,0), @Korekta = ISNULL(CZD_Korekta,0) from dbo.CtiZlecenieDok where CZD_TrnId = @id
--cechy
declare @cechaKod varchar(40) = '' SELECT  @cechaKod = KNF_Wartosc FROM  CtiKonfiguracja where KNF_Numer = 5
declare @cechaId int = -1 select @cechaId = DeA_DeAId from CDN.DefAtrybuty where DeA_Kod = @cechaKod
declare @cechaKod2 varchar(40) = '' SELECT  @cechaKod2 = KNF_Wartosc FROM  CtiKonfiguracja where KNF_Numer = 19
declare @cechaId2 int = -1 select @cechaId2 = DeA_DeAId from CDN.DefAtrybuty where DeA_Kod = @cechaKod2

declare @66 INT = 0 SELECT @66 = ISNULL(KNF_Wartosc,1) FROM  CtiKonfiguracja where KNF_Numer = 66--20160704 kontrola zasobów przyjêtych dokumentem MM

declare @prmPrc int = 0 SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> '' --61.      Parametr – Przydzielenie do ZP
--1.  Brak przydzielenia
--2.  Przydzielenie pracownika i ilość
--3.  Przydzielenie pracownik bez ilości
--4.  Przydzielenie Działy -- tylko jeden dział
if @prmPrc in(1,4) set @PrcId = 0 

            
CREATE TABLE #TmpM (TwrId INT, IloscW DECIMAL(15, 4), IdDost INT)      


if (@Korekta > 0)--korekta
begin
      insert into #TmpM (TwrId, IloscW, IdDost)
      SELECT  Twz_TwrId as TwrId
            , case when  sum(TwZ_Ilosc) < sum(Ilosc) then sum(TwZ_Ilosc) else sum(Ilosc) end as IloscDostepna
            , TrS_TrSIdDost as DostId
      FROM 
            (--wszystkie dostawy uzyte w MM 
            select sum(trs_ilosc) as Ilosc, TrS_TrSIdDost as DstID
            from cdn.TraNag  
                  INNER JOIN cdn.TraElem on TrE_TrNId = TrN_TrNID 
                  INNER JOIN cdn.TraSElem on TrS_TrEId = TrE_TrEID 
            where TrN_TrNID = @Korekta and TrN_Bufor=0 and TrS_Ilosc > 0           -- operacja z jednego magazynu
            group by TrS_TrSIdDost
            ) MMDst
            INNER JOIN cdn.TwrZasoby on ISNULL(TwZ_TrSIdDost,0) =DstID
            INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
            LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
            LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
      WHERE TwZ_MagId = @MagId     AND (TrN_Korekta=0 or TrN_Korekta is null)
      group by Twz_TwrId,TrS_TrSIdDost
end
else if (@Korekta = 0  and (@typMM = 6 and @66 = 1))
begin
      --dostepne towary dla dokumentu TRN na danych magazynach, sprawdznie kokretnych dostaw
      insert into #TmpM (TwrId, IloscW, IdDost)
      SELECT  Twz_TwrId as TwrId
            , case when  sum(TwZ_Ilosc) < sum(Ilosc) then sum(TwZ_Ilosc) else sum(Ilosc) end as IloscDostepna
            , TrS_TrSIdDost as DostId
      FROM (
                  select sum(Ilosc) as Ilosc , DstID
                  from dbo.CTIProdukcjaDostawyNaZP(@zpId)--ZP
                  where doktyp in (4,6) and--przychód i rozchód produktu
                  (
                        (@PrcId = 0)--dostawy bez pracownika
                        or
                        ((@PrcId > 0) and PrcId = @PrcId)--PRC--konkretny pracownik
                  )
                  group by DstID
            ) ZpDst
            INNER JOIN cdn.TwrZasoby on ISNULL(TwZ_TrSIdDost,0) = DstID
            INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
            LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
            LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
      WHERE TwZ_MagId = @MagId     AND (TrN_Korekta=0 or TrN_Korekta is null)
      group by Twz_TwrId, TrS_TrSIdDost
end   
else
begin
      insert into #TmpM (TwrId, IloscW, IdDost)
      SELECT  Twz_TwrId as TwrId, sum(TwZ_Ilosc) as IloscDostepna, TrS_TrSIdDost as DostId
      FROM CDN.TraElem treMM
            INNER JOIN cdn.TwrZasoby on treMM.TrE_TwrId = TwZ_TwrId
            INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
            LEFT JOIN cdn.TraElem treDst ON treDst.TrE_TrEId = TrS_TrEId
            LEFT JOIN cdn.TraNag ON TrN_TrNId = treDst.TrE_TrNId
      WHERE treMM.tre_trnid = @id and    TwZ_MagId = @MagId      AND (TrN_Korekta=0 or TrN_Korekta is null)
      group by Twz_TwrId, TrS_TrSIdDost
end




      SELECT  TwrId,Twr_kod as TwrKod, IdDost as DostId
                  , ISNULL(TrN_NumerPelny,'Dokument usuniety') AS NumerPelny
                  , IloscW as IloscDostepna
                  , @cechaId as DeAId
				  , dbo.CtiProdukcjaPobierzCeche(TrS_TrSId, 1) as CechaWartosc  
                  , Cecha = @cechaKod
                  , case ISNULL(twa_deaid,0) when 0 then 2 else 1 end as Typ ,twr_jm as JM
                  , dbo.CtiProdukcjaPobierzCeche(TrS_TrSId, 1) as CechaWartoscWidok
                  , dbo.CtiProdukcjaPobierzCeche(TrS_TrSId, 2) as Cecha2Wartosc
                  , Cecha2 = @cechaKod2
                  , trn_dataope as DostData
      from #TmpM
            INNER JOIN cdn.TraSElem ON IdDost = TrS_TrSId AND TrS_Typ=1
            INNER JOIN CDN.Towary ON TwrId = Twr_GIDNumer 
            LEFT JOIN cdn.TraSElemCechy ON IdDost = ISNULL(TsC_TrSId,0)
            LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId 
            LEFT JOIN cdn.TraNag ON TrN_TrNId =TrE_TrNId
            LEFT JOIN CDN.TwrAtrybuty on TwA_TwrID = TwrId and TwA_DeAId = ( select DISTINCT TwA_DeAId from cdn.twratrybuty where TwA_DeAId = @cechaId)
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaDostepneSurowce2]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaDostepneSurowce2]  @ZPid INT, @PRCid INT,@ilosc decimal(15,4)
AS

declare @THId INT = 0
declare @MagId INT = 0
declare @cecha varchar(40) = ''
select @MagId = knf_wartosc from CtiKonfiguracja  where knf_numer = 8
--magazyn pracownika
--SELECT DISTINCT @MagId = case ISNULL(KnM_MagSId,0) when 0 then @MagId else KnM_MagSId  end FROM dbo.CtiKonfigMag WHERE KnM_PraId= @PRCid
declare @prmPrc int = 0
SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> ''	

--magazyn pracownika/działu
select DISTINCT @MagId = case ISNULL(CDZ_MagSId,0) when 0 then @MagId else CDZ_MagSId  end
FROM 	dbo.CtiZleceniePrc 
	INNER JOIN dbo.CtiDzial on ISNULL(CZO_PrcTyp,80) = 81 and CDZ_ID = czo_prcid
	INNER JOIN dbo.CtiZlecenieNag on czn_id = CZO_CZNId
WHERE  czo_cznid = @ZPid  and ISNULL(CZO_PrcTyp,80) = 81 and CZO_PrcId = @PRCid

select DISTINCT @MagId = case ISNULL(KnM_MagSId,0) when 0 then @MagId else KnM_MagSId  end
FROM 	dbo.CtiZleceniePrc 
	INNER JOIN dbo.CtiZlecenieNag on czn_id = CZO_CZNId
	LEFT JOIN dbo.CtiKonfigMag  ON czo_prcid = KnM_PraId
WHERE  czo_cznid = @ZPid    and ISNULL(CZO_PrcTyp,80) = 80 and CZO_PrcId = @PRCid







--cecha
SELECT @cecha = KNF_Wartosc FROM  CtiKonfiguracja where KNF_Numer = 5
--id technologii
select @THId = czn_ctnid from dbo.CtiZlecenieNag where czn_id = @ZPid

CREATE TABLE #TmpM (LP INT, TwrId INT, TwrKod VARCHAR(40), IloscDostepna DECIMAL(15, 4), CechaWartosc VARCHAR(40)
,cecha VARCHAR(40),TYP	INT, Jm NVARCHAR(20),nrCechy int,IdDost INT, IloscWykorzystana DECIMAL(15, 4), DstDokNumer nvarchar(40))


insert into #TmpM (LP, TwrId , TwrKod , IloscDostepna , CechaWartosc ,cecha , TYP , Jm ,nrCechy,IdDost,IloscWykorzystana ,DstDokNumer)
select  ROW_NUMBER() OVER (ORDER BY Twrkod) , TwrId , TwrKod , IloscDostepna , CechaWartosc ,cecha , TYP	, Jm ,nrCechy,IdDost ,0,DstDokNumer from
(


	--dostepne towary dla danego kodu dostawy na danych magazynach
		SELECT TwrId = Twz_TwrId ,TwrKod = Twr_kod --DostId = TrS_TrSIdDost, 
			--ISNULL(TrN_NumerPelny,'Dokument usuniety') AS NumerPelny, 
			, IloscDostepna = sum(TwZ_Ilosc) -- Wartosc = TwZ_Wartosc,--  Cena = TwZ_Cena,
				--DeAId = ISNULL(TsC_Cecha1_DeAId,0) , 
			, CechaWartosc = case when ISNULL(tsc_cecha1_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(tsc_cecha1_Wartosc,0),0), 23)
				else ISNULL(tsc_cecha1_Wartosc,'') end, cecha = ISNULL(TsC_Cecha1_Kod,'')
			, TYP = case ISNULL(twa_deaid,0) when 0 then 2 else 1 end, jm = twr_jm ,nrCechy= 1
				--,case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end as IdDost
			, DstDokNumer =ISNULL(TrN_NumerPelny,'Dokument usuniety')
			, ISNULL(TrS_TrSIdDost,0)  as IdDost
		FROM cdn.TwrZasoby
			JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
			LEFT JOIN cdn.TraSElemCechy ON TsC_TrSId = TrS_TrSId
			LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
			LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
			LEFT JOIN CDN.Towary ON Twz_TwrId = Twr_GIDNumer 
			left join dbo.CtiTechnolElem on CTE_CTNId = @THId and CTE_TwrId = Twz_TwrId
			LEFT JOIN CDN.TwrAtrybuty on TwA_TwrID = Twz_TwrId and TwA_DeAId = 
				( select DISTINCT TwA_DeAId from cdn.twratrybuty INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId where  DeA_Kod = @cecha)
		WHERE TwZ_MagId = @MagId AND (TrN_Korekta=0 or TrN_Korekta is null)
			and ((ISNULL(twa_deaid,0) = 0) or ((ISNULL(twa_deaid,0) > 0) and (ISNULL(TsC_Cecha1_Kod,'') = @cecha or  ISNULL(TsC_Cecha1_Kod,'') = '')) )
			and  CTE_CTNId = @THId
		group by --case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end,
			Twz_TwrId,Twr_kod,ISNULL(TsC_Cecha1_Wartosc,''),ISNULL(TsC_Cecha1_Kod,''),twa_deaid,twr_jm,ISNULL(tsc_cecha1_format,0),ISNULL(tsc_cecha1_Wartosc,0)
			, ISNULL(TrN_NumerPelny,'Dokument usuniety') 
			, ISNULL(TrS_TrSIdDost,0) 
		
		union
		SELECT TwrId = Twz_TwrId ,TwrKod = Twr_kod
			, IloscDostepna = sum(TwZ_Ilosc)
			, CechaWartosc = case when ISNULL(tsc_cecha2_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(tsc_cecha2_Wartosc,0),0), 23)
			else ISNULL(tsc_cecha2_Wartosc,'') end, cecha = ISNULL(TsC_Cecha2_Kod,'')
			, TYP = case ISNULL(twa_deaid,0) when 0 then 2 else 1 end, jm = twr_jm,nrCechy= 2 
			--,case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end as IdDost
			, DstDokNumer =ISNULL(TrN_NumerPelny,'Dokument usuniety')
			, ISNULL(TrS_TrSIdDost,0)  as IdDost
		FROM cdn.TwrZasoby
			JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
			LEFT JOIN cdn.TraSElemCechy ON TsC_TrSId = TrS_TrSId
			LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
			LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
			LEFT JOIN CDN.Towary ON Twz_TwrId = Twr_GIDNumer 
			left join dbo.CtiTechnolElem on CTE_CTNId = @THId and CTE_TwrId = Twz_TwrId
			LEFT JOIN CDN.TwrAtrybuty on TwA_TwrID = Twz_TwrId and TwA_DeAId = 
				( select DISTINCT TwA_DeAId from cdn.twratrybuty INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId where  DeA_Kod = @cecha)
		WHERE TwZ_MagId = @MagId AND (TrN_Korekta=0 or TrN_Korekta is null)
			and ( (ISNULL(twa_deaid,0) > 0) and ISNULL(TsC_Cecha2_Kod,'') = @cecha )
			and  CTE_CTNId = @THId
		group by  --case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end,
			Twz_TwrId,Twr_kod,ISNULL(TsC_Cecha2_Wartosc,''),ISNULL(TsC_Cecha2_Kod,''),twa_deaid,twr_jm,ISNULL(tsc_cecha2_format,0),ISNULL(tsc_cecha2_Wartosc,0)
			, ISNULL(TrN_NumerPelny,'Dokument usuniety') 
			, ISNULL(TrS_TrSIdDost,0)  
		union
		SELECT TwrId = Twz_TwrId ,TwrKod = Twr_kod
			, IloscDostepna = sum(TwZ_Ilosc)
			, CechaWartosc = case when ISNULL(tsc_cecha3_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(tsc_cecha3_Wartosc,0),0), 23)
				else ISNULL(tsc_cecha3_Wartosc,'') end , cecha = ISNULL(TsC_Cecha3_Kod,'')
			, TYP = case ISNULL(twa_deaid,0) when 0 then 2 else 1 end, jm = twr_jm,nrCechy= 3 
				--,case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end as IdDost
			, DstDokNumer =ISNULL(TrN_NumerPelny,'Dokument usuniety')
			, ISNULL(TrS_TrSIdDost,0)  as IdDostn
		FROM cdn.TwrZasoby
			JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
			LEFT JOIN cdn.TraSElemCechy ON TsC_TrSId = TrS_TrSId
			LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
			LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
			LEFT JOIN CDN.Towary ON Twz_TwrId = Twr_GIDNumer 
			left join dbo.CtiTechnolElem on CTE_CTNId = @THId and CTE_TwrId = Twz_TwrId
			LEFT JOIN CDN.TwrAtrybuty on TwA_TwrID = Twz_TwrId and TwA_DeAId = 
				( select DISTINCT TwA_DeAId from cdn.twratrybuty INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId where  DeA_Kod = @cecha)
		WHERE TwZ_MagId = @MagId AND (TrN_Korekta=0 or TrN_Korekta is null)
			and ( (ISNULL(twa_deaid,0) > 0) and ISNULL(TsC_Cecha3_Kod,'') = @cecha )
			and  CTE_CTNId = @THId
		group by  --case when ISNULL(twa_deaid,0) = 0 then  0 else ISNULL(TwZ_TrSIdDost,0) end,
			Twz_TwrId,Twr_kod,ISNULL(TsC_Cecha3_Wartosc,''),ISNULL(TsC_Cecha3_Kod,''),twa_deaid,twr_jm,ISNULL(tsc_cecha3_format,0),ISNULL(tsc_cecha3_Wartosc,0)
			, ISNULL(TrN_NumerPelny,'Dokument usuniety') 
			, ISNULL(TrS_TrSIdDost,0)
		
) a
 order by typ,TwrKod,IdDost asc


--kursor po tym co ma być
--jężeli nie ma dodać z zerowa ilością


DECLARE kursor cursor FOR
    SELECT  CTE_TwrId, (@ilosc*CTE_Ilosc)/ISNULL(ctn_ilosc,1), twr_kod, twr_JM
	FROM  dbo.CtiTechnolNag INNER JOIN dbo.CtiTechnolElem on ctn_id = Cte_ctnid
	INNER JOIN cdn.towary on CTE_TwrId = twr_twrid
	where CTE_CTNId =@THId and cte_typ in (1,2)
DECLARE @CTE_TwrId int, @CTE_Ilosc decimal(15,4), @twr_kod varchar(40), @twr_JM nvarchar(20)
    OPEN kursor

FETCH kursor   INTO @CTE_TwrId , @CTE_Ilosc,@twr_kod , @twr_JM

while @@FETCH_STATUS = 0
begin
        declare @iloscWykorzystana decimal(15,4)=  @CTE_Ilosc   
            DECLARE kursor2 cursor FOR
				SELECT  lp,IloscDostepna FROM #TmpM where TwrId = @CTE_TwrId order by lp
			DECLARE @lp int,@IloscDostepna decimal(15,4) OPEN kursor2
			FETCH kursor2 	INTO @lp,@IloscDostepna 

			while @@FETCH_STATUS = 0
			begin
				if(@iloscWykorzystana>0)
				begin
					if(@IloscDostepna<=@iloscWykorzystana)
					begin
						update #TmpM set IloscWykorzystana = @IloscDostepna where lp = @lp
						set @iloscWykorzystana = @iloscWykorzystana-@IloscDostepna
					end
					else
					begin
						update #TmpM set IloscWykorzystana = @iloscWykorzystana where lp = @lp
						set @iloscWykorzystana = 0
					end
				end    
						        
				FETCH kursor2 INTO @lp,@IloscDostepna 
			end
			CLOSE kursor2 deallocate     kursor2

        if(@iloscWykorzystana>0)
		begin 
		   insert into #TmpM (LP, TwrId , TwrKod , IloscDostepna , CechaWartosc ,cecha , TYP , Jm ,nrCechy,IdDost,IloscWykorzystana )
			select  MAX(LP)+1 , @CTE_TwrId , @twr_kod , 0 , '' ,'' , 2	, @twr_JM ,1,0 ,@iloscWykorzystana from #TmpM
        end  
            
       FETCH kursor   INTO @CTE_TwrId , @CTE_Ilosc,@twr_kod , @twr_JM
end

CLOSE kursor
deallocate     kursor




	SELECT   TwrId , TwrKod , IloscDostepna , CechaWartosc ,cecha , TYP , Jm 
		, case when ISNULL(cecha,'') = '' then 0 else nrCechy end as nrCechy
		, IdDost,IloscWykorzystana , ISNULL(DstDokNumer,'') as DstDokNumer
		, replace(Twr_Nazwa,char(10), ' ')  as Nazwa
	FROM #TmpM INNER JOIN CDN.Towary on  TwrId = twr_twrid  where IloscWykorzystana > 0
	order by case when IloscWykorzystana > 0 then 0 else 1 end ,TwrKod,IdDost

DROP TABLE #TmpM

GO
```

# [dbo].[CTIProdukcjaDuplikatySesji]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaDuplikatySesji]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaDuplikatySesji]
@id int
AS
BEGIN
	--wycinanie starszych sesji duplikatów
	DECLARE kursor cursor FOR
			SELECT KnS_COId as OpeId,max(KnS_Id) as SesjaId FROM dbo.CtiKonfigSesja
			where KnS_Id <> @id --poza obecną sesją
			group by KnS_COId, KnS_Typ
			having count(*)> 1
	DECLARE @OpeId int, @SesjaId int OPEN kursor

	FETCH kursor INTO @OpeId,@SesjaId

	while @@FETCH_STATUS = 0
	begin
		delete from CtiKonfigSesja where  KnS_COId = @OpeId and KnS_Id <> @SesjaId
	            
		FETCH kursor INTO @OpeId, @SesjaId
	end

	CLOSE kursor deallocate     kursor

END
GO
```

# [dbo].[CTIProdukcjaDzialL]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaDzialL]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaDzialL] 
@ZP int--0 wszystkie-- idZP pracownicy powiazani z ZP
AS
	declare @MagazynDomyslySurId int = 0
	declare @MagazynDomyslySur varchar(40) = '' 
	declare @MagazynDomyslyProdId int =0
	declare @MagazynDomyslyProd varchar(40) = '' 
	select @MagazynDomyslySurId = knf_wartosc from CtiKonfiguracja  where knf_numer = 8
	select @MagazynDomyslyProdId = knf_wartosc from CtiKonfiguracja  where knf_numer = 9
	select @MagazynDomyslySur = mag_symbol from cdn.magazyny where mag_magid = @MagazynDomyslySurId
	select @MagazynDomyslyProd = mag_symbol from cdn.magazyny where mag_magid = @MagazynDomyslyProdId

	SELECT DISTINCT CDZ_ID as ID, CDZ_Kod as Kod--DZIAL
		, case isnull(CDZ_MagSId,0) when 0 then '' else (select Mag_Symbol from cdn.magazyny where Mag_MagId =CDZ_MagSId) end  as MagSur
		, CDZ_MagSId As MagSurId
		, case isnull(CDZ_MagPId,0) when 0 then '' else (select Mag_Symbol from cdn.magazyny where Mag_MagId =CDZ_MagPId) end  as MagProd
		, CDZ_MagPId As MagProdId
	FROM dbo.CtiDzial
		LEFT JOIN dbo.CtiZleceniePrc ON CDZ_ID = isnull(CZO_PrcId,0)
	WHERE ((@ZP<>0  and isnull(CZO_CZNId,0) = @ZP)OR (@ZP = 0))
	order by CDZ_Kod
GO
```

# [dbo].[CTIProdukcjaGanttUpdate]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaGanttUpdate]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaGanttUpdate]
@Title varchar(20),
@datStart varchar(20),
@datStop varchar(20)
AS
BEGIN
declare @Start datetime set @Start = convert(datetime, @datStart,120)
declare @End datetime set @End= convert(datetime, @datStop,120)

	declare @Start2 datetime
	declare @End2 datetime
	declare @komunikat varchar(200) = ''
	declare @ID int
	declare @status int
	select @ID = CZN_ID,@status = dbo.CTIProdukcjaStatusy(2, CZN_ID),
	@Start2 = CZN_CzasStart,@End2 = CZN_CzasEnd
	from dbo.CtiZlecenieNag where CZN_NrPelny = @Title
	
	if (@Start2 = @Start and @End2 = @End)
	begin
		set @komunikat ='-'
	end
	else
	begin
		if @status = 0
		begin
			update dbo.CtiZlecenieNag  set CZN_CzasStart= @Start, CZN_CzasEnd = @End
			from  dbo.CtiZlecenieNag where  CZN_ID = @ID
			set @komunikat = @Title+' - zapis zakończony poprawnie'
		end
		else
		begin
			set @komunikat = @Title+' - zapis niemożliwy zlecenie potwierdzone'
		end
	end
	
	select @komunikat as komunikat
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CtiProdukcjaHistoriaDostawySurowca]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CtiProdukcjaHistoriaDostawySurowca]
 @TwrID INT,
 @Typ INT,--typ 1 - surowiec--typ 2 produkt
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmWszystkie int

AS

declare @dane table (
	ID int IDENTITY (1, 1)  ,
	TrNId int,
	trEId int,
	twrid INT,
	twrkod NVARCHAR(50),
	NumerPelny NVARCHAR(30),
	NumerString NVARCHAR(30),
	NumerNr int,
	Korekta int,
	typdok  NVARCHAR(3),
	typdokumentu int,
	Data DateTime,
	Podmiot NVARCHAR(50),
	jm NVARCHAR(20),
	iloscjm DECIMAL(15,4),
	wartoscnetto DECIMAL(15,4),
	magazyn NVARCHAR(50),
	twrnazwa NVARCHAR(255),
	trs_trsiddost int,
	bufor int,
	vanid int,
	dekid int,
	Cecha1_DeAId int ,
	Cecha1_Kod NVARCHAR(20),
	Cecha1_Format smallint ,
	Cecha1_Wartosc NVARCHAR(255) ,
	Cecha2_DeAId int ,
	Cecha2_Kod NVARCHAR(20),
	Cecha2_Format smallint ,
	Cecha2_Wartosc NVARCHAR(255) ,
	CZN_ID int,
	CZN_NrPelny varchar(20),
	ProduktId int,
	ProduktKod NVARCHAR(50),
	listaPW varchar(1000),
	listaWZ varchar(1000),
	Cecha3_DeAId int ,
	Cecha3_Kod NVARCHAR(20),
	Cecha3_Format smallint ,
	Cecha3_Wartosc NVARCHAR(255),
	DokumentDostawy NVARCHAR(30)
)

	declare @KodCechy1 varchar(200) = (select top 1 knf_wartosc from dbo.CtiKonfiguracja where knf_numer = 5)
	declare @KodCechy2 varchar(200) = (select top 1 knf_wartosc from dbo.CtiKonfiguracja where knf_numer = 19)
	declare @idcechy1 int;
	declare @idcechy2 int;

	select DISTINCT @idcechy1 = TwA_DeAId from cdn.twratrybuty 
		INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId 
		INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 5 and DeA_Kod = KNF_Wartosc 
		where TwA_CzyPrzenosicNaDostawy = 1 and TwA_DeAId <> 0

	select DISTINCT @idcechy2 = TwA_DeAId from cdn.twratrybuty 
		INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId 
		INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 19 and DeA_Kod = KNF_Wartosc 
		where TwA_CzyPrzenosicNaDostawy = 1 and TwA_DeAId <> 0


if	(@typ = 1) 
begin
	insert into @dane(
		TrnId, TrEId,twrid,twrkod,NumerPelny, NumerString, NumerNr, Korekta,typdok,typdokumentu,Data,
		Podmiot,jm,iloscjm,wartoscnetto, magazyn, twrnazwa,trs_trsiddost,bufor,vanid,dekid,
		Cecha1_DeAId,Cecha1_Kod,Cecha1_Format,Cecha1_Wartosc,
		Cecha2_DeAId,Cecha2_Kod,Cecha2_Format,Cecha2_Wartosc,
		CZN_ID,CZN_NrPelny,ProduktId,ProduktKod,listaPW,listaWZ,
		Cecha3_DeAId,Cecha3_Kod,Cecha3_Format,Cecha3_Wartosc
		, DokumentDostawy
	)
	select distinct TrN_TrNId, TrE_TrEId,surowiec.twr_twrid, surowiec.twr_kod , Trn_NumerPelny, Trn_NumerString, Trn_NumerNr, TrN_Korekta , CDN.TypDokumentu(trn_typdokumentu),TrN_TypDokumentu,  trn_dataWys , Pod_kod ,
		 surowiec.twr_jm , case when Trn_TypDokumentu = 312 then ABS(trs_ilosc) else trs_ilosc end  as  trs_ilosc,
		 case when Trn_TypDokumentu = 312 then ABS(trs_wartosc)else case when trs_typ = 4 and trs_wartosc =0 then tre_wartoscnetto else trs_wartosc end end as trs_wartosc,
		 mag_symbol,replace(surowiec.Twr_Nazwa,char(10), ' '),trs_trsiddost,trn_bufor,trn_vanid,trn_dekid
		
		,@idcechy1, @KodCechy1, tsd_cecha1_format, dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1)
		
		, TsD_Cecha2_DeAId , TsD_Cecha2_Kod, TsD_Cecha2_Format, TsD_Cecha2_Wartosc
		, CZN_ID, CZN_NrPelny
		, produkty.twr_twrid as ProduktId, produkty.twr_kod as ProduktKod
		, dbo.CTIProdukcjaListaPWzZP(ISNULL(CZN_ID,0),ISNULL(produkty.twr_twrid,0)) as listaPW
		, dbo.CTIProdukcjaListaWZzZP(ISNULL(CZN_ID,0),ISNULL(produkty.twr_twrid,0)) as listaWZ 
		, TsD_Cecha3_DeAId , TsD_Cecha3_Kod, TsD_Cecha3_Format, TsD_Cecha3_Wartosc
		, ISNULL(TsD_NumerPelny,Trn_NumerPelny)  as DokumentDostawy
	from cdn.traselem
		LEFT JOIN cdn.TraSElemDost on TsD_TrSIdDost = TrS_TrSIdDost and TsD_TrEId = TrS_TrEId
		join cdn.towary surowiec on trs_twrid = surowiec.twr_twrid
		join cdn.traelem on trs_treid = tre_treid
		join cdn.tranag on trn_trnid = tre_trnid
		join cdn.podmiotyview on pod_podmiottyp = trn_podmiottyp and pod_podid = trn_podid
		join cdn.magazyny on mag_magid = trs_magid
		LEFT JOIN dbo.CtiZlecenieDok on CZD_TrnId = TrN_TrNId and czd_typ <> 4
		LEFT JOIN dbo.CtiZlecenieNag on CZD_CZNId = CZN_ID
		LEFT JOIN cdn.towary as produkty on produkty.twr_twrid = CZN_TwrId
	where Trn_TypDokumentu IN (301,302,305,303,304,306,307, 312)
		and (surowiec.Twr_TwrID = @TwrID OR @TwrID=0)
		AND (TrN_Bufor in (0,1))
		and ISNULL(czd_typ,0) <> 4
		AND (TrN_Korekta in (0,1) or (Trn_TypDokumentu=312 and trs_ilosc>0))
			and (
				(ISNULL(@prmWszystkie,0) =0) 
				or
				(@prmWszystkie =1 and CONVERT(VARCHAR, ISNULL(trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)

end
else
begin
	insert into @dane(
		TrnId, TrEId,twrid,twrkod,NumerPelny, NumerString, NumerNr, Korekta,typdok,typdokumentu,Data,
		Podmiot,jm,iloscjm,wartoscnetto, magazyn, twrnazwa,trs_trsiddost,bufor,vanid,dekid,
		Cecha1_DeAId,Cecha1_Kod,Cecha1_Format,Cecha1_Wartosc,
		Cecha2_DeAId,Cecha2_Kod,Cecha2_Format,Cecha2_Wartosc,
		CZN_ID,CZN_NrPelny,ProduktId,ProduktKod,listaPW,listaWZ
		,Cecha3_DeAId,Cecha3_Kod,Cecha3_Format,Cecha3_Wartosc
		, DokumentDostawy
	)
	select distinct TrN_TrNId, TrE_TrEId,surowiec.twr_twrid, surowiec.twr_kod , Trn_NumerPelny, Trn_NumerString, Trn_NumerNr, TrN_Korekta , CDN.TypDokumentu(trn_typdokumentu),TrN_TypDokumentu,  trn_dataWys , Pod_kod ,
		 surowiec.twr_jm , case when Trn_TypDokumentu = 312 then ABS(trs_ilosc) else trs_ilosc end  as  trs_ilosc,
		 case when Trn_TypDokumentu = 312 then ABS(trs_wartosc)else case when trs_typ = 4 and trs_wartosc =0 then tre_wartoscnetto else trs_wartosc end end as trs_wartosc,
		 mag_symbol,replace(surowiec.Twr_Nazwa,char(10), ' '),trs_trsiddost,trn_bufor,trn_vanid,trn_dekid

		 ,@idcechy1, @KodCechy1, tsd_cecha1_format, dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1)
		, TsD_Cecha2_DeAId , TsD_Cecha2_Kod, TsD_Cecha2_Format, TsD_Cecha2_Wartosc
		, CZN_ID, CZN_NrPelny
		, produkty.twr_twrid as ProduktId, produkty.twr_kod as ProduktKod
		, dbo.CTIProdukcjaListaPWzZP(ISNULL(CZN_ID,0),ISNULL(produkty.twr_twrid,0)) as listaPW
		, dbo.CTIProdukcjaListaWZzZP(ISNULL(CZN_ID,0),ISNULL(produkty.twr_twrid,0)) as listaWZ
		, TsD_Cecha3_DeAId , TsD_Cecha3_Kod, TsD_Cecha3_Format, TsD_Cecha3_Wartosc
		, ISNULL(TsD_NumerPelny,Trn_NumerPelny)  as DokumentDostawy
from cdn.traselem
		--left join cdn.traselemcechy on tsc_trsid = trs_trsid
		LEFT JOIN cdn.TraSElemDost on TsD_TrSIdDost = TrS_TrSIdDost and TsD_TrEId = TrS_TrEId
		join cdn.towary surowiec on trs_twrid = surowiec.twr_twrid
		join cdn.traelem on trs_treid = tre_treid
		join cdn.tranag on trn_trnid = tre_trnid
		join cdn.podmiotyview on pod_podmiottyp = trn_podmiottyp and pod_podid = trn_podid
		join cdn.magazyny on mag_magid = trs_magid
		LEFT JOIN dbo.CtiZlecenieDok on CZD_TrnId = TrN_TrNId
		LEFT JOIN dbo.CtiZlecenieNag on CZD_CZNId = CZN_ID
		LEFT JOIN cdn.towary as produkty on produkty.twr_twrid = CZN_TwrId
	where (Trn_TypDokumentu = 304)
		AND (TrN_Bufor in (0,1))
		and ISNULL(czd_typ,0) <> 4
		AND TrN_Bufor <> -1 AND TrN_Korekta in (0,1)
		and ((trs_trsiddost in 
		(select distinct trs_trsiddost
		from cdn.traselem
		left join cdn.traselemcechy on tsc_trsid = trs_trsid
		join cdn.traelem on trs_treid = tre_treid
		join cdn.tranag on trn_trnid = tre_trnid
		LEFT JOIN dbo.CtiZlecenieDok on CZD_TrnId = TrN_TrNId
		LEFT JOIN dbo.CtiZlecenieNag on CZD_CZNId = CZN_ID
		LEFT JOIN cdn.towary as produkty on produkty.twr_twrid = CZN_TwrId
		where (Trn_TypDokumentu IN (301,302,305,303,304,306,307 ) )
		and (CZN_TwrId = @TwrID)
		AND (TrN_Bufor in (0,1))
		and ISNULL(czd_typ,0) <> 4
		AND TrN_Bufor <> -1 AND TrN_Korekta in (0,1)))
			and (
				(ISNULL(@prmWszystkie,0) =0) 
				or
				(@prmWszystkie =1 and CONVERT(VARCHAR, ISNULL(trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)
		)
end


--SET NOCOUNT OFF
select distinct TrNId, TrEId, twrid,twrkod as 'kod',Korekta,typdok,typdokumentu, data, podmiot
	, jm as 'jm', iloscjm , wartoscnetto
	, 1 as 'poziom', magazyn,TwrNazwa as 'TwrNazwa', NumerPelny, NumerString, NumerNr
	, trs_trsiddost,bufor as bufor ,vanid as vanid,dekid
	, Cecha1_Kod as 'CechaKod1'  
	, case when ISNULL(cecha1_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(cecha1_Wartosc,0),0), 23) else ISNULL(cecha1_Wartosc,'') end as wartosc1,Cecha1_Format as format1
	, Cecha2_Kod as 'CechaKod2'  
	, case when ISNULL(cecha2_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(cecha2_Wartosc,0),0), 23) else ISNULL(cecha2_Wartosc,'') end as wartosc2
	, Cecha2_Format as format2
	, CZN_ID, CZN_NrPelny,ProduktId,ProduktKod,listaPW,listaWZ,Cecha3_Kod as 'CechaKod3' , Cecha3_Wartosc as wartosc3
	, DokumentDostawy
from @dane
order by kod,poziom,data,trs_trsiddost
--drop table @dane
GO
```

# [dbo].[CTIProdukcjaKnfNrWlasny]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaKnfNrWlasny]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaKnfNrWlasny]
 @typ int,
 @id int 

AS
BEGIN

	declare @def1 nvarchar(1000) = ''
	declare @def2 nvarchar(1000) = ''
	declare @def3 nvarchar(2000) = ''

	SELECT Top 1 
		 @def1= CASE ISNULL(KNW_DefinicjaSQL1,'') WHEN '' THEN '0'            ELSE ISNULL(KNW_DefinicjaSQL1,'') END	
		,@def2= CASE ISNULL(KNW_DefinicjaSQL2,'') WHEN '' THEN '1=1'          ELSE ISNULL(KNW_DefinicjaSQL2,'') END	
		,@def3= CASE ISNULL(KNW_DefinicjaSQL3,'') WHEN '' THEN 'Select '''''  ELSE ISNULL(KNW_DefinicjaSQL3,'') END	
	FROM dbo.CtiSlowniki 
		left join dbo.CtiKnfNrWlasny on S_Id = KNW_DokTyp and ISNULL(KNW_Domyslny,0)=1
	WHERE S_Typ = 1 and S_Id = @typ order by isnull(KNW_ID,0) desc


	declare @zapytanie nvarchar(4000)='

	declare @id int = '+cast(@id as varchar)+'
	declare @typ int = '+cast(@typ as varchar)+'
	declare @OstatniNumer int = 0

	select top 1 @OstatniNumer = isnull('+@def1+',0)
	from CtiProdukcjaNrWlasny where Typ = @typ and ID <> @id 
	and 
	( 
	'+@def2+'
	)

	'+@def3+'

	'


	execute (@zapytanie) 
END

GO
```

# [dbo].[CTIProdukcjaKonfigOgraniczeniaAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaKonfigOgraniczeniaAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaKonfigOgraniczeniaAdd]
@IdList int,
--72 lista wyrobów
--71 lista surowców
--80 Lista pracowników
--308 Lista dok RO
@Parametr int,
--0-Grupa domyślna CDN.TwrGrupy
--1-AtrybutTwr CDN.DefAtrybuty
--2-Kategoria CDN.Kategorie
--3-Wydział CDN.Dzialy
--4-Stanowisko CDN.DaneKadMod
--5-Symbol
--6-AtrybutDok CDN.DokAtrybuty
--7-Data
@WartoscTxt varchar(250),
@WartoscAtrTxt nvarchar(100)
As
begin

	if (@Parametr = 0)--"Grupa domyślna"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr, twg_TwgId,TwG_Kod,@WartoscAtrTxt FROM CDN.TwrGrupy 
		WHERE ((TwG_GIDTyp = -16) AND (TwG_GIDTyp = -16)) and TwG_Kod = @WartoscTxt
	end
	else if (@Parametr = 1)--"TWR Atrybut"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr, DEA_DEAId,DeA_Kod,@WartoscAtrTxt FROM CDN.DefAtrybuty 
		WHERE DeA_Typ =1 and DeA_Kod = @WartoscTxt
	end
	else if (@Parametr = 2)--"Kategoria"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr,Kat_Katid ,Kat_KodSzczegol,@WartoscAtrTxt FROM CDN.Kategorie  
		WHERE  Kat_KodSzczegol = @WartoscTxt
		
	end
	else if (@Parametr = 3)--"Wydział"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr, DZL_DZlID,DZL_Kod,@WartoscAtrTxt FROM CDN.Dzialy A LEFT JOIN CDN.Lokalizacje B 
		ON B.Lok_LokId = A.DZL_LokId  where DZL_Kod <> 'brak'  and DZL_Kod = @WartoscTxt
	end
	else if (@Parametr = 4)--"Stanowisko"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr, DKM_DKMId,DKM_Nazwa,@WartoscAtrTxt FROM CDN.DaneKadMod 
		WHERE (DKM_Rodzaj = 1)  and DKM_Nazwa = @WartoscTxt
	end
	else if (@Parametr = 5)--"Symbol"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		values ( @IdList,@Parametr, 0,@WartoscTxt,@WartoscAtrTxt)
	end
	else if (@Parametr = 6)--"DOK Atrybut"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		SELECT distinct @IdList,@Parametr, DEA_DEAId,DeA_Kod,@WartoscAtrTxt FROM CDN.DefAtrybuty 
		WHERE DeA_Typ = 4 and DeA_Kod = @WartoscTxt
	end
	else if (@Parametr = 7)--"Data wystawienia od"
	begin
		Insert Into dbo.CtiKonfigOgraniczenia (KnO_IdList, KnO_Parametr, KnO_WartoscId, KnO_WartoscTxt, KnO_WartoscAtrTxt)
		values ( @IdList,@Parametr, 0,@WartoscTxt,@WartoscAtrTxt)
	end

end
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaKonfiguracjaCenyKal]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
/*

	declare @typ int = 72
		-- zasob - 70
		-- surowiec - 71
		-- wyrob/produkt - 72
		-- odpad - 73
		-- usługa kooperacja - 74
		-- narzut wydziałowy - 75
	declare @id int = 0 
	declare @ilosc decimal(15,4) = 2
	declare @cenaReczna decimal(15,4) = 0 -- przekazana cena  wprowadzana ręczna
	declare @TH int = 0--cena ewidencyjna, zasoby, surowce, usługi
	declare @RW int = 0--dla surowców
	declare @PW int = 0--dla odpadów
	declare @ZP int = 0--dla zasobów
	declare @wartosc decimal(15,4) = 0
	exec dbo.CTIProdukcjaKonfiguracjaCenyKal @typ, @id, @ilosc, @cenaReczna, @TH, @PW, @RW,@ZP,  @wartosc OUTPUT
	select @wartosc

*/

create procedure [dbo].[CTIProdukcjaKonfiguracjaCenyKal]
  @typ int
	-- zasob - 70
	-- surowiec - 71
	-- wyrob/produkt - 72
	-- odpad - 73
	-- usługa kooperacja - 74
	-- narzut wydziałowy - 75

, @id int
, @ilosc decimal(15,4)
, @cenaReczna decimal(15,4)		-- przekazana cena  wprowadzana ręczna, wartość pola specjalnego
, @TH int		--wyrob - cena ewidencyjna z technologii
, @RW int		--faktyczne zużycie surowców
, @PW int		--faktyczne zużycie odpadów
, @ZP int		--planowane zużycie
, @wartosc decimal(15,4) = 0 OUTPUT
, @opcjaCenZasobow int = NULL
as
begin
set @wartosc = 0 

	declare @wartoscPrzekazana decimal(15,4) = 0 

	if (@typ = 72) -----------------------------------------------wyrob
	begin
		declare @35 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =35) --Parametr cenaEwidencyjna_cena 
	
		declare @54 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =54) --Parametr cenaStala  
			declare @55 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =55) --Parametr cenyStalej  
			--1 --Parametr wprowadzenie ręcznie ceny na PW 
			--2 --Parametr cena pobrana z pola specjalnego
			declare @56 decimal(15,4)   = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =56) --Wartosc cenyStalej  
		declare @36 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =36) --Parametr cenaKalkulacyjna 
			declare @37 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =37) --Parametr doliczajSurowice 
			declare @40 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =40) --Parametr doliczajOdpady  
			declare @43 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =43) --Parametr doliczajZasoby  
			declare @48 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =48) --Parametr doliczajNarzuty  
			declare @51 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =51) --Parametr kosztyKooperacji   
			declare @58 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =58) --Parametr Kooperacja z ZP   

			if (@54 = 1)---------
			begin
				
				if(@55 in (1,2)) select @wartosc = isnull(@cenaReczna,0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@54 = 1')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))

			end
			else if(@36 = 1)--------------kalkulacja ceny
			begin
				if(@37 = 1) -----------------------------------------------surowce
				begin

					set @wartoscPrzekazana = 0
					declare @380 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =380) 
						--Parametr Surowce 
						--1.	Parametr - ilosc Według technologii
						--2.	Parametr - ilosc Według faktycznego zużycia
					if(@380 = 1)
					begin
						exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 71, 1, @TH, @wartoscPrzekazana OUTPUT, @ilosc
						set @wartosc = @wartosc +  isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					end
					else if(@380 = 2)
					begin
						exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 71, 5, @RW, @wartoscPrzekazana OUTPUT, @ilosc
						set @wartosc = @wartosc + isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					end
					else
						set @wartosc = @wartosc +0

					PRINT('@37 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end

				if(@40 = 1) -----------------------------------------------odpad
				begin
					set @wartoscPrzekazana = 0
					declare @410 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =410) --Parametr Odpady 
						--1.	Parametr - Według technologii
						--2.	Parametr - Według faktycznego zużycia
					if(@410 = 1)
					begin
						exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 73, 1, @TH, @wartoscPrzekazana OUTPUT, @ilosc
						set @wartosc = @wartosc - isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					end
					else if(@410 = 2)
					begin
						exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 73, 4, @PW, @wartoscPrzekazana OUTPUT, @ilosc
						set @wartosc = @wartosc -  isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					end
					else
						set @wartosc = @wartosc +0

					PRINT('@40 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end
				
				if(@51 = 1) -----------------------------------------------usługa
				begin
					set @wartoscPrzekazana = 0
					declare @520 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =520) --Parametr Kooperacji 
						--1.	Parametr - Według technologii
						--2.	Parametr - Według faktycznego zużycia
					if(@520 = 1)
					begin
						exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 74, 1, @TH, @wartoscPrzekazana OUTPUT, @ilosc
						set @wartosc = @wartosc +   isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					end
					else
						set @wartosc = @wartosc +0
					--else if(@520 = 2)
					--begin
					--	exec dbo.CTIProdukcjaKonfiguracjaCenyPozycjiDok 74, 5, @PW, @wartoscPrzekazana OUTPUT
					--	set @wartosc = @wartosc +  isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)
					--end

					PRINT('@51 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end
				
				if(@43 = 1) -----------------------------------------------zasob
				begin
					set @wartoscPrzekazana = 0
					exec dbo.CTIProdukcjaKonfiguracjaCenyKal 70, 0, @ilosc, 0, 0,0,0, @ZP, @wartoscPrzekazana OUTPUT, @opcjaCenZasobow
					set @wartosc = @wartosc +   isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana)),0)

					PRINT('@43 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end

				if(@48 = 1) -----------------------------------------------narzut
				begin
					declare @49 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =49) --Parametr Narzuty 
						--1 --Parametr w zależności od wartości wyrobu gotowego
						--2 --Parametr uzależniony od masy wyrobu gotowego
						--3 --Parametr przypisany w kartotece wyrobu gotowego
						--4 --Parametr Przypisany w polu specjalnym:
					--declare @50 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =50) --Wartosc Narzuty 
					if (@49 = 3)
					begin
						set @wartoscPrzekazana = 0
						set @wartoscPrzekazana = isnull((select isnull(ctwr_narzut,0) from dbo.CtiTowary where ctwr_twrid = @id) ,0)
						set @wartosc = @wartosc +  isnull(CONVERT(decimal(15,4),( @wartoscPrzekazana * @ilosc)) ,0)
					end
					else if (@49 = 4)
					begin
						set @wartosc = @wartosc +  isnull(CONVERT(decimal(15,4),( @cenaReczna * @ilosc)) ,0)
					end
					else
						set @wartosc = @wartosc +0

					PRINT('@48 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end
				
				if(@58 = 1) -----------------------------------------------kooperacja z ZP
				begin
					--set @wartoscPrzekazana = 0
					--set @wartoscPrzekazana = isnull((select sum(isnull(cze_ilosc,0)*isnull(cze_cena,0)) as Wartosc from dbo.CtiZlecenieElem where cze_cznid = @ZP) ,0)
					--set @wartosc = @wartosc +  isnull(CONVERT(decimal(15,4),@wartoscPrzekazana) ,0)

					PRINT('@58 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end


			end else--if(@35=1)----------------cena ewidencyjna
			begin
				if (@35 = 3) -------- z ZP
				begin
					-- domyslna
					declare @1012 int = (select isnull(KNF_Wartosc,1) from dbo.CtiKonfiguracja where KNF_Numer = 1012)
					if (@1012 <= 0)
					begin
						set @1012 = 1;
					end
	
					-- wybrana
					declare @wybranaCenaZP int = (select isnull(CZN_CenaID, @1012) from dbo.CtiZlecenieNag WHERE CZN_ID = @ZP)
					set @wartosc = isnull((select isnull(TwC_Wartosc,0) from cdn.TwrCeny where TwC_TwrID = @id and TwC_TwCNumer = @wybranaCenaZP),0)
					set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)	
					
					PRINT('@35 = 3')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))			
				end
				else if (@35 = 2)---------z cennika
				begin
					declare @350 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =350) --wartosc ceny ewidencyjnej
					set @wartosc = isnull((select isnull(TwC_Wartosc,0) from cdn.TwrCeny where TwC_TwrID = @id and TwC_TwCNumer = @350),0)
					set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

					PRINT('@35 = 2')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end
				else-- z technologii
				begin
					select @wartosc = isnull(CTN_CenaEwidencyjna,0) from dbo.CtiTechnolNag where ctn_id = @TH
					set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

					PRINT('@35 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
				end
			end
	end
	else if (@typ = 71)-----------------------------------------------surowce
	begin
		declare @38 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =38) --Parametr Surowce 
			--1 --Parametr Koszty według cena RW
			--2 --Parametr Koszty według cena według cennika OPTIMA 
			--3 --Parametr Koszty według stała cena 
			--4 --Parametr Koszty według stała cena z kartoteki 
		declare @39 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =39) --Wartosc Surowce 


			if(@38 = 1)--
			begin
				set @wartosc = isnull((select sum(TrE_WartoscNetto)  from cdn.TraElem where TrE_TypDokumentu = 304 and TrE_TrNId = @RW) ,0)

				PRINT('@38 = 1')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@38 = 2)
			begin
				set @wartosc = isnull((select isnull(TwC_Wartosc,0) from cdn.TwrCeny where TwC_TwrID = @id and TwC_TwCNumer = @39),0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@38 = 2')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@38 = 3)
			begin
				set @wartosc = isnull(CONVERT(decimal(15,4),( @39 * @ilosc)) ,0)

				PRINT('@38 = 3')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@38 = 4)
			begin
				set @wartosc = isnull((select isnull(CTwr_Cena,0) from dbo.CtiTowary where CTwr_TwrId = @id) ,0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@38 = 4')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end


	end
	else if (@typ = 73) -----------------------------------------------odpad
	begin
		declare @41 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =41) --Parametr Odpady  
			--1 --Parametr Koszty według cena PW
			--2 --Parametr Koszty według cena według cennika OPTIMA 
			--3 --Parametr Koszty według stała cena 
			--4 --Parametr Koszty według stała cena z kartoteki 
			--5.	Parametr - wagi
			--6.	Parametr - ewidencyjna cena odpadu
		declare @42 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =42) --Wartosc Odpady  

			if(@41 = 1)
			begin
				set @wartosc = isnull((select sum(TrE_WartoscNetto) from cdn.TraElem where TrE_TypDokumentu = 303 and TrE_TrNId = @PW),0)

				PRINT('@41 = 1')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@41 = 2)
			begin
				set @wartosc = isnull((select isnull(TwC_Wartosc,0) from cdn.TwrCeny where TwC_TwrID = @id and TwC_TwCNumer = @42),0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@41 = 2')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@41 = 3)
			begin
				set @wartosc = isnull(CONVERT(decimal(15,4),( @42 * @ilosc)) ,0)

				PRINT('@41 = 3')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@41 = 4)
			begin
				set @wartosc = isnull((select isnull(CTwr_Cena,0) from dbo.CtiTowary where CTwr_TwrId = @id),0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@41 = 4')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@410 = 5)--system wag, koszty przechodza na odpady/wyrowy
			begin
				set @wartosc = isnull((select sum(TrE_WartoscNetto)  from cdn.TraElem where TrE_TypDokumentu = 304 and TrE_TrNId = @RW) ,0)

				PRINT('@410 = 5')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
	end
	else if (@typ = 70) -----------------------------------------------zasob
	begin
		declare @44 int  = isnull(@opcjaCenZasobow, (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =44)) --Parametr  
			--1.	Parametr - Według zlecenia
			--2.	Parametr - Według faktycznego zużycia
		if(@44 = 1)
		begin
			declare @45 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =45) --Parametr KosztyCzasuPracy   
			declare @46 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =46) --Parametr KosztyUzbrojenia 
			declare @47 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =47) --Parametr KosztyRozbrojenia 
			if(@45 = 1)
			begin
				
					set @wartosc = @wartosc + isnull(

					(
						select sum (
							CASE WHEN CZS_Dzielnik = 0 THEN
							CZS_CzasPracy * CZS_KosztPracy	
							ELSE
							(CEILING(@ilosc/CZS_Dzielnik) * CZS_CzasPracy * CZS_KosztPracy)
							END
						)
						from dbo.CtiZasobDok inner join dbo.CtiZlecenieNag on czn_id =CZS_CTNID 
						where CZS_DokTyp = 2 and CZS_CTNID = @ZP)

					,0)

					PRINT('@45 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			if(@46 = 1)
			begin
					set @wartosc = @wartosc + isnull((select sum(CZS_CzasUzbrojenia * CZS_KosztUzbrojenia)
					from dbo.CtiZasobDok where CZS_DokTyp = 2 and CZS_CTNID = @ZP),0)

					PRINT('@46 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			if(@47 = 1)
			begin
					set @wartosc = @wartosc + isnull((select sum(CZS_CzasRozbrojenia  * CZS_KosztRozbrojenia)
					from dbo.CtiZasobDok where CZS_DokTyp = 2 and CZS_CTNID = @ZP),0)

					PRINT('@47 = 1')
					PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end

			PRINT('Obliczanie kosztu zasobów')
			PRINT('ZP: ' + CAST(@ZP AS VARCHAR(MAX)))
			PRINT('Ilosc: ' + CAST(@ilosc AS VARCHAR(MAX)))
			PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
		end
		else if(@44 = 2)
		begin
				
				set @wartosc = @wartosc + (select sum(KosztZasobu) from
				(
					select 	sum(isnull((
							 case when CONVERT(VARCHAR,isnull(ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then isnull(ZZs_CzasMin,0)
							  when CONVERT(VARCHAR,isnull(ZZs_DataDo,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then  CONVERT(decimal(15,4),DATEDIFF(second, ZZs_DataOd,GETDATE())/60.0000)
							  else CONVERT(decimal(15,4),(isnull(DATEDIFF(SECOND, ZZs_DataOd,ZZs_DataDo) ,0))/60.0000)  end 
							),0)) --całkowity czas zasobu w minutach
							*
							isnull(
								(
								 select top 1 CZS_KosztPracy 
							     from dbo.CtiZasobDok 
								 where CZS_DokTyp = 2 and CZS_CTNID = @ZP and CZS_CZID = ZZs_CZID
								)
								, (
								 select top 1 CZ_KosztPracy 
								 from dbo.CtiZasob 
								 where CZ_ID = ZZs_CZID)
								) -- koszt w sekundach/minutach/godzinach
							*
							case (
							isnull(
								(
									select top 1 czs_jmczasu
									from dbo.ctizasobdok
									where CZS_DokTyp = 2 and CZS_CTNID = @ZP and CZS_CZID = ZZs_CZID
								),
								(
									select top 1 cz_jmczasu 
									from dbo.CtiZasob 
									where CZ_ID = ZZs_CZID
								)
							) )
							when 0 then 60 -- jak zasób ma koszt w sekundach, to nasz wynik mnożymy
							when 1 then 1 -- minut nie mnożyny
							when 2 then 1 -- godzin nie mnożymy
							end

							/
							
							case (
							isnull(
								(
									select top 1 czs_jmczasu
									from dbo.ctizasobdok
									where CZS_DokTyp = 2 and CZS_CTNID = @ZP and CZS_CZID = ZZs_CZID
								),
								(
									select top 1 cz_jmczasu 
									from dbo.CtiZasob 
									where CZ_ID = ZZs_CZID
								)
							) )
							when 0 then 1 -- sekund nie dzielimy
							when 1 then 1 -- minut nie dzielimy
							when 2 then 60 -- jak zasób ma koszt w godzinach, to nasz wynik dzielimy
							end

						as KosztZasobu
					from dbo.CtiZlecenieZasob  where ZZs_CZNID = @ZP and ZZs_DokTyp = 2
					group by ZZs_CZID--zasoby
				) k)

				PRINT('@44 = 2')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
		end
	end
	--else if (@typ = 75) -----------------------------------------------narzut
	--begin 
	--	declare @49 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =49) --Parametr Narzuty 
	--		--1 --Parametr w zależności od wartości wyrobu gotowego
	--		--2 --Parametr uzależniony od masy wyrobu gotowego
	--		--3 --Parametr przypisany w kartotece wyrobu gotowego
	--		--4 --Parametr Przypisany w polu specjalnym:
	--	declare @50 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =50) --Wartosc Narzuty 
	--end
	else if (@typ = 74) -----------------------------------------------usluga,kooperacja
	begin
		declare @52 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =52) --Parametr Kooperacji   
			--1 --Parametr Koszty według cena zakupu OPTIMA - 
			--2 --Parametr Koszty według cena według cennika OPTIMA 
			--3 --Parametr Koszty według stała cena 
			--4 --Parametr Koszty według stała cena z kartoteki 
		declare @53 decimal(15,4)  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =53) --Wartosc Kooperacji   

			if(@52 = 2)
			begin
				set @wartosc = isnull((select isnull(TwC_Wartosc,0) from cdn.TwrCeny where TwC_TwrID = @id and TwC_TwCNumer = @53),0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@52 = 2')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			if(@52 = 3)
			begin
				set @wartosc = CONVERT(decimal(15,4),( @53 * @ilosc)) 

				PRINT('@52 = 3')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
			else if(@52 = 4)
			begin
				set @wartosc = isnull((select isnull(CTwr_Cena,0) from dbo.CtiTowary where CTwr_TwrId = @id),0)
				set @wartosc = isnull(CONVERT(decimal(15,4),( @wartosc * @ilosc)) ,0)

				PRINT('@52 = 4')
				PRINT('Wartosc: ' + CAST(@wartosc AS VARCHAR(MAX)))
			end
	end
	else
	begin
		set @wartosc = 0
	end







		
	--	select @typ as Typ, @id as ID, @ilosc as Ilosc, @TH as TH, @ZP as ZP,@wartosc as Wartosc

END

GO
```

# [dbo].[CTIProdukcjaKonfiguracjaCenyPozycjiDok]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaKonfiguracjaCenyPozycjiDok]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaKonfiguracjaCenyPozycjiDok]
@typ int,
	--zasob - 70
	--surowiec - 71
	--odpad - 73
	--usługa kooperacja - 74
@typDok int,
	--TH - 1
	--ZP - 2 -- dla zasobów
	--PW - 4 -- dla odpadu
	--RW - 5 -- dla surowca
@idDok int,
@wartosc decimal(15,4) = 0 OUTPUT,
@iloscWyrobu decimal(15,4)
as
begin
set @wartosc = 0
declare @THilosc int = 1
select @THilosc = isnull(CTN_ilosc,1) from dbo.CtiTechnolNag where CTN_ID = @idDok

CREATE TABLE #lista (ID INT, TwrId INT, Ilosc DECIMAL(15, 4), Cena DECIMAL(15, 4), Wartosc DECIMAL(15, 4))
	if(@typDok = 1)--------------------------------------------------TH
	begin
		if(@typ = 71)------ surowce
		begin
			insert into #lista (ID, TwrId , Ilosc, Cena, Wartosc)
			SELECT ROW_NUMBER() OVER (ORDER BY CTE_TwrId), CTE_TwrId, (@iloscWyrobu*sum(CTE_Ilosc))/@THilosc,0,0  
			from dbo.CtiTechnolElem where CTE_CTNId = @idDok and CTE_Typ in (1,2) group by CTE_TwrId
		end
		else if(@typ = 73)------ odpady
		begin
			insert into #lista (ID, TwrId , Ilosc, Cena, Wartosc)
			SELECT ROW_NUMBER() OVER (ORDER BY CTE_TwrId), CTE_TwrId, (@iloscWyrobu*sum(CTE_Ilosc))/@THilosc,0,0  
			from dbo.CtiTechnolElem where CTE_CTNId = @idDok and CTE_Typ = 3 group by CTE_TwrId
		end
		else if(@typ = 74)------ usługi
		begin
			insert into #lista (ID, TwrId , Ilosc, Cena, Wartosc)
			SELECT ROW_NUMBER() OVER (ORDER BY CTE_TwrId), CTE_TwrId, (@iloscWyrobu*sum(CTE_Ilosc))/@THilosc,0,0  
			from dbo.CtiTechnolElem where CTE_CTNId = @idDok and CTE_Typ = 4 group by CTE_TwrId
		end
	end
	else if(@typDok = 5)---------------------------------------------RW
	begin
		declare @38 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =38) --Parametr Surowce 
		if(@38 = 1) --wartość RW
		begin
			set @wartosc = (select sum(TrE_WartoscNetto)  from cdn.TraElem where TrE_TypDokumentu = 304 and TrE_TrNId = @idDok);
		end
		else
		begin
			insert into #lista (ID, TwrId , Ilosc, Cena, Wartosc)
			select ROW_NUMBER() OVER (ORDER BY tre_twrid), tre_twrid, sum(Tre_Ilosc),sum(TrE_WartoscNetto),0   
			from cdn.TraElem where TrE_TypDokumentu = 304 and TrE_TrNId = @idDok group by tre_twrid
		end
	end
	else if(@typDok = 4)---------------------------------------------PW
	begin
		declare @41 int  = (select isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where KNF_Numer =41) --Parametr Odpady  
		if(@41 = 1)
		begin
			set @wartosc = (select sum(TrE_WartoscNetto)  from cdn.TraElem where TrE_TypDokumentu = 303 and TrE_TrNId = @idDok)
		end
		else
		begin
			insert into #lista (ID, TwrId , Ilosc, Cena, Wartosc)
			select ROW_NUMBER() OVER (ORDER BY tre_twrid), tre_twrid, sum(Tre_Ilosc),sum(TrE_WartoscNetto),0   
			from cdn.TraElem where TrE_TypDokumentu = 303 and TrE_TrNId = @idDok group by tre_twrid
		end
	end
	--else if(@typDok = 2)---------------------------------------------ZP
	--begin
	--	if(@typ = 70)------ zasoby
	--	begin
	--		select * from dbo.CtiZasobDok where CZS_DokTyp = 2 and CZS_CTNID = @idDok
	--	end
	--end
	else
		set @wartosc =0

if exists (select id from #lista)--jeżeli są jakieś pozycje
begin

	declare @wartoscPrzekazana decimal(15,4)= 0
	declare @TH int = 0 set @TH = case when @typDok = 1 then @idDok else 0 end
	declare @ZP int = 0 set @ZP = case when @typDok = 2 then @idDok else 0 end
	declare @PW int = 0 set @PW = case when @typDok = 4 then @idDok else 0 end
	declare @RW int = 0 set @RW = case when @typDok = 5 then @idDok else 0 end
		DECLARE kursor cursor
		LOCAL STATIC READ_ONLY FORWARD_ONLY
		FOR SELECT  ID, TwrId , Ilosc, Cena  from #lista
		DECLARE @id INT, @TwrId INT, @Ilosc DECIMAL(15, 4), @Cena DECIMAL(15, 4)
		OPEN kursor
		FETCH NEXT FROM kursor INTO @id, @TwrId , @Ilosc, @Cena
				while @@FETCH_STATUS = 0
				begin
					set @wartoscPrzekazana = 0;
					exec dbo.CTIProdukcjaKonfiguracjaCenyKal @typ, @TwrId, @ilosc, @Cena, @TH,@RW,@PW, @ZP, @wartoscPrzekazana OUTPUT
					update #lista set Wartosc = @wartoscPrzekazana where ID = @id
            
					FETCH NEXT FROM kursor INTO @id, @TwrId , @Ilosc, @Cena
				end
		CLOSE kursor
		deallocate     kursor

set @wartosc = (SELECT sum(Wartosc) FROM #lista)
end




DROP TABLE #lista
end
GO
```

# [dbo].[CTIProdukcjaMontazHistList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaMontazHistList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaMontazHistList]  @NrStanowiska int
AS

SELECT  top 1  CMH_ID, CMH_KodWprow, CMH_Status, CMH_Data, CMH_DataKontr, CMH_NrStanowiska, CMH_PrcId, CMH_Kod, CMH_Dzialanie, CMH_CZNID, 
CMH_TwrIdWprow, CMH_TwrLp, CMH_TwrIdPopr, CMH_DokStatus, CMH_DokInfo, isnull(a.Twr_kod,'') as Kod, isnull(b.Twr_kod,'') as KodPopr,
Pri_kod, pri_Nazwisko + ' '+ PRI_Imie1 as Pracownik,  CMS_Nazwa as Stanowisko,CMH_PrdKod,  CZN_NrPelny as Zlecenie,
case CMH_Status when 1 then 'OK' else 'BLAD' end as Status,
case CMH_Status when 1 then 

	case CMH_Dzialanie 
	when 1 then 'Zakonczenie pracy' 
	when 2 then 
	
		case CMH_DokStatus 
		when 0 then 'Zakonczenie montazu:
Przygotowanie dokumentow... proszę czekać' 
		when 1 then 'Zakonczenie montazu:
Tworzenie dokumentow' 
		when 2 then 'Zakonczenie montazu: 
Stworzono RW' 
		when 3 then 'Zakonczenie montazu: 
Produkt '+CMH_PrdKod
		when 4 then 'Zakonczenie montazu: 
Blad RW'
		when 5 then 'Zakonczenie montazu: 
Blad PW' 
		when 7 then 'Zakonczenie montazu: 
Oczekiwanie na kod wyrobu gotowego' 
		else 'Zakonczenie montazu '+CMH_PrdKod end

	when 3 then 'Cofniecie montazu: 
Produkt ' + isnull(a.Twr_kod,'')+' 
Nazwa '+isnull(replace(a.Twr_Nazwa,char(10), ' ')  ,'')
	when 4 then 'Rozpoczecie pracy' 
    when 5 then 'Montaz:
Dodano ' + isnull(a.Twr_kod,'')+' 
Nazwa ' + isnull(replace(a.Twr_Nazwa,char(10), ' '),'')+' 
Zlecenie '+CZN_NrPelny 
	else '' end

when 2 then 'BLAD' 
when 3 then 'Brak stanowiska' 
when 4 then 'Brak kodu pracownika' 
when 5 then 'Montaz: 
Nie znaleziono pierwszego elementu' 
when 6 then 'Montaz: 
Nie znaleziono elementu' 
when 7 then 'Blad montazu, bledny element:
Wymagany '+isnull(b.Twr_kod,'') 
when 8 then 'Montaz: 
Element cofniety'

when 9 then 'Brak elementu do cofniecia' 
when 10 then 'Blad Zakonczenia montazu: 
Brak elementow'
when 11 then 'Blad Zakonczenia montazu: 
Brak elementow nie wszystkie elementy zostaly wprowadzone'
when 12 then 'Blad Zakonczenie pracy: 
nie zakończono montazu'
when 13 then 'Blad Zakonczenia montazu: 
Blad unikalnosci kodu wyrobu gotowego - ponow' 
when 14 then 'Blad Zakonczenia montazu: 
Blad KODU!!' 
else 'BLAD' end as StatusInfo
FROM  CtiMontazHist
LEFT JOIN cdn.Towary a on a.twr_twrid  = CMH_TwrIdWprow
LEFT JOIN cdn.Towary b on b.twr_twrid  = CMH_TwrIdPopr
LEFT JOIN CDN.Pracidx on PRI_Typ = 1 and PRI_Archiwalny < 1 and CMH_PrcId = PRI_PraId
LEFT JOIN dbo.CtiMontazStanowiska on CMH_NrStanowiska = cms_Nr 
LEFT JOIN dbo.CtiZlecenieNag on CMH_CZNID = czn_id
where CMH_NrStanowiska =@NrStanowiska and CMH_Status >0 order by CMH_ID desc
GO
```

# [dbo].[CTIProdukcjaMontazHistUpdate]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaMontazHistUpdate]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaMontazHistUpdate]
AS
BEGIN


Declare @CMHStatus int  = 0 
--0-kod wprowadzony 
--1-poprawny kod 
-->1-bledny kod
--3-nie znaleziono stanowiska
--4-nie znaleziono pracownika
--5-nie znaleziono pierwszego elementu - brak zasobu
--6-nie znaleziono elementu
--7-bledny element
--8-Element cofniety
--9-Brak elementu do cofniecia
--10-Zakonczenie montazu - Brak elementow - nie wprowadzono elementow 
--11-Zakonczenie montazu - Brak elementow - nie wszystkie elementy zostaly wprowadzone
--12-Zakonczenie pracy - nie zakonczono montazu
--13-Zakonczenia montazu - Blad unikalnosci kodu wyrobu gotowego - ponow
--14-Zakonczenia montazu - Blad KODU!!
Declare @CMHNrStanowiska int = 0
Declare @CMHPrcId int  = 0
Declare @CMHKod varchar(50) = ''
Declare @CMHDzialanie int = 0
--0 - brak - np nie ma historii operacji
Declare @ZakonczeniePracy  varchar(10)		= '000000001' --1 000000001 Zakonczenie pracy
Declare @ZakonczenieMontazu  varchar(10)	= '000000002' --2 000000002 Zakonczenie montazu
Declare @CofniecieMontazu  varchar(10)		= '000000003' --3 000000003 Cofniecie montowanego produktu
--4 Rozpoczecie pracy-Polaczenie stanowiska z pracownikem
--5 Montazu
--7 Kod wyrobu

Declare @CMHTwrIdWprow int = 0
Declare @CMHTwrIdPopr int = 0
Declare @CMHTwrLp int = 0
Declare @CMHCZNID int = 0
Declare @CMHDokStatus int = 7 
--0- Zakonczenie montazu - Tworzenie dokumentu
--1- Zakonczenie montazu - Tworzenie dokumentu
--2- Zakonczenie montazu - Stworzono RW 
--3- Zakonczenie montazu - Stworzono PW
--4- Zakonczenie montazu - Blad RW
--5- Zakonczenie montazu - Blad PW
--7- Zakonczenie montazu - Oczekiwanie na kod wyrobu gotowego


Declare @CMHDokInfo nvarchar(250) = ''

	DECLARE kursor cursor FOR
				SELECT CMH_ID, CMH_KodWprow FROM CtiMontazHist where CMH_Status = 0 order by cmh_id
	DECLARE @CMHID int, @CMHKodWprow nvarchar(250)
		OPEN kursor

	FETCH kursor                 
		INTO  @CMHID , @CMHKodWprow

	while @@FETCH_STATUS = 0
	begin
		set @CMHStatus = 1
		set @CMHNrStanowiska  = 0
		set @CMHPrcId  = 0
		set @CMHKod = ''
		set @CMHDzialanie = 0
		set @CMHTwrIdWprow = 0
		set @CMHTwrIdPopr = 0
		set @CMHTwrLp = 0
		set @CMHCZNID = 0
		set @CMHDokStatus = 7 
		set @CMHDokInfo = ''
		set @CMHTwrLp = 0 
	--numer stanowista
		SELECT @CMHKod =  RIGHT(@CMHKodWprow,LEN(@CMHKodWprow)-2),@CMHNrStanowiska = CMS_Nr 
		FROM CtiMontazStanowiska where CMS_Kod = LEFT(@CMHKodWprow,2) 
	--brak Blad
		if @CMHNrStanowiska = 0 begin set @CMHStatus = 3 end else 
		begin
		--spradzenie czy na danym stanowisku jest ktos zalogowany(czy ktos sie logowal)i czy sie wylogowal
		--czyli czy ostatnia funkcja poprawnie wykonana to kod zakonczenia
			SELECT top 1 @CMHDzialanie = isnull(CMH_Dzialanie,0), @CMHPrcId = CMH_PrcId , @CMHTwrLp = isnull(CMH_TwrLp,0),@CMHCznId=CMH_CznId
			FROM CtiMontazHist where CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status = 1 and CMH_Dzialanie <> 3  order by cmh_id desc
		
				--magazyn surowców pracownika
				declare @MagazynDomyslySurId int = 0
				select @MagazynDomyslySurId = knf_wartosc from CtiKonfiguracja  where knf_numer = 8
				declare @MagazynDomyslySurPrcId int = 0
				SELECT @MagazynDomyslySurPrcId = case isnull(D.KnM_MagSId,0) when 0 then @MagazynDomyslySurId else KnM_MagSId  end 
				FROM CDN.Pracidx AS A LEFT JOIN
				dbo.CtiKonfigMag D ON A.PRI_PraId = D.KnM_PraId
				where (A.PRI_Typ = 1) AND (A.PRI_Archiwalny < 1) and PRI_PraId =@CMHPrcId
				--kod cechy
				declare @cecha varchar(50) = ''
				SELECT @cecha = isnull(KNF_Wartosc,'') FROM  CtiKonfiguracja where KNF_Numer = 5
						
						
		--jezeli nikt nie jest zalogowany to nastepny kod jest kodem pracownika
			if (@CMHDzialanie = 0 or @CMHDzialanie = 1)
			begin
			--znalezienie pracownika o podanym kodzie
				set @CMHPrcId = 0
				SELECT distinct @CMHPrcId = PRI_PraId FROM CDN.Pracidx 
				LEFT JOIN CDN.PracKartyRcp ON CDN.PracKartyRcp.PKR_PrcId = PRI_PraId 
				where PRI_Typ = 1 and PRI_Archiwalny < 1 and isnull(PKR_Numer, '') <> '' 
				and isnull(PKR_Numer, '') = @CMHKod
				
				if @CMHPrcId = 0 begin set @CMHStatus = 4 end else 
				begin set @CMHDzialanie = 4 end--4-nie znaleziono pracownik
			end
			else --jezeli osoba jest zalogowana to nastepny kod jest towarem lub funkcja
			begin
			--spr czy poprzednim poprawnym kodem jest zakonczenie montazu z CMH_Status = 7
					declare @IDZakonczeniaMontazuADDkod int = 0
					
					SELECT @IDZakonczeniaMontazuADDkod = CMH_ID FROM (
					SELECT  LP = ROW_NUMBER() OVER (ORDER BY cmh_id DESC) ,
					CMH_ID, CMH_KodWprow, CMH_Status, CMH_Data, CMH_DataKontr, 
					CMH_NrStanowiska, CMH_PrcId, CMH_Kod, CMH_Dzialanie, CMH_CZNID, 
					CMH_TwrIdWprow, CMH_TwrLp, CMH_TwrIdPopr, CMH_DokStatus, CMH_DokInfo, CMH_DokRWID, CMH_DokPWID
					FROM CtiMontazHist 
					where CMH_ID < @CMHID AND  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status = 1
					) A WHERE LP = 1 AND CMH_Dzialanie = 2 AND CMH_DokStatus = 7  and CMH_Status = 1
				IF @IDZakonczeniaMontazuADDkod > 0
				BEGIN
					set @CMHDzialanie = 7
					--spr unikalnosci kodu CMH_Status = 13 - błąd
					if exists(SELECT CMD_ID FROM CtiMontazDok INNER JOIN cdn.tranag on trn_trnid = CMD_PWID 
where isnull(CMD_PWID,0)<>0 and trn_anulowany = 0 and  CMD_PrdKod = @CMHKod)
					begin
						set @CMHStatus = 13	
					end
					else--poprawny 
					begin
						-- dodanie kodu -- usuniecie obecnego wpisu-- update poprzedniego CMH_DokStatus = 0
						BEGIN TRY
						BEGIN TRAN T1
							 INSERT INTO dbo.CtiMontazDok(CMD_PrdKod,CMD_CMHId,CMD_Status,CMD_DataKoniec) VALUES(@CMHKod,@IDZakonczeniaMontazuADDkod,0,getdate())
							 DELETE FROM dbo.CtiMontazHist where CMH_ID = @CMHID
							 DELETE FROM dbo.CtiMontazHist where  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Dzialanie = 7
							 UPDATE dbo.CtiMontazHist SET CMH_DokStatus = 0 where  CMH_ID = @IDZakonczeniaMontazuADDkod
							 COMMIT TRAN T1
						END TRY
						BEGIN CATCH
						   RollBack TRAN T1
							set @CMHStatus = 14	
						END CATCH
					end
				END ELSE
			--spr czy kod odpowiada funkcjom
				if @CMHKod = @ZakonczeniePracy		--!!!!!!!!!!!!Zakonczenie PRACY
				begin
					if @CMHDzialanie = 2 or @CMHDzialanie = 4
					begin
						set @CMHTwrLp = 0 -- wyczyscic LP
						set @CMHCznId = 0 -- wyczyscic ID
						set @CMHStatus = 1 --mozna zakonczenie pracy
					end
					else
						set @CMHStatus = 12		
				
					Set @CMHDzialanie = 1 
				end else
				if @CMHKod = @ZakonczenieMontazu	--!!!!!!!!!!!!Zakonczenie montazu
				begin 
					declare @DzialanieZakonczenie int = 0
					declare @idRozpoczeciaMontazu int = 0
					declare @iloscWymagana int = 0
					declare @iloscWprowadzona int = 0
					Set @CMHDzialanie = 2 
					SELECT  top 1   @DzialanieZakonczenie = CMH_Dzialanie FROM  CtiMontazHist
					where  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status =1  and CMH_Dzialanie <> 3 order by CMH_ID desc
					if @DzialanieZakonczenie <> 5
						set @CMHStatus = 10	--ostatnim elementem nie jest Montazu -- Zakonczenie montazu - Brak elementow
					else
					begin --id pierszego elementu
						SELECT  top 1  @idRozpoczeciaMontazu=CMH_ID FROM  CtiMontazHist
						where  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status =1  
						and CMH_Dzialanie = 5 and CMH_TwrLp = 1 order by CMH_ID desc 
						--ilosc elementow od id pierwszego
						select  @iloscWprowadzona = count(*) FROM  CtiMontazHist 
						where  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status =1  and CMH_Dzialanie = 5 and CMH_ID >= @idRozpoczeciaMontazu
						--ilosc elementow wymaganych
						select @iloscWymagana = count(*) from CtiZlecenieNag INNER JOIN CtiTechnolElem on cte_ctnid = czn_ctnid 
						where cte_typ = 1 and czn_id = @CMHCznId
						
						if @iloscWprowadzona <> @iloscWymagana 
						begin set @CMHStatus = 11 end --11-Zakonczenie montazu - Brak elementow - nie wszystkie elementy zostaly wprowadzone
						else begin
							set @CMHTwrLp = 0 -- wyczyscic LP
							set @CMHCznId = 0 -- wyczyscic ID
						end
					end
				end else
				if @CMHKod = @CofniecieMontazu		--!!!!!!!!!!!!COFNIECIE montazu 
				begin 
					Set @CMHDzialanie = 3 
					declare @DzialanieCofniecie int = 0
					declare @idCofniecie int = 0
					SELECT  top 1  @idCofniecie=CMH_ID, @DzialanieCofniecie = CMH_Dzialanie ,
					@CMHTwrIdPopr = CMH_TwrIdPopr, @CMHTwrIdWprow = CMH_TwrIdWprow FROM  CtiMontazHist
					where  CMH_NrStanowiska = @CMHNrStanowiska and CMH_Status =1  and CMH_Dzialanie <> 3 order by CMH_ID desc 
					--czy ostatni wpis to Montazu elementu
					if @DzialanieCofniecie = 5
					begin --COFNIECIE elementu
						UPDATE dbo.CtiMontazHist
						SET CMH_Status = 8
						WHERE CMH_ID = @idCofniecie
					end
					else --brak elementu do cofniecia
						set @CMHStatus = 9
				end else
				begin								--!!!!!!!!!!!!Montazu 
				 set @CMHDzialanie = 5
				--sprawdzenie czy towar jest pierwszym dodanym lub czy poprzedni zostal zakoczony
					set @CMHTwrLp= @CMHTwrLp+1
				-- jezeli pierwszy sprawdzenie nr obudowy i przypisanie nr zlecenia-- najstarsze zlecenie, nie zrealizowane w calosci dla danego pracownika!!!!!!!!!!!!
					if @CMHTwrLp = 1
					begin

						--zasób na magazynie pracownika - czy istnieje
						SELECT  @CMHTwrIdPopr = Twz_TwrId, @CMHTwrIdWprow = Twz_TwrId
						--TwrId = Twz_TwrId,DostId = TrS_TrSIdDost,isnull(TsC_Cecha1_Wartosc,'')as Wartosc
						FROM cdn.TwrZasoby
						INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
						LEFT JOIN cdn.TraSElemCechy ON TsC_TrSId = TrS_TrSId
						--LEFT JOIN CDN.TraElem TraElemsProdukcja ON  TraElemsProdukcja.TrE_TwrId = Twz_TwrId
						WHERE TwZ_MagId = @MagazynDomyslySurPrcId and isnull(TsC_Cecha1_Kod,'')= @cecha
						and isnull(TsC_Cecha1_Wartosc,'') = @CMHKod


						--zlecenie w którym jest dana dostawa, spr jaki to pracownik
						select @CMHCznId=CZN_ID--, isnull(TsD_Cecha1_Wartosc,'') as wartosc,
						--sum(case isnull(czd_korekta,0) when 0 then Tsd_ilosc else (-1*Tsd_ilosc) end) as Ilosc --minus to co skorygowane
						from dbo.CtiZlecenieNag 
						INNER JOIN CtiZlecenieDok on CZD_CZNId = CZN_ID
						INNER JOIN cdn.tranag on trn_trnid = CZD_TrnId
						INNER JOIN cdn.TraElem on CZD_TrnId = TrE_TrNId 
						LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID 
						INNER JOIN dbo.CtiTechnolElem on czn_ctnid = cte_ctnid and cte_twrid = TrE_TwrId --pozycja
						where 
						dbo.CTIProdukcjaStatusy(2, CZN_ID) not in (0,3,4) --ZP nie bufor,zakończony,anulowany
						and  dbo.CTIProdukcjaStatusy(24, CZN_ID) <> 2 --ZP nie wyprodukowano w całości
						and dbo.CTIProdukcjaStatusy(3, CZD_ID) = 2 --MM wyciągnięte z bufora
						and isnull(TsD_Cecha1_Wartosc,'') <> '' --tylko towary z cechą
						and isnull(TsD_Cecha1_Kod,'')= @cecha --konkretna cecha
						and CZD_PracownikId = @CMHPrcId-- zlecenie do realizacji dla danego pracownika
						and cte_lp = 1 --cecha musi byc pierwszym elementem w technologii
						and isnull(TsD_Cecha1_Wartosc,'') = @CMHKod
						and tre_twrid = @CMHTwrIdPopr
						group by CZN_ID,TsD_Cecha1_Wartosc
						having sum(case isnull(czd_korekta,0) when 0 then Tsd_ilosc else (-1*Tsd_ilosc) end) >0


						if @CMHCznId = 0 begin set @CMHStatus = 5 end --nie znaleziono pierwszego elementu - brak zasobu


							
					end else
					begin
						------Declare @KodPopr varchar(50) = ''
						------jaki ma być kolejny element
						----select distinct @CMHTwrIdPopr = TrE_TwrId, @CMHCznId = czn_id 
						----from cdn.traelem LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID
						----INNER JOIN dbo.CtiZlecenieDok on 
						----CZD_Status >0 
						----and  isnull(czd_korekta,0) = 0
						----and case isnull(CZD_zastepstwoId,0) when 0 then CZD_PracownikId else CZD_zastepstwoId end = @CMHPrcId
						----INNER JOIN dbo.ctiZlecenieNag on czd_cznid = czn_id and CZN_Status >0 
						----INNER JOIN dbo.CtiTechnolElem on czn_ctnid = cte_ctnid and cte_twrid = TrE_TwrId 
						----and tre_trnid =  czd_trnid
						----where isnull(TsD_Cecha1_Wartosc,'') <> ''  and cte_lp = @CMHTwrLp and czn_id = @CMHCznId
						
						
						--jaki ma byc kolejny element
						select @CMHTwrIdPopr = TrE_TwrId, @CMHCznId = czn_id 
						from dbo.CtiZlecenieNag 
						INNER JOIN CtiZlecenieDok on CZD_CZNId = CZN_ID
						INNER JOIN cdn.tranag on trn_trnid = CZD_TrnId
						INNER JOIN cdn.TraElem on CZD_TrnId = TrE_TrNId 
						LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID 
						INNER JOIN dbo.CtiTechnolElem on czn_ctnid = cte_ctnid and cte_twrid = TrE_TwrId --pozycja
						where 
						dbo.CTIProdukcjaStatusy(2, CZN_ID) not in (0,3,4) --ZP nie bufor,zakończony,anulowany
						and  dbo.CTIProdukcjaStatusy(24, CZN_ID) <> 2 --ZP nie wyprodukowano w całości
						and dbo.CTIProdukcjaStatusy(3, CZD_ID) = 2 --MM wyciągnięte z bufora
						and isnull(TsD_Cecha1_Wartosc,'') <> '' --tylko towary z cechą
						and isnull(TsD_Cecha1_Kod,'')= @cecha --konkretna cecha
						--and CZD_PracownikId = @CMHPrcId-- zlecenie do realizacji dla danego pracownika
						and cte_lp = @CMHTwrLp --cecha musi byc kolejny elementem w technologii
						--and isnull(TsD_Cecha1_Wartosc,'') = @CMHKod
						--and tre_twrid = @CMHTwrIdPopr
						and CZN_ID = @CMHCznId
						group by CZN_ID,TrE_TwrId
						having sum(case isnull(czd_korekta,0) when 0 then Tsd_ilosc else (-1*Tsd_ilosc) end) >0
						
						if @CMHTwrIdPopr = 0 begin set @CMHStatus = 6 end else--6-nie znaleziono elementu
						begin
							declare @ilosc decimal(15,4) = 0
							--czy ten towar istnije na magazynie
							SELECT @ilosc = twz_ilosc
							FROM cdn.TwrZasoby
							INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
							LEFT JOIN cdn.TraSElemCechy ON TsC_TrSId = TrS_TrSId
							WHERE TwZ_MagId = @MagazynDomyslySurPrcId and isnull(TsC_Cecha1_Kod,'')= @cecha
							and isnull(TsC_Cecha1_Wartosc,'') = @CMHKod and Twz_TwrId = @CMHTwrIdPopr
							if @ilosc = 0 begin set @CMHStatus = 6 end else--6-brak zasobów na magazynie
							begin
								select distinct @CMHTwrIdWprow = TrE_TwrId, @CMHCznId = czn_id   
								from cdn.traelem LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID
								INNER JOIN dbo.CtiZlecenieDok on 
								CZD_Status >0 
								and  isnull(czd_korekta,0) = 0 
								and case isnull(CZD_zastepstwoId,0) when 0 then CZD_PracownikId else CZD_zastepstwoId end = @CMHPrcId
								INNER JOIN dbo.ctiZlecenieNag on czd_cznid = czn_id and CZN_Status >0 
								INNER JOIN dbo.CtiTechnolElem on czn_ctnid = cte_ctnid and cte_twrid = TrE_TwrId 
								and tre_trnid =  czd_trnid
								where isnull(TsD_Cecha1_Wartosc,'') <> ''  and cte_lp = @CMHTwrLp and czn_id = @CMHCznId and TsD_Cecha1_Wartosc = @CMHKod
							end
						end
					end
					if @CMHTwrIdPopr <> @CMHTwrIdWprow begin set @CMHStatus = 7 end--7-Bladny element
				end
			end
		end
		
				UPDATE dbo.CtiMontazHist
				SET CMH_Status = @CMHStatus
				  ,CMH_NrStanowiska = @CMHNrStanowiska
				  ,CMH_PrcId = @CMHPrcId
				  ,CMH_Kod = @CMHKod
				  ,CMH_Dzialanie = @CMHDzialanie
				  ,CMH_TwrIdWprow = @CMHTwrIdWprow
				  ,CMH_TwrIdPopr = @CMHTwrIdPopr
				  ,CMH_TwrLp = @CMHTwrLp
				  ,CMH_DataKontr = GETDATE()
				  ,CMH_CZNID=@CMHCZNID
				  ,CMH_DokStatus=@CMHDokStatus
				  ,CMH_DokInfo=@CMHDokInfo
				WHERE CMH_ID = @CMHID
	            
				FETCH kursor                 
					INTO  @CMHID , @CMHKodWprow
	end

	CLOSE kursor
	deallocate     kursor

	
	
	


END
GO
```

# [dbo].[CTIProdukcjaPracMagAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaPracMagAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaPracMagAdd]
@idPrac int,
@MagSur varchar(40),
@MagProd varchar(40)
AS
BEGIN
declare @MagSurId int = 0
declare @MagProdId int = 0
if @idPrac <> 0
begin
	if	@MagSur <> ''
	begin
		select @MagSurId = Mag_MagId from cdn.magazyny where 
		Mag_NieAktywny = 0 and Mag_Symbol = @MagSur
	end
	if	@MagProd <> ''
	begin
		select @MagProdId = Mag_MagId from cdn.magazyny where 
		Mag_NieAktywny = 0 and Mag_Symbol = @MagProd
	end


		IF EXISTS
		(select KnM_PraId FROM CtiKonfigMag where KnM_PraId = @idPrac)
		begin
			if	@MagSur = ''
			begin
				UPDATE [dbo].[CtiKonfigMag]
				SET [KnM_MagPId] =@MagProdId
				where KnM_PraId =@idPrac
			end
			else if	@MagProd = ''
			begin
				UPDATE [dbo].[CtiKonfigMag]
				SET [KnM_MagSId] = @MagSurId
				where KnM_PraId =@idPrac
			end
			else
			begin
				UPDATE [dbo].[CtiKonfigMag]
				SET [KnM_MagSId] = @MagSurId,[KnM_MagPId] =@MagProdId
				where KnM_PraId =@idPrac
			end
		end
		else
		begin
			INSERT INTO [dbo].[CtiKonfigMag]([KnM_PraId],[KnM_MagSId],[KnM_MagPId])
			VALUES(@idPrac,@MagSurId,@MagProdId)
		end
end    	
END
GO
```

# [dbo].[CTIProdukcjaPracMagList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaPracMagList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaPracMagList] 
@ZP int, --305 --0 wszystkie-- idZP pracownicy powiazani z ZP
@Typ int, --0 wszystkie -- 80 ograniczenie na pracowniku
@DzialID int = NULL
AS

declare  @CzyJestOgraniczenieWydzialu int = 0 SELECT @CzyJestOgraniczenieWydzialu = count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 3
declare  @CzyJestOgraniczenieKategorii int = 0 SELECT  @CzyJestOgraniczenieKategorii = count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 2
declare  @CzyJestOgraniczenieStanowisko int = 0 SELECT @CzyJestOgraniczenieStanowisko = count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 4

declare @MagazynDomyslySurId int = 0 select @MagazynDomyslySurId = knf_wartosc from CtiKonfiguracja  where knf_numer = 8
declare @MagazynDomyslySur varchar(40) = '' select @MagazynDomyslySur = mag_symbol from cdn.magazyny where mag_magid = @MagazynDomyslySurId
declare @MagazynDomyslyProdId int =0 select @MagazynDomyslyProdId = knf_wartosc from CtiKonfiguracja  where knf_numer = 9
declare @MagazynDomyslyProd varchar(40) = '' select @MagazynDomyslyProd = mag_symbol from cdn.magazyny where mag_magid = @MagazynDomyslyProdId



	SELECT DISTINCT A.PRI_PraId as ID , A.PRI_Kod as PrcKod, 
		(SELECT TOP 1 PKR_Numer FROM CDN.PracKartyRcp WHERE PKR_PrcId = A.PRI_PraId AND GETDATE() BETWEEN PKR_OkresOd AND PKR_OkresDo) AS RCP,
		case ISNULL(D.KnM_MagSId,0) when 0 then @MagazynDomyslySur else (select Mag_Symbol from cdn.magazyny where Mag_MagId = D.KnM_MagSId  ) end As MagSur,
		case ISNULL(D.KnM_MagSId,0) when 0 then @MagazynDomyslySurId else KnM_MagSId  end As MagSurId,
		case ISNULL(D.KnM_MagPId,0) when 0 then @MagazynDomyslyProd else (select Mag_Symbol from cdn.magazyny where Mag_MagId = D.KnM_MagPId  ) end As MagProd, 
		case ISNULL(D.KnM_MagPId,0) when 0 then @MagazynDomyslyProdId else KnM_MagPId  end As MagProdId,
		PRI_Imie1 as Imie, PRI_Nazwisko as Nazwisko,
		DZL_Kod as Wydzial, PRI_Pesel as Pesel, ISNULL(DKM_Nazwa,'') as Stanowisko
	FROM CDN.Pracidx AS A 
		LEFT JOIN CDN.PracKod AS B ON B.PRA_PraId = A.PRI_PraId 
		LEFT JOIN CDN.PracEtaty AS C ON C.PRE_PraId = B.PRA_PraId 
		LEFT JOIN dbo.CtiKonfigMag D ON A.PRI_PraId = D.KnM_PraId
		LEFT JOIN CDN.Dzialy ON PRE_DzlId = CDN.Dzialy.DZL_DzlId
		LEFT JOIN cdn.danekadmod on C.PRE_ETADkmIdStanowisko = DKM_DkmId 
		LEFT JOIN dbo.CtiZleceniePrc ON A.PRI_PraId = ISNULL(CZO_PrcId,0)
	where (A.PRI_Typ = 1) AND (A.PRI_Archiwalny < 1) 
	   and Convert(DATETIME,getdate(),120) >= PRE_DataOd --wydziały
	   and Convert(DATETIME,getdate(),120) <= PRE_DataDo --wydziały
	   and ((@ZP<>0  and ISNULL(CZO_CZNId,0) = @ZP)OR 
		 (@ZP = 0
	  --ograniczenia
		 and ((@CzyJestOgraniczenieKategorii = 0) or (@CzyJestOgraniczenieKategorii <> 0 and (ISNULL(PRE_KatId,0) in (SELECT distinct KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 2))))
		 and ((@CzyJestOgraniczenieWydzialu = 0) or (@CzyJestOgraniczenieWydzialu <> 0 and (ISNULL(PRE_DzlId,0) in (SELECT distinct KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 3)))) 
		 and ((@CzyJestOgraniczenieStanowisko = 0)	or (@CzyJestOgraniczenieStanowisko <> 0 and (ISNULL(C.PRE_ETADkmIdStanowisko,0) in (SELECT distinct KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 4)))) 
		  ))
		and (@DzialID IS NULL OR EXISTS (
			SELECT * FROM dbo.CtiDzialPRC
			WHERE A.PRI_PraId = DzP_PrcId
			AND DzP_CDZID = @DzialID
		))

	order by  A.PRI_Kod
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaPWlist]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaPWlist]
 @ID int,
 @IDZPZ int,
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmData int--0 bez --1 ograniczenie po dacie
AS
BEGIN
--czy poszczególne parametry maja być widoczne
declare @DokTyp int = 4 -- PW
--declare @prmMontaz int = 0 SELECT  @prmMontaz = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 25 and ISNULL(KNF_Wartosc,'') <> ''
declare @prmPrc int = 0 SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> ''
if @prmPrc in (2,3,4) begin set @prmPrc = 1 end else begin set @prmPrc = 0 end
declare @prmPole1 int = 0 SELECT   @prmPole1 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 1 and CAU_TypDok = @DokTyp
declare @prmPole2 int = 0 SELECT   @prmPole2 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 2 and CAU_TypDok = @DokTyp
declare @prmPole3 int = 0 SELECT   @prmPole3 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 3 and CAU_TypDok = @DokTyp


		SELECT DISTINCT 
			e.TrN_TrNID as ID, ISNULL(CtiZlecenieNag.CZN_ID, 0) AS CZN_ID
			, ISNULL(e.TrN_NumerPelny, '') AS DokumentWydania
			, ISNULL(a.CO_Kod, '') AS CZD_OpeWyst, ISNULL(CtiZlecenieDok.CZD_DataWyst, GETDATE())AS CZD_DataWyst
			, ISNULL(b.CO_Kod, '') AS CZD_OperModyf, ISNULL(CtiZlecenieDok.CZD_DataModyf, GETDATE())   AS CZD_DataModyf
			, ISNULL(CtiZlecenieNag.CZN_TwrId, 0) AS ZasobID 
			, ISNULL(CDN.Towary.Twr_Kod, '') AS ZasobKod
			, ISNULL(dbo.CTIProdukcjaZlecenieNumer(CtiZlecenieNag.CZN_ID), '') AS CZN_Numer
			, ISNULL(CZO_PrcKod, '') AS PRI_Kod --pracownik
			, dbo.CTIProdukcjaStatusy(3, CtiZlecenieDok.CZD_ID) AS CZD_Status
			, dbo.CTIProdukcjaStatusyNazwy(3, dbo.CTIProdukcjaStatusy(3, CtiZlecenieDok.CZD_ID)) AS StatusNazwa 
			--, ISNULL(CMD_PrdKod, '') AS ProdKod--montaz
			,''  AS ProdKod
			, ISNULL(c.Mag_Symbol, '') AS MagDoc 
			, ISNULL(d.Mag_Symbol, '') AS MagZrd
			, case @prmPole1 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) WHERE (Lp = 1)) else '' end AS Pole1
			, case @prmPole2 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) WHERE (Lp = 2))  else '' end AS Pole2
			, case @prmPole3 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) WHERE (Lp = 3))  else '' end AS Pole3
		FROM CtiZlecenieDok 
			INNER JOIN CtiZlecenieNag ON CtiZlecenieDok.CZD_CZNId = CtiZlecenieNag.CZN_ID 
			INNER JOIN CDN.TraNag AS e ON CtiZlecenieDok.CZD_TrnId = e.TrN_TrNID 
			LEFT JOIN CtiOperatorzy AS a ON CtiZlecenieDok.CZD_OpeWyst = a.CO_ID 
			LEFT JOIN CtiOperatorzy AS b ON CtiZlecenieDok.CZD_OperModyf = b.CO_ID 
			LEFT JOIN dbo.CtiZleceniePrc ON  CZO_CZNId = CZN_ID and CZO_PrcId = CZD_PracownikId and @prmPrc = 1
			LEFT JOIN CDN.Magazyny AS c ON e.TrN_MagDocId = c.Mag_MagId 
			LEFT JOIN CDN.Magazyny AS d ON e.TrN_MagZrdId = d.Mag_MagId 
			--LEFT JOIN CtiMontazDok ON e.TrN_TrNID = CtiMontazDok.CMD_PWId and @prmMontaz = 1
			LEFT JOIN CDN.Towary ON CtiZlecenieNag.CZN_TwrId = CDN.Towary.Twr_TwrId
			LEFT JOIN dbo.CtiZlecenieZElem ON CZN_ID = ISNULL(CZZE_CZNID,0)
		where e.TrN_TypDokumentu in  (303)  and e.trn_anulowany = 0 
		and
		(
			((@ID = 0) or (@ID <> 0 and CZN_ID = @ID ))
			and
			((@IDZPZ = 0) or (@IDZPZ <> 0 and ISNULL(CZZE_CZZNID,0) = @IDZPZ ))
			and
			(
				 (ISNULL(@prmData,0) =0) 
				 or
				(@prmData =1 and
				CONVERT(VARCHAR, ISNULL(e.trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)
		)
		ORDER BY e.TrN_TrNID
END

GO
```

# [dbo].[CTIProdukcjaPWRWlist]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaPWRWlist]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaPWRWlist]
 @ID int,
 @IDZPZ int,
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmData int--0 bez --1 ograniczenie po dacie
AS
BEGIN
--czy poszczególne parametry maja być widoczne
declare @DokTyp int = 4 --produkcja PW
--declare @prmMontaz int = 0 SELECT  @prmMontaz = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 25 and ISNULL(KNF_Wartosc,'') <> ''
declare @prmPrc int = 0 SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> ''
if @prmPrc in (2,3,4) begin set @prmPrc = 1 end else begin set @prmPrc = 0 end

		SELECT DISTINCT 
			e.TrN_TrNID as ID, ISNULL(CtiZlecenieNag.CZN_ID, 0) AS CZN_ID
			, ISNULL(e.TrN_NumerPelny, '') AS DokumentWydania
			, ISNULL(a.CO_Kod, '') AS CZD_OpeWyst, ISNULL(CtiZlecenieDok.CZD_DataWyst, GETDATE())AS CZD_DataWyst
			, ISNULL(b.CO_Kod, '') AS CZD_OperModyf, ISNULL(CtiZlecenieDok.CZD_DataModyf, GETDATE())   AS CZD_DataModyf
			, ISNULL(CtiZlecenieNag.CZN_TwrId, 0) AS ZasobID 
			, ISNULL(CDN.Towary.Twr_Kod, '') AS ZasobKod
			, ISNULL(dbo.CTIProdukcjaZlecenieNumer(CtiZlecenieNag.CZN_ID), '') AS CZN_Numer
			, COALESCE(CZO_PrcKod, pv.Kod, '') AS PRI_Kod --pracownik
			, case TrN_Bufor 
					when 0	then 2--zamknięte
					when -1 then 4 --anulowane
					else 0	end--bufor
				AS CZD_Status
			, dbo.CTIProdukcjaStatusyNazwy(3, case TrN_Bufor 
					when 0	then 2--zamknięte
					when -1 then 4 --anulowane
					else 0	end) AS StatusNazwa 
			--, ISNULL(CMD_PrdKod, '') AS ProdKod--montaz
			,''  AS ProdKod
			, ISNULL(c.Mag_Symbol, '') AS MagDoc 
			, ISNULL(d.Mag_Symbol, '') AS MagZrd
	,(SELECT TOP 1 CASE CAE_Wartosc
					WHEN 'True' THEN 'Tak'
					WHEN 'False' THEN 'Nie'
					ELSE isnull(CAE_Wartosc, '') END 
					FROM dbo.CtiAtrybutyElem where ((CAE_IdDok = e.TrN_TrNID 
						and CAE_TypDok = (SELECT case e.TrN_TypDokumentu
						when 304 then 5
						when 303 then 4 end)
					)) and  CAE_CANLp = 1) as pole1

			,(SELECT TOP 1 CASE CAE_Wartosc
					WHEN 'True' THEN 'Tak'
					WHEN 'False' THEN 'Nie'
					ELSE isnull(CAE_Wartosc, '') END 
					FROM dbo.CtiAtrybutyElem where ((CAE_IdDok = e.TrN_TrNID 
						and CAE_TypDok = (SELECT case e.TrN_TypDokumentu
						when 304 then 5
						when 303 then 4 end)
					)) and  CAE_CANLp = 2) as pole2
			,(SELECT TOP 1 CASE CAE_Wartosc
					WHEN 'True' THEN 'Tak'
					WHEN 'False' THEN 'Nie'
					ELSE isnull(CAE_Wartosc, ' ') END 
					FROM dbo.CtiAtrybutyElem where ((CAE_IdDok = e.TrN_TrNID 
						and CAE_TypDok = (SELECT case e.TrN_TypDokumentu
						when 304 then 5
						when 303 then 4 end)
					)) and  CAE_CANLp = 3) as pole3					

		FROM CtiZlecenieDok 
			INNER JOIN CtiZlecenieNag ON CtiZlecenieDok.CZD_CZNId = CtiZlecenieNag.CZN_ID 
			INNER JOIN CDN.TraNag AS e ON CtiZlecenieDok.CZD_TrnId = e.TrN_TrNID 
			LEFT JOIN CtiOperatorzy AS a ON CtiZlecenieDok.CZD_OpeWyst = a.CO_ID 
			LEFT JOIN CtiOperatorzy AS b ON CtiZlecenieDok.CZD_OperModyf = b.CO_ID 
			LEFT JOIN dbo.CtiZleceniePrc ON  CZO_CZNId = CZN_ID and CZO_PrcId = CZD_PracownikId and @prmPrc = 1
			LEFT JOIN CDN.Magazyny AS c ON e.TrN_MagDocId = c.Mag_MagId 
			LEFT JOIN CDN.Magazyny AS d ON e.TrN_MagZrdId = d.Mag_MagId 
			--LEFT JOIN CtiMontazDok ON e.TrN_TrNID = CtiMontazDok.CMD_PWId and @prmMontaz = 1
			LEFT JOIN CDN.Towary ON CtiZlecenieNag.CZN_TwrId = CDN.Towary.Twr_TwrId
			LEFT JOIN dbo.CtiZlecenieZElem ON CZN_ID = ISNULL(CZZE_CZNID,0)
			LEFT JOIN CDN.PracownicyView pv ON pv.PracId = CZD_PracownikID
		where e.TrN_TypDokumentu in  (303,304)  and e.trn_anulowany = 0 
		and
		(
			((@ID = 0) or (@ID <> 0 and CZN_ID = @ID ))
			and
			((@IDZPZ = 0)or (@IDZPZ <> 0 and ISNULL(CZZE_CZZNID,0) = @IDZPZ ) )
			and
			(
				(ISNULL(@prmData,0) = 0) 
				or
				(@prmData =1 and
				CONVERT(VARCHAR, ISNULL(e.trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)
		)

		ORDER BY e.TrN_TrNID
END

GO
```

# [dbo].[CtiProdukcjaRaportBrakow]

```sql
/****** Object:  StoredProcedure [dbo].[CtiProdukcjaRaportBrakow]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create  PROCEDURE [dbo].[CtiProdukcjaRaportBrakow]
	@OpeId INT,
	@MagId INT,
	@TwrGrupaId  INT,
	@datZasobu  varchar(20),
	@DostawcaId  INT,
	@UwzgledniajRezerwacje TINYINT,
	@UwzgledniajZamowienia TINYINT,
	@UwzgledniajStanyTowarow TINYINT,
	@RodzajNormatywu TINYINT,
	@TypAlgorytmu TINYINT,
	@SprzedOd  varchar(20),
	@SprzedDo  varchar(20),
	@NiePobierajListyBrakow TINYINT,
	@TowaryId    TEXT,
	@WybraneROId TEXT
AS
BEGIN

declare @SprzedazOd datetime set @SprzedazOd = convert(datetime, @SprzedOd,120)
declare @SprzedazDo datetime set @SprzedazDo = convert(datetime, @SprzedDo,120)
declare @DataZasobu datetime set @DataZasobu = convert(datetime, @datZasobu,120)

  SET NOCOUNT ON
 -- Notowania walut
  DECLARE
  @TylkoWgId     INT,
  @TwGNumer      INT,
  @NrCeny        INT,
  @WalutaKoncowa    NVARCHAR(3),
  @sqltxt			NVARCHAR(1024),
  @FullKnfName      NVARCHAR(512),
  @WalOstrzegaj     NVARCHAR(3),
  @TrN_DataDokChar  NVARCHAR(25),
  @KursLWalutaKoncowa                DECIMAL(15,4),
  @KursMWalutaKoncowa                DECIMAL(5,0)
--pobranie notowania
 SELECT @TrN_DataDokChar    = CONVERT(NVARCHAR(25), @DataZasobu, 102)
 SET @FullKnfName = CDN.GetFullKnfName()
 SELECT @TrN_DataDokChar    = CONVERT(NVARCHAR(25), @DataZasobu, 102)
 -- 1 = "Pobieraj ostatnio ustalony kurs waluty (nie ostrzegaj o zmianie kursu)"
 IF EXISTS(SELECT Fir_Wartosc FROM Cdn.Firma WHERE Fir_Numer=1540 )
  SELECT @WalOstrzegaj=Fir_Wartosc FROM Cdn.Firma WHERE Fir_Numer=1540
 ELSE
  SELECT @WalOstrzegaj=1
 SET @sqltxt = 'EXEC ' + @FullKnfName + '.CDN.PobierzNotowania '''', 2,'''+ @TrN_DataDokChar+''','+ @WalOstrzegaj+''
 CREATE TABLE #Notowania  ( Systemowa TINYINT, Waluta NVARCHAR(3) PRIMARY KEY, KursNumer INT, KursL DECIMAL(15,4), KursM DECIMAL(5,0), Brak TINYINT )
 INSERT INTO  #Notowania  ( Systemowa,         Waluta ,           KursNumer,     KursL,               KursM             , Brak         )
 EXEC sp_executesql @sqltxt
 IF NOT EXISTS (SELECT Fir_Wartosc FROM Cdn.Firma WHERE Fir_Numer = 1849) SELECT @WalutaKoncowa = '' ELSE SELECT @WalutaKoncowa = Fir_Wartosc FROM Cdn.Firma WHERE Fir_Numer = 1849
 IF @WalutaKoncowa = ''
  SELECT @KursLWalutaKoncowa = 1, @KursMWalutaKoncowa =1
 ELSE
  SELECT @KursLWalutaKoncowa = KursL, @KursMWalutaKoncowa =KursM FROM #Notowania WHERE Waluta = @WalutaKoncowa
    DECLARE @opeZakazCenZak INT;
 SET @opeZakazCenZak = 0;
 SET @sqltxt  = ' SELECT TOP 1 Ope_ZakazCenyZak FROM ' + @FullKnfName + '.CDN.Operatorzy WHERE Ope_OpeId = ' + CONVERT(NVARCHAR(20), @OpeId)
         -- przerzucenie wyniku do zmiennej poprzez tabelę tymczasową
    CREATE TABLE #Ope (OpeParam INT)
    INSERT #Ope EXEC sp_executeSQL @sqltxt
    SELECT TOP 1  @opeZakazCenZak = OpeParam  FROM #Ope
    DROP TABLE #Ope
 IF @DostawcaId Is NULL SET @DostawcaId = 0
 IF DATALENGTH(@TowaryId) = 0
  SET @TylkoWgId = 0
 ELSE
  SET @TylkoWgId = 1
 CREATE TABLE #RaportBrakow (
     ID     INT NOT NULL PRIMARY KEY,
     Kod    NVARCHAR(255) NOT NULL,
     Nazwa    NVARCHAR(255) NOT NULL,
     EAN    NVARCHAR(255) NOT NULL,
     [Status]   NVARCHAR(3) NOT NULL,
     NumerKat   NVARCHAR(40) NOT NULL,
  DostawcaID   INT NOT NULL,
  Dostawca   NVARCHAR(255) NOT NULL,
  KodDostawcy  NVARCHAR(255) NOT NULL,
     Ilosc    DECIMAL(20,4) NOT NULL,
     Rezerwacje   DECIMAL(20,4) NOT NULL,
     UwzgRezerwacje  DECIMAL(20,4) NOT NULL,
     Zamowienia   DECIMAL(20,4) NOT NULL,
     UwzgZamowienia  DECIMAL(20,4) NOT NULL,
     Braki    DECIMAL(20,4) NOT NULL,
     BrakiPozostale  DECIMAL(20,4) NOT NULL,
     IloscSprzedana  DECIMAL(20,4) NOT NULL,
     IloscZamawiana  DECIMAL(20,4) NOT NULL,
     IloscZamawianaJM DECIMAL(20,4) NOT NULL,
  JMZam    NVARCHAR(255) NOT NULL,
  JM     NVARCHAR(255) NOT NULL,
  JMCalkowite  TINYINT NOT NULL,
  CenaNettoWal  DECIMAL(20,2) NOT NULL,
  CenaNettoWalDisp NVARCHAR(20) NOT NULL,
  WalutaSymbol  NVARCHAR(3) NOT NULL,
  KursL    DECIMAL(15,4) NOT NULL,
  KursM    DECIMAL(5,0) NOT NULL,
  CenaNetto   DECIMAL(20,2) NOT NULL,
  WartoscNetto  DECIMAL(20,2) NOT NULL,
  WartoscNettoDisp NVARCHAR(20) NOT NULL,
  BrakKursuWaluty TINYINT NOT NULL,
  IloscMin           DECIMAL (15,4) NOT NULL,
  IloscMinJM         NVARCHAR (255)   NOT NULL ,
  IloscMax           DECIMAL (15,4) NOT NULL ,
  IloscMaxJM         NVARCHAR (255)   NOT NULL ,
  IloscZam           DECIMAL (15,4) NOT NULL ,
  IloscZamJM         NVARCHAR (255)   NOT NULL
 )
 IF @magId = 0
 BEGIN
  INSERT INTO #RaportBrakow
  SELECT Twr_TwrId,Twr_Kod,replace(Twr_Nazwa,char(10), ' ')  ,Twr_EAN,
   CASE WHEN (Twr_Typ = 0 AND Twr_Produkt = 0 AND Twr_Kaucja = 0) THEN 'UP' WHEN (Twr_Typ = 0 AND Twr_Produkt = 1 AND Twr_Kaucja = 0) THEN 'UZ' WHEN (Twr_Typ = 1 AND Twr_Produkt = 0 AND Twr_Kaucja = 0) THEN 'TP' WHEN (Twr_Typ = 1 AND Twr_Produkt = 1 AND Twr_Kaucja = 0) THEN 'TZ' WHEN (Twr_Typ = 0 AND Twr_Produkt = 0 AND Twr_Kaucja = 1) THEN 'UPO' WHEN (Twr_Typ = 0 AND Twr_Produkt = 1 AND Twr_Kaucja = 1) THEN 'UZO' WHEN (Twr_Typ = 1 AND Twr_Produkt = 0 AND Twr_Kaucja = 1) THEN 'TPO' WHEN (Twr_Typ = 1 AND Twr_Produkt = 1 AND Twr_Kaucja = 1) THEN 'TZO' END,
   Twr_NumerKat,
   isnull(Twr_KntId,0),
   isnull((SELECT TOP 1 Knt_Kod FROM cdn.Kontrahenci WHERE Knt_KntId = Twr_KntId),''), --Dostawca
   Twr_KodDostawcy,
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Ilosc FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0)), -- ilosc
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Rezerwacje FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0)), -- Rezerwacje
   CASE WHEN @UwzgledniajRezerwacje=1 THEN
    Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Rezerwacje FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0))
   ELSE 0 END, --  UwzgRezerwacje
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Zamowienia FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0)), -- Zamowienia
   CASE WHEN @UwzgledniajZamowienia=1 THEN
    Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Zamowienia FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0))
   ELSE 0 END, -- UwzgZamowienia
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Braki FROM cdn.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId Is Null ORDER BY TwI_Data DESC), 0)), --  Braki
   0, -- BrakiPozostale
   CASE WHEN @TypAlgorytmu=1 THEN cdn.IloscSprzedana(Twr_TwrId, @MagId, @SprzedazOd, @SprzedazDo)
    WHEN @TypAlgorytmu=2 THEN cdn.IloscNaWybranychRezerwacjach(Twr_TwrId, @MagId, @WybraneROId)
   ELSE 0 END, --IloscSprzedana
   0, -- IloscZamawiana,
   0, -- IloscZamawianaJM,
   CASE Twr_IloscZam
    WHEN  0  THEN Twr_JM
    ELSE Twr_IloscZamJM
   END,  --ZamJM
   Twr_JM, -- JM,
   Twr_JMCalkowite,
   TwC_Wartosc, -- CenaNettoWal,
   '0', -- CenaNettoWalDisp
   TwC_Waluta,  -- Waluta,
   CASE WHEN N.Brak = 1 THEN 1
   ELSE isnull(N.KursL,1) END, --KursL
   CASE WHEN N.Brak = 1 THEN 1
   ELSE isnull(N.KursM,1) END, -- KursM
   CASE WHEN N.Brak = 1 OR N.KursL IS NULL OR N.KursM IS NULL OR N.KursL = 0 OR N.KursM = 0 THEN 0
   ELSE
    Round( TwC_Wartosc * (N.KursL / N.KursM ), 2)
   END, -- CenaNetto
   0,  -- WartoscNetto
   '0', -- WartoscNettoDisp
   isnull(N.Brak, 0),  -- BrakKursuWaluty
   Twr_IloscMin,
   Twr_IloscMinJM,
   Twr_IloscMax,
   Twr_IloscMaxJM,
   Twr_IloscZam,
   Twr_IloscZamJM
   FROM cdn.Towary
   JOIN cdn.TwrCeny ON TwC_TwrId = Twr_TwrId AND TwC_TwCNumer = 1
   LEFT JOIN #Notowania N ON N.Waluta = TwC_Waluta
   WHERE Twr_Typ = 1 AND Twr_Nieaktywny = 0
   AND (@NiePobierajListyBrakow=0 OR TwR_TwRId=-1)
   AND (@TylkoWgId=1 OR @DostawcaId=0 OR TwR_KntId = @DostawcaId)
   AND (@TylkoWgId=1 OR @TwrGrupaId=-1 OR (Twr_GidNumer IN (SELECT GidNumer FROM CDN.TwrInTwG(@TwrGrupaId))))
   AND (@TylkoWgId=0 OR TwR_TwRId IN (SELECT NextIdInt FROM CDN.Text2TableInt(@TowaryId)))
 END
 ELSE
 BEGIN
  INSERT INTO #RaportBrakow
  SELECT Twr_TwrId,Twr_Kod,replace(Twr_Nazwa,char(10), ' ')  ,Twr_EAN,
   CASE WHEN (Twr_Typ = 0 AND Twr_Produkt = 0 AND Twr_Kaucja = 0) THEN 'UP' WHEN (Twr_Typ = 0 AND Twr_Produkt = 1 AND Twr_Kaucja = 0) THEN 'UZ' WHEN (Twr_Typ = 1 AND Twr_Produkt = 0 AND Twr_Kaucja = 0) THEN 'TP' WHEN (Twr_Typ = 1 AND Twr_Produkt = 1 AND Twr_Kaucja = 0) THEN 'TZ' WHEN (Twr_Typ = 0 AND Twr_Produkt = 0 AND Twr_Kaucja = 1) THEN 'UPO' WHEN (Twr_Typ = 0 AND Twr_Produkt = 1 AND Twr_Kaucja = 1) THEN 'UZO' WHEN (Twr_Typ = 1 AND Twr_Produkt = 0 AND Twr_Kaucja = 1) THEN 'TPO' WHEN (Twr_Typ = 1 AND Twr_Produkt = 1 AND Twr_Kaucja = 1) THEN 'TZO' END,
   Twr_NumerKat,
   isnull(Twr_KntId,0),
   isnull((SELECT TOP 1 Knt_Kod FROM cdn.Kontrahenci WHERE Knt_KntId = Twr_KntId),''), --Dostawca
   Twr_KodDostawcy,
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Ilosc FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0)), -- ilosc
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Rezerwacje FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0)), -- Rezerwacje
   CASE WHEN @UwzgledniajRezerwacje=1 THEN
    Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Rezerwacje FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0))
   ELSE 0 END, --  UwzgRezerwacje
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Zamowienia FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0)), -- Zamowienia
   CASE WHEN @UwzgledniajZamowienia=1 THEN
    Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Zamowienia FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0))
   ELSE 0 END, -- UwzgZamowienia
   Convert( DECIMAL(20,4), isnull((SELECT TOP 1 TwI_Braki FROM cdn.TwrIlosci LEFT JOIN cdn.Magazyny ON Mag_MagId = TwI_MagId WHERE TwI_TwrId = Twr_TwrId AND TwI_Data <= @DataZasobu AND TwI_MagId = @MagId ORDER BY TwI_Data DESC), 0)), --  Braki
   0, -- BrakiPozostale
   CASE WHEN @TypAlgorytmu=1 THEN cdn.IloscSprzedana(Twr_TwrId, @MagId, @SprzedazOd, @SprzedazDo)
    WHEN @TypAlgorytmu=2 THEN cdn.IloscNaWybranychRezerwacjach(Twr_TwrId, @MagId, @WybraneROId)
   ELSE 0 END, --IloscSprzedana
   0, -- IloscZamawiana,
   0, -- IloscZamawianaJM,
   CASE Twr_IloscZam
    WHEN  0  THEN Twr_JM
    ELSE Twr_IloscZamJM
   END,  --ZamJM
   Twr_JM, -- JM,
   Twr_JMCalkowite,
   TwC_Wartosc, -- CenaNettoWal,
   '0', -- CenaNettoWalDisp
   TwC_Waluta,  -- Waluta,
   CASE WHEN N.Brak = 1 THEN 1
   ELSE isnull(N.KursL,1) END, --KursL
   CASE WHEN N.Brak = 1 THEN 1
   ELSE isnull(N.KursM,1) END, -- KursM
   CASE WHEN N.Brak = 1 OR N.KursL IS NULL OR N.KursM IS NULL OR N.KursL = 0 OR N.KursM = 0 THEN 0
   ELSE
    Round( TwC_Wartosc * (N.KursL / N.KursM ), 2)
   END, -- CenaNetto
   0,  -- WartoscNetto
   '0', -- WartoscNettoDisp
   isnull(N.Brak, 0),  -- BrakKursuWaluty
   Twr_IloscMin,
   Twr_IloscMinJM,
   Twr_IloscMax,
   Twr_IloscMaxJM,
   Twr_IloscZam,
   Twr_IloscZamJM
   FROM cdn.Towary
   JOIN cdn.TwrCeny ON TwC_TwrId = Twr_TwrId AND TwC_TwCNumer = 1
   LEFT JOIN #Notowania N ON N.Waluta = TwC_Waluta
   WHERE Twr_Typ = 1 AND Twr_Nieaktywny = 0
   AND (@NiePobierajListyBrakow=0 OR TwR_TwRId=-1)
   AND (@TylkoWgId=1 OR @DostawcaId=0 OR TwR_KntId = @DostawcaId)
   AND (@TylkoWgId=1 OR @TwrGrupaId=-1 OR (Twr_GidNumer IN (SELECT GidNumer FROM CDN.TwrInTwG(@TwrGrupaId))))
   AND (@TylkoWgId=0 OR TwR_TwRId IN (SELECT NextIdInt FROM CDN.Text2TableInt(@TowaryId)))
  END
  --SELECT * FROM #RaportBrakow
  IF @TylkoWgId = 1  -- dodawanie towaru do raportu przez użytkownika
  BEGIN
   UPDATE #RaportBrakow  SET IloscZamawiana = 1 FROM  #RaportBrakow
   -- dopasuj ilość zamawianą do zamawiać po
   UPDATE #RaportBrakow SET
    IloscZamawianaJM = cdn.DopasujDoZamawiacPo(cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, IloscZamawiana, JM, JMZam),  T.Twr_IloscZam, 1)
   FROM #RaportBrakow
   JOIN  cdn.Towary T ON ID = T.Twr_TwrId
   WHERE IloscZamawiana > 0
  END
  ELSE
  BEGIN
   -- ustal ilość zamawianą w jednostkach podstawowych
   IF @TypAlgorytmu > 0
    DELETE #RaportBrakow WHERE IloscSprzedana <= 0
   UPDATE #RaportBrakow SET
     IloscZamawiana = CASE @TypAlgorytmu
         WHEN 1 THEN  IloscSprzedana - (Ilosc + UwzgZamowienia)
         WHEN 2 THEN CASE WHEN @UwzgledniajStanyTowarow = 0 THEN IloscSprzedana - UwzgZamowienia
            ELSE IloscSprzedana - (Ilosc + UwzgZamowienia)
            END
         ELSE   UwzgRezerwacje - (Ilosc + UwzgZamowienia)
          END
   IF @RodzajNormatywu = 0   -- nie uwzględniaj stanu minimalny ani maksymalnego
   BEGIN
    -- dopasuj ilość zamawianą do zamawiać po
    UPDATE #RaportBrakow SET
     IloscZamawianaJM = cdn.DopasujDoZamawiacPo(cdn.PrzeliczIloscJMDlaTowaru(ID, IloscZamawiana, JM, JMZam), T.Twr_IloscZam, 1)
    FROM #RaportBrakow
    JOIN  cdn.Towary T ON ID = T.Twr_TwrId
    WHERE IloscZamawiana > 0
   END
   -- uwzględnij stan minimalny lub maksymalny
   IF @RodzajNormatywu = 1   -- uwzględnij stan minimalny
   BEGIN
    IF @TypAlgorytmu = 0
     UPDATE #RaportBrakow SET
     IloscZamawiana   = cdn.PrzeliczIloscJMDlaTowaru(ID, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM) + UwzgRezerwacje - Ilosc - UwzgZamowienia
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMin > 0 AND (Ilosc + IloscZamawiana-UwzgRezerwacje) <  cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM)
    IF @TypAlgorytmu = 1  -- wg sprzedaży
     UPDATE #RaportBrakow SET
      IloscZamawiana   = cdn.PrzeliczIloscJMDlaTowaru(ID, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM) - (Ilosc + UwzgZamowienia)
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMin > 0 AND (IloscZamawiana + Ilosc + UwzgZamowienia)<  cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM)
    IF @TypAlgorytmu = 2    -- wg RO
     UPDATE #RaportBrakow SET
      IloscZamawiana   = cdn.PrzeliczIloscJMDlaTowaru(ID, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM) - (Ilosc + UwzgZamowienia)
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMin > 0 AND (IloscZamawiana + Ilosc + UwzgZamowienia)<  cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMin, T.Twr_IloscMinJM, T.Twr_JM)
    -- dopasuj ilość zamawianą do zamawiać po
    UPDATE #RaportBrakow SET
     IloscZamawianaJM = cdn.DopasujDoZamawiacPo(cdn.PrzeliczIloscJMDlaTowaru(ID, IloscZamawiana, JM, JMZam), T.Twr_IloscZam, 1)
    FROM #RaportBrakow
    JOIN  cdn.Towary T ON ID = T.Twr_TwrId
    WHERE IloscZamawiana > 0
   END
   IF @RodzajNormatywu = 2   -- uwzględnij stan maksymalny
   BEGIN
    IF @TypAlgorytmu = 0
    BEGIN
     UPDATE #RaportBrakow SET
      IloscZamawiana  = cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMax, T.Twr_IloscMaxJM, T.Twr_JM) + IloscZamawiana
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMax > 0
    END
    IF @TypAlgorytmu = 1
    BEGIN
     UPDATE #RaportBrakow SET
      IloscZamawiana = cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMax, T.Twr_IloscMaxJM, T.Twr_JM) - (Ilosc + UwzgZamowienia)
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMax > 0  AND (IloscZamawiana + Ilosc + UwzgZamowienia) >  cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMax, T.Twr_IloscMaxJM, T.Twr_JM)
    END
    IF @TypAlgorytmu = 2    -- wg RO
    BEGIN
     UPDATE #RaportBrakow SET
      IloscZamawiana = cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMax, T.Twr_IloscMaxJM, T.Twr_JM) - (Ilosc + UwzgZamowienia)
     FROM  #RaportBrakow
     JOIN  cdn.Towary T ON ID = T.Twr_TwrId
     WHERE T.Twr_IloscMax > 0  AND (IloscZamawiana + Ilosc + UwzgZamowienia) >  cdn.PrzeliczIloscJMDlaTowaru(T.Twr_TwrId, T.Twr_IloscMax, T.Twr_IloscMaxJM, T.Twr_JM)
    END
    -- dopasuj ilość zamawianą do zamawiać po (dopasuj z -1)
    UPDATE #RaportBrakow SET
     IloscZamawianaJM = cdn.DopasujDoZamawiacPo(cdn.PrzeliczIloscJMDlaTowaru(ID, IloscZamawiana, JM, JMZam), T.Twr_IloscZam, -1)
    FROM #RaportBrakow
    JOIN  cdn.Towary T ON ID = T.Twr_TwrId
    WHERE IloscZamawiana > 0
   END
  END
  -- Ustal cenę dla jednostki, w której zamawiany jest towar (najpierw jednostka zamawiana, potem podstawowa)
  UPDATE #RaportBrakow SET CenaNettoWal = cdn.PrzeliczIloscJMDlaTowaru(ID, CenaNettoWal, JMZam, JM)
  UPDATE #RaportBrakow SET
   IloscZamawiana = cdn.PrzeliczIloscJMDlaTowaru(ID, IloscZamawianaJM, JMZam, JM),  -- ustal ilość w jednostkach podstawowych
   CenaNetto =
    CASE WHEN BrakKursuWaluty = 1 THEN 0
    ELSE
     Round( CenaNettoWal * (KursL / KursM ), 2)
    END -- ustal cenę netto w walucie systemowej
  FROM #RaportBrakow
  WHERE IloscZamawiana > 0
  UPDATE #RaportBrakow SET
   BrakiPozostale  = CASE WHEN (Braki - Zamowienia) < 0 THEN 0 ELSE (Braki - Zamowienia) END, -- BrakiPozostale
   WartoscNetto = Convert( DECIMAL(20,2),CenaNetto * IloscZamawianaJM)
  FROM #RaportBrakow
  UPDATE #RaportBrakow SET
   CenaNettoWalDisp  =
   CASE WHEN @opeZakazCenZak = 1  THEN REPLICATE('-', 15)
   ELSE      Convert(NVARCHAR(20), CenaNettoWal) END, -- WartoscNettoDisp
   WartoscNettoDisp  =
   CASE WHEN @opeZakazCenZak = 1  THEN REPLICATE('-', 15)
     WHEN BrakKursuWaluty = 1  THEN '-kurs-'
   ELSE
    Convert(NVARCHAR(20), WartoscNetto)
   END -- WartoscNettoDisp
  FROM #RaportBrakow
  SET NOCOUNT OFF
  --cti
  SELECT *,(IloscZamawiana - dbo.CTIProdukcjaTwrIloscDoRealizacji2(ID)) as IloscZamawiana2 , dbo.CTIProdukcjaTwrIloscDoRealizacji2(ID) as ZP
  FROM #RaportBrakow WHERE (IloscZamawiana - dbo.CTIProdukcjaTwrIloscDoRealizacji2(ID)) > 0 ORDER BY Kod
END
GO
```

# [dbo].[CTIProdukcjaROlist]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaROlist]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaROlist]--Rezerwacja odbiorcy
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmData int, --0 bez --1 ograniczenie po dacie,
 @tylkoWyjeteZBufora int = 1
AS
BEGIN
declare  @DokTyp int = 308
	declare  @CzyJestOgraniczenieSymbolu	int = 0 SELECT @CzyJestOgraniczenieSymbolu	= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @DokTyp and KnO_Parametr = 5
	declare  @CzyJestOgraniczenieAtrybutu	int = 0 SELECT @CzyJestOgraniczenieAtrybutu	= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @DokTyp and KnO_Parametr = 6
	declare  @CzyJestOgraniczenieDaty		int = 0 SELECT @CzyJestOgraniczenieDaty		= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @DokTyp and KnO_Parametr = 7

		SELECT DISTINCT 
			e.TrN_TrNID as ID
			, ISNULL(e.TrN_NumerPelny, '') AS Dokument
			, Pod_Kod as Kontrahent
			, e.trn_datadok as DataWystawienia
			, dbo.CTIProdukcjaStatusy(308, e.TrN_TrNID) AS Status
			, dbo.CTIProdukcjaStatusyNazwy(308, dbo.CTIProdukcjaStatusy(308, e.TrN_TrNID)) AS StatusNazwa 
			, ISNULL(d.Mag_Symbol, '') AS MagZrd
			, e.TrN_NumerObcy as NumerObcy
			, e.TrN_Opis as Opis
		FROM CDN.TraNag e
			LEFT JOIN CDN.Magazyny AS d ON e.TrN_MagZrdId = d.Mag_MagId 
			LEFT JOIN CDN.PodmiotyView Pod ON Pod.Pod_PodId = TrN_PodId AND Trn_PodmiotTyp = Pod.Pod_PodmiotTyp
		where e.TrN_TypDokumentu = @DokTyp  and e.trn_anulowany = 0 
			and((@tylkoWyjeteZBufora = 1 AND TrN_Bufor = 0) OR (@tylkoWyjeteZBufora = 0))
			and((ISNULL(@prmData,0) = 0) or(@prmData =1 and CONVERT(VARCHAR, ISNULL(e.trn_datadok, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))	)
			and ( (@CzyJestOgraniczenieAtrybutu = 0)    or ( @CzyJestOgraniczenieAtrybutu <> 0	and TrN_TrNID in ( select distinct AtrOgr.DAt_TrNId from cdn.DokAtrybuty AtrOgr INNER JOIN dbo.CtiKonfigOgraniczenia on  KnO_IdList = @DokTyp and KnO_Parametr = 6 and  AtrOgr.DAt_DeAId = KnO_WartoscId and AtrOgr.DAt_WartoscTxt = KnO_WartoscAtrTxt)))
		and ( (@CzyJestOgraniczenieDaty = 0)		or ( @CzyJestOgraniczenieDaty <> 0		
				and CONVERT(VARCHAR, ISNULL(e.trn_datadok, GETDATE()), 23) >=
				CONVERT(VARCHAR, ISNULL((SELECT Max(KnO_WartoscTxt) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @DokTyp and KnO_Parametr = 7), GETDATE()), 23)
				))
			and ( (@CzyJestOgraniczenieSymbolu = 0) or ( @CzyJestOgraniczenieSymbolu <> 0		
				AND SUBSTRING(e.TrN_NumerPelny,0,CHARINDEX('/',e.TrN_NumerPelny) ) IN
				(SELECT dISTINCT KnO_WartoscTxt FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @DokTyp and KnO_Parametr = 5)
				))
		ORDER BY e.TrN_TrNID
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaROlistZP]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaROlistZP]
 @idRO int,
 @idPozycji int 
AS
BEGIN
declare @DokTyp int = 2 -- ZP
declare @prmPole1 int = 0
declare @prmPole2 int = 0
declare @prmPole3 int = 0
declare @obslugaWydania int = 0
declare @obslugaKontroli int = 0 

select @obslugaKontroli= isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where knf_numer =63
select @obslugaWydania = isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where knf_numer=62
SELECT @prmPole1 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 1 and CAU_TypDok = @DokTyp
SELECT @prmPole2 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 2 and CAU_TypDok = @DokTyp
SELECT @prmPole3 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 3 and CAU_TypDok = @DokTyp


	SELECT  CZN_ID AS ID, CZN_NrPelny AS CZN_Numer, isnull(CZN_CTNId,0) as CZN_CTNId
		,isnull(CZN_TwrId,0) as CZN_TwrId, isnull(Twr_Kod,'') AS CZN_TwrKod, isnull(CZN_Ilosc, 0) AS Ilosc
		,isnull(CZN_OpeWystaw,0) as CZN_OpeWystaw,isnull(CZN_OperModyf,0) as CZN_OperModyf
		,isnull(CZN_DataWystaw, GETDATE()) AS CZN_DataWystaw ,isnull(CZN_DataModyf, GETDATE()) AS CZN_DataModyf
		,isnull(a.CO_Kod,'') AS Wystawil, isnull(b.CO_Kod,'') AS Modyfikowal 
		,isnull(CTN_Nazwa,'') AS CTN_Nazwa,isnull(CZN_Opis, '') AS CZN_Opis 
		,isnull(CtiTechnolNag.CTN_Cecha, 0) AS ctn_cecha
		,dbo.CTIProdukcjaStatusy(2, CZN_ID) AS CZN_Status
		,0--dbo.CTIProdukcjaStatusy(23, CZN_ID) 
		AS StatusNumer1
		,0--dbo.CTIProdukcjaStatusy(24, CZN_ID) 
		AS StatusNumer2
		,0--dbo.CTIProdukcjaStatusy(26, CZN_ID) 
		AS StatusNumer3
		,case @obslugaWydania when 1 then dbo.CTIProdukcjaStatusyNazwy(23, dbo.CTIProdukcjaStatusy(23, CZN_ID)) else '' end AS Status1
		,dbo.CTIProdukcjaStatusyNazwy(24, dbo.CTIProdukcjaStatusy(24, CZN_ID)) AS Status2
		,case @obslugaKontroli when 1 then dbo.CTIProdukcjaStatusyNazwy(25, dbo.CTIProdukcjaStatusy(26, CZN_ID)) else '' end AS Status3
		, case @prmPole1 when 1 then (SELECT TOP (1) Wartosc
		FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0,0) 
		AS CTIProdukcjaTowaryPolaSpecjalne_3
		WHERE (Lp = 1)) else '' end AS Pole1
		, case @prmPole2 when 1 then (SELECT TOP (1) Wartosc
		FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0,0) 
		AS CTIProdukcjaTowaryPolaSpecjalne_2
		WHERE (Lp = 2))  else '' end AS Pole2
		, case @prmPole3 when 1 then (SELECT TOP (1) Wartosc
		FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0,0) 
		AS CTIProdukcjaTowaryPolaSpecjalne_1
		WHERE (Lp = 3))  else '' end AS Pole3

		,isnull((select top 1 czze_czznid from dbo.CtiZlecenieZElem where czze_cznid = CZN_ID),0) as czznid
		,isnull(CZN_CzasStart, GETDATE()) AS CzasStart,isnull(CZN_CzasEnd, GETDATE()) AS CzasEnd
	FROM dbo.CtiZlecenieNag 
		LEFT JOIN dbo.CtiOperatorzy AS a ON isnull(CZN_OpeWystaw,0) = a.CO_ID 
		LEFT JOIN dbo.CtiOperatorzy AS b ON isnull(CZN_OperModyf,0) = b.CO_ID 
		LEFT JOIN dbo.CtiTechnolNag ON isnull(CZN_CTNId,0) = CTN_ID 
		LEFT JOIN CDN.Towary ON isnull(CZN_TwrId,0) = Twr_TwrId
	where
		isnull(CZN_DokZwiazTyp,0) = 308
		and ((isnull(@idPozycji,0) = 0 and 1=1) OR (isnull(@idPozycji,0) != 0 and isnull(czn_DokZwiazTreId,0) = @idPozycji))
		and ((isnull(@idRO,0) = 0 and 1=1) OR (isnull(@idRO,0) != 0 and isnull(czn_DokZwiazId,0) = @idRO))

	ORDER BY 1
END

GO
```

# [dbo].[CTIProdukcjaRWlist]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaRWlist]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaRWlist]
 @ID int,
 @IDZPZ int,
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmData int--0 bez --1 ograniczenie po dacie
AS
BEGIN
--czy poszczególne parametry maja być widoczne
declare @DokTyp int = 5 -- RW
--declare @prmMontaz int = 0 SELECT  @prmMontaz = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 25 and ISNULL(KNF_Wartosc,'') <> ''
declare @prmPrc int = 0 SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> ''
if @prmPrc in (2,3,4) begin set @prmPrc = 1 end else begin set @prmPrc = 0 end
declare @prmPole1 int = 0 SELECT   @prmPole1 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 1 and CAU_TypDok = @DokTyp
declare @prmPole2 int = 0 SELECT   @prmPole2 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 2 and CAU_TypDok = @DokTyp
declare @prmPole3 int = 0 SELECT   @prmPole3 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 3 and CAU_TypDok = @DokTyp

		SELECT DISTINCT 
			e.TrN_TrNID as ID, ISNULL(CtiZlecenieNag.CZN_ID, 0) AS CZN_ID
			, ISNULL(e.TrN_NumerPelny, '') AS DokumentWydania
			, ISNULL(a.CO_Kod, '') AS CZD_OpeWyst, ISNULL(CtiZlecenieDok.CZD_DataWyst, GETDATE())AS CZD_DataWyst
			, ISNULL(b.CO_Kod, '') AS CZD_OperModyf, ISNULL(CtiZlecenieDok.CZD_DataModyf, GETDATE())   AS CZD_DataModyf
			, ISNULL(CtiZlecenieNag.CZN_TwrId, 0) AS ZasobID 
			, ISNULL(CDN.Towary.Twr_Kod, '') AS ZasobKod
			, ISNULL(dbo.CTIProdukcjaZlecenieNumer(CtiZlecenieNag.CZN_ID), '') AS CZN_Numer
			, ISNULL(CZO_PrcKod, '') AS PRI_Kod --pracownik
			, dbo.CTIProdukcjaStatusy(3, CtiZlecenieDok.CZD_ID) AS CZD_Status
			, dbo.CTIProdukcjaStatusyNazwy(3, dbo.CTIProdukcjaStatusy(3, CtiZlecenieDok.CZD_ID)) AS StatusNazwa 
			--, ISNULL(CMD_PrdKod, '') AS ProdKod--montaz
			,''  AS ProdKod
			, ISNULL(c.Mag_Symbol, '') AS MagDoc 
			, ISNULL(d.Mag_Symbol, '') AS MagZrd
			, case @prmPole1 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) 
				AS CTIProdukcjaTowaryPolaSpecjalne_3
				WHERE (Lp = 1)) else '' end AS Pole1
			, case @prmPole2 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) 
				AS CTIProdukcjaTowaryPolaSpecjalne_2
				WHERE (Lp = 2))  else '' end AS Pole2
			, case @prmPole3 when 1 then (SELECT TOP (1) Wartosc
				FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, e.TrN_TrNID, 0, 0,0) 
				AS CTIProdukcjaTowaryPolaSpecjalne_1
				WHERE (Lp = 3))  else '' end AS Pole3
		FROM CtiZlecenieDok 
			INNER JOIN CtiZlecenieNag ON CtiZlecenieDok.CZD_CZNId = CtiZlecenieNag.CZN_ID 
			INNER JOIN CDN.TraNag AS e ON CtiZlecenieDok.CZD_TrnId = e.TrN_TrNID 
			LEFT JOIN CtiOperatorzy AS a ON CtiZlecenieDok.CZD_OpeWyst = a.CO_ID 
			LEFT JOIN CtiOperatorzy AS b ON CtiZlecenieDok.CZD_OperModyf = b.CO_ID 
			LEFT JOIN dbo.CtiZleceniePrc ON  CZO_CZNId = CZN_ID and CZO_PrcId = CZD_PracownikId and @prmPrc = 1
			LEFT JOIN CDN.Magazyny AS c ON e.TrN_MagDocId = c.Mag_MagId 
			LEFT JOIN CDN.Magazyny AS d ON e.TrN_MagZrdId = d.Mag_MagId 
			--LEFT JOIN CtiMontazDok ON e.TrN_TrNID = CtiMontazDok.CMD_PWId and @prmMontaz = 1
			LEFT JOIN CDN.Towary ON CtiZlecenieNag.CZN_TwrId = CDN.Towary.Twr_TwrId
			LEFT JOIN dbo.CtiZlecenieZElem ON CZN_ID = ISNULL(CZZE_CZNID,0)
		where e.TrN_TypDokumentu in  (304)  and e.trn_anulowany = 0 
		and
		(
			((@ID = 0) or (@ID <> 0 and CZN_ID = @ID ))
			and
			((@IDZPZ = 0) or (@IDZPZ <> 0 and ISNULL(CZZE_CZZNID,0) = @IDZPZ ))
			and
			(
				 (ISNULL(@prmData,0) =0) 
				 or
				(@prmData =1 and
				CONVERT(VARCHAR, ISNULL(e.trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)
		)
		ORDER BY e.TrN_TrNID
END

GO
```

# [dbo].[CTIProdukcjaTechnologiaAddEdit]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaAddEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaTechnologiaAddEdit]
   @id int output
 , @Opis nvarchar(254)
 , @Ope int
 , @Status int
 , @TwrId int
 , @cenaEwid decimal(15,4) 
 , @NumerObcy varchar(40)
 , @ilosc int
 , @NrWlasny nvarchar(50)
 , @Jm nvarchar(40) = NULL
 , @DomyslneMagazyny tinyint = 0
 , @zlecenieID int = NULL
 , @szablon int = NULL
 , @generateRWAfterResource int = 0
AS
BEGIN
declare @numerTech int = 0
declare @miesiac int = month(GETDATE())
declare @rok int = year(GETDATE())
declare @nrpelny varchar(20) =''
declare @Seria varchar(5) = ''
declare @konfiguracja varchar(1) = ''

select @konfiguracja = isnull(KNF_Wartosc,'') from dbo.CtiKonfiguracja where KNF_Numer = 72

if (@Jm IS NULL)
begin
	SET @Jm = (SELECT TOP 1 Twr_JM FROM CDN.Towary WHERE Twr_TwrId = @TwrId);
end


if @konfiguracja = '1'--uwzględnienie miesiąca w numeracji
begin
	select @numerTech = isnull(Max(CTN_Numer),0)+1  from dbo.CtiTechnolNag
	where CTN_Rok = @rok and CTN_Miesiac = @miesiac --miesiac
	select  @nrpelny='TH/'+cast(replace(str(@numerTech,4),' ','0')as char(4)) +'/'+cast(replace(str(@miesiac,2),' ','0')as char(2)) +'/'+cast(@rok as varchar(4))
end
else
begin
	select @numerTech = isnull(Max(CTN_Numer),0)+1  from dbo.CtiTechnolNag
	where CTN_Rok = @rok --rok
	select  @nrpelny='TH/'+cast(replace(str(@numerTech,5),' ','0')as char(5)) +'/'+cast(@rok as varchar(4))
end


	declare @TwrKod varchar(100) = ''
	select @TwrKod=Twr_Kod from cdn.Towary where Twr_TwrId = @TwrId
	if @TwrId =0 and  @id = 0
	begin
		INSERT INTO dbo.CtiTechnolNag
		(CTN_Opis, CTN_TwrId, CTN_TwrKod, CTN_Nazwa, CTN_OpeWystaw, CTN_DataWystaw, CTN_OperModyf, CTN_DataModyf, CTN_Status,CTN_Numer,CTN_Miesiac,CTN_Rok,CTN_Seria,CTN_Cecha,CTN_CenaEwidencyjna,CTN_NumerObcy,CTN_Ilosc, CTN_TwrJm, CTN_DomyslneMagSur, CTN_ZpZrodloweID,
		CTN_Szablon, CTN_GenerateRWAfterResource)
		VALUES ('',0,'',@nrpelny,@Ope,GETDATE(),@Ope,GETDATE(),0,@numerTech,@miesiac,@rok,@Seria,0,@cenaEwid,@NumerObcy,@ilosc,@Jm, @DomyslneMagazyny, @zlecenieID,
		isnull(@szablon, 0), @generateRWAfterResource)
		Select  @id = SCOPE_IDENTITY() 
	end
	if @TwrId >0 and  @id = 0
	begin
		INSERT INTO dbo.CtiTechnolNag
		(CTN_Opis, CTN_TwrId, CTN_TwrKod, CTN_Nazwa, CTN_OpeWystaw, CTN_DataWystaw, CTN_OperModyf, CTN_DataModyf, CTN_Status,CTN_Numer,CTN_Miesiac,CTN_Rok,CTN_Seria,CTN_Cecha,CTN_CenaEwidencyjna,CTN_NumerObcy,CTN_Ilosc, CTN_TwrJm, CTN_DomyslneMagSur, CTN_ZpZrodloweID,
		CTN_Szablon, CTN_GenerateRWAfterResource)
		VALUES (@Opis,@TwrId,@TwrKod,@nrpelny,@Ope,GETDATE(),@Ope,GETDATE(),@Status,@numerTech,@miesiac,@rok,@Seria,0,@cenaEwid,@NumerObcy,@ilosc,@Jm, @DomyslneMagazyny, @zlecenieID,
		isnull(@szablon, 0), @generateRWAfterResource)
		Select @id = SCOPE_IDENTITY() 
	end

	if @id > 0
	begin
		declare @cecha int = 0

		select @cecha = isnull(a.twa_deaid,0) from cdn.towary
		LEFT JOIN CDN.TwrAtrybuty a on a.TwA_TwrID = Twr_TwrId and a.TwA_DeAId = 
		( 
			select DISTINCT TwA_DeAId from cdn.twratrybuty
			INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId
			INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 5 and DeA_Kod = KNF_Wartosc 
			where TwA_CzyPrzenosicNaDostawy = 1 ) 
		where twr_twrid = @TwrId

		
		if (@cecha <= 0)
		begin
			--jeżeli nie ma pierwszej cechy to sprawdzamy czy jest druga

			select @cecha = isnull(a.twa_deaid,0) from cdn.towary
			LEFT JOIN CDN.TwrAtrybuty a on a.TwA_TwrID = Twr_TwrId and a.TwA_DeAId = 
			( 
				select DISTINCT TwA_DeAId from cdn.twratrybuty
				INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId
				INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 19 and DeA_Kod = KNF_Wartosc 
				where TwA_CzyPrzenosicNaDostawy = 1 ) 
			where twr_twrid = @TwrId
		end

		if(@cecha > 0)
		begin
			set @cecha = 1
		end

		UPDATE dbo.CtiTechnolNag
	    SET CTN_Opis = isnull(@Opis,CTN_Opis)
		  , CTN_TwrId = Isnull(@TwrId,CTN_TwrId)
		  , CTN_TwrKod = Isnull(@TwrKod,CTN_TwrKod)
		  , CTN_OperModyf = @Ope
		  , CTN_DataModyf = GETDATE()
		  , CTN_Status = Isnull(@Status,CTN_Status)
		  , CTN_Cecha = Isnull(@cecha,CTN_Cecha)
		  , CTN_CenaEwidencyjna = @cenaEwid
		  , CTN_Ilosc = Isnull(@ilosc,CTN_Ilosc)
		  , CTN_TwrJm = @Jm
		  , CTN_DomyslneMagSur = @DomyslneMagazyny
		  , CTN_ZpZrodloweID = ISNULL(@zlecenieID, CTN_ZpZrodloweID)
		  , CTN_Szablon = ISNULL(@szablon, CTN_Szablon)
		  , CTN_GenerateRWAfterResource = @generateRWAfterResource
	   WHERE CTN_ID = @id
	   
	   --usunięcie elementów z cechą jeżeli technologia nie jest z cechą
		if(@cecha = 0  )
		begin
			delete from dbo.CtiTechnolElem where CTE_CTNId  = @id and CTE_Typ = 1
		end

		--numer wlasny
		if not exists (SELECT isnull(DNW_Numer,0) FROM dbo.CtiDokNrWlasny where DNW_DokTyp = 1 and DNW_DokID = @id)
			insert dbo.CtiDokNrWlasny (DNW_DokTyp, DNW_DokID, DNW_Numer) values (1,@id,@NrWlasny)
		else
			update  dbo.CtiDokNrWlasny set DNW_Numer = @NrWlasny  where DNW_DokTyp = 1 and DNW_DokID = @id
	end
	select @id as ID
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaDuplikat]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaTechnologiaDuplikat]
@idTechnologii int,
@idOperatora int,
@NrWlasny nvarchar(50),
@Szablon int = NULL
AS
BEGIN

declare @numerTech int = 0
declare @miesiac int = month(GETDATE())
declare @rok int = year(GETDATE())
declare @nrpelny varchar(20) = ''
declare @Seria varchar(5) = ''

select @numerTech = isnull(Max(CTN_Numer),0)+1  from dbo.CtiTechnolNag
where CTN_Rok = @rok --and isnull(CTN_Seria,'') = '' and CTN_Miesiac = @miesiac --tylko rok
select  @nrpelny='TH/'+cast(replace(str(@numerTech,5),' ','0')as char(5)) +'/'+cast(@rok as varchar(4))

	declare @NoweId int = 0

	INSERT INTO dbo.CtiTechnolNag
	(CTN_Opis, CTN_TwrId, CTN_TwrKod, CTN_Nazwa, CTN_OpeWystaw, CTN_DataWystaw, CTN_OperModyf, CTN_DataModyf, CTN_Status, CTN_Numer, CTN_Miesiac, CTN_Rok, CTN_Seria, Ctn_ilosc, CTN_CenaEwidencyjna, CTN_TwrJM, CTN_DomyslneMagSur, CTN_Szablon)
	select '', 0, '', @nrpelny, @idOperatora, GETDATE(), @idOperatora, GETDATE(), 0, @numerTech, @miesiac, @rok, @Seria, isnull(ctn_ilosc,1), CTN_CenaEwidencyjna, CTN_TwrJM, CTN_DomyslneMagSur, ISNULL(@Szablon, CTN_Szablon)
	from [dbo].[CtiTechnolNag] where CTN_ID = @idTechnologii
	Select @NoweId = SCOPE_IDENTITY() 
	
	if @NoweId>0
	begin

	--kopiowanie działów
	INSERT INTO dbo.CtiTechnolDzialy
	(CTDZ_TechnolID, CTDZ_DokTyp, CTDZ_DzialID)
	SELECT 
	@NoweId, 1, CTDZ_DzialID
	FROM dbo.CtiTechnolDzialy
	WHERE CTDZ_TechnolID = @idTechnologii
	AND CTDZ_DokTyp = 1;


	--elementy
	DECLARE kursorElem cursor FOR
		 select CTE_ID as ID, case CTE_Typ when 3 then 73 when 4 then 74 else 71 end as TYP  from dbo.CtiTechnolElem where CTE_CTNId = @idTechnologii order by CTE_Lp
	DECLARE @idElem int, @typElem int OPEN kursorElem

	FETCH NEXT FROM kursorElem   INTO @idElem, @typElem
	while @@FETCH_STATUS = 0
	begin
		INSERT INTO dbo.CtiTechnolElem (CTE_CTNId, CTE_Lp, CTE_TwrId, CTE_TwrKod, CTE_Ilosc, CTE_Typ, CTE_CenaWaga, CTE_Cena, CTE_Magazyn, CTE_Opis, CTE_Staly, CTE_DzialID)
		select @NoweId, CTE_Lp, CTE_TwrId, CTE_TwrKod, CTE_Ilosc, CTE_Typ, CTE_CenaWaga, CTE_Cena, CTE_Magazyn, CTE_Opis, CTE_Staly, CTE_DzialID from dbo.CtiTechnolElem where CTE_CTNId = @idTechnologii and CTE_ID = @idElem
		declare @NoweIdElam int = SCOPE_IDENTITY() 

			if(@typElem not in (1,2))--cechy dla elementów--nie surowce, 1 to TH
			begin
				INSERT INTO CtiTechnolCecha (CTC_CTNID, CTC_Typ, CTC_TwrId, CTC_Def, CTC_NrCechy)
				SELECT @NoweIdElam, @typElem, CTC_TwrId, CTC_Def, CTC_NrCechy FROM CtiTechnolCecha where CTC_CTNID = @idElem  and isnull(CTC_Typ,1) = @typElem
			end     


		--kopiowanie planowanych cech
		INSERT INTO dbo.CtiPlanowanaCecha
		(CPC_TypDokumentu, CPC_IdElementu, CPC_IdTowaru, CPC_AtrybutID, 
		CPC_AtrybutTyp, CPC_AtrybutKod, CPC_Komparator, CPC_Wartosc, CPC_TypDaty) 
		SELECT
		CPC_TypDokumentu, @NoweIdElam, CPC_IdTowaru, CPC_AtrybutID, 
		CPC_AtrybutTyp, CPC_AtrybutKod, CPC_Komparator, CPC_Wartosc, CPC_TypDaty
		FROM dbo.CtiPlanowanaCecha
		WHERE CPC_TypDokumentu = 1 AND CPC_IdElementu = @idElem;


		FETCH NEXT FROM kursorElem  INTO @idElem, @typElem
	end

	CLOSE kursorElem  deallocate kursorElem

		--cecha dla TH
		INSERT INTO CtiTechnolCecha (CTC_CTNID, CTC_Typ, CTC_TwrId, CTC_Def, CTC_NrCechy)
		SELECT top 1 @NoweId,1,  CTC_TwrId, CTC_Def, CTC_NrCechy FROM CtiTechnolCecha where CTC_CTNID = @idTechnologii and isnull(CTC_Typ,1) = 1 and CTC_NrCechy = 1 order by isnull(CTC_Typ,99)
		INSERT INTO CtiTechnolCecha (CTC_CTNID, CTC_Typ, CTC_TwrId, CTC_Def, CTC_NrCechy)
		SELECT top 1 @NoweId,1,  CTC_TwrId, CTC_Def, CTC_NrCechy FROM CtiTechnolCecha where CTC_CTNID = @idTechnologii and isnull(CTC_Typ,1) = 1 and CTC_NrCechy = 2 order by isnull(CTC_Typ,99)

		--zasoby 'główne'
		insert into dbo.CtiZasobDok
			(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia,CZS_DokTyp
			, CZS_PwBlokada, CZS_RwBlokada, CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita)
		SELECT CZS_CZID, CZS_CZGID, @NoweId, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia, CZS_DokTyp
			, isnull(CZS_PwBlokada,0), isnull(CZS_RwBlokada,0), CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp
			, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita
		FROM dbo.CtiZasobDok 
			LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
			INNER JOIN  dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID 
			where CZS_DokTyp = 1
			and CZS_ctnid =@idTechnologii
			and CZS_ParentID IS NULL
		order by isnull(CZS_Lp,0), CZS_ID

		--zasoby 'pracownicy'
		insert into dbo.CtiZasobDok
			(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia,CZS_DokTyp
			, CZS_PwBlokada, CZS_RwBlokada, CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita)
		SELECT stary.CZS_CZID, stary.CZS_CZGID, @NoweId, stary.CZS_Kod, stary.CZS_JMCzasu 
			, stary.CZS_CzasPracy,stary.CZS_CzasUzbrojenia, stary.CZS_CzasRozbrojenia, stary.CZS_KosztPracy 
			, stary.CZS_KosztUzbrojenia, stary.CZS_KosztRozbrojenia, stary.CZS_DokTyp
			, isnull(stary.CZS_PwBlokada,0), isnull(stary.CZS_RwBlokada,0), stary.CZS_Ilosc, stary.CZS_CzasStart, stary.CZS_CzasEnd
			, stary.CZS_Lp
			, stary.CZS_Opis, stary.CZS_PrcId, stary.CZS_Dzielnik, stary.CZS_PlanowanieCzasuPracy, stary.CZS_CzasPlanowany, stary.CZS_WieleOsob, stary.CZS_ZasobPrcId
			, (
				SELECT TOP 1 nowy.CZS_ID 
				FROM dbo.CtiZasobDok nowy 
				WHERE nowy.CZS_DokTyp = 1
				AND nowy.CZS_CTNID = @NoweId
				AND nowy.CZS_LP = parent.CZS_LP
			), stary.CZS_UtworzonoPracownika, stary.CZS_PWOdpad, stary.CZS_RealizacjaCalkowita
		FROM dbo.CtiZasobDok stary
			JOIN dbo.CtiZasobDok parent ON stary.CZS_ParentID = parent.CZS_ID
			LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and stary.CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob on CZ_ID = stary.CZS_CZID
			INNER JOIN  dbo.CtiZasobGrupy on stary.CZS_CZGID = CZG_ID 
			where stary.CZS_DokTyp = 1
			and stary.CZS_ctnid =@idTechnologii
			and stary.CZS_ParentID IS NOT NULL
		order by isnull(stary.CZS_Lp,0), stary.CZS_ID

		--numer wlasny
		if not exists (SELECT isnull(DNW_Numer,0) FROM dbo.CtiDokNrWlasny where DNW_DokTyp = 1 and DNW_DokID = @NoweId)
			insert dbo.CtiDokNrWlasny (DNW_DokTyp, DNW_DokID, DNW_Numer) values (1, @NoweId, @NrWlasny)
		else
			update  dbo.CtiDokNrWlasny set DNW_Numer = @NrWlasny  where DNW_DokTyp = 1 and DNW_DokID = @NoweId
	end
	select @NoweId as ID
END
GO
```

# [dbo].[CTIProdukcjaTechnologiaElemAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaElemAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaTechnologiaElemAdd]
  @DokId					int
, @idSurowca				int
, @LogujacyId				int
, @KodSurowca				varchar(40)
, @typSurowca				int
, @ilosc					decimal(15,4) = 1
, @agregacjaSurowcow		tinyint = 0
, @magazyn					int = null
, @opis						varchar(max) = null
, @staly					int = 0
, @cena						decimal(15, 4) = 0
AS
BEGIN
--@typSurowca
--1 - Surowiec z cechą - jeżeli włączona obsługa cech
--2 - Surowiec
--3 - Odpad
--4 - Praca/koszt/Usługa

	UPDATE dbo.CtiTechnolNag SET  CTN_OperModyf =@LogujacyId, CTN_DataModyf = GETDATE() from  dbo.CtiTechnolNag WHERE CTN_ID = @DokId
	if (isnull(@ilosc,0)=0) 
		set @ilosc = 1

	if (@agregacjaSurowcow = 1 and exists (select * from dbo.CtiTechnolElem WHERE CTE_TwrId = @idSurowca AND CTE_Typ = @typSurowca AND CTE_CTNId = @DokId))
	begin
		update dbo.CtiTechnolElem SET CTE_Ilosc = CTE_Ilosc + @ilosc, CTE_Cena=@cena WHERE CTE_TwrId = @idSurowca AND CTE_Typ = @typSurowca AND CTE_CTNId = @DokId
	end
	else
	begin
		insert into dbo.CtiTechnolElem (CTE_CTNId, CTE_Lp, CTE_TwrId, CTE_TwrKod, CTE_Ilosc, CTE_typ , CTE_CenaWaga, CTE_Cena, CTE_Magazyn, CTE_Opis, CTE_Staly)
		select @DokId, (COUNT(*)+1) as CTE_Lp, @idSurowca, @KodSurowca, @ilosc,@typSurowca,0, @cena, @magazyn, @opis, @staly
		from CtiTechnolElem where CTE_CTNId = @DokId and CTE_typ = @typSurowca
	end

	
	
END
GO
```

# [dbo].[CTIProdukcjaTechnologiaList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaTechnologiaList]
 @status int,
 @zp int,
 @idWyrobu int = 0
AS
BEGIN
declare @DokTyp int = 1 --produkcja TH
declare @prmPole1 int = 0 SELECT @prmPole1 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 1 and CAU_TypDok = @DokTyp
declare @prmPole2 int = 0 SELECT @prmPole2 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 2 and CAU_TypDok = @DokTyp
declare @prmPole3 int = 0 SELECT @prmPole3 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 3 and CAU_TypDok = @DokTyp
declare @twrid int = 0
if ISNULL(@zp,0) <> 0 select @twrid = CZN_TwrId from dbo.CtiZlecenieNag where czn_id = @zp and CZN_DokZwiazTyp = 308

	SELECT DISTINCT
		CTN_ID as ID, CTN_Opis, CTN_TwrId, Twr_Kod as CTN_TwrKod, CTN_Nazwa
		, CTN_OpeWystaw, CTN_DataWystaw, CTN_OperModyf
		, ISNULL(CTN_DataModyf, GETDATE()) AS CTN_DataModyf, a.CO_Kod AS Wystawil, b.CO_Kod AS Modyfikowal
		, case  when CTN_DodatkowyStatus <> -1 THEN CTN_DodatkowyStatus ELSE CTN_Status END AS CTN_Status
		, CASE WHEN ISNULL(CTD_ID, 0) = 0 THEN '' ELSE 'Domyślna' END AS domyslna
		, case @prmPole1 when 1 then (SELECT TOP (1) Wartosc
			FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CTN_ID, 0,0,0,0,0,0) AS CTIProdukcjaTowaryPolaSpecjalne_3
			WHERE (Lp = 1))   else '' end AS Pole1
		, case @prmPole2 when 1 then (SELECT TOP (1) Wartosc
			FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CTN_ID, 0,0,0,0,0,0) AS CTIProdukcjaTowaryPolaSpecjalne_2
			WHERE (Lp = 2))   else '' end AS Pole2
		, case @prmPole3 when 1 then (SELECT TOP (1) Wartosc
			FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CTN_ID, 0,0,0,0,0,0) AS CTIProdukcjaTowaryPolaSpecjalne_1
			WHERE (Lp = 3))   else '' end AS Pole3
		, CTN_ID, ISNULL(CTN_NumerObcy,'') as NumerObcy
		, replace(Twr_Nazwa,char(10), ' ')  as TwrNazwa
		, ISNULL(DNW_Numer,CTN_Nazwa) as NumerDokWlasny
	FROM dbo.CtiTechnolNag 
		INNER JOIN cdn.towary ON twr_twrid = CTN_TwrId 
		LEFT JOIN dbo.CtiOperatorzy AS a ON CTN_OpeWystaw = a.CO_ID 
		LEFT JOIN dbo.CtiOperatorzy AS b ON CTN_OperModyf = b.CO_ID 
		LEFT JOIN dbo.CtiTechnolDomyslna  on CTN_TwrId = CTD_TwrId and ctd_Id = ctn_id
		LEFT JOIN dbo.CtiDokNrWlasny on CTN_ID = isnull(DNW_DokID,0) and  1 = isnull(DNW_DokTyp,0) 
	where ((ISNULL(@twrid,0)=0)or(ISNULL(@twrid,0) = CTN_TwrId))
	and
	(
		(ISNULL(@idWyrobu,0)=0)or(ISNULL(@idWyrobu,0) = CTN_TwrId)
	)
	and 
	(
		(@status = -1)
		or
		(@status = 0 and CTN_Status= 0)
		or 
		(@status = 1 and CTN_Status not in (0,4))
		or 
		(@status = 4 and CTN_Status = 4)
	)
	order by 1 desc
END


GO
```

# [dbo].[CTIProdukcjaTechnologiaZAddEdit]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaZAddEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaTechnologiaZAddEdit]
   @id			int output 
 , @Opis		nvarchar(254)
 , @Ope			int
 , @Status		int
 , @TwrId		int
AS
BEGIN

declare @numerTech int = 0
declare @miesiac int =month(GETDATE())
declare @rok int =  year(GETDATE())
declare @nrpelny varchar(20) =''
declare @Seria varchar(5) = ''

select @numerTech = isnull(Max(CTZN_Numer),0)+1  from dbo.CtiTechnolZNag where CTZN_Rok = @rok --and isnull(CTZN_Seria,'') = '' and CTZN_Miesiac = @miesiac --tylko rok
select  @nrpelny='THZ/'+cast(replace(str(@numerTech,5),' ','0')as char(5)) +'/'+cast(@rok as varchar(4))

	declare @TwrKod varchar(20) = '' select @TwrKod=Twr_Kod from cdn.Towary where Twr_TwrId = @TwrId

	if @TwrId =0 and  @id = 0
	begin
		INSERT INTO dbo.CtiTechnolZNag
		(CTZN_Opis, CTZN_TwrId, CTZN_TwrKod, CTZN_Nazwa, CTZN_OpeWystaw, CTZN_DataWystaw, CTZN_OperModyf, CTZN_DataModyf, CTZN_Status,CTZN_Numer,CTZN_Miesiac,CTZN_Rok,CTZN_Seria,CTZN_Cecha)
		VALUES ('',0,'',@nrpelny,@Ope,GETDATE(),@Ope,GETDATE(),0,@numerTech,@miesiac,@rok,@Seria,0)
		Select SCOPE_IDENTITY() as ID 
	end
	if @TwrId >0 and  @id = 0
	begin
		INSERT INTO dbo.CtiTechnolZNag
		(CTZN_Opis, CTZN_TwrId, CTZN_TwrKod, CTZN_Nazwa, CTZN_OpeWystaw, CTZN_DataWystaw, CTZN_OperModyf, CTZN_DataModyf, CTZN_Status,CTZN_Numer,CTZN_Miesiac,CTZN_Rok,CTZN_Seria,CTZN_Cecha)
		VALUES (@Opis,@TwrId,@TwrKod,@nrpelny,@Ope,GETDATE(),@Ope,GETDATE(),@Status,@numerTech,@miesiac,@rok,@Seria,0)
		Select SCOPE_IDENTITY() as ID 
	end

	if @id > 0
	begin
		declare @cecha int = 0
		--select @cecha = count(*) from dbo.CtiTechnolElem where  CTE_CTNId = @id and CTE_Typ = 1
		select @cecha = isnull(a.twa_deaid,0) from cdn.towary
		LEFT JOIN CDN.TwrAtrybuty a on a.TwA_TwrID = Twr_TwrId and a.TwA_DeAId = 
		( select DISTINCT TwA_DeAId from cdn.twratrybuty
		INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId
		INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 5 and DeA_Kod = KNF_Wartosc 
		where  TwA_CzyPrzenosicNaDostawy = 1 ) where twr_twrid = @TwrId

		if(@cecha >0 ) begin set @cecha = 1 end

		UPDATE dbo.CtiTechnolZNag
	    SET CTZN_Opis = @Opis
		  , CTZN_TwrId = @TwrId
		  , CTZN_TwrKod = @TwrKod
		  , CTZN_OperModyf = @Ope
		  , CTZN_DataModyf = GETDATE()
		  , CTZN_Status = @Status
		  , CTZN_Cecha = @cecha
	   WHERE CTZN_ID = @id
	end

END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTechnologiaZDuplikat]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaTechnologiaZDuplikat]
@idTechnologii INT,
@idOperatora INT
AS
BEGIN
DECLARE @numerTech INT = 0
DECLARE @miesiac INT =month(GETDATE())
DECLARE @rok INT =  year(GETDATE())
DECLARE @nrpelny VARCHAR(20) =''
declare @Seria VARCHAR(5) = ''

SELECT @numerTech = isnull(Max(CTZN_Numer),0)+1  FROM dbo.CtiTechnolZNag
WHERE CTZN_Rok = @rok --and isnull(CTZN_Seria,'') = '' and CTZN_Miesiac = @miesiac --tylko rok
SELECT  @nrpelny='THZ/'+cast(replace(str(@numerTech,5),' ','0')AS CHAR(5)) +'/'+cast(@rok AS VARCHAR(4))



	DECLARE @NoweId INT
	SET @NoweId = 0
	
	INSERT INTO dbo.CtiTechnolZNag
			(CTZN_Opis, CTZN_TwrId, CTZN_TwrKod, CTZN_Nazwa, CTZN_OpeWystaw, CTZN_DataWystaw, CTZN_OperModyf, CTZN_DataModyf, CTZN_Status,CTZN_Numer,CTZN_Miesiac,CTZN_Rok,CTZN_Seria,CTZN_Cecha)
		SELECT CTZN_Opis, CTZN_TwrId, CTZN_TwrKod, @nrpelny, @idOperatora, GETDATE(), @idOperatora, GETDATE(), 0,@numerTech,@miesiac,@rok,@Seria,CTZN_Cecha
		FROM dbo.CtiTechnolZNag WHERE CTZN_ID = @idTechnologii
	SELECT @NoweId = SCOPE_IDENTITY() 

	--skladniki
	IF @NoweId > 0
	BEGIN
		--kopiowanie działów
		INSERT INTO dbo.CtiTechnolDzialy
		(CTDZ_TechnolID, CTDZ_DokTyp, CTDZ_DzialID)
		SELECT 
		@NoweId, 3, CTDZ_DzialID
		FROM dbo.CtiTechnolDzialy
		WHERE CTDZ_TechnolID = @idTechnologii
		AND CTDZ_DokTyp = 3;

		INSERT INTO dbo.CtiTechnolZElem (CTZE_CTZNID, CTZE_CTNID, CTZE_Lp, CTZE_Typ, CTZE_Ilosc, CTZE_Opis, CTZE_DzialID, CTZE_Potwierdzanie)
		SELECT @NoweId, CTZE_CTNID, CTZE_Lp, CTZE_Typ, CTZE_Ilosc, CTZE_Opis, CTZE_DzialID, CTZE_Potwierdzanie
		FROM dbo.CtiTechnolZElem
		WHERE CTZE_CTZNID = @idTechnologii;

		--zasoby 'główne'
		INSERT INTO dbo.CtiZasobDok
			(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia,CZS_DokTyp
			, CZS_PwBlokada, CZS_RwBlokada, CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita)
		SELECT CZS_CZID, CZS_CZGID, @NoweId, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia, CZS_DokTyp
			, isnull(CZS_PwBlokada,0), isnull(CZS_RwBlokada,0), CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp
			, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita
		FROM dbo.CtiZasobDok 
			LEFT JOIN dbo.CtiKonfigJM ON CKJ_Typ = 1 AND CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob ON CZ_ID = CZS_CZID
			INNER JOIN  dbo.CtiZasobGrupy ON CZS_CZGID = CZG_ID 
			WHERE CZS_DokTyp = 3
			AND CZS_ctnid =@idTechnologii
			AND CZS_ParentID IS NULL
		ORDER BY isnull(CZS_Lp,0), CZS_ID

		--zasoby 'pracownicy'
		INSERT INTO dbo.CtiZasobDok
			(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_Kod, CZS_JMCzasu 
			, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
			, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia,CZS_DokTyp
			, CZS_PwBlokada, CZS_RwBlokada, CZS_Ilosc, CZS_CzasStart, CZS_CzasEnd
			, CZS_Lp, CZS_Opis, CZS_PrcId, CZS_Dzielnik, CZS_PlanowanieCzasuPracy, CZS_CzasPlanowany, CZS_WieleOsob, CZS_ZasobPrcId
			, CZS_ParentID, CZS_UtworzonoPracownika, CZS_PWOdpad, CZS_RealizacjaCalkowita)
		SELECT stary.CZS_CZID, stary.CZS_CZGID, @NoweId, stary.CZS_Kod, stary.CZS_JMCzasu 
			, stary.CZS_CzasPracy,stary.CZS_CzasUzbrojenia, stary.CZS_CzasRozbrojenia, stary.CZS_KosztPracy 
			, stary.CZS_KosztUzbrojenia, stary.CZS_KosztRozbrojenia, stary.CZS_DokTyp
			, isnull(stary.CZS_PwBlokada,0), isnull(stary.CZS_RwBlokada,0), stary.CZS_Ilosc, stary.CZS_CzasStart, stary.CZS_CzasEnd
			, stary.CZS_Lp
			, stary.CZS_Opis, stary.CZS_PrcId, stary.CZS_Dzielnik, stary.CZS_PlanowanieCzasuPracy, stary.CZS_CzasPlanowany, stary.CZS_WieleOsob, stary.CZS_ZasobPrcId
			, (
				SELECT TOP 1 nowy.CZS_ID 
				FROM dbo.CtiZasobDok nowy 
				WHERE nowy.CZS_DokTyp = 3
				AND nowy.CZS_CTNID = @NoweId
				AND nowy.CZS_LP = parent.CZS_LP
			), stary.CZS_UtworzonoPracownika, stary.CZS_PWOdpad, stary.CZS_RealizacjaCalkowita
		FROM dbo.CtiZasobDok stary
			JOIN dbo.CtiZasobDok parent ON stary.CZS_ParentID = parent.CZS_ID
			LEFT JOIN dbo.CtiKonfigJM ON CKJ_Typ = 1 AND stary.CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob ON CZ_ID = stary.CZS_CZID
			INNER JOIN  dbo.CtiZasobGrupy ON stary.CZS_CZGID = CZG_ID 
			WHERE stary.CZS_DokTyp = 3
			AND stary.CZS_ctnid =@idTechnologii
			AND stary.CZS_ParentID IS NOT NULL
		ORDER BY isnull(stary.CZS_Lp,0), stary.CZS_ID

		INSERT INTO dbo.CtiAtrybutyElem (CAE_TypDok, CAE_IdDok, CAE_CANLp, CAE_CAHid, CAE_Typ, CAE_Wartosc)
		SELECT CAE_TypDok, @NoweId, CAE_CANLp, CAE_CAHid, CAE_Typ, CAE_Wartosc
		FROM dbo.CtiAtrybutyElem
		WHERE CAE_TypDok = 22 and CAE_IdDok = @idTechnologii;
	END
	
	SELECT @NoweId AS ID
END
GO
```

# [dbo].[CTIProdukcjaTowaryList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTowaryList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaTowaryList] 
 @magazyn		varchar(100),
 @typ			int,			-- 0 lista główna wszystkie - 72 wyroby - 71 surowce  -- mechanizm nieaktywny dla surowców
 @TwrTyp		int = 0,		-- 0 ignoruj filtr --Ograniczenie do:  1 towar -- 2 usługa 
 @prmCzyCecha	int = 0,		-- 0 ignoruj filtr --Ograniczenie do: 1 bez cech -- 2 surowce z cechą
 @TH			int = 0			-- dodatkowe zamienniki z TH -- mechanizm nieaktywny
AS
BEGIN
	declare  @magId							int = 0 SET @magId = ISNULL((SELECT TOP 1 Mag_MagId FROM CDN.Magazyny WHERE Mag_Symbol = @magazyn), 0)
	declare  @CzyJestOgraniczenieGrupy		int = 0 SELECT @CzyJestOgraniczenieGrupy		= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 0
	declare  @CzyJestOgraniczenieKategorii	int = 0 SELECT @CzyJestOgraniczenieKategorii	= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 2
	declare  @CzyJestOgraniczenieAtrybutu	int = 0 SELECT @CzyJestOgraniczenieAtrybutu		= count(*) FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 1

	DECLARE @idAtrCechy		int = 0 
	select DISTINCT @idAtrCechy = TwA_DeAId		from cdn.twratrybuty INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 5 and DeA_Kod = KNF_Wartosc where  TwA_CzyPrzenosicNaDostawy = 1 and TwA_DeAId <> 0
	DECLARE @idAtrCechy2	int = 0 
	select DISTINCT @idAtrCechy2 = TwA_DeAId	from cdn.twratrybuty INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 19 and DeA_Kod = KNF_Wartosc where  TwA_CzyPrzenosicNaDostawy = 1 and TwA_DeAId <> 0


	SELECT DISTINCT 
		Twr_TwrId as ID, Twr_kod as Kod, replace(Twr_Nazwa,char(10), ' ')  as Nazwa, twg_kod as Grupa
		,  CASE Twr_Typ WHEN 0 THEN CASE Twr_Produkt WHEN 0 THEN CASE WHEN Twr_Kaucja = 1 AND 0 = 1 THEN 'UPO' ELSE 'UP' END ELSE CASE WHEN Twr_Kaucja = 1 AND 0 = 1 THEN 'UZO' ELSE 'UZ' END END 
			WHEN 1 THEN CASE Twr_Produkt WHEN 0 THEN CASE WHEN Twr_Kaucja = 1 AND 0 = 1 THEN 'TPO' ELSE 'TP' END ELSE CASE WHEN Twr_Kaucja = 1 AND 0 = 1 THEN 'TZO' ELSE 'TZ' END END END AS TypS
		, Twr_NumerKat as NumerKatalogowy, Twr_EAN as EAN
		, Twr_KodDostawcy as KodUDostawcy, ISNULL(C.Knt_Kod,'') as Dostawca, ISNULL(Twr_ProducentKod,'') as KodProducenta,ISNULL(F.Prd_Kod,'') as Producent
		, TwC_Wartosc as Wartosc, CASE WHEN ISNULL(TwC_Waluta, '') = '' THEN 'PLN' ELSE Twc_Waluta END AS Waluta, Twr_JM as Jm
		, case  when ISNULL(a1.TwA_deaid,0) > 0  then '+' else '' end as Cecha, case  when ISNULL(a2.TwA_deaid,0) > 0 then '+' else '' end as Cecha2
		, ISNULL((SELECT TOP 1 TwI_Ilosc FROM CDN.TwrIlosci WHERE TwI_TwrId = Twr_TwrId AND TwI_MagId = @magId AND TwI_Data <= GETDATE()), 0) AS Ilosc--zasob na magazynie
	FROM CDN.Towary A  
		LEFT JOIN CDN.TwrCeny B			ON TwC_TwrID = Twr_TwrId AND Twr_TwCNumer = TwC_TwCNumer  
		LEFT JOIN CDN.Kontrahenci C		ON C.Knt_KntId = Twr_KntId 
		LEFT JOIN CDN.Producenci F		ON F.Prd_PrdId = A.Twr_PrdId  
		INNER JOIN cdn.twrgrupy			ON Twr_TwGGIDNumer = twg_gidnumer and twg_gidtyp = -16
		LEFT JOIN CDN.TwrAtrybuty a1	ON @idAtrCechy > 0 and ISNULL(a1.TwA_DeAId,0) = @idAtrCechy and  ISNULL(a1.TwA_TwrID,0) = Twr_TwrId  
		LEFT JOIN CDN.TwrAtrybuty a2	ON @idAtrCechy2 > 0 and ISNULL(a2.TwA_DeAId,0) = @idAtrCechy2 and ISNULL(a2.TwA_TwrID,0) = Twr_TwrId  
	where Twr_NieAktywny = 0
			and	( (ISNULL(@TwrTyp,0) = 0 )				or ( ISNULL(@TwrTyp,0) <> 0 and ((ISNULL(@TwrTyp,0) = 1 and Twr_Typ = 1) or (ISNULL(@TwrTyp,0) = 2 and Twr_Typ = 0))))
			and ( (ISNULL(@prmCzyCecha,0) = 0)			or ( ISNULL(@prmCzyCecha,0) <> 0 and ((ISNULL(@prmCzyCecha,0) = 1 and ISNULL(a1.TwA_deaid,0) = 0) or (ISNULL(@prmCzyCecha,0) = 2 and ISNULL(a1.TwA_deaid,0) <> 0))))
			and ( (@CzyJestOgraniczenieGrupy = 0)		or ( @CzyJestOgraniczenieGrupy <> 0 and twg_TwgId in (SELECT KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 0)))
			and ( (@CzyJestOgraniczenieKategorii = 0)   or ( @CzyJestOgraniczenieKategorii <> 0 and ((ISNULL(Twr_KatId,0)	in	(SELECT KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 2))or(ISNULL(Twr_KatZakId,0) in	(SELECT KnO_WartoscID FROM dbo.CtiKonfigOgraniczenia where KnO_IdList = @typ and KnO_Parametr = 2)))))
			and ( (@CzyJestOgraniczenieAtrybutu = 0)    or ( @CzyJestOgraniczenieAtrybutu <> 0 and twr_twrid in ( select distinct AtrOgr.TwA_twrid from cdn.TwrAtrybuty AtrOgr INNER JOIN dbo.CtiKonfigOgraniczenia on  KnO_IdList = @typ and KnO_Parametr = 1 and  AtrOgr.TwA_deaId = KnO_WartoscId and AtrOgr.TwA_WartoscTxt = KnO_WartoscAtrTxt)))
			--and ( (ISNULL(@TH,0) = 0 )	or ((ISNULL(@TH,0) <> 0 ) and twr_twrid in (SELECT distinct Zam_ZamTwrID FROM dbo.CtiTechnolNag  INNER JOIN dbo.CtiTechnolElem on ctn_id = Cte_ctnid INNER JOIN cdn.Zamienniki on CTE_TwrId = Zam_TwrID	where CTE_CTNId =@TH and cte_typ in (1,2))))--tylko surowce--20170227 ZAMIENNIKI 
	order by 2
	return;
END
GO
```

# [dbo].[CTIProdukcjaTworzenieKorektySpr]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaTworzenieKorektySpr]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaTworzenieKorektySpr]  
@id INT --id dokumentu z którego ma być tworzona korekat
AS
--sprawdzenie czy można stworzyć korektę do danego dokumentu--czy istnieją jakieś zasoby
declare @MagId INT select @MagId = TrN_MagDocId from cdn.TraNag INNER JOIN cdn.Magazyny on TrN_MagZrdId = Mag_MagId where  TrN_TrNId = @id

--declare @66 INT = 0 SELECT @66 = ISNULL(KNF_Wartosc,1) FROM  CtiKonfiguracja where KNF_Numer = 66--20160704 kontrola zasobów przyjętych dokumentem MM
declare @zpId INT select @zpId = ISNULL(CZD_CZNId,0) from dbo.CtiZlecenieDok where CZD_TrnId = @id

select count(*) as ilosc from
	(
		--wszystkie dostępne dostawy z danego magazynu, no towary z korygowanego dokumentu

		SELECT  Twz_TwrId as TwrId
			, sum(TwZ_Ilosc) as IloscDostepna
			, TrS_TrSIdDost as DostId
		FROM 
			(--wszystkie dostawy uzyte w MM 
				select sum(trs_ilosc) as Ilosc, TrS_TrSIdDost as DstID
				from cdn.TraNag  
					INNER JOIN cdn.TraElem on TrE_TrNId = TrN_TrNID 
					INNER JOIN cdn.TraSElem on TrS_TrEId = TrE_TrEID 
				where TrN_TrNID = @id and TrN_Bufor=0 
				and TrS_Ilosc > 0		-- operacja z jednego magazynu
				group by TrS_TrSIdDost
			) MMDst
			INNER JOIN cdn.TwrZasoby on ISNULL(TwZ_TrSIdDost,0) =DstID
			INNER JOIN cdn.TraSElem ON TwZ_TrSIdDost = TrS_TrSIdDost AND TrS_Typ=1
			LEFT JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
			LEFT JOIN cdn.TraNag ON TrN_TrNId = TrE_TrNId
		WHERE TwZ_MagId = @MagId	AND (TrN_Korekta=0 or TrN_Korekta is null)
		group by Twz_TwrId,TrS_TrSIdDost

			
	) a 



GO
```

# [dbo].[CtiProdukcjaTwrDst]

```sql
/****** Object:  StoredProcedure [dbo].[CtiProdukcjaTwrDst]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CtiProdukcjaTwrDst]
 @TwrID INT,
 @Typ INT, --312 --MM 303 --PW --304 RW 
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmWszystkie int
AS
declare @cechaKod varchar(40) = '' SELECT  @cechaKod = KNF_Wartosc FROM  CtiKonfiguracja where KNF_Numer = 5
declare @cechaId int = -1 select @cechaId = DeA_DeAId from CDN.DefAtrybuty where DeA_Kod = @cechaKod

select ID
		, Dokument
		, MagZrd, MagDoc
		, TsD_Ilosc as Ilosc
		, dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1) as NrDostawy
		, ISNULL(TsD_NumerPelny,'') as DokumentDostawy
		, DokumentPW
		, DokumentRW
		, DokumentZlecenia
		, @cechaKod as CechaKod 
		, Data 
from
(
	select a.trn_trnid as ID
		, a.trn_numerPelny as Dokument
		, ISNULL(d.Mag_Symbol,'') as MagZrd, ISNULL(c.Mag_Symbol,'') as MagDoc
		, ISNULL(f.trn_numerPelny,'') as DokumentPW
		, ISNULL(h.trn_numerPelny,'') as DokumentRW
		, ISNULL(CZN_NrPelny,'') as DokumentZlecenia
		, a.trn_datawys as Data
		, TrE_TrEID
	from cdn.traelem 
		INNER JOIN cdn.tranag a on a.trn_trnid = tre_trnid
		LEFT JOIN cdn.Magazyny c on a.trn_MagDocId = c.Mag_MagId
		LEFT JOIN cdn.Magazyny d on a.trn_MagZrdId = d.Mag_MagId
		LEFT JOIN dbo.CtiZlecenieDok on CZD_TrnId = a.trn_TrNID
		LEFT JOIN dbo.CtiZlecenieNag ON CZD_CZNId = CZN_ID
		LEFT JOIN dbo.CtiMontazDok e ON a.trn_typDokumentu = 304 and a.trn_trnid = e.CMD_RWId
		LEFT JOIN cdn.tranag f on f.trn_trnid = ISNULL(e.CMD_PWId,0)
		LEFT JOIN dbo.CtiMontazDok g ON a.trn_typDokumentu = 303 and a.trn_trnid = g.CMD_PWId
		LEFT JOIN cdn.tranag h on h.trn_trnid = ISNULL(g.CMD_RWId,0)
	where TrE_TwrId = @TwrID and a.trn_typDokumentu = @typ and
		 (
				(ISNULL(@prmWszystkie,0) =0) 
				or
				(@prmWszystkie =1 and CONVERT(VARCHAR, ISNULL(a.trn_datawys, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23))
			)

) a 
		LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID
		where cast(TsD_Ilosc as int) > 0
		order by Data
GO
```

# [dbo].[CTIProdukcjaUpdateStatusyZP]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaUpdateStatusyZP]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
/*
	Funkcja ustawia status przekazany w typie dla wskazanego zlecenia
	Id - id zlecenia prostego
	Typ:
	0 - Wszytkie
	2 - Zlecenia ZP
	6 - Przeliczanie ilości zrealizowanej i update do tabeli CtiProdukcjaStatus
	23 - MM Wydanie do produkcji
	24 - Produkcja PW
	26 - MM Kontrola
*/

CREATE PROCEDURE [dbo].[CTIProdukcjaUpdateStatusyZP] (@typ INT, @id INT) 
AS 
 BEGIN 
	set nocount on;
	DECLARE @status int = 0

/* Status zlecenia
	Zwracane wartości:
	0 - Bufor
	1 - Potwierdzone
	2 - W realizacji
	3 - Zamknięte
	4 - (?)
*/
	IF	@typ = 2 OR @typ = 0
	BEGIN 

		EXEC dbo.CTIProdukcjaUpdateStatusyZP 6, @id

		SELECT @status = CZN_Status FROM CtiZlecenieNag WHERE CZN_Id = @id
		IF @status not in (0, 3, 4)
		BEGIN
			DECLARE @automatycznyStatusZP INT = (SELECT CAST(ISNULL(KNF_Wartosc,1) AS INT) FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 64 AND ISNULL(KNF_Wartosc,'') <> '')
			DECLARE @MM INT = (SELECT CAST(ISNULL(KNF_Wartosc,1) AS INT) FROM dbo.CtiKonfiguracja WHERE KNF_Numer=63 AND ISNULL(KNF_Wartosc,'') <> '')
			IF @automatycznyStatusZP = 1 -- Czy włączony automatyczne blokowanie ZP
			BEGIN 
				IF dbo.CTIProdukcjaStatusy(26,@id) = 2 --Czy wszystko jest skontrolowane
					BEGIN SET @status = 3 END
				ELSE IF (SELECT TOP 1 IloscZrealizowana FROM dbo.CtiProdukcjaStatus WHERE CPS_DokID = @id AND CPS_DokTyp = 2) >= ISNULL((SELECT TOP 1 czn_ilosc FROM dbo.CtiZlecenieNag WHERE czn_id = @id), 0)
				AND @MM <> 1
				-- Czy zlecenie w 100% zrealizowane
					BEGIN SET @status = 3 END 
				ELSE IF EXISTS (
					SELECT * FROM dbo.CtiZlecenieDok
					JOIN cdn.TraNag ON CZD_TrnId = TrN_TrNID
					WHERE trn_bufor >= 0 AND czd_cznid = @id
					) 
					-- Czy istnieje dokument w buforze powiązany ze zleceniem
					BEGIN SET @status = 2 END
				ELSE
					BEGIN SET @status = 1 END
			END
		END
		UPDATE dbo.CtiZlecenieNag SET CZN_Status = @status WHERE CZN_ID = @id

		UPDATE dbo.CtiTechnolNag SET CTN_DodatkowyStatus = 2 where CTN_ID = (SELECT TOP 1 CZN_CTNId FROM dbo.CtiZlecenieNag WHERE CZN_ID = @id)
		UPDATE dbo.CtiTechnolNag SET CTN_DodatkowyStatus = 3 WHERE CTN_ID = (SELECT TOP 1 CZN_CTNId FROM dbo.CtiZlecenieNag WHERE CZN_ID = @id AND CZN_Status in (1,2,3))
	END

/*Status Wydania
	Zwracane wartości:
	-1 - Nie mamy włączonego wydania
	0 - Nie ma dokumentu MM (Brak)
	1 - Wydano częściowo
	2 - Wydano wszystko
	3 - jest jakikolwiek dokument w bufORze
*/
	IF	@typ = 23 OR @typ = 0 
	BEGIN
		DECLARE @obslugaWydania INT = (SELECT cast(isnull(KNF_Wartosc,1) AS INT) FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 62 AND ISNULL(KNF_Wartosc,'') <> '')
		IF @obslugaWydania = 1 
		BEGIN
				DECLARE @MMStatus INT = (SELECT top 1 TrN_Bufor FROM dbo.CtiZlecenieDok INNER JOIN cdn.TraNag ON CZD_TrnId = TrN_TrNID WHERE czd_typ = 3 AND CZD_CZNId = @id and TrN_Anulowany=0 order by TrN_Bufor desc)
				IF @MMStatus IS NULL -- Czy wygenerowano jakikolwiek dokument
					SET @status = 0
				ELSE 
				BEGIN
					IF @MMStatus = 1 -- Czy  jakikolwiek dokument jest w buforze
						SET @status = 3
					ELSE 
					BEGIN
						IF (select COUNT(CZE_ID) from dbo.ctizlecenieelem where cze_cznid=@id and cze_ilosc>0)=0
							SET @status = 2
						ELSE 
						BEGIN
							DECLARE @result DECIMAL(15,4);
							EXEC dbo.CTIProdukcjaCountShortageMMWyd @ZPid = @id, @resultStatus = @result OUTPUT;
							SET @status = @result;
						END
					END
				END
		END
		ELSE 
			BEGIN SET @status = -1 END
		UPDATE dbo.CtiProdukcjaStatus SET CPS_StatusMMWydane = @status, DataModyfikacji = GETDATE() WHERE CPS_DokID = @id AND CPS_DokTyp = 2
	END

/* Status Produkcji
	Zwracane wartości:
	0 - nic nie przyjęte (brak)
	1 - cokolwiek przyjęto (zrealizowane częściowo)
	2 - wszystko lub więcej przyjęte (zrealizowane)
	3 - (w realizacji)
*/
	IF @typ = 24 OR @typ = 0 
	BEGIN
		
		IF @typ <> 0
		BEGIN
			EXEC dbo.CTIProdukcjaUpdateStatusyZP 6, @id
		END

		DECLARE @sumaZrealizowanych DECIMAL(15,4) = 0
		DECLARE @iloscZamowiona DECIMAL(15,4) = ISNULL((SELECT TOP 1 CZN_Ilosc FROM dbo.CtiZlecenieNag WHERE CZN_Id = @id), 0)
		DECLARE @statusZP int = ISNULL((SELECT TOP 1 CZN_Status FROM dbo.CtiZlecenieNag WHERE CZN_Id = @id), 0)
		SET @sumaZrealizowanych = ISNULL((SELECT IloscZrealizowana FROM dbo.CtiProdukcjaStatus WHERE CPS_DokID = @id AND CPS_DokTyp = 2), 0)
		/* do wyciągania z tabelki */
		IF (@sumaZrealizowanych >= @iloscZamowiona)
			BEGIN SET @status = 2 END
		ELSE IF (@sumaZrealizowanych > 0)
			BEGIN SET @status = 1 END
		ELSE IF EXISTS (
			SELECT DISTINCT TrN_TrNID
		FROM dbo.CtiZlecenieDok
			INNER JOIN CDN.TraNag ON CZD_TrnId = TrN_TrNID 
		WHERE CZD_Typ = 4 
			AND ISNULL(CZD_Odpady, 0) = 0 --bez PW na odpady
			AND CZD_CZNId = @id
			AND TrN_TypDokumentu = 303
			AND TrN_Anulowany = 0)
			BEGIN SET @status = 3 END
		else
			BEGIN SET @status = 0 END
		UPDATE dbo.CtiProdukcjaStatus SET CPS_StatusPWWG = @status, DataModyfikacji = GETDATE() WHERE CPS_DokID = @id AND CPS_DokTyp = 2 
	END

/* Status Kontroli
	Zwracane wartości:
	-1 - brak obsługi
	0 - brak kontroli
	1 - wydano częściowo
	2 - wydano całkowicie
	3 - dokument MM w buforze
*/
	IF	@typ = 26 OR @typ = 0
	BEGIN
		DECLARE @obslugaKontroli int = (SELECT cast(isnull(KNF_Wartosc,1) as int) FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 63 and isnull(KNF_Wartosc,'') <> '')
		IF @obslugaKontroli = 1 -- Czy mamy wałączone MM kontrola
		BEGIN
			DECLARE @Bufor int;
			DECLARE @sumaWydana DECIMAL(15,4);
			SELECT top 1 @Bufor = TrN_Bufor FROM dbo.CtiZlecenieDok INNER JOIN cdn.TraNag ON CZD_TrnId = TrN_TrNID
			WHERE czd_typ = 6 AND CZD_CZNId = @id and Trn_anulowany=0 order by TrN_Bufor desc
 
			IF @Bufor IS NULL --Czy istnieje jakikolwiek dokument czd_typ = 6
				BEGIN SET @status = 0 END
			ELSE 
			BEGIN
						
				IF @Bufor=1 --Czy istnieje dokument w buforze
					BEGIN SET @status = 3 END
				ELSE
				BEGIN
					SET @sumaWydana = (SELECT SUM(czd_ilosc) FROM dbo.CtiZlecenieDok INNER JOIN cdn.TraNag ON CZD_TrnId = TrN_TrNID
			WHERE czd_typ = 6 AND CZD_CZNId = @id and trn_bufOR = 0 and Trn_anulowany=0)
					DECLARE @sumaDoWydania DECIMAL(15,4) = (SELECT SUM(ISNULL(czn_ilosc,1)) FROM dbo.ctiZlecenieNag WHERE czn_id = @id)

					IF @sumaWydana >= @sumaDoWydania
						BEGIN 
							SET @status = 2 
							UPDATE dbo.CtiZlecenieNag SET CZN_Status = 3 WHERE CZN_ID = @id
						END --Wydano wszystkie pozycje
					ELSE
						BEGIN SET @status = 1 END--Wydano czesciowo
				END
			END
		END 
		ELSE 
			BEGIN SET @status = -1 END
		UPDATE dbo.CtiProdukcjaStatus SET CPS_StatusMMKontrola = @status, DataModyfikacji = GETDATE() WHERE CPS_DokID = @id AND CPS_DokTyp = 2
	END

	--Przeliczanie ilości dostępnej dla ZP
	IF @typ = 6
	BEGIN
		DECLARE @pracownik int = 0
		DECLARE @przeliczJM int = 0
				IF (@pracownik IS NULL or isnull(@pracownik,0) <= 0)	SET @pracownik = 0;
		ELSE IF ((SELECT TOP 1 KNF_Wartosc FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 61) <> '2')
			SET @pracownik = 0;

		DECLARE @Ret decimal(15,4) = 0;
		DECLARE @idWyrobu int = 0;
		DECLARE @jmWyrobu varchar(25);

		SELECT TOP 1 @idWyrobu = CZN_TwrId, @jmWyrobu = CZN_TwrJM FROM dbo.CtiZlecenieNag WHERE CZN_ID = @id;

		DECLARE @pw TABLE (id int)
		INSERT INTO @pw (id)
		SELECT DISTINCT TrN_TrNID
		FROM dbo.CtiZlecenieDok
			INNER JOIN CDN.TraNag ON CZD_TrnId = TrN_TrNID 
		WHERE CZD_Typ = 4 
			AND ISNULL(CZD_Odpady, 0) = 0 --bez PW na odpady
			AND CZD_CZNId = @id
			AND TrN_TypDokumentu = 303
			AND TrN_Anulowany = 0
			AND ( @pracownik = 0 OR	CZD_PracownikId = @pracownik )
		GROUP BY TrN_TrNID

		DECLARE @IloscPw DECIMAL(15,4) = ISNULL((
			SELECT SUM(isnull(TrE_Ilosc,0)) 
			FROM CDN.TraElem
			WHERE TrE_TrNId in (SELECT id FROM @pw)
				AND TrE_TwrId = @idWyrobu
			), 0);


		DECLARE @IloscKorekty decimal(15,4) = ISNULL((
			SELECT SUM(isnull(TrE_Ilosc,0)) 
			FROM @pw
				INNER JOIN CDN.TraNag ON TrN_ZwrId = id
				INNER JOIN CDN.TraElem  ON TrE_TrNID = TrN_TrNID 
			WHERE TrN_TypDokumentu = 303 
				AND TrN_Anulowany = 0
				AND TrE_TwrId = @idWyrobu 
			), 0);
	
		SELECT  @Ret = @IloscPw + @IloscKorekty FROM @pw

		IF (@przeliczJM = 1)
		BEGIN
			SET @Ret =  isnull(@Ret,0) * ISNULL((
				SELECT TOP 1 TwJZ_JMPrzelicznikL / TwJZ_JMPrzelicznikM
				FROM CDN.TwrJMZ
				WHERE TwJZ_TwrID = @idWyrobu AND TwJZ_JM = @jmWyrobu
			), 1);
		END

	UPDATE dbo.CtiProdukcjaStatus SET IloscZrealizowana = isnull(@Ret,0), DataModyfikacji = GETDATE() WHERE CPS_DokID = @id AND CPS_DokTyp = 2

	END

	IF @typ=6 and (select KNF_Wartosc from dbo.CtiKonfiguracja where KNF_Numer=1142)=1
	BEGIN
		DECLARE @StatusID INT;
		DECLARE @TwrID INT;
		SELECT top 1 @StatusID = CZN_DokZwiazTreID, @TwrID=CZN_TwrId 
		FROM dbo.CtiZlecenieNag
		WHERE CZN_ID=@id AND CZN_DokZwiazTyp=308
		EXEC dbo.CTIProdukcjaUpdateStatusyZPProdukcja @StatusID, @TwrID
	END
 END 
GO
```

# [dbo].[CTIProdukcjaUpdateStatusyZPProdukcja]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaUpdateStatusyZPProdukcja]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
/*
	Funkcja wylicza status według zasad
	0 - "zaplanować produkcję" w momencie kiedy jest przynajmniej jedna pozycja która ma stan magazynowy mniejszy niż ilość na RO
	1 - "" - ilość zgodna
	2 - "zaplanowane częściowo" (czyli część poszła na ZP z RO)
	3 - "zaplanowane w całości" (zaplanowano wszystkie pozycje z RO)
	4 - "wyprodukowano częściowo" (jeżeli jest jakaś ilość na PW)
	5 - "wyprodukowano w całości" (PW na całość)
	Id - id RO
*/
CREATE PROCEDURE [dbo].[CTIProdukcjaUpdateStatusyZPProdukcja] (@treid INT, @twrid INT) 
AS 
 BEGIN 
	SET NOCOUNT ON;
	IF @treid IS NULL
	RETURN;
	DECLARE @RoEnum INT;
	DECLARE @RoKnf INT = ISNULL((select KNF_Wartosc from dbo.CtiKonfiguracja where KNF_Numer=1060),0);
	DECLARE @NewAmount DECIMAL(18,4);
	DECLARE @PW_a DECIMAL(18,4) = ISNULL((
        SELECT SUM(e.TrE_ilosc) 
        from dbo.CtiZlecenieNag
        INNER JOIN CtiZlecenieDok ON CZD_CZNId = CZN_ID
        INNER JOIN CDN.TraNag AS n ON CZD_TrnId = n.TrN_TrNID
        INNER JOIN CDN.TraElem AS e ON n.TrN_TrNID = e.TrE_TrNID AND e.TrE_TwrId = CZN_TwrId												
		WHERE CZD_Typ = 4 AND CZD_Odpady = 0 and n.trn_anulowany = 0 AND CZN_DokZwiazTyp = 308 and CZD_Uszkodzenia = 0
		AND CZN_DokZwiazTreID = @treid), 0);
	DECLARE @PW_kor DECIMAL(18,4) = ISNULL((
        SELECT SUM(e.TrE_ilosc) 
        from dbo.CtiZlecenieNag
        INNER JOIN CtiZlecenieDok ON CZD_CZNId = CZN_ID
        INNER JOIN CDN.TraNag AS n ON CZD_TrnId = n.TrN_TrNID
		INNER JOIN CDN.TraNag AS k ON n.TrN_TrNID = k.TrN_ZwrID
        INNER JOIN CDN.TraElem AS e ON k.TrN_TrNID = e.TrE_TrNID AND e.TrE_TwrId = CZN_TwrId												
		WHERE CZD_Typ = 4 AND CZD_Odpady = 0 and k.trn_anulowany = 0 AND CZN_DokZwiazTyp = 308 and CZD_Uszkodzenia = 0 and k.Trn_Rodzaj = 303001
		AND CZN_DokZwiazTreID = @treid), 0);
	WITH amount as
	(SELECT roeInner.TrE_TreId, SUM(ISNULL(zp.CZN_Ilosc, 0)) AS zp, SUM(ISNULL(zzp.CZZN_Ilosc, 0)) AS zpz
        FROM cdn.traelem roeInner 
        LEFT JOIN dbo.CtiZlecenieZNag zzp ON roeInner.tre_treid = zzp.CZZN_DokZwiazTreId AND zzp.CZZN_DokZwiazTyp = 308
        LEFT JOIN dbo.CtiZlecenieNag zp ON (zp.CZN_DokZwiazTyp = 308 AND zp.CZN_DokZwiazTreID = roeInner.tre_treid AND zzp.CZZN_ID IS NULL) 
            OR (zzp.CZZN_DokZwiazTyp = 308 AND CZN_CZZNIdGlowne = CZZN_ID AND zzp.CZZN_ID IS NOT NULL)
        where roeInner.Tre_treid = @treid
		GROUP BY roeInner.TrE_TreId
    ),   stock AS (
        SELECT SUM(main.twi_ilosc - main.twi_rezerwacje)/COUNT(*) AS available, @treid as treid
        FROM CDN.TwrIlosci main 
        JOIN cdn.traelem ON tre_twrid = main.TwI_TwrId and main.TwI_MagId = TrE_MagId 
        WHERE main.TwI_TwrId = @twrid and tre_treid=@treid											   
          AND main.TwI_Data = (
              SELECT MAX(inner_twi.TwI_Data) 
              FROM CDN.TwrIlosci AS inner_twi 
              WHERE inner_twi.Twi_TwrId = main.Twi_TwrId 
                AND inner_twi.TwI_MagId = main.TwI_MagId
          )	  
		group by tre_treid
    ),
	countStatus AS (
	select top 1
		CASE WHEN sum(ISNULL(roe.TrE_Ilosc,0))<=sum(ISNULL(a.zp, ISNULL(a.zpz,0))) and sum(ISNULL(roe.TrE_Ilosc,0))<=sum(@PW_a + @PW_kor) THEN 5
		WHEN @PW_a+@PW_kor>0 THEN 4
		WHEN sum(ISNULL(roe.TrE_Ilosc,0))<=sum(ISNULL(a.zp, ISNULL(a.zpz,0))) THEN 3
		WHEN sum(ISNULL(a.zp, ISNULL(a.zpz,0)))>0 THEN 2
		WHEN @RoKnf=0 OR (@RoKnf*ISNULL((SELECT available FROM stock WHERE roe.tre_treid=treid),0)-sum(ISNULL(roe.TrE_Ilosc,0))/count(*))>=0 THEN 1
		ELSE 0
		END as currentStatus, sum(ISNULL(roe.TrE_Ilosc,0))/count(*) as additionalAmount
		from cdn.traelem roe
        LEFT JOIN amount a ON roe.tre_treid = a.TrE_TreId
		where roe.Tre_TypDokumentu=308 and roe.tre_treid = @treid 
		group by roe.Tre_treid
		order by currentStatus asc
	)

	SELECT @RoEnum = currentStatus, @NewAmount= additionalAmount
	FROM countStatus 
	IF @RoEnum IS NULL
	BEGIN
		SET @RoEnum = 0;
	END	
	IF @NewAmount IS NULL
	BEGIN
		SET @NewAmount = 0;
	END

	MERGE dbo.CtiProdukcjaStatusROElem AS tgt
	USING (
		SELECT 
			@treid AS CPSRE_TreID
		WHERE EXISTS (
			SELECT 1 
			FROM [cdn].[TraElem] 
			WHERE [Tre_treid] = @treid
		)
	) AS src
	ON tgt.CPSRE_TreID = src.CPSRE_TreID
	WHEN MATCHED THEN
		UPDATE SET 
			CPSRE_Status = @RoEnum, 
			CPSRE_Amount = @NewAmount
	WHEN NOT MATCHED BY TARGET AND src.CPSRE_TreID IS NOT NULL THEN
		INSERT (CPSRE_TreID, CPSRE_Status, CPSRE_Amount)
		VALUES (src.CPSRE_TreID, @RoEnum, @NewAmount);
END
GO
```

# [dbo].[CTIProdukcjaWydanieElemList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaWydanieElemList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaWydanieElemList]   @id INT
AS

select 
	b.tre_treid as treid, a.TrE_TrNId as TrNId,a.TrE_TwrKod as TwrKod, replace(a.TrE_TwrNazwa,char(10), ' ')as TwrNazwa
		, 

	CASE WHEN CZD_Typ = 6 THEN
	COALESCE ( CZD_Ilosc, 0 ) 
	ELSE
	COALESCE( sum(cze_ilosc) , 0 ) 
	END

	as IloscZlecenie 
	,sum(a.TrE_Ilosc) as Ilosc
	, case ISNULL(twa_deaid,0) when 0 then sum(a.TrE_Ilosc) else sum(a.iloscDostawa) end  as iloscDostawa
	, TrE_Jm AS jm ,case ISNULL(twa_deaid,0) when 0 then 2 else 1 end  as TYP
	, case ISNULL(twa_deaid,0) when 0 then 'surowiec' else 'surowiec z cechą' end  as TypNazwa
	, Twr_EAN as EAN, twr_twrid as twrid 
from 
	(
		select TrE_TrNId,TrE_TwrId,TrE_TwrKod,TrE_TwrNazwa,cast(TrE_Ilosc as decimal(15,4)) as TrE_Ilosc,
		cast(ISNULL(sum(TsD_Ilosc),0)as decimal(15,4))  as iloscDostawa 
		from cdn.traelem LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID
		where TrE_TrNId = @id group by TrE_TrEID,TrE_TrNId,TrE_TwrKod,TrE_TwrNazwa,TrE_Ilosc,TrE_TwrId
	) a
	INNER JOIN dbo.CtiZlecenieDok on  a.tre_trnid =CZD_TrnId
	INNER JOIN cdn.TraElem b on b.TrE_TrNId = CZD_TrnId and a.TrE_TwrId = b.TrE_TwrId
	LEFT JOIN dbo.CtiZlecenieElem on CZD_CZNId = CZE_CZNId AND CZE_TwrId = b.TrE_TwrId
	INNER JOIN cdn.towary on a.TrE_TwrId = twr_twrid
	LEFT JOIN CDN.TwrAtrybuty on TwA_TwrID = a.TrE_TwrId and TwA_DeAId = 
	( 
		select DISTINCT TwA_DeAId
		from cdn.twratrybuty
		INNER JOIN cdn.defatrybuty on TwA_DeAId = DeA_DeAId
		INNER JOIN dbo.CtiKonfiguracja on KNF_Numer = 5 and DeA_Kod = KNF_Wartosc 
		where  TwA_CzyPrzenosicNaDostawy = 1 
	)
where a.tre_trnid = @id
group by a.TrE_TrNId,a.TrE_TwrKod,a.TrE_TwrNazwa,a.TrE_TwrId,b.tre_treid,TrE_Jm,ISNULL(twa_deaid,0),Twr_EAN,twr_twrid, czd_typ, czd_ilosc
order by a.TrE_TwrKod
GO
```

# [dbo].[CTIProdukcjaWykorzystaneDostawyList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaWykorzystaneDostawyList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaWykorzystaneDostawyList]   @id INT
AS

select TwrId, TwrKod,DostId,NumerPelny,sum(case when iloscDostawa> 0 then iloscDostawa else 0 end) as iloscDostawa,tsd_wartosc
    , DeAId = Cecha1DeAId,CechaWartosc,cecha= Cecha1Kod,JM,CechaWartoscWidok, Cecha2Wartosc, DostData, TsD_TsDID
from 
(
        select TraElemG.TrE_TwrId as TwrId, TraElemG.TrE_TwrKod as TwrKod
            , case when ISNULL(TsD_TrSIdDost,0) = 0 then  ISNULL(TraSElemG.TrS_trsiddost,0) else ISNULL(TsD_TrSIdDost,0) end as DostId
            , case when ISNULL(TsD_NumerPelny,'') = '' then ISNULL(TraNagDST.TrN_NumerPelny,'') else ISNULL(TsD_NumerPelny,'') end as NumerPelny
            --,cast(ISNULL(TsD_Ilosc,TraElemG.TrE_ilosc) as decimal(15,4)) as iloscDostawa
            , cast(ISNULL(TsD_Ilosc,ISNULL(TraSElemG.TrS_ilosc,0)) as decimal(15,4))  as iloscDostawa
            , ISNULL(tsd_wartosc,0) as tsd_wartosc

            , ISNULL(TsD_Cecha1_DeAId,0) as Cecha1DeAId, ISNULL(TsD_Cecha1_Kod,'') as Cecha1Kod 
            , dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1) as CechaWartosc
            , TraElemG.TrE_JM as JM
            , dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 1) as CechaWartoscWidok
            , dbo.CtiProdukcjaPobierzCecheDostawy(tsd_tsdid, 2) as Cecha2Wartosc
            ,  ISNULL(trn_dataOpe,'') as DostData
            , tsd_tsdid
        from cdn.TraElem as TraElemG 
            --jest cecha
            LEFT JOIN cdn.TraSElemDost on TraElemG.TrE_TrEID = ISNULL(TsD_TrEID,0)
            LEFT JOIN cdn.TraSElem as TraSElemG on ISNULL(TraElemG.TrE_TrEId,0)  = ISNULL(TraSElemG.TrS_TrEId,0)
            --nie ma cechy
            LEFT JOIN  cdn.TraSElem as TraSElemDST ON ISNULL(TraSElemDST.trs_trsid,0) = ISNULL(TraSElemG.TrS_trsiddost,0)
            LEFT JOIN  cdn.TraElem as TraElemDST ON  ISNULL(TraElemDST.TrE_TrEId,0) = ISNULL(TraSElemDST.TrS_TrEId,0)
            LEFT JOIN  cdn.TraNag as TraNagDST ON ISNULL(TraNagDST.TrN_TrNId,0) = ISNULL(TraElemDST.TrE_TrNId,0) 
        where TraElemG.TrE_trnid = @id
        group by TraElemG.TrE_TwrId,TraElemG.TrE_TwrKod,TraElemG.TrE_JM--dane towaru
            , ISNULL(TsD_TrSIdDost,0),ISNULL(TraSElemG.TrS_trsiddost,0),ISNULL(TsD_TrSIdDost,ISNULL(TraSElemG.TrS_trsiddost,0))--id dostawy
            , ISNULL(TsD_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,ISNULL(TsD_NumerPelny,'')) --numer dokumentu
            , ISNULL(TsD_Ilosc,ISNULL(TraSElemG.TrS_ilosc,0)) --ilość
            , ISNULL(tsd_wartosc,0)--cena
            , ISNULL(TsD_Cecha1_DeAId,0), ISNULL(TsD_Cecha1_Kod,'') ,ISNULL(TsD_Cecha1_Wartosc,'') --cecha
            , ISNULL(TsD_cecha1_format,0),ISNULL(TsD_cecha1_Wartosc,0), ISNULL(TsD_Cecha2_Wartosc,'') 
            , ISNULL(trn_dataOpe,'')
            , tsd_tsdid
    union 

        select TrE_TwrId as TwrId, TrE_TwrKod as TwrKod,TsD_TrSIdDost as DostId, TsD_NumerPelny as NumerPelny
            , cast(TsD_Ilosc as  decimal(15,4)) as iloscDostawa,tsd_wartosc
            , ISNULL(TsD_Cecha1_DeAId,0) as Cecha1DeAId, ISNULL(TsD_Cecha1_Kod,'') as Cecha1Kod,ISNULL(TsD_Cecha1_Wartosc,'') as CechaWartosc
            , tre_JM as JM
            , case when ISNULL(TsD_cecha1_format,0) = 4 then CONVERT(VARCHAR, dbo.CtiProdukcjaTSToDate(ISNULL(TsD_cecha1_Wartosc,0),0), 23)
                else ISNULL(TsD_cecha1_Wartosc,'') end as CechaWartoscWidok
            , Cecha2Wartosc = ISNULL(TsD_Cecha2_Wartosc,'') 
            , ISNULL(trn_dataOpe,'') as DostData
            , tsd_tsdid
        from cdn.tranag INNER JOIN cdn.TraElem on TrN_TrNID = TrE_TrNId
            LEFT JOIN cdn.TraSElemDost on TrE_TrEID = TsD_TrEID
        where TrN_Bufor = 0 and TrN_DDfId = 24 and TrN_ZwrId = @id 
) a
group by TwrId, TwrKod,DostId,NumerPelny,tsd_wartosc,Cecha1DeAId, Cecha1Kod,CechaWartosc,JM,CechaWartoscWidok,Cecha2Wartosc,DostData,tsd_tsdid
having sum(case when iloscDostawa> 0 then iloscDostawa else 0 end) <> 0
order by DostId ,TwrKod asc
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaWyslijDoMaszyny]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE  [dbo].[CTIProdukcjaWyslijDoMaszyny]   
 @ZPid int

AS 

declare @nrzbiornika varchar(100)
SELECT  top 1 @nrzbiornika=CAE_Wartosc 	FROM dbo.CtiAtrybutyElem inner join  dbo.CtiAtrybutyElemHist on isnull(CAE_CAHid,0) = CAH_ID where CAH_Nazwa='Kontrahent' and (CAE_TypDok = 2 and CAE_IdDok = @ZPid) 

INSERT INTO dbo.CtiMaszyna
	(  Mas_DokID,Mas_DokElemID, Mas_DokNumer, Mas_Ilosc, Mas_KodSurowca, Mas_NazwaSurowca,  Mas_NRZbiornika, Mas_AtrSurowca, Mas_Status) 
	select czn_id,cze_id,czn_nrpelny,CZE_Ilosc,CZE_TwrKod,Twr_Nazwa,coalesce(@nrzbiornika,''),'',0
	from dbo.CtiZlecenieNag 
	join dbo.CtiZlecenieElem on cze_typ in (1,2) and cze_cznid=czn_id
	left outer join cdn.towary on twr_twrid = CZE_TwrId
	left outer join dbo.CtiMaszyna on czn_id=Mas_DokID and CZE_ID=Mas_DokElemID
	where czn_id = @ZPid and mas_id is null

INSERT INTO dbo.CtiMaszynaZasoby
	( Maz_CZNID, Maz_CZSID, Maz_CzasZadany) 
select CZS_CTNID,CZS_ID,CZS_CzasPracy*CZN_Ilosc*60 from [dbo].[CtiZasobDok] 
join CtiZlecenieNag on CZN_ID=CZS_CTNID
	left outer join dbo.CtiMaszynaZasoby on Maz_CZSID=CZS_ID
	where CZS_CTNID = @ZPid and CZS_DokTyp = 2 and Maz_ID is null
GO
```

# [dbo].[CTIProdukcjaZasobDokAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobDokAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZasobDokAdd]
  @DokId					int --ID technologii / zlecenia
, @idZasobu					int --Z tabeli dbo.CtiZasob
, @LogujacyId				int --ID operatora
, @DokTyp					int --1 = TH, 2 = ZP
, @idElementu				INT OUTPUT --Zwracane ID
, @CzasPracy				DECIMAL(15,4) = NULL
, @CzasUzbrojenia			DECIMAL(15,4) = NULL
, @CzasRozbrojenia			DECIMAL(15,4) = NULL
, @KosztPracy				DECIMAL(15,4) = NULL
, @KosztUzbrojenia			DECIMAL(15,4) = NULL
, @KosztRozbrojenia			DECIMAL(15,4) = NULL
, @Ilosc					DECIMAL(15,4) = NULL
, @Dzielnik					DECIMAL(15,4) = NULL
, @JMCzasu					int = NULL
, @PlanowanieCzasuPracy		int = NULL
, @BlokadaPW				int = NULL
, @BlokadaRW				int = NULL
, @CzasPlanowany			DECIMAL(15,4) = NULL
, @WieleOsob				int = NULL
AS
BEGIN

	if (@DokTyp = 1)
		UPDATE dbo.CtiTechnolNag SET CTN_OperModyf =@LogujacyId, CTN_DataModyf = GETDATE() from dbo.CtiTechnolNag WHERE CTN_ID =@DokId and @DokTyp = 1
	else
		UPDATE dbo.CtiZlecenieNag SET CZN_OperModyf = @LogujacyId, CZN_DataModyf = GETDATE() from dbo.CtiZlecenieNag WHERE CZN_ID =@DokId and @DokTyp = 2

	insert into dbo.CtiZasobDok
	(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_Kod, CZS_JMCzasu, CZS_CzasPracy,CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_Ilosc
	, CZS_KosztPracy , CZS_KosztUzbrojenia, CZS_KosztRozbrojenia, CZS_PlanowanieCzasuPracy, 

CZS_DokTyp, CZS_PwBlokada, CZS_RwBlokada, CZS_Dzielnik, CZS_CzasPlanowany, CZS_WieleOsob, CZS_Lp)
	SELECT CZ_ID, CZ_CZGID, @DokId, CZ_Kod, ISNULL(@JMCzasu, CZ_JMCzasu), ISNULL(@CzasPracy, CZ_CzasPracy), ISNULL(@CzasUzbrojenia, CZ_CzasUzbrojenia), ISNULL(@CzasRozbrojenia, CZ_CzasRozbrojenia), ISNULL(@Ilosc, CZ_IloscDomyslna)
	, ISNULL(@KosztPracy, CZ_KosztPracy), ISNULL(@KosztUzbrojenia, CZ_KosztUzbrojenia), ISNULL(@KosztRozbrojenia, CZ_KosztRozbrojenia), ISNULL(@PlanowanieCzasuPracy, CZ_PlanowanieCzasuPracy), 

@DokTyp, ISNULL(@BlokadaPW, 0), ISNULL(@BlokadaRW, 0), COALESCE(@Dzielnik, CZ_Dzielnik, 0), isnull(@CzasPlanowany, CZ_CzasPlanowany), isnull(@WieleOsob, CZ_WieleOsob), isnull((select max(isnull(CZS_Lp,0))+1 
	from dbo.CtiZasobDok where  CZS_DokTyp=@DokTyp and CZS_CTNID=@DokId),1)
	FROM dbo.CtiZasob where CZ_ID = @idZasobu


	SET @idElementu = SCOPE_IDENTITY() 
	if (@DokTyp = 2) --ZP
	begin
		declare @czasStart datetime = ISNULL((SELECT TOP 1 CZN_CzasStart FROM dbo.CtiZlecenieNag WHERE CZN_ID = @DokId), GETDATE());
		declare @iloscZp int = ISNULL((SELECT TOP 1 CZN_Ilosc FROM dbo.CtiZlecenieNag WHERE CZN_ID = @DokId), 1);
		declare @DzielnikFinal int = COALESCE(@Dzielnik, (SELECT TOP 1 CZ_Dzielnik FROM dbo.CtiZasob WHERE CZ_ID = @idZasobu), 0);
		declare @WieleOsobFinal int = COALESCE(@WieleOsob, (SELECT TOP 1 CZ_WieleOsob FROM dbo.CtiZasob WHERE CZ_ID = @idZasobu), 1);

		UPDATE dbo.CtiZasobDok SET CZS_CzasStart = @czasStart, CZS_WieleOsob = @WieleOsobFinal,
		CZS_CzasEnd = dbo.CTIProdukcjaZPZasobDataEnd(@idZasobu, @czasStart, dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, @iloscZp, COALESCE(@DzielnikFinal,@iloscZp,0), CZS_JMCzasu, CZS_ID)) 
		WHERE CZS_ID = @idElementu;
	end

END


GO
```

# [dbo].[CTIProdukcjaZasobDokEdit]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobDokEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaZasobDokEdit]
   @DokTyp int
 , @DokId int--TechnologiiZlecenia
 , @idElementu int
 , @LogujacyId int
 , @CzasPracy decimal(15, 4)
 , @CzasUzbrojenia decimal(15, 4)
 , @CzasRozbrojenia decimal(15, 4)
 , @KosztPracy decimal(15, 4)
 , @KosztUzbrojenia decimal(15, 4)
 , @KosztRozbrojenia decimal(15, 4)
 , @JMCzasu int
 , @idZasobu int output
 , @ilosc decimal(15, 4)
 , @CzasCalkowity decimal(15, 4)
 , @BlokadaPw int
 , @BlokadaRw int
 , @datStart varchar(23)
 , @datEnd varchar(23)
 , @opis nvarchar(4000)
 , @prcId int
 , @lp int = null
 , @dzielnik decimal(15,4) = null
 , @PlanowanieCzasuPracy int = null
 , @CzasPlanowany decimal(15,4) = null
 , @WieleOsob int = null
 , @zasobPrcId int = null
 , @PWOdpad int = null
 , @RealizacjaCalkowita int = null
AS
BEGIN
declare @czasStart datetime 
declare @czasEnd datetime	


	if(@lp <= 0)
		set @lp = null;

	if(@idElementu=0) begin	exec dbo.CTIProdukcjaZasobDokAdd @DokId, @idZasobu, @LogujacyId, @DokTyp, @idElementu OUTPUT end

	if (@DokTyp = 2)--ZP
	begin
		if isnull(@datStart,'') = ''--z ZP
		begin
			select @czasStart = CZN_CzasStart FROM dbo.CtiZasobDok INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
			INNER JOIN dbo.CtiZlecenieNag  on CZS_CTNID = czn_id where CZS_ID = @idElementu
		end
		else
		begin set @czasStart = convert(datetime, @datStart,120)	end


		select @czasEnd = dbo.CTIProdukcjaZPZasobDataEnd(CZ_ID,ISNULL(@czasStart,getdate())
			, dbo.CTIProdukcjaZPZasobCzasCalkowity(@CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, isnull(CZN_Ilosc,1), COALESCE(@dzielnik,CZ_Dzielnik,0), @JMCzasu, @idElementu)) 
		FROM dbo.CtiZasobDok INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
			INNER JOIN dbo.CtiZlecenieNag  on CZS_CTNID = czn_id where CZS_ID = @idElementu

	end

	Update dbo.CtiZasobDok SET
		  CZS_JMCzasu = ISNULL(@JMCzasu,CZS_JMCzasu)
		, CZS_CzasPracy = isnull(@CzasPracy,CZS_CzasPracy)
		, CZS_CzasUzbrojenia = isnull(@CzasUzbrojenia,CZS_CzasUzbrojenia)
		, CZS_CzasRozbrojenia = isnull(@CzasRozbrojenia,CZS_CzasRozbrojenia)
		, CZS_KosztPracy = isnull(@KosztPracy,CZS_KosztPracy)
		, CZS_KosztUzbrojenia = isnull(@KosztUzbrojenia ,CZS_KosztUzbrojenia)
		, CZS_KosztRozbrojenia = isnull(@KosztRozbrojenia ,CZS_KosztRozbrojenia)
		, CZS_ilosc  = @ilosc 
		, CZS_CzasStart = @czasStart
		, CZS_CzasEnd = @czasEnd
		, CZS_PwBlokada = ISNULL(@BlokadaPw,0)
		, CZS_RwBlokada = ISNULL(@BlokadaRw,0)
		, CZS_Opis = isnull(@opis,CZS_Opis)
		, CZS_PrcId = isnull(@prcId,CZS_PrcId)
		, CZS_Lp = isnull(@lp,CZS_Lp)
		, CZS_Dzielnik = isnull(@dzielnik, CZS_Dzielnik)
		, CZS_PlanowanieCzasuPracy = isnull(@PlanowanieCzasuPracy, CZS_PlanowanieCzasuPracy)
		, CZS_CzasPlanowany = isnull(@CzasPlanowany, CZS_CzasPlanowany)
		, CZS_WieleOsob = isnull(@WieleOsob, CZS_WieleOsob)
		, CZS_ZasobPrcId = isnull(@zasobPrcId, CZS_ZasobPrcId)
		, CZS_PWOdpad = ISNULL(@PWOdpad,0)
		, CZS_RealizacjaCalkowita = isnull(@RealizacjaCalkowita,1)
	from dbo.CtiZasobDok
	where CZS_ID = @idElementu

END
GO
```

# [dbo].[CTIProdukcjaZasobDokstepnyAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobDokstepnyAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZasobDokstepnyAdd]
@idZasobu int,
@calyDzien int,
@dataOd varchar(20),
@dataDo varchar(20),
@godzinaDo varchar(20),
@godzinaOd varchar(20),
@pn int, @wt int, @sr int, @cz int, @pt int, @sob int, @ndz int, @holiday int = 0
AS
BEGIN
	--declare @idZasobu int = 2
	--declare @calyDzien int = 1
	--declare @dateOd datetime = getdate() --nie może być mniejsza od datydo
	--declare @dateDo datetime = getdate()
	--declare @dateGodzinaOd datetime = '2015-11-01 8:00'
	--declare @dateGodzinaDo datetime = '2015-11-01 16:00' --nie może być mniejsza od czasu do chyba że północ
	--declare @pn int = 1 -- 2
	--declare @wt int = 0 -- 3
	--declare @sr int = 0 -- 4
	--declare @cz int = 0 -- 5
	--declare @pt int = 0 -- 6
	--declare @sob int = 0 -- 7
	--declare @ndz int = 0 -- 1


	declare @i int = 0 
	declare @do int = 0 
	declare @dw int = 0 
	declare @Data datetime
	declare @dateOd datetime
	declare @dateDo datetime
	declare @dateGodzinaDo datetime
	declare @dateGodzinaOd datetime

	set @dateOd = CONVERT(datetime,CONVERT(VARCHAR, convert(datetime,@dataOd), 23)+' 00:00')
	set @dateDo = CONVERT(datetime,CONVERT(VARCHAR, convert(datetime,@dataDo), 23)+' 00:00')

	if @calyDzien = 1
	begin 
		set @dateGodzinaOd = CONVERT(datetime,CONVERT(VARCHAR, @dateOd, 23)+' 00:00')
		set @dateGodzinaDo = CONVERT(datetime,CONVERT(VARCHAR, @dateDo, 23)+' 00:00')
	end
	else
	begin
		set @dateGodzinaOd = CONVERT(datetime,CONVERT(VARCHAR, @dateOd, 23) +' '+ LEFT(CONVERT(VARCHAR, convert(datetime,@godzinaOd), 24),5))
		set @dateGodzinaDo = CONVERT(datetime,CONVERT(VARCHAR, @dateDo, 23) +' '+ LEFT(CONVERT(VARCHAR, convert(datetime,@godzinaDo), 24),5))
	end

		--select @dateOd, @dateDo, @dateGodzinaOd, @dateGodzinaDo

	if(@dateOd <= @dateDo) and 
		((isnull(@dateGodzinaOd,getdate()) <= isnull(@dateGodzinaDo,getdate())) OR (LEFT(CONVERT(VARCHAR, @dateGodzinaDo, 24),5)  = '00:00'))
	begin
		set @i = 0 --start
		SELECT @do = DATEDIFF(day, @dateOd,@dateDo)
			while @i <= @do
			begin
				set @Data = @dateOd + @i
				set @dw = datepart(dw, @data)

				if (@dw = 1 and @ndz = 1) OR (@dw = 2 and @pn = 1) OR (@dw = 3 and @wt = 1) OR (@dw = 4 and @sr = 1)
					OR (@dw = 5 and @cz = 1) OR (@dw = 6 and @pt = 1) OR (@dw = 7 and @sob = 1) OR (@holiday = 1 and EXISTS(
					select 1 from dbo.CtiHolidays where YEAR(CH_Date)=YEAR(@Data) and MONTH(CH_Date)=MONTH(@Data) and DAY(CH_Date)=DAY(@Data)))
				begin
					if (@holiday = 0 and EXISTS(
					select 1 from dbo.CtiHolidays where YEAR(CH_Date)=YEAR(@Data) and MONTH(CH_Date)=MONTH(@Data) and DAY(CH_Date)=DAY(@Data)))
					BEGIN
						set @i= @i + 1
						CONTINUE;
					END
					if @calyDzien = 1
					begin 
						set @dateGodzinaOd = CONVERT(datetime,CONVERT(VARCHAR, @Data, 23)+' 00:00')
						set @dateGodzinaDo = CONVERT(datetime,CONVERT(VARCHAR, @Data, 23)+' 00:00')
					end
					else
					begin
						set @dateGodzinaOd = CONVERT(datetime,CONVERT(VARCHAR, @Data, 23)+' '+LEFT(CONVERT(VARCHAR, @dateGodzinaOd, 24),5))
						set @dateGodzinaDo = CONVERT(datetime,CONVERT(VARCHAR, @Data, 23)+' '+LEFT(CONVERT(VARCHAR, @dateGodzinaDo, 24),5))
					end

						if isnull(@calyDzien,0) = 0-- czy istnieje większy przedział
						begin
							declare @NowaGodzinaOd datetime = @dateGodzinaOd
							declare @NowaGodzinaDo datetime = @dateGodzinaDo

							select @NowaGodzinaOd = isnull(Min(ZsD_CzasOd),@dateGodzinaOd)
							from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
								 and ( ZsD_CzasOd <= @dateGodzinaOd) 
								 and ( ZsD_CzasDo >= @dateGodzinaOd) 
								 and ( ZsD_CzasDo <= @dateGodzinaDo)
							select @NowaGodzinaDo = isnull(Max(ZsD_CzasDo),@dateGodzinaDo)
							from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
								 and ( ZsD_CzasDo >= @dateGodzinaDo) 
								 and ( ZsD_CzasOd >= @dateGodzinaOd)
								 and ( ZsD_CzasOd <= @dateGodzinaDo)
							if not(@NowaGodzinaOd = @dateGodzinaOd and @NowaGodzinaDo = @dateGodzinaDo)
							begin
								delete from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
									and ( ZsD_CzasOd <= @dateGodzinaOd) 
									and ( ZsD_CzasDo >= @dateGodzinaOd) 
									and ( ZsD_CzasDo <= @dateGodzinaDo)
								delete	from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
									and ( ZsD_CzasDo >= @dateGodzinaDo) 
									and ( ZsD_CzasOd >= @dateGodzinaOd)
									and ( ZsD_CzasOd <= @dateGodzinaDo)
								set @dateGodzinaOd = @NowaGodzinaOd
								set @dateGodzinaDo = @NowaGodzinaDo
							end
						end







					if --jeżeli wpis jeszcze nie istnieje
						not exists(select * from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 1)-- cały dzień
						and--zakres dat
						not exists(select * from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0 and  ZsD_CzasOd = @dateGodzinaOd and ZsD_CzasDo = @dateGodzinaDo)

					begin 

						if(
								((isnull(@calyDzien,0) = 0)
								and
								not exists(select ZsD_CZID, ZsD_Data, ZsD_CzasOd, ZsD_CzasDo, ZsD_CalyDzien 
											from dbo.CtiZasobDostepny 
											where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0	 and (  @dateGodzinaOd >= ZsD_CzasOd ) and (@dateGodzinaDo <= ZsD_CzasDo ))
								)or
								(isnull(@calyDzien,0) = 1)
						  )
						begin

							if isnull(@calyDzien,0) = 1
							begin  -- cały dzień --usunięcie dodanych pozycji z zakresem dat
								delete from   dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
							end
							else--zakres dat --usunięcie zaskresów pomiędzy dany zakres
							begin
								delete from dbo.CtiZasobDostepny where zsd_data = @data and zsd_czid = @idZasobu and isnull(ZsD_CalyDzien,0) = 0
									 and (ZsD_CzasOd between @dateGodzinaOd and @dateGodzinaDo) and (ZsD_CzasDo between @dateGodzinaOd and @dateGodzinaDo )
							end


								if not exists(select * from dbo.CtiZasobDostepny where 
									ZsD_CZID = @idZasobu 
									and ZsD_Data = @Data
									and ZsD_CzasOd = @dateGodzinaOd
									and ZsD_CzasDo = @dateGodzinaDo
									and isnull(ZsD_CalyDzien,0) = @calyDzien)
								begin
									insert into dbo.CtiZasobDostepny (ZsD_CZID, ZsD_Data, ZsD_CzasOd, ZsD_CzasDo, ZsD_CalyDzien)
									values (@idZasobu, @Data, @dateGodzinaOd, @dateGodzinaDo, @calyDzien)
								end
							end
						end
					end
			
				set @i= @i + 1
			end
	end

end
GO
```

# [dbo].[CTIProdukcjaZasobDokstepnyList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobDokstepnyList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaZasobDokstepnyList]
@idZasobu int,
@archiwalne int
AS
BEGIN
	SET LANGUAGE English--SET LANGUAGE Polish
	SELECT ZsD_ID as ID, ZsD_CZID as CZID
	, year(ZsD_Data) as Rok
	, case datename(dw,ZsD_Data) 
		when 'Monday' then 'Poniedziałek'
		when 'Tuesday' then 'Wtorek'
		when 'Wednesday' then 'Środa'
		when 'Thursday' then 'Czwartek'
		when 'Friday' then 'Piątek'
		when 'Saturday' then 'Sobota'
		when 'Sunday' then 'Niedziela'
		else datename(dw,ZsD_Data) end
	  as DzienTygodnia
	, case datepart(mm, ZsD_Data)
		when 1 then 'Styczeń'
		when 2 then 'Luty'
		when 3 then 'Marzec'
		when 4 then 'Kwiecień'
		when 5 then 'Maj'
		when 6 then 'Czerwiec'
		when 7 then 'Lipiec'
		when 8 then 'Sierpień'
		when 9 then 'Wrzesień'
		when 10 then 'Październik'
		when 11 then 'Listopad'
		when 12 then 'Grudzień'
		else datename(mm,ZsD_Data) end
	  as nazwaMiesiaca
	, CONVERT(VARCHAR, ZsD_Data, 23)  as Data
	, LEFT(CONVERT(VARCHAR, ZsD_CzasOd, 8),5)  as CzasOd
	, LEFT(CONVERT(VARCHAR, ZsD_CzasDo, 8),5)  as CzasDo
	, isnull(ZsD_CalyDzien,0)  as CalyDzien
	FROM dbo.CtiZasobDostepny 
	where ZsD_CZID = @idZasobu
		and
		(
			(isnull(@archiwalne,0) = 1 and 1=1 )--wszystkie zasoby
			or
			(isnull(@archiwalne,0) = 0 and CONVERT(datetime,CONVERT(VARCHAR, ZsD_Data, 23)+' 00:00') >= CONVERT(datetime,CONVERT(VARCHAR, getdate(), 23)+' 00:00'))
		)
	order by   CONVERT(VARCHAR, ZsD_Data, 23), LEFT(CONVERT(VARCHAR, ZsD_CzasOd, 8),5)
END


GO
```

# [dbo].[CTIProdukcjaZasobList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZasobList]
@BezTypu int,
@DzialID int = null,
@TylkoTyp int = null,
@BezID int = null,
@AkceptujBezDzialu int = null,
@IDCSU int = null
AS
BEGIN
	SELECT CZ_ID AS ID, CZG_Kod AS Grupa, CZ_Kod AS Kod
	, case isnull(cz_typ,0) when 80 then 'Pracownicy' when 76 then 'Czynność' else 'Zasób' end as Typ
	, isnull(cz_typ,0) as CznTyp
	, (SELECT STUFF(
	( 
		SELECT DISTINCT ', ' + CDZ_Kod
		FROM dbo.CtiDzial
		LEFT OUTER JOIN dbo.CtiZasobDzialy ON CDZ_ID = ZDZ_DzialID
		WHERE ZDZ_ZasobID = CZ_ID
		FOR XML PATH('')
	), 1, 1, ''))
	
	AS [Dział], ISNULL(SzS_DokID, -1) as 'Wykorzystany'
	FROM dbo.CtiZasob 
		LEFT JOIN dbo.CtiKonfigJM ON CtiKonfigJM.CKJ_Typ = 1 
			AND CtiZasob.CZ_JMCzasu = CtiKonfigJM.CKJ_Lp 
		INNER JOIN dbo.CtiZasobGrupy ON CtiZasob.CZ_CZGID = CtiZasobGrupy.CZG_ID
		left join dbo.CtiZasobSlownik on @IDCSU IS NOT NULL and Szs_CsUID=@IDCSU and SzS_DokID=CZ_ID

	where ((isnull(@BezTypu,0)=0) or (isnull(@BezTypu,0) <> 0 and isnull(cz_typ,0)<>isnull(@BezTypu,0)))
	and (@DzialID IS NULL 
		OR EXISTS (SELECT * FROM dbo.CtiZasobDzialy WHERE ZDZ_ZasobID = CZ_ID AND ZDZ_DzialID = @DzialID) 
		OR (@AkceptujBezDzialu = 1 AND NOT EXISTS (SELECT * FROM dbo.CtiZasobDzialy WHERE ZDZ_ZasobID = CZ_ID)))
	and (isnull(@TylkoTyp, -1) = -1 OR CZ_Typ = @TylkoTyp)
	and (isnull(@BezID, 0) <> CZ_ID)
	ORDER BY Grupa, Kod
	
END
GO
```

# [dbo].[CTIProdukcjaZasobyPlanList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZasobyPlanList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZasobyPlanList]
   @typ int						-- = 0
 , @datOd varchar(23)			--'2016-09-08 11:00:00.000' --data start
 , @datDo varchar(23)			-- 
 , @iloscZP decimal(15,4)		--przekazana ilość
 , @cznid int					-- konkretne zlecenie
 , @czid int					-- konkretny zasob
 , @idTH int					--
 , @Zasob VARCHAR(20)			--
 , @GrupaZasobu  VARCHAR(20)	--
 , @Zlecenie VARCHAR(20)		--
AS
BEGIN

declare @datastart datetime set @datastart = convert(datetime, @datOd,120)
declare @dataEnd datetime set @dataEnd= convert(datetime, @datDo,120)


if @typ > 0
begin--------------------------plan zasobów-------------------------
		SELECT   ROW_NUMBER() OVER (ORDER BY ISNULL(czn_id,0), ISNULL(cz_id,0),ISNULL(czs_id,0)) as Lp, 0 as LP2 		
			, 0 as Progress
			, ISNULL(CZS_ID,czn_id) as ID, ISNULL(CZS_CZID,0) as CZS_CZID, ISNULL(czs_ilosc,0) as Ilosc
			, ISNULL(CZG_Kod,'') as Grupa, ISNULL(CZ_Kod,'') As Kod, czn_id as dokId, CZN_NrPelny as NrPelny, Twr_Kod AS TwrKod 
			, ISNULL(CZS_CzasPracy,0) as CzasPracy, ISNULL(CZS_CzasUzbrojenia,0)  as CzasUzbrojenia, ISNULL(CZS_CzasRozbrojenia,0)  as CzasRozbrojenia
			, ISNULL(CKJ_Nazwa,'') as JM
			, ISNULL(CZS_KosztPracy,0) as KosztPracy  , ISNULL(CZS_KosztUzbrojenia,0)  as KosztUzbrojenia, ISNULL(CZS_KosztRozbrojenia,0)  as KosztRozbrojenia
    
			--obliczanie czasu
			, ISNULL(CZ_Ilosc,0)  as iloscZP
			, COALESCE(CZS_Dzielnik,CZN_Ilosc,0) as Dzielnik
			, 1 as Mnoznik --przeliczanie występuje przed przeliczeniem
			, convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZN_Ilosc, COALESCE(CZS_Dzielnik,CZN_Ilosc,0), CZS_JMCzasu, CZS_ID)/60.0) --w minutach
			as CzasCalkowity
			--daty
			, ISNULL(CZS_CzasStart,CZN_CzasStart) as DataStart
			, ISNULL(CZS_CzasEnd,CZN_CzasEnd) as DataEnd

			, ISNULL(CZ_Typ,0) as Typ
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_PwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzeniePw
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_RwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzenieRw
			
			, isnull(CZS_PrcId,0) as PrcId, isnull(PRA_Kod,'') as PrcKod,  isnull(CZS_Opis, '') as Opis , isnull(CZS_Lp,1) as Czs_LP
		FROM dbo.CtiZlecenieNag 
			LEFT JOIN dbo.CtiZasobDok  on ISNULL(CZS_DokTyp,0) = 2 and ISNULL(CZS_CTNID,0) = czn_id and  @typ = 1
			LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp and  @typ = 1
			LEFT JOIN dbo.CtiZasob on CZ_ID = CZS_CZID and @typ = 1
			LEFT JOIN dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID  and @typ = 1
			LEFT JOIN CDN.Towary ON CZN_TwrId = Twr_TwrId
			LEFT JOIN CDN.PracKod on pra_praid =  isnull(CZS_PrcId,0) 
		where 
			(
				(ISNULL(CZS_CzasEnd,CZN_CzasEnd) between @dataStart and @dataEnd)
				or
				 (ISNULL(CZS_CzasStart,CZN_CzasStart) between @dataStart  and @dataEnd)
				or 
				(ISNULL(CZS_CzasStart,CZN_CzasStart) < @dataStart and ISNULL(CZS_CzasEnd,CZN_CzasEnd)  > @dataEnd)
			)
			and	
			( 
				((@Zasob = '' and 1=1) or (@Zasob <> '' and ISNULL(CZ_Kod,'')=@Zasob))
				and ((@GrupaZasobu = ''  and 1=1) or (@GrupaZasobu <> '' and ISNULL(CZG_Kod,'')=@GrupaZasobu))
				and ((@Zlecenie = ''  and 1=1) or (@Zlecenie <> '' and CZN_NrPelny=@Zlecenie))
			)
		order by ISNULL(czn_id,0), ISNULL(cz_id,0), isnull(CZS_Lp,1), ISNULL(czs_id,0)
end
else if  @czid > 0
begin--------------------------Szczegoly zasobu-------------------------

		SELECT ROW_NUMBER() OVER (ORDER BY id) as Lp, 0 as LP2 ,0 as Progress
			, ID, CZS_CZID, Ilosc
			, Grupa, Kod, dokId, NrPelny, TwrKod
			, CzasPracy, CzasUzbrojenia, CzasRozbrojenia
			, JM
			, KosztPracy , KosztUzbrojenia, KosztRozbrojenia
			--obliczanie czasu
			, iloscZP
			, Dzielnik
			, Mnoznik 
			, CzasCalkowity
			--daty
			, DataStart
			, DataEnd

			, Typ, TworzeniePw, TworzenieRw
			, PrcId, PrcKod, Opis, Czs_LP
		from
		(
			SELECT CZS_ID as ID, CZS_CZID, ISNULL(czs_ilosc,0) as Ilosc
				, CZG_Kod as Grupa,CZ_Kod As Kod, czn_id as dokId, CZN_NrPelny as NrPelny, Twr_Kod AS TwrKod 
				, ISNULL(CZS_CzasPracy,0) as CzasPracy, CZS_CzasUzbrojenia as CzasUzbrojenia , CZS_CzasRozbrojenia as CzasRozbrojenia
				, CKJ_Nazwa as JM
				, CZS_KosztPracy as KosztPracy, CZS_KosztUzbrojenia as KosztUzbrojenia, CZS_KosztRozbrojenia as KosztRozbrojenia
		
				--obliczanie czasu
				, CZN_Ilosc as iloscZP
				--, COALESCE(CZS_Dzielnik,CZN_Ilosc,0) as Dzielnik
				--, case when CZN_Ilosc%COALESCE(CZS_Dzielnik,CZN_Ilosc,0) > 0 
				--then convert(int,(CZN_Ilosc/COALESCE(CZS_Dzielnik,CZN_Ilosc,0)))+1 
				--else convert(int,CZN_Ilosc/COALESCE(CZS_Dzielnik,CZN_Ilosc,0)) end as Mnoznik
				, 1 as Dzielnik
				, 1 as Mnoznik

				
				, convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZN_Ilosc, COALESCE(CZS_Dzielnik,CZN_Ilosc,0), CZS_JMCzasu, CZS_ID)/60.0 ) --w minutach
				as CzasCalkowity

				--data
				, case when CZN_Status = 0 then CZN_CzasStart else ISNULL(CZS_CzasStart,CZN_CzasStart) end as DataStart
				, case when CZN_Status = 0 then 
				 dbo.CTIProdukcjaZPZasobDataEnd(CZ_ID,ISNULL(CZS_CzasStart,CZN_CzasStart),dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZN_Ilosc, COALESCE(CZS_Dzielnik,CZN_Ilosc,0), CZS_JMCzasu, CZS_ID)) 
				else ISNULL(CZS_CzasEnd,CZN_CzasEnd) end
				
				as DataEnd

				, ISNULL(CZ_Typ,0) as Typ
				, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_PwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzeniePw
				, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_RwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzenieRw
		
				, isnull(CZS_PrcId,0) as PrcId, isnull(PRA_Kod,'') as PrcKod,  isnull(CZS_Opis, '') as Opis , isnull(CZS_Lp,1) as Czs_LP
			FROM dbo.CtiZasobDok  
				LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
				INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
				INNER JOIN dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID 
				INNER JOIN dbo.CtiZlecenieNag on CZS_DokTyp = 2 and CZS_CTNID = czn_id
				LEFT JOIN CDN.Towary ON CZN_TwrId = Twr_TwrId
				LEFT JOIN CDN.PracKod on pra_praid =  isnull(CZS_PrcId,0) 
			where CZS_DokTyp = 2 and CZS_CZID = @czid
		) a
		where 
		(
			(DataEnd between @dataStart and @dataEnd)
			or
			 (DataStart between @dataStart  and @dataEnd)
			or 
			(DataStart < @dataStart and DataEnd  > @dataEnd)
		)
		 order by isnull(CZS_Lp,1),ID 
end
------------------------------------------------------Na Zleceniu----------------------------------------------------
else if @idTH > 0 
begin--------------------------Niezapisane ZP, zasoby zgodnie z TH--------------------------
	
	SELECT ROW_NUMBER() OVER (ORDER BY czs_id) as Lp, 0 as LP2 ,0 as Progress
		, CZS_ID as ID, CZS_CZID, ISNULL(czs_ilosc,0) as Ilosc
		, CZG_Kod as Grupa, CZ_Kod As Kod, CTN_ID as dokId, CTN_Numer as NrPelny, Twr_Kod AS TwrKod 
		, (ISNULL(CZS_CzasPracy,0)) as CzasPracy
		, (CZS_CzasUzbrojenia) as CzasUzbrojenia
		, (CZS_CzasRozbrojenia) as CzasRozbrojenia
		, CKJ_Nazwa as JM
		, CZS_KosztPracy as KosztPracy , CZS_KosztUzbrojenia as KosztUzbrojenia
		, CZS_KosztRozbrojenia as KosztRozbrojenia

		--, COALESCE(CZS_Dzielnik,@iloscZP,0) as Dzielnik
		--, case when ISNULL(@iloscZP,0)%COALESCE(CZS_Dzielnik,@iloscZP,0) > 0 
		--	then convert(int,(ISNULL(@iloscZP,0)/COALESCE(CZS_Dzielnik,@iloscZP,0)))+1 
		--	else convert(int,ISNULL(@iloscZP,0)/COALESCE(CZS_Dzielnik,@iloscZP,0)) 
		--	end as Mnoznik
		, 1 as Dzielnik
		, 1 as Mnoznik

		, convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, @iloscZP, COALESCE(CZS_Dzielnik,@iloscZP,0), CZS_JMCzasu, CZS_ID)/60.0) --w minutach
			as CzasCalkowity

		--daty
		, ISNULL(@datastart,getdate()) as DataStart
		, dbo.CTIProdukcjaZPZasobDataEnd(CZ_ID,ISNULL(@datastart,getdate()),dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, @iloscZP, COALESCE(CZS_Dzielnik,@iloscZP,0), CZS_JMCzasu, CZS_ID)) 
		as DataEnd
		, ISNULL(CZ_Typ,0) as Typ
		, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_PwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzeniePw
		, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_RwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzenieRw
		
		, isnull(CZS_PrcId,0) as PrcId, isnull(PRA_Kod,'') as PrcKod,  isnull(CZS_Opis, '') as Opis , isnull(CZS_Lp,1) as Czs_LP
	FROM dbo.CtiZasobDok 
		LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
		INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
		INNER JOIN dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID 
		INNER JOIN dbo.CtiTechnolNag on CZS_DokTyp = 1 and CZS_CTNID = CTN_ID
		LEFT JOIN CDN.Towary ON ctn_TwrId = Twr_TwrId
		LEFT JOIN CDN.PracKod on pra_praid =  isnull(CZS_PrcId,0) 
	where CZS_DokTyp = 1 --TH
		and CZS_ctnid =@idTH
	order by isnull(CZS_Lp,1),ID 
end
else 
begin--------------------------Zapisane ZP--------------------------
	declare @bufor int = 0 
	select @bufor = ISNULL(CZN_Status,0)
		, @iloscZP = ISNULL(@iloscZP,CZN_Ilosc) 
		, @dataStart = ISNULL(@dataStart,CZN_CzasStart) 
		, @dataEnd =  ISNULL(@dataEnd,CZN_CzasEnd)
	from dbo.CtiZlecenieNag where czn_id =@cznid


	if @bufor = 0 --------------------------nie potwierdzone zlecenie --------------------------
	begin

		SELECT ROW_NUMBER() OVER (ORDER BY czs_id) as Lp, 0 as LP2 ,0 as Progress
			, CZS_ID as ID, CZS_CZID, ISNULL(czs_ilosc,0) as Ilosc
			, CZG_Kod as Grupa, CZ_Kod As Kod, czn_id as dokId, CZN_NrPelny as NrPelny, Twr_Kod AS TwrKod 
			, ISNULL(CZS_CzasPracy,0) as CzasPracy, CZS_CzasUzbrojenia as CzasUzbrojenia, CZS_CzasRozbrojenia as CzasRozbrojenia
			, CKJ_Nazwa as JM
			, CZS_KosztPracy as KosztPracy , CZS_KosztUzbrojenia as KosztUzbrojenia, CZS_KosztRozbrojenia as KosztRozbrojenia

			--obliczanie czasu
			, @iloscZP as iloscZP
			, 1 as Dzielnik
			, 1 as Mnoznik
			--, COALESCE(CZS_Dzielnik,@iloscZP,0) as Dzielnik
			--, case when @iloscZP%COALESCE(CZS_Dzielnik,@iloscZP,0) > 0 
			--	then convert(int,(@iloscZP/COALESCE(CZS_Dzielnik,@iloscZP,0)))+1 
			--	else convert(int,@iloscZP/COALESCE(CZS_Dzielnik,@iloscZP,0)) end 
			-- as Mnoznik
			, convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, @iloscZP, COALESCE(CZS_Dzielnik,@iloscZP,0), CZS_JMCzasu, CZS_ID)/60.0) --w minutach
			 as CzasCalkowity
			 --daty
			, ISNULL(CZS_CzasStart,CZN_CzasStart) as DataStart
			, ISNULL(CZS_CzasEnd,CZN_CzasEnd) as DataEnd
			, ISNULL(CZ_Typ,0) as Typ
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_PwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzeniePw
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_RwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzenieRw
		
			, isnull(CZS_PrcId,0) as PrcId, isnull(PRA_Kod,'') as PrcKod,  isnull(CZS_Opis, '') as Opis, isnull(CZS_Lp,1) as Czs_LP 
		FROM dbo.CtiZasobDok  
			LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
			INNER JOIN dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID 
			INNER JOIN dbo.CtiZlecenieNag on CZS_DokTyp = 2 and CZS_CTNID = czn_id
			LEFT JOIN CDN.Towary ON CZN_TwrId = Twr_TwrId
			LEFT JOIN CDN.PracKod on pra_praid =  isnull(CZS_PrcId,0) 
		where CZS_DokTyp = 2 --ZP
			and CZS_ctnid =@cznid
		order by isnull(CZS_Lp,1),ID 

	end
	else --------------------------potwierdzone zlecenie --------------------------
	begin

		SELECT ROW_NUMBER() OVER (ORDER BY czs_id) as Lp, 0 as LP2 ,0 as Progress
			, CZS_ID as ID, CZS_CZID, ISNULL(czs_ilosc,0) as Ilosc
			, CZG_Kod as Grupa, CZ_Kod As Kod, czn_id as dokId, CZN_NrPelny as NrPelny, Twr_Kod AS TwrKod 
			, ISNULL(CZS_CzasPracy,0) as CzasPracy, CZS_CzasUzbrojenia as CzasUzbrojenia, CZS_CzasRozbrojenia as CzasRozbrojenia
			, CKJ_Nazwa as JM
			, CZS_KosztPracy as KosztPracy , CZS_KosztUzbrojenia as KosztUzbrojenia, CZS_KosztRozbrojenia as KosztRozbrojenia
    
			--obliczanie czasu
			, CZ_Ilosc as iloscZP
			, COALESCE(CZS_Dzielnik,CZN_Ilosc,0) as Dzielnik
			, 1 as Mnoznik --przeliczanie występuje przed przeliczeniem
			--20170412
			, convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZN_Ilosc, COALESCE(CZS_Dzielnik,CZN_Ilosc,0), CZS_JMCzasu, CZS_ID)/60.0) --w minutach
			 as CzasCalkowity

			--daty
			, ISNULL(CZS_CzasStart,CZN_CzasStart) as DataStart
			, ISNULL(CZS_CzasEnd,CZN_CzasEnd) as DataEnd

			, ISNULL(CZ_Typ,0) as Typ
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_PwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzeniePw
			, case when ISNULL(CZ_Typ,0) = 76 then	case when ISNULL(CZS_RwBlokada,0) = 0 then 1 else 0 end else 0 end as TworzenieRw

			, isnull(CZS_PrcId,0) as PrcId, isnull(PRA_Kod,'') as PrcKod,  isnull(CZS_Opis, '') as Opis, isnull(CZS_Lp,1) as Czs_LP 
		FROM dbo.CtiZasobDok  
			LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
			INNER JOIN dbo.CtiZasob on CZ_ID = CZS_CZID
			INNER JOIN dbo.CtiZasobGrupy on CZS_CZGID = CZG_ID 
			INNER JOIN dbo.CtiZlecenieNag on CZS_DokTyp = 2 and CZS_CTNID = czn_id
			LEFT JOIN CDN.Towary ON CZN_TwrId = Twr_TwrId
			LEFT JOIN CDN.PracKod on pra_praid =  isnull(CZS_PrcId,0) 
		where CZS_DokTyp = 2 
			and CZS_ctnid = @cznid
	order by isnull(CZS_Lp,1),ID 

	end
end
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieAddEdit]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaZlecenieAddEdit]
	  @id						decimal(15,4)
	, @Ilosc					decimal(15,4)
	, @Ope						int
	, @Status					int--2 potwierdzone -- 3 zamknięte
	, @CTNId					int--Technologia
	, @opis						nvarchar(254)
	, @datStart					varchar(23)
	, @datStop					varchar(23)
	, @ZasobyKopia				tinyint
	, @NrWlasny					nvarchar(50)
	, @idZleceniaProstego		INT OUTPUT
	, @idWyrobuGotowego			INT = 0
	, @Jm						nvarchar(40) = NULL
	, @lp						INT = 0
	, @cenaID					INT = NULL
	, @pominCzasZasobow			INT = NULL
	, @domyslneMagSur			INT = NULL
	, @uwzglednijDateZPNaZasobach INT = 0
	, @generateRWAfterResource int = -1
AS
BEGIN
declare @dataStart datetime
declare @dataStop datetime 

if @generateRWAfterResource = -1
	set @generateRWAfterResource = (SELECT TOP 1 ISNULL(CTN_GenerateRWAfterResource,0) from dbo.ctitechnolnag where ctn_id = @CTNId);
if @generateRWAfterResource IS NULL
	set @generateRWAfterResource = 0;

BEGIN TRY
if @datStart = '' set @dataStart = DATEADD(HOUR, 8, CONVERT(DATETIME, CONVERT(DATE, GETDATE()))) else set @dataStart = convert(datetime, @datStart,120)
END TRY
BEGIN CATCH
set @dataStart = DATEADD(HOUR, 8, CONVERT(DATETIME, CONVERT(DATE, GETDATE())))
END CATCH

BEGIN TRY
if @datStop = ''  set @dataStop = @dataStart else set @dataStop= convert(datetime, @datStop,120)
END TRY
BEGIN CATCH
set @dataStop = @dataStart
END CATCH

declare @numerZlc int = 0
declare @miesiac int = month(GETDATE())
declare @rok int =  year(GETDATE())
declare @nrpelny varchar(20) =''
declare @Seria varchar(5) = ''
declare @TwrId  int = 0
declare @TwrKod varchar(20) = ''
declare @NoweLP int = 0
declare @konfiguracja varchar(1) = ''
select @konfiguracja = ISNULL(KNF_Wartosc,'') from dbo.CtiKonfiguracja where KNF_Numer = 73
select @NoweLP = ISNULL(MAX(CZN_LP), 0) + 1 FROM dbo.CtiZlecenieNag

DECLARE @czyIstniejeUstawienie INT 
SELECT @czyIstniejeUstawienie = KNF_Wartosc FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 1139
if (@Jm IS NULL)
BEGIN
	SET @Jm = (SELECT TOP 1 CTN_TwrJm FROM dbo.CtiTechnolNag WHERE CTN_ID = @CTNId);
END

if (@Jm IS NULL)
BEGIN
	SET @Jm = (SELECT TOP 1 Twr_JM FROM CDN.Towary WHERE Twr_TwrId = @idWyrobuGotowego);
END

if @konfiguracja = '1'--uwzględnienie miesiąca w numeracji
begin	
	select @numerZlc = ISNULL(Max(CZN_Numer),0)+1  from dbo.CtiZlecenieNag where CZN_Rok = @rok and CZN_Miesiac = @miesiac --miesiac
	select @TwrKod=ctn_twrkod,@TwrId=ctn_twrid from dbo.CtiTechnolNag where CTN_ID = @CTNId
	select @nrpelny='ZP/'+cast(replace(str(@numerZlc,4),' ','0')as char(4)) +'/'+cast(replace(str(@miesiac,2),' ','0')as char(2)) +'/'+cast(@rok as varchar(4))
end
else
begin
	select @numerZlc = ISNULL(Max(CZN_Numer),0)+1  from dbo.CtiZlecenieNag where CZN_Rok = @rok ---tylko rok
	select @TwrKod=ctn_twrkod, @TwrId=ctn_twrid from dbo.CtiTechnolNag where CTN_ID = @CTNId
	select @nrpelny='ZP/'+cast(replace(str(@numerZlc,5),' ','0')as char(5)) +'/'+cast(@rok as varchar(4))
end
if @idWyrobuGotowego > 0
begin
	select @TwrKod=Twr_Kod, @TwrId=Twr_TwrId from cdn.towary WHERE Twr_TwrId = @idWyrobuGotowego
end

if isnull(@id,0) = 0
begin
	PRINT('Nie wskazano ID istniejacego ZP, tworzenie nowego');

	if isnull(@CTNId,0) <= 0
	begin
		PRINT('Nie wskazano ID technologii');

		INSERT INTO dbo.CtiZlecenieNag 
		(CZN_Numer,CZN_Miesiac,CZN_Rok,CZN_Seria, CZN_OpeWystaw, CZN_DataWystaw, CZN_OperModyf, CZN_DataModyf, CZN_Status,CZN_NrPelny,CZN_Opis,CZN_CzasStart,CZN_CzasEnd, CZN_Ilosc, CZN_TwrJm, CZN_LP, CZN_CenaID, CZN_PominCzasZasobow, CZN_DomyslneMagSur, CZN_UwzglednijDateZPNaZasobach, CZN_GenerateRWAfterResource)
		VALUES (@numerZlc,@miesiac,@rok,@Seria,@Ope,GETDATE(),@Ope,GETDATE(),0,@nrpelny,@opis,@dataStart,@dataStop,@Ilosc,@Jm,@NoweLP, @cenaID, ISNULL(@pominCzasZasobow, 0), ISNULL(@domyslneMagSur, 0), @uwzglednijDateZPNaZasobach, @generateRWAfterResource)      
		Select @idZleceniaProstego = SCOPE_IDENTITY()

		if (@idWyrobuGotowego > 0)
		begin
			UPDATE dbo.CtiZlecenieNag SET CZN_TwrId = @idWyrobuGotowego, CZN_TwrKod = @TwrKod WHERE CZN_ID = @idZleceniaProstego
		end
	end
	else if isnull(@CTNId,0) >0
	begin
		DECLARE @IloscTH decimal(15,4) = 1;
		SET @IloscTH = (SELECT TOP 1 CTN_ilosc FROM dbo.CtiTechnolNag WHERE CTN_ID = @CTNId);

		IF (@domyslneMagSur IS NULL)
			SET @domyslneMagSur = (SELECT TOP 1 CTN_DomyslneMagSur FROM dbo.CtiTechnolNag WHERE CTN_ID = @CTNId);

		INSERT INTO dbo.CtiZlecenieNag 
		(CZN_Numer,CZN_Miesiac,CZN_Rok,CZN_Seria, CZN_TwrId, CZN_TwrKod, CZN_Ilosc, CZN_OpeWystaw, CZN_DataWystaw, CZN_OperModyf, CZN_DataModyf, CZN_Status,czn_ctnid,CZN_NrPelny,CZN_Opis,CZN_CzasStart,CZN_CzasEnd,CZN_TwrJm, CZN_LP, CZN_CenaID, CZN_PominCzasZasobow, CZN_DomyslneMagSur, CZN_UwzglednijDateZPNaZasobach, CZN_GenerateRWAfterResource)
		VALUES (@numerZlc,@miesiac,@rok,@Seria,@TwrId,@TwrKod,@Ilosc,@Ope,GETDATE(),@Ope,GETDATE(),@Status,@CTNId,@nrpelny,@opis,@dataStart,@dataStop,@Jm, @NoweLP, @cenaID, ISNULL(@pominCzasZasobow, 0), ISNULL(@domyslneMagSur, 0), @uwzglednijDateZPNaZasobach, @generateRWAfterResource)
		Select @idZleceniaProstego = SCOPE_IDENTITY()

		if isnull(@ZasobyKopia,0) > 0
		begin--zasoby
			declare @przeliczanieZasobow varchar(max) = ISNULL((SELECT TOP 1 KNF_Wartosc FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 1008), '1')

			insert into dbo.CtiZasobDok
				(CZS_CZID, CZS_CZGID, CZS_CTNID, CZS_DokTyp, CZS_Kod, CZS_JMCzasu, CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy 
				, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia, czs_ilosc, CZS_PwBlokada, CZS_RwBlokada 
				, CZS_CzasStart, CZS_CzasEnd
				, CZS_Lp, CZS_Opis
				, CZS_Dzielnik, CZS_PlanowanieCzasuPracy )
			SELECT CZS_CZID, CZS_CZGID, @idZleceniaProstego, 2, CZS_Kod, CZS_JMCzasu
				, CASE WHEN @przeliczanieZasobow IN ('True', 'TRUE', '1') THEN CZS_CzasPracy * (@Ilosc / @IloscTH) ELSE ISNULL(CZS_CzasPracy,0) END CZS_CzasPracy
				, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, CZS_KosztPracy, CZS_KosztUzbrojenia, CZS_KosztRozbrojenia
				, CASE WHEN @przeliczanieZasobow IN ('True', 'TRUE', '1') THEN ISNULL(czs_ilosc,0) * (@Ilosc / @IloscTH) ELSE ISNULL(CZS_Ilosc,0) END czs_ilosc
				, ISNULL(CZS_PwBlokada,0), ISNULL(CZS_RwBlokada,0)
				, @dataStart, dbo.CTIProdukcjaZPZasobDataEnd(CZS_CZID,@dataStart,dbo.CTIProdukcjaZPZasobCzasCalkowity(CZS_CzasPracy, CZS_CzasUzbrojenia, CZS_CzasRozbrojenia, @Ilosc, COALESCE(CZS_Dzielnik,@Ilosc,0), CZS_JMCzasu, CZS_ID)) 
				, ROW_NUMBER() OVER (PARTITION BY CZS_DokTyp, CZS_CTNID ORDER BY  isnull(CZS_Lp,0),CZS_ID) as Lp--20170810 LP
				, CZS_Opis--20180125
				, CZS_Dzielnik, CZS_PlanowanieCzasuPracy
			FROM dbo.CtiZasobDok LEFT JOIN dbo.CtiKonfigJM on CKJ_Typ = 1 and CZS_JMCzasu = CKJ_Lp
				INNER JOIN dbo.CtiZasob on CZS_CZID = CZ_ID
			where CZS_DokTyp = 1 and CZS_ctnid = @CTNId
			order by isnull(CZS_Lp,0), CZS_ID--20170810 LP
			SELECT @dataStop=MAX(CZS_CzasEnd) FROM dbo.CtiZasobDok WHERE CZS_CTNID = @idZleceniaProstego AND CZS_DokTyp=2
		end

			
	end

	--przepisanie surowców z TH
end
else
begin

	PRINT('Wskazano ID istniejacego ZP, aktualizacja');

	UPDATE dbo.CtiZlecenieNag SET	
		  CZN_Ilosc		= @Ilosc
		, CZN_CTNId		= @CTNId
		, CZN_OperModyf = @Ope
		, CZN_Status	= @Status
		, CZN_TwrId		= @TwrId
		, CZN_TwrKod	= @TwrKod
		, CZN_DataModyf = GETDATE()
		, CZN_Opis		= @opis
		, CZN_CzasStart = @dataStart
		, CZN_CzasEnd	= @dataStop
		, CZN_TwrJm		= @Jm
		, CZN_LP		= @lp
		, CZN_CenaID	= @cenaID
		, CZN_PominCzasZasobow = ISNULL(@pominCzasZasobow, CZN_PominCzasZasobow)
		, CZN_DomyslneMagSur = ISNULL(@domyslneMagSur, CZN_DomyslneMagSur)
		, CZN_UwzglednijDateZPNaZasobach = @uwzglednijDateZPNaZasobach
		, CZN_GenerateRWAfterResource = @generateRWAfterResource
	WHERE CZN_ID = @id;

	UPDATE dbo.CtiZlecenieZElem
	SET CZZE_Ilosc = @Ilosc
	WHERE CZZE_CZNID = @id
	AND CZZE_Typ = 2;

	set @idZleceniaProstego = @id
end

--Nowy mechanizm przypisywania statusu na technologii
UPDATE dbo.CtiTechnolNag SET CTN_DodatkowyStatus = 2 WHERE CTN_ID = ISNULL(@CTNId, -1)

if @idZleceniaProstego > 0	--numer wlasny
begin
	PRINT('Ustawienie numeru wlasnego');

	if not exists (SELECT isnull(DNW_Numer,0) FROM dbo.CtiDokNrWlasny where DNW_DokTyp = 2 and DNW_DokID = @idZleceniaProstego)
		insert dbo.CtiDokNrWlasny (DNW_DokTyp, DNW_DokID, DNW_Numer) values (2,@idZleceniaProstego,@NrWlasny)
	else
		update  dbo.CtiDokNrWlasny set DNW_Numer = @NrWlasny  where DNW_DokTyp = 2 and DNW_DokID = @idZleceniaProstego

	IF ISNULL(@czyIstniejeUstawienie, 0) > 0
	BEGIN
	 IF(not exists(SELECT TOP 1 CPS_ID FROM dbo.CtiProdukcjaStatus where CPS_DokID = @idZleceniaProstego))
		INSERT INTO dbo.CtiProdukcjaStatus (CPS_DokID, CPS_DokTyp, CPS_StatusMMWydane, CPS_StatusMMKontrola, CPS_StatusPWWG, IloscZrealizowana, DataModyfikacji)
									VALUES (@idZleceniaProstego, 2, -1, -1, -1, 0, GETDATE())

		EXEC dbo.CTIProdukcjaUpdateStatusyZP 0, @idZleceniaProstego

	END
end

Select @idZleceniaProstego as ID 
END
GO
```

# [dbo].[CTIProdukcjaZlecenieCofnij]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieCofnij]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieCofnij]
@id int
AS
BEGIN
declare @info varchar(200) = ''
declare @status int = 0;

	select @info = dbo.CTIProdukcjaZlecenieNumer(CZN_ID), @status = CZN_Status from dbo.CtiZlecenieNag where CZN_ID = @id
	DECLARE @automatycznyStatusZP INT = (SELECT CAST(ISNULL(KNF_Wartosc,1) AS INT) FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 64 AND ISNULL(KNF_Wartosc,'') <> '')
	DECLARE @MM INT = (SELECT CAST(ISNULL(KNF_Wartosc,0) AS INT) FROM dbo.CtiKonfiguracja WHERE KNF_Numer = 63 AND ISNULL(KNF_Wartosc,'') <> '')
	DECLARE @ZPamount DECIMAL = (select sum(CZN_Ilosc) from dbo.CtiZlecenieNag where czn_id =@id)
	IF @status = 3
	BEGIN
		IF @automatycznyStatusZP = 1 and @MM = 1 and (SELECT dbo.CTIProdukcjaZlecenieIloscSkontrolowana(@id))>=@ZPamount
			BEGIN
				set @info += ' - w związku z włączoną automatyczną blokadą zleceń (MM na maksymalną ilość) nie da się odblokować zlecenia.'
			END
		ELSE IF @automatycznyStatusZP = 1 and @MM = 0 and (SELECT dbo.CTIProdukcjaZlecenieIloscZrealizowana(@id,0,0))>=@ZPamount
			BEGIN
				set @info += ' - w związku z włączoną automatyczną blokadą zleceń (PW na maksymalną ilość) nie da się odblokować zlecenia.'
			END
		ELSE
			BEGIN
					Update dbo.CtiZlecenieNag set czn_status = 2 where CZN_ID = @id
					set @info += ' - pomyślnie odblokowano.'
			END
	END
	ELSE IF @status not in (0,3)
	BEGIN
		IF not exists (select CZD_CZNId from dbo.CtiZlecenieDok JOIN CDN.TraNag ON CZD_TrnId = TrN_TrNID where CZD_CZNId = @id AND TrN_Anulowany = 0)--dokumenty optima
		begin
			if not exists (select ZZs_CZNID from dbo.CtiZlecenieZasob where ZZs_CZNID = @id and ZZs_DokTyp = 2)--realizacja zasobu
			begin
				 Update dbo.CtiZlecenieNag set czn_status = 0 where CZN_ID = @id
				 set @info += ' - pomyślnie otwarto.'
			end
			else set @info += ' - błąd: istnieją realizacje zasobu.'
		end
		else set @info += ' - błąd: istnieje powiązanie z dokumentami w Optimie.'
	END

	select @info as Info
END

GO
```

# [dbo].[CTIProdukcjaZlecenieDel]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieDel]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieDel]
@id int,
@walidacjaStatusu int = 1
AS
BEGIN
declare @status int
DECLARE @SavePointName NVARCHAR(MAX) = 'SavePoint';

select @status =dbo.CTIProdukcjaStatusy(2,CZN_ID)  from  dbo.CtiZlecenieNag  where CZN_ID =@id
if @status=0 or @walidacjaStatusu = 0
begin
	BEGIN TRY
        -- Sprawdzenie, czy istnieje aktywna transakcja
        IF @@TRANCOUNT > 0
        BEGIN
            SAVE TRANSACTION @SavePointName;
        END
		 declare @CZZNID int = 0
		 
		  
		if exists (select isnull(CZD_ID,0) from dbo.CtiZlecenieDok LEFT JOIN cdn.tranag on czd_trnid = trn_trnid where isnull(trn_trnid,0) = 0 and CZD_CZNId = @id )
			begin
				delete FROM CtiZlecenieDokElem where CZE_TrnId in   
				(select isnull(czd_trnid,0) from dbo.CtiZlecenieDok LEFT JOIN cdn.tranag on czd_trnid = trn_trnid where isnull(trn_trnid,0) = 0 and CZD_CZNId = @id)

				delete from dbo.CtiZlecenieDok where CZD_ID in 
				 (select isnull(CZD_ID,0) from dbo.CtiZlecenieDok LEFT JOIN cdn.tranag on czd_trnid = trn_trnid where isnull(trn_trnid,0) = 0 and CZD_CZNId = @id)
		  end
		  
		IF EXISTS (SELECT CZD_TrNID FROM dbo.CtiZlecenieDok JOIN CDN.TraNag ON CZD_TrnID = TrN_TrNID WHERE TrN_Bufor < 0 AND CZD_CZNID = @id)
		BEGIN
			DELETE FROM dbo.CtiZlecenieDokElem WHERE CZE_TrNID IN
			(SELECT CZD_TrNID FROM dbo.CtiZlecenieDok JOIN CDN.TraNag ON CZD_TrnID = TrN_TrNID WHERE TrN_Bufor < 0 AND CZD_CZNID = @id);

			DELETE FROM dbo.CtiZlecenieDok WHERE CZD_TrNID IN
			(SELECT CZD_TrNID FROM dbo.CtiZlecenieDok JOIN CDN.TraNag ON CZD_TrnID = TrN_TrNID WHERE TrN_Bufor < 0 AND CZD_CZNID = @id)
		END


		SELECT  @CZZNID = CZZE_CZZNID FROM  CtiZlecenieZElem WHERE CZZE_CZNID = @id
		DELETE from  dbo.CtiZlecenieZasobHist WHERE ZsH_CZNID = @id 
		DELETE FROM  dbo.CtiZlecenieZasob WHERE ZZs_CZNID = @id
		DELETE FROM	 dbo.CtiZalaczniki WHERE CZAL_DokTyp = 4 AND CZAL_DokId IN (SELECT CZS_ID FROM dbo.CtiZasobDok WHERE CZS_DokTyp = 2 AND CZS_CTNID = @id)
		DELETE FROM  dbo.CtiZasobDok WHERE CZS_CTNID = @id and CZS_DokTyp = 2 
		DELETE FROM  dbo.CtiProdukcjaStatus WHERE CPS_DokId = @id AND CPS_DokTyp = 2 

		IF EXISTS(SELECT czn_ctnid FROM dbo.CtiZlecenieNag WHERE CZN_CTNId in (SELECT TOP 1 CZN_CTNId FROM dbo.CtiZlecenieNag WHERE CZN_ID = @id))
			UPDATE dbo.CtiTechnolNag SET CTN_DodatkowyStatus = -1 WHERE ctn_id = (SELECT TOP 1 CZN_CTNId FROM dbo.CtiZlecenieNag WHERE CZN_ID = @id)
		
-----powiązania z ZPZ--170829
		DECLARE kursor cursor FOR
				SELECT distinct CZZE_CZNID FROM dbo.CtiZlecenieZElem
		DECLARE @idZPZ INT OPEN kursor
		FETCH NEXT FROM kursor   INTO @idZPZ
		while @@FETCH_STATUS = 0
		BEGIN
			delete FROM CtiZlecenieZElem where CZZE_CZZNID = @idZPZ and CZZE_CZNID =  @id and isnull(CZZE_Typ,2) = 2	

			--przenumerowanie
			update dbo.CtiZlecenieZElem Set CZZE_Lp= lp from
			(	select ROW_NUMBER() OVER (PARTITION BY CZZE_CZZNID ORDER BY  isnull(CZZE_Lp,0),CZZE_ID) as Lp, CZZE_ID as Id from  dbo.CtiZlecenieZElem  where CZZE_CZZNID = @idZPZ  
			) a where CZZE_ID=Id
			and isnull(CZZE_Typ,2) = 2	

			 FETCH NEXT FROM kursor  INTO @idZPZ
		END
		CLOSE kursor  deallocate kursor
-------
		 DELETE FROM  dbo.CtiDokNrWlasny WHERE DNW_DokTyp = 2 AND DNW_DokID = @id
		 DELETE FROM  dbo.CtiZleceniePrc where  CZO_CZNId = @id
		 DELETE FROM  dbo.CtiZlecenieNag where CZN_ID = @id
		 DELETE FROM  dbo.CtiZalaczniki WHERE CZAL_DokTyp = 1 AND CZAL_DokId = @id


--- załączniki do zasobów na zleceniu
		

		
		
	END TRY
	BEGIN CATCH
        -- Sprawdzenie, czy istnieje aktywna transakcja
        IF @@TRANCOUNT > 0
        BEGIN
            IF @@TRANCOUNT = 1
            BEGIN
                ROLLBACK TRANSACTION;
            END
            ELSE
            BEGIN
                ROLLBACK TRANSACTION @SavePointName;
            END
        END

        -- Przekazanie informacji o błędzie
        DECLARE @ErrorMessage NVARCHAR(4000);
        DECLARE @ErrorSeverity INT;
        DECLARE @ErrorState INT;

        SELECT 
            @ErrorMessage = ERROR_MESSAGE(),
            @ErrorSeverity = ERROR_SEVERITY(),
            @ErrorState = ERROR_STATE();

        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH
END
ELSE
BEGIN
	RAISERROR('Status zlecenia  nie pozwala na jego usunięcie.', 16, 1);
END
END

GO
```

# [dbo].[CTIProdukcjaZlecenieList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieList]
 @dataOd		varchar(20),
 @dataDo		varchar(20),
 @prmData		int,			--0 bez --1 ograniczenie po dacie
 @prmZamkniete	int,			--plus zamkniďż˝te
 @prmStatusy	int = 1,			--20170824
 @prmNiepotwierdzone int = 1,
 @prmDoRealizacji int = 0
AS
BEGIN
declare @DokTyp int = 2 -- ZP
declare @prmPole1 int = 0 SELECT @prmPole1 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 1 and CAU_TypDok = @DokTyp
declare @prmPole2 int = 0 SELECT @prmPole2 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 2 and CAU_TypDok = @DokTyp
declare @prmPole3 int = 0 SELECT @prmPole3 = CAU_Aktywny FROM  dbo.CtiAtrybutyUprawnienia where cau_canlp = 3 and CAU_TypDok = @DokTyp
declare @obslugaWydania	int = 0 select @obslugaWydania = isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where knf_numer = 62
declare @obslugaKontroli int = 0 select @obslugaKontroli = isnull(KNF_Wartosc,0) from dbo.CtiKonfiguracja where knf_numer = 63
declare @przekraczanieRealizacji int = 0 select @przekraczanieRealizacji = case when KNF_Wartosc IN ('True', 'TRUE', '1') then 1 else 0 end from dbo.CtiKonfiguracja where knf_numer = 1016

SELECT a.* FROM
(
		SELECT  CZN_ID AS ID, dbo.CTIProdukcjaZlecenieNumer(CZN_ID) AS CZN_Numer, isnull(CZN_CTNId,0) as CZN_CTNId
			, isnull(CZN_TwrId,0) as CZN_TwrId, isnull(Twr_Kod,'') AS CZN_TwrKod, ISNULL(CZN_Ilosc, 0) AS Ilosc
			, isnull(CZN_OpeWystaw,0) as CZN_OpeWystaw,isnull(CZN_OperModyf,0) as CZN_OperModyf
			, ISNULL(CZN_DataWystaw, GETDATE()) AS CZN_DataWystaw ,ISNULL(CZN_DataModyf, GETDATE()) AS CZN_DataModyf
			, ISNULL(a.CO_Kod,'') AS Wystawil, ISNULL(b.CO_Kod,'') AS Modyfikowal 
			, isnull(CTN_Nazwa,'') AS CTN_Nazwa,ISNULL(CZN_Opis, '') AS CZN_Opis 
			, ISNULL(CtiTechnolNag.CTN_Cecha, 0) AS ctn_cecha
			, CZN_Status as CZN_Status1, case  when CZN_Status in (0,3,4) then CZN_Status else dbo.CTIProdukcjaStatusy(2, CZN_ID) end AS CZN_Status
			--statusy
			, case when @prmStatusy = 1 then 
				case @obslugaWydania when 1 then dbo.CTIProdukcjaStatusyNazwy(23, (case  when CTN_Status in(0,3) then CZN_Status else dbo.CTIProdukcjaStatusy(23, CZN_ID) end )) else '' end 
				else '' end AS Status1
			, case when @prmStatusy = 1 then 
			dbo.CTIProdukcjaStatusyNazwy(24, (case  when CTN_Status in (0,3,4) then CZN_Status else dbo.CTIProdukcjaStatusy(24, CZN_ID) end)) 
			else '' end AS Status2
			, case when @prmStatusy = 1 then case @obslugaKontroli when 1 then dbo.CTIProdukcjaStatusyNazwy(25,(case  when CTN_Status in(0,3) then CZN_Status else dbo.CTIProdukcjaStatusy(26, CZN_ID) end ) ) else '' end 
			else '' end AS Status3
			--pola specjalne
			, case @prmPole1 when 1 then (SELECT TOP (1) Wartosc FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0, 0) WHERE (Lp = 1)) else '' end AS Pole1
			, case @prmPole2 when 1 then (SELECT TOP (1) Wartosc FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0, 0) WHERE (Lp = 2))  else '' end AS Pole2
			, case @prmPole3 when 1 then (SELECT TOP (1) Wartosc FROM dbo.CTIProdukcjaTowaryPolaSpecjalneNew(CtiZlecenieNag.CZN_CTNId, CtiZlecenieNag.CZN_ID, 0, 0, 0, 0, 0) WHERE (Lp = 3))  else '' end AS Pole3
			, isnull((select top 1 czze_czznid from dbo.CtiZlecenieZElem where czze_cznid = CZN_ID and isnull(CZZE_Typ,2)=2),0) as czznid
			, ISNULL(CZN_CzasStart, GETDATE()) AS CzasStart, ISNULL(CZN_CzasEnd, GETDATE()) AS CzasEnd
			, replace(Twr_Nazwa,char(10), ' ')			as TwrNazwa
			, ISNULL(NrZp.DNW_Numer,CZN_NrPelny)		as NumerDokWlasny
			, ISNULL(NrTH.DNW_Numer,CTN_Nazwa)			as NumerDokWlasnyTH
			, case isnull(CZN_DokZwiazTyp,0) when 0 then '' when 9 then isnull((select TOP 1 CBN_NrPelny from dbo.CtiBrakiElem INNER JOIN  DBO.CtiBrakiNag ON CBE_CBNid = CBN_ID where  CBE_CZNID = czn_id and isnull(CBE_Typ,2) = 2),'') --BRK
			  else isnull((select TOP 1 trn_numerpelny from cdn.TraElem as tre INNER JOIN cdn.TraNag	ON tre.TrE_TrNId = TrN_TrNID WHERE isnull(CZN_DokZwiazTyp,0) >0 and tre.TrE_TypDokumentu = CZN_DokZwiazTyp and tre.TrE_TrNId = CZN_DokZwiazID and tre.TrE_TrEID = CZN_DokZwiazTreID ),'')--ro
			  end as DokPowz
			, dbo.CTIProdukcjaZlecenieIloscZrealizowana(CZN_ID, 0, 0) AS IloscZrealizowana
			, ISNULL((SELECT TOP 1 TrN_NumerObcy FROM CDN.TraNag WHERE TrN_TrNID = CZN_DokZwiazID AND TrN_TypDokumentu = 308), '') AS NumerObcyZamowienia
		FROM dbo.CtiZlecenieNag 
			LEFT JOIN dbo.CtiOperatorzy AS a		ON ISNULL(CZN_OpeWystaw,0) = a.CO_ID 
			LEFT JOIN dbo.CtiOperatorzy AS b		ON ISNULL(CZN_OperModyf,0) = b.CO_ID 
			LEFT JOIN dbo.CtiTechnolNag				ON ISNULL(CZN_CTNId,0) = CTN_ID 
			LEFT JOIN CDN.Towary					ON ISNULL(CZN_TwrId,0) = Twr_TwrId
			LEFT JOIN dbo.CtiDokNrWlasny as NrZp	ON CZN_ID = isnull(NrZp.DNW_DokID,0) and  2 = isnull(NrZp.DNW_DokTyp,0) 
			LEFT JOIN dbo.CtiDokNrWlasny as NrTH	ON ISNULL(CZN_CTNID,0) = isnull(NrTH.DNW_DokID,0) and  1 = isnull(NrTH.DNW_DokTyp,0) 
		where
			(
				((isnull(@prmData,0) = 0) or (@prmData = 1 and CONVERT(VARCHAR, ISNULL(CZN_DataWystaw, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)))
				and
				((@prmZamkniete = 1) or (isnull(@prmZamkniete,0) = 0 and ((case  when CTN_Status in(0,3) then CZN_Status else dbo.CTIProdukcjaStatusy(2, CZN_ID) end) < 3)))
				and
				((@prmNiepotwierdzone = 1) or ((isnull(@prmNiepotwierdzone,0)) = 0) and (czn_status > 0))
			)
) a
WHERE
    ((@prmDoRealizacji = 0) or (@przekraczanieRealizacji = 1) or (Ilosc > IloscZrealizowana))
ORDER BY 1 DESC
    
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieListShort]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieListShort]
 @typ int,--24 PW -- 26 Kontrola
 @dataOd varchar(20),
 @dataDo varchar(20),
 @prmWszystkie int = 1,
 @prmDoRealizacji int = 0
AS
BEGIN

	declare @przekraczanieRealizacji int = 0 select @przekraczanieRealizacji = case when KNF_Wartosc IN ('True', 'TRUE', '1') then 1 else 0 end from dbo.CtiKonfiguracja where knf_numer = 1016

	SELECT 
	a.ID,
	a.NrPelny,
	a.KodWyrobu
	FROM
	(
		SELECT 
		CZN_ID as ID,  
		dbo.CTIProdukcjaZlecenieNumer(CZN_ID) as NrPelny, 
		CZN_TwrKod as KodWyrobu,
		CZN_Ilosc AS Ilosc, 
		dbo.CTIProdukcjaZlecenieIloscZrealizowana(CZN_ID, 0, 0) AS IloscZrealizowana

		FROM dbo.CtiZlecenieNag
			--LEFT JOIN dbo.CtiDokNrWlasny on CZN_ID = isnull(DNW_DokID,0) and  2 = isnull(DNW_DokTyp,0) 
		where CZN_Status in (1,2) 
			and CONVERT(VARCHAR, ISNULL(CZN_DataWystaw, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)
			--and dbo.CTIProdukcjaStatusy(2, CZN_ID) in (1,2) -- czy zlecenie nie zamknięte i nie w buforze
			--and dbo.CTIProdukcjaStatusy(@typ, CZN_ID) <> 2 -- nie wykorzystano
	) a
	WHERE 
		((@prmDoRealizacji = 0) or (@przekraczanieRealizacji = 1) or (Ilosc > IloscZrealizowana))
	ORDER BY a.ID DESC

END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieListShortKontrola]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaZlecenieListShortKontrola]
@dataOd VARCHAR(20),
@dataDo VARCHAR(20)
AS
BEGIN
	SELECT 
		a.ID,
		a.NrPelny,
		a.KodWyrobu
		FROM
		(
			SELECT 
				CZN_ID AS ID,  
				dbo.CTIProdukcjaZlecenieNumer(CZN_ID) AS NrPelny, 
				CZN_TwrKod AS KodWyrobu,
				CZN_Ilosc AS Ilosc, 
				dbo.CTIProdukcjaZlecenieIloscZrealizowana(CZN_ID, 0, 0) AS IloscZrealizowana
			FROM dbo.CtiZlecenieNag
			WHERE CZN_Status in (1,2) 
				AND CONVERT(VARCHAR, ISNULL(CZN_DataWystaw, GETDATE()), 23)  
					BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)
				AND dbo.CTIProdukcjAStatusy(24, CZN_ID) in (1,2) --produkcja (zrealizowano czesciowo/zrealizowano)
				AND dbo.CTIProdukcjAStatusy(26, CZN_ID) <> 2 -- kontrola (nie skontrolowano wszystkiego)
		) a
		ORDER BY a.ID DESC
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZleceniePowiazaneDokDost]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create PROCEDURE [dbo].[CTIProdukcjaZleceniePowiazaneDokDost]
 @CZNId int,@dokTyp int --312 MM -- 304 RW -- 303 PW
AS
BEGIN

	if(@dokTyp =306 ) --WZ
	begin

		select TraNagG.TrN_trnid as DokId, TraNagG.TrN_typdokumentu as DokTyp, TraNagG.TrN_NumerPelny AS DokNumerPelny --dane dokumentu głównego
			, TraElemG.TrE_TwrId as TwrId, TraElemG.TrE_TwrKod as TwrKod,TraElemG.TrE_JM as JM --dane towaru
			, case when ISNULL(TsD_TrSIdDost,0) = 0 then ISNULL(TraSElemG.TrS_trsiddost,0) else ISNULL(TsD_TrSIdDost,0) end as DostId--id dostawy --NULL - dla dokumentów nie generujących dostaw
			, case when ISNULL(TsD_TrSIdDost,0) = 0 then ISNULL(TraNagDST.TrN_NumerPelny,'') else ISNULL(TsD_NumerPelny,'') end as DostNumerPelny--numer dokumentu
			, ISNULL(TraNagDST.TrN_TrNID,0) as DostDokID
			, cast(COALESCE(TsD_Ilosc,TraSElemG.TrS_ilosc,TraElemG.TrE_ilosc,0) as decimal(15,4)) as DostIlosc --ilość
			, COALESCE(TsD_wartosc,traSElemG.TrS_wartosc,TraElemG.TrE_WartoscNetto,0) as Cena --cena
			, ISNULL(TsD_Cecha1_DeAId,0) as Cecha1DeAId, ISNULL(TsD_Cecha1_Kod,'') as Cecha1Kod ,ISNULL(TsD_Cecha1_Wartosc,'') as Cecha1Wartosc--cecha
			, ISNULL(KNTDost.knt_kod,'') as DostKntKod --Kontrahent dostawy
			, ISNULL(TraNagDST.trn_dataOpe,'') as DostData --Data dostawy
			, ISNULL(KnTG.knt_kod,'') as DokKntKod --Kontrahent dok
			, ISNULL(TraNagG.trn_dataOpe,'') as DostDok --Data dok
		from cdn.TraNag as TraNagG
			INNER JOIN cdn.TraElem as TraElemG on TraElemG.TrE_trnid = TraNagG.TrN_trnid
			--jest cecha
			LEFT JOIN cdn.TraSElemDost on TraElemG.TrE_TrEID = ISNULL(TsD_TrEID,0)
			LEFT JOIN cdn.TraSElem as TraSElemG on ISNULL(TraElemG.TrE_TrEId,0)  = ISNULL(TraSElemG.TrS_TrEId,0)
			--nie ma cechy
			LEFT JOIN  cdn.TraSElem as TraSElemDST ON ISNULL(TraSElemDST.trs_trsid,0) = ISNULL(TraSElemG.TrS_trsiddost,0)
			LEFT JOIN  cdn.TraElem as TraElemDST ON  ISNULL(TraElemDST.TrE_TrEId,0) = ISNULL(TraSElemDST.TrS_TrEId,0)
			LEFT JOIN  cdn.TraNag as TraNagDST ON ISNULL(TraNagDST.TrN_TrNId,0) = ISNULL(TraElemDST.TrE_TrNId,0) 
			--dokumenty które powstały z produkcji
			INNER JOIN dbo.CtiZlecenieDok ON CZD_TrnId = TraNagDST.TrN_trnid  
			--kontrahent
			LEFT JOIN cdn.kontrahenci as KnTDost on  ISNULL(TraNagDST.TrN_PodID,0) = ISNULL(KNTDost.knt_kntid,0)
			LEFT JOIN cdn.kontrahenci as KnTG on  ISNULL(TraNagG.TrN_PodID,0) = ISNULL(KNTG.knt_kntid,0)
		where-- TraNagG.TrN_typdokumentu <> 312 -- bez MM 
			TraNagG.TrN_typdokumentu in( @dokTyp,304)
			AND @CZNId = CZD_CZNId
		group by TraNagG.TrN_trnid, TraNagG.TrN_typdokumentu, TraNagG.TrN_NumerPelny--dane dokumentu głównego
			, TraElemG.TrE_TwrId,TraElemG.TrE_TwrKod,TraElemG.TrE_JM,TraElemG.TrE_ilosc--dane towaru
			, ISNULL(TsD_TrSIdDost,0),ISNULL(TraSElemG.TrS_trsiddost,0)--id dostawy
			, ISNULL(TsD_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,ISNULL(TsD_NumerPelny,'')) --numer dokumentu
			, COALESCE(TsD_Ilosc,TraSElemG.TrS_ilosc,TraElemG.TrE_ilosc,0)--ilość
			, COALESCE(TsD_wartosc,traSElemG.TrS_wartosc,TraElemG.TrE_WartoscNetto,0)--cena
			, ISNULL(TsD_Cecha1_DeAId,0), ISNULL(TsD_Cecha1_Kod,'') ,ISNULL(TsD_Cecha1_Wartosc,'') --cecha
			, ISNULL(KNTDost.knt_kod,'') --Kontrahent dostawy
			, ISNULL(TraNagDST.trn_dataOpe,'') --Data dostawy
			, ISNULL(KnTG.knt_kod,'') --Kontrahent dok
			, ISNULL(TraNagG.trn_dataOpe,'')--Data dok
			, ISNULL(TraNagDST.TrN_TrNID,0)
	end
	else
	begin
		select TraNagG.TrN_trnid as DokId, TraNagG.TrN_typdokumentu as DokTyp, TraNagG.TrN_NumerPelny AS DokNumerPelny --dane dokumentu głównego
			, TraElemG.TrE_TwrId as TwrId, TraElemG.TrE_TwrKod as TwrKod,TraElemG.TrE_JM as JM --dane towaru
			, case when ISNULL(TsD_TrSIdDost,0) = 0 then ISNULL(TraSElemG.TrS_trsiddost,0) else ISNULL(TsD_TrSIdDost,0) end as DostId--id dostawy --NULL - dla dokumentów nie generujących dostaw
			, ISNULL(TraNagDST.TrN_NumerPelny,'') as DostNumerPelny--numer dokumentu
			, ISNULL(TraNagDST.TrN_TrNID,0) as DostDokID
			, cast(COALESCE(TsD_Ilosc,TraSElemG.TrS_ilosc,TraElemG.TrE_ilosc,0) as decimal(15,4)) as DostIlosc --ilość
			, COALESCE(TsD_wartosc,TraSElemG.TrS_wartosc,TraElemG.TrE_WartoscNetto) as Cena --cena
			, ISNULL(TsD_Cecha1_DeAId,0) as Cecha1DeAId, ISNULL(TsD_Cecha1_Kod,'') as Cecha1Kod ,ISNULL(TsD_Cecha1_Wartosc,'') as Cecha1Wartosc--cecha
			, ISNULL(KNTDost.knt_kod,'') as DostKntKod --Kontrahent dostawy
			, ISNULL(TraNagDST.trn_dataOpe,'') as DostData --Data dostawy
			, ISNULL(KnTG.knt_kod,'') as DokKntKod --Kontrahent dok
			, ISNULL(TraNagG.trn_dataOpe,'') as DostDok --Data dok
			
		from cdn.TraNag as TraNagG
			INNER JOIN cdn.TraElem as TraElemG on TraElemG.TrE_trnid = TraNagG.TrN_trnid
			--jest cecha
			LEFT JOIN cdn.TraSElem as TraSElemG on ISNULL(TraElemG.TrE_TrEId,0)  = ISNULL(TraSElemG.TrS_TrEId,0)
			--nie ma cechy
			LEFT JOIN  cdn.TraSElem as TraSElemDST ON ISNULL(TraSElemDST.trs_trsid,0) = ISNULL(TraSElemG.TrS_trsiddost,0)
			LEFT JOIN cdn.TraSElemDost on TraSElemDST.TrS_TrSID = ISNULL(TsD_TrSID,0) 

			LEFT JOIN  cdn.TraElem as TraElemDST ON  ISNULL(TraElemDST.TrE_TrEId,0) = ISNULL(TraSElemDST.TrS_TrEId,0)
			LEFT JOIN  cdn.TraNag as TraNagDST ON ISNULL(TraNagDST.TrN_TrNId,0) = ISNULL(TraElemDST.TrE_TrNId,0) 
			--dokumenty które powstały z produkcji
			INNER JOIN dbo.CtiZlecenieDok ON CZD_TrnId = TraNagG.TrN_trnid  
			--kontrahent
			LEFT JOIN cdn.kontrahenci as KnTDost on  ISNULL(TraNagDST.TrN_PodID,0) = ISNULL(KNTDost.knt_kntid,0)
			LEFT JOIN cdn.kontrahenci as KnTG on  ISNULL(TraNagG.TrN_PodID,0) = ISNULL(KNTG.knt_kntid,0)
		where
			TraNagG.TrN_typdokumentu = @dokTyp
			AND CZD_CZNId = @CZNId
		group by TraNagG.TrN_trnid, TraNagG.TrN_typdokumentu, TraNagG.TrN_NumerPelny--dane dokumentu głównego
			, TraElemG.TrE_TwrId,TraElemG.TrE_TwrKod,TraElemG.TrE_JM,TraElemG.TrE_ilosc--dane towaru
			, ISNULL(TsD_TrSIdDost,0),ISNULL(TraSElemG.TrS_trsiddost,0),ISNULL(TsD_TrSIdDost,ISNULL(TraSElemG.TrS_trsiddost,0))--id dostawy
			, ISNULL(TsD_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,''),ISNULL(TraNagDST.TrN_NumerPelny,ISNULL(TsD_NumerPelny,'')) --numer dokumentu
			, COALESCE(TsD_Ilosc,TraSElemG.TrS_ilosc,TraElemG.TrE_ilosc,0)--ilość
			, COALESCE(TsD_wartosc,TraSElemG.TrS_wartosc,TraElemG.TrE_WartoscNetto)--cena
			, ISNULL(TsD_Cecha1_DeAId,0), ISNULL(TsD_Cecha1_Kod,'') ,ISNULL(TsD_Cecha1_Wartosc,'') --cecha
			, ISNULL(KNTDost.knt_kod,'') --Kontrahent dostawy
			, ISNULL(TraNagDST.trn_dataOpe,'') --Data dostawy
			, ISNULL(KnTG.knt_kod,'') --Kontrahent dok
			, ISNULL(TraNagG.trn_dataOpe,'')--Data dok
			, ISNULL(TraNagDST.TrN_TrNID,0)
			, ISNULL(TraNagDST.Trn_numerpelny, '')
	end
END
GO
```

# [dbo].[CTIProdukcjaZleceniePRC]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZleceniePRC]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZleceniePRC]
 @ID int,
 @PrcId int
AS
BEGIN

declare @MagazynDomyslySur varchar(40) = '' 
declare @MagazynDomyslyProd varchar(40) = '' 
select @MagazynDomyslySur = mag_symbol from cdn.magazyny INNER JOIN CtiKonfiguracja on knf_numer = 8 and knf_wartosc = mag_magid 
select @MagazynDomyslyProd = mag_symbol from cdn.magazyny INNER JOIN CtiKonfiguracja on knf_numer = 9 and knf_wartosc = mag_magid 
declare @typPrzypisania int = 0
select @typPrzypisania = cast(knf_wartosc as int) from dbo.ctikonfiguracja where knf_numer = 61

	SELECT ID, Status, CZOID, CZNId, PrcId , RCP, PrcKod, ilosc, StatusNazwa, StatusNazwa2, StatusNazwa3
		, case isnull(MagSur,0) when 0 then @MagazynDomyslySur else 
			(select Mag_Symbol from cdn.magazyny where Mag_MagId = MagSur  ) end As MagSur
		, case isnull(MagProd,0) when 0 then @MagazynDomyslyProd  else 
			(select Mag_Symbol from cdn.magazyny where Mag_MagId = MagProd  ) end As MagProd
		, Status1, Status2, Status3
	from
	(
		select CZO_CZNId as ID
			, dbo.CTIProdukcjaStatusy(2, CZO_CZNId) AS Status
			, CZO_ID as CZOID, CZO_CZNId as CZNId, CZO_PrcId as PrcId 
			, '' as RCP 
			, CZO_PrcKod as PrcKod, CZO_Ilosc as ilosc
			, isnull(dbo.CTIProdukcjaStatusyNazwy(23, dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,3)),'') AS StatusNazwa 
			, isnull(dbo.CTIProdukcjaStatusyNazwy(24, dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,4)),'') AS StatusNazwa2 
			, isnull(dbo.CTIProdukcjaStatusyNazwy(25, dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,6)),'') AS StatusNazwa3
			, isnull(dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,3),0) AS Status1
			, isnull(dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,4),0) AS Status2
			, isnull(dbo.CTIProdukcjaStatusPracownika(CZO_CZNId, CZO_PrcId,6),0) AS Status3
			, isnull(KnM_MagSId,CDZ_MagSId) As MagSur
			, isnull(KnM_MagPId,CDZ_MagPId) As MagProd
		FROM dbo.CtiZleceniePrc
			LEFT JOIN dbo.CtiKonfigMag ON isnull(CZO_PrcTyp,80) = 80 and CZO_PrcId = KnM_PraId
			LEFT JOIN dbo.CtiDzial on isnull(CZO_PrcTyp,80) = 81 and CDZ_ID = czo_prcid
		where ((@ID = 0 and 1=1)or(@ID <> 0 and CZO_CZNId = @ID ) )
			and  ((@PrcId = 0 and 1=1)or(@PrcId <> 0 and CZO_PrcId = @PrcId ))
			and (
				(@typPrzypisania IN (2,3) and isnull(CZO_PrcTyp, 80) in (80,90)) or
				(@typPrzypisania = 4 and isnull(CZO_PrcTyp, 80) = 81)
			)
		
	)a ORDER BY PrcKod


end
GO
```

# [dbo].[CTIProdukcjaZlecenieWydruk]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieWydruk]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieWydruk]
@id int
AS
BEGIN

declare @EUR DECIMAL(18,4)
declare @USD DECIMAL(18,4)
--kurs waluty
	declare @Dzien NVARCHAR(10) = CONVERT(VARCHAR, GETDATE(), 23)
	CREATE TABLE #TmpM (Notowanie DECIMAL(18,4),WNa_symbol varchar(5) )
	insert into #TmpM (Notowanie,WNa_symbol)
	exec CDN.TabelkaWalut @Dzien
	select @EUR = Notowanie from #TmpM where WNa_symbol = 'EUR'
	select @USD = Notowanie from #TmpM where WNa_symbol = 'USD'
	drop TABLE #TmpM



 SELECT CTE_ID,
 CASE TwC_Waluta  WHEN 'EUR' THEN TwC_Wartosc * @EUR / 1 WHEN 'PLN' 
 THEN TwC_Wartosc * 1 / 1 WHEN 'USD' THEN TwC_Wartosc * @USD / 1 END AS TwC_Wartosc
 FROM CDN.TwrCeny  
 INNER JOIN CDN.Towary  ON TwC_TwrID=Twr_TwrId
 INNER JOIN dbo.CtiTechnolElem ON Twr_TwrId=CTE_TwrId
 INNER JOIN dbo.CtiTechnolNag ON CTE_CTNId=CTN_ID
 
 WHERE  CTN_ID=@id AND 
 TwrCeny.TwC_TwCNumer=1
 ORDER BY CTE_Typ, CTE_Lp
 
end
GO
```

# [dbo].[CTIProdukcjaZlecenieWyrobNaMagazynie]

# [dbo].[CTIProdukcjaZlecenieZasobyAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZasobyAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieZasobyAdd]
 @ZZsID int, --jeżeli edycja
 @CZNID int,--Id zlecenia
 @CZID int, --ID zasobu
 @datOd varchar(25),
 @datDo varchar(25),
 @czasMin decimal(15,4), --czas w minutach
 @prcId int, --pracownik
 @ilosc decimal(15,4), 
 @opeId int, --operator
 @Opis  nvarchar(250)
AS
BEGIN
declare  @prcKod nvarchar(40) 
select @prcKod = isnull(PRA_Kod,'') from CDN.PracKod where pra_praid =  isnull(@prcId,0) 
 
declare @dataOd datetime set @dataOd = convert(datetime, @datOd,121)
declare @dataDo datetime set @dataDo= convert(datetime, @datDo,121)

	if @ZZsID = 0 
	begin

		insert into CtiZlecenieZasob  (ZZs_CZNID, ZZs_CZID, ZZs_CTNID, ZZs_DataOd, ZZs_DataDo, ZZs_CzasMin, ZZs_PrcID, ZZs_PrcKod
		, ZZs_OpeID, ZZs_Data, ZZs_Ilosc, ZZs_Opis, ZZs_DokTyp)
		select @CZNID, @CZID,CZN_CTNId, @DataOd, @DataDo, @czasMin , @prcId, @prcKod
		, @opeId, getdate(),@ilosc,@Opis, 2
		from CtiZlecenieNag where CZN_ID = @CZNID
		select @ZZsID = SCOPE_IDENTITY()

	end  
	else--update
	begin

			insert into CtiZlecenieZasobHist 
				(ZsH_ZZsID, ZsH_Data, ZsH_CZNID, ZsH_CZID, ZsH_OpeId
				, ZsH_PrcId, ZsH_DataOd, ZsH_DataDo, ZsH_CzasMin, ZsH_Ilosc)
			select ZZs_ID, ZZs_Data, ZZs_CZNID, ZZs_CZID, ZZs_OpeID
				, ZZs_PrcID, ZZs_DataOd, ZZs_DataDo, ZZs_CzasMin, ZZs_Ilosc
			from CtiZlecenieZasob where ZZs_ID = @ZZsID 

		update CtiZlecenieZasob set 
			ZZs_CZID = @CZID
			, ZZs_DataOd = @DataOd
			, ZZs_DataDo = @DataDo
			, ZZs_CzasMin = @czasMin
			, ZZs_PrcId = @prcId
			, ZZs_PrcKod  = @prcKod
			, ZZs_Ilosc =  @ilosc
			, ZZs_OpeId = @opeId
			, ZZs_Data = getdate()
			, ZZs_Opis =@Opis
		where ZZs_ID = @ZZsID
	end  

	select @ZZsID as Id
END

GO
```

# [dbo].[CTIProdukcjaZlecenieZasobyList]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZasobyList]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieZasobyList]
 @CZN_ID int,--342
 @ZZs_ID int --pozycja
AS
BEGIN
	select *,
	--isnull(convert(varchar, cast(((WyliczonyCzasMin/60))as int))+' Godz '+convert(varchar, cast((isnull(WyliczonyCzasMin,0)%60) as int))+' Min '
	--	  +convert(varchar, cast(((isnull(WyliczonyCzasMin,0)*60)%60) as int)) +'s' ,'')  as CzasTekst

	--CONVERT(varchar, RazemGodzin) + ' Godz ' +
	--CONVERT(varchar, RazemMinut - (60*RazemGodzin)) + ' Min ' +
	--CONVERT(varchar, RazemSekund - (60*RazemMinut)) + ' s' AS CzasTekst

	--CASE WHEN (RazemSekund >= 3600)
	--THEN RazemSekund 

	CONVERT(varchar, RazemSekund / 3600) + ' Godz ' +
	CONVERT(varchar, RazemSekund / 60 % 60) + ' Min ' +
	CONVERT(varchar, RazemSekund % 60) + ' s' AS CzasTekst
	 from
	(
		select cz_kod as ZasobKod, CZG_Kod as GrupaKod, dbo.CTIProdukcjaZlecenieNumer(CZN_ID) as ZlecenieNumer 
		, ZZs_ID as ID, ZZs_CZNID as CZNID, ZZs_CZID as CZID, ZZs_CTNID as CTNID,isnull(ZZs_PrcId,0) as PrcId
		, case when CONVERT(VARCHAR,isnull(ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then '' else CONVERT(VARCHAR,isnull(ZZs_DataOd,'') , 120) end  as DataOd
		 , case when CONVERT(VARCHAR,isnull(ZZs_DataDo,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then '' else CONVERT(VARCHAR,isnull(ZZs_DataDo,'') , 120) end  as DataDo
		, isnull(ZZs_CzasMin,0) as CzasMin  
	--	,isnull(DATEDIFF(minute, ZZs_DataOd,ZZs_DataDo) ,0) as WyliczonyCzasMin
		, isnull(CZ_KontrolaDostepnosci,0) as KontrolaDostepnosci
		--,  isnull(CDN.PracKartyRcp.PKR_Numer, '') as RCP
		, isnull(isnull(PRA_Kod,ZZs_PrcKod),'') as PrcKod
		, case 
		 when CONVERT(VARCHAR,isnull(ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then isnull(ZZs_CzasMin,0)
		 when CONVERT(VARCHAR,isnull(ZZs_DataDo,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then 
		 CONVERT(decimal(15,4),DATEDIFF(second, ZZs_DataOd,GETDATE())/60.0000)
		  else   CONVERT(decimal(15,4),(isnull(DATEDIFF(SECOND, ZZs_DataOd,ZZs_DataDo) ,0))/60.0000)  end as WyliczonyCzasMin
		,isnull(ZZs_Ilosc,0) as Ilosc, isnull(ZZs_Opis,'') as Opis,
		DATEDIFF(second, isnull(ZZs_DataOd,'1900-01-01 00:00:00'), isnull(ZZs_DataDo,GETDATE())) AS RazemSekund
	from dbo.CtiZasob 
			INNER JOIN dbo.CtiZasobGrupy on cz_czgid = CZG_ID
			INNER JOIN dbo.CtiZlecenieZasob on ZZs_CZID = CZ_ID
			INNER JOIN dbo.CtiZlecenieNag on ZZs_CZNID = CZN_ID
			--LEFT JOIN CDN.PracKartyRcp ON isnull(PKR_PrcId,0) = isnull(ZZs_PrcId,0)
			LEFT JOIN CDN.PracKod on pra_praid =  isnull(ZZs_PrcId,0) 
	where ZZs_CZNID = @CZN_ID
	and ZZs_DokTyp = 2
			and ((isnull(@ZZs_ID,0)=0)or (isnull(@ZZs_ID,0)<>0 and ZZs_ID = isnull(@ZZs_ID,0)))
	)a
END

GO
```

# [dbo].[CTIProdukcjaZlecenieZasobyMaszynaAdd]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZasobyMaszynaAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaZlecenieZasobyMaszynaAdd] 
 @CZNID int,--Id zlecenia
 @MARID int,--Id reazlizacji
 @czas int, --czas w sekundach
 @ilosc decimal(15,4), -- ilosc
 @opeId int --operator
AS

--ID zasobu
Declare @ID_Zasobu int
set @ID_Zasobu = 1

--ID pracownika z zasobu
declare @ID_PRAC int
set @ID_PRAC = coalesce((select top 1 zsp_prcid from  dbo.CtiZasobPRC where zsp_czid = @ID_Zasobu ),0)

--ustawianie dat na podstawie podanych sekund oraz daty rozpoczecia z zasobów ZP
declare @dataOd datetime 
set @dataOd = coalesce((select CZS_CzasStart from [dbo].[CtiZasobDok] join CtiZlecenieNag on CZN_ID=CZS_CTNID where CZS_CTNID = @CZNID and CZS_DokTyp = 2 and CZS_CZID=@ID_Zasobu),GETDATE())
declare @dataDo datetime set @dataDo= convert(datetime, DATEADD(ss,@czas,@dataOd),121)

-- konwersja sekund do minut
declare @czasx int 
set @czasx = ROUND(@czas/60,0)

--dodawanie wpisu do zealizacji ZP
exec dbo.CTIProdukcjaZlecenieZasobyAdd 0,@CZNID,@ID_Zasobu,@dataOd,@dataDo,@czasx,1,1,@opeId,'Zaimportowano z maszyny'

--ustawianie realizacji z maszyn jako zaimportowanych
update dbo.CtiMaszynaRealizacja set mar_status = 1 where mar_id = @MARID
GO
```

# [dbo].[CTIProdukcjaZlecenieZDel]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZDel]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[CTIProdukcjaZlecenieZDel]
@id INT,
@walidacjAStatusu INT = 1
AS
BEGIN
DECLARE @status INT
DECLARE @tempid  INT
DECLARE @templp INT
SELECT @status =dbo.CTIProdukcjAStatusy(10,CZZN_ID)  FROM  dbo.CtiZlecenieZNag WHERE CZZN_ID =@id
IF @status=0 OR @walidacjAStatusu = 0
BEGIN
DECLARE @SavePointName NVARCHAR(MAX) = 'SavePoint';

	BEGIN TRY
        -- Sprawdzenie, czy istnieje aktywna transakcja
        IF @@TRANCOUNT > 0
        BEGIN
            SAVE TRANSACTION @SavePointName;
        END
		DECLARE kursor CURSOR FOR
				SELECT DISTINCT CZZE_CZZNID FROM dbo.CtiZlecenieZElem WHERE ISNULL(CZZE_Typ,2) = 22	
		DECLARE @idZPZ INT OPEN kursor
		FETCH NEXT FROM kursor INTO @idZPZ
		WHILE @@FETCH_STATUS = 0
		BEGIN	
			DELETE FROM dbo.CtiZlecenieZElem WHERE CZZE_CZZNID = @idZPZ AND CZZE_CZNID = @id
			UPDATE dbo.CtiZlecenieZElem SET CZZE_Lp = lp FROM
			(	
				SELECT ROW_NUMBER() OVER (PARTITION BY CZZE_CZZNID ORDER BY  ISNULL(CZZE_Lp,0),CZZE_ID) AS Lp, CZZE_ID AS Id FROM  dbo.CtiZlecenieZElem  WHERE CZZE_CZZNID = @idZPZ	AND ISNULL(CZZE_Typ,2) = 22	
			) a 
			WHERE CZZE_ID = Id
			FETCH NEXT FROM kursor INTO @idZPZ
		END
		CLOSE kursor DEALLOCATE kursor

        --Usuwanie powiązanych pól dodatkowych
        DELETE FROM CtiAtrybutyElem WHERE CAE_TypDok = 23 AND CAE_IdDok = @id

        --Usuwanie powiązań na ZP i ZPZ
        UPDATE CtiZlecenieNag SET CZN_CZZNIdNadrzedne = null, CZN_CZZNIdGlowne = null, CZN_CZZNNrPelnyGlowne = null, CZN_CZZNNrPelnyNadrzedne = null WHERE CZN_CZZNIdNadrzedne = @id OR CZN_CZZNIdGlowne = @id
        UPDATE CtiZlecenieZNag SET CZZN_CZZNIdNadrzedne = null, CZZN_CZZNIdGlowne = null, CZZN_CZZNNrPelnyGlowne = null, CZZN_CZZNNrPelnyNadrzedne = null WHERE CZZN_CZZNIdNadrzedne = @id OR CZZN_CZZNIdGlowne = @id

        --Usuwanie wpisów dot. ZPZ w kolumnach edytowalnych
        DELETE CPKW
        FROM dbo.CtiProdukcjaKolumnyWartosci AS CPKW
        JOIN dbo.CtiProdukcjaKolumny AS CPK ON CPKW.CKW_CPKID = CPK.CPK_ID
        WHERE CPK.CPK_Tabela = 1 AND CPKW.CKW_ObiektID = @id

        --Usuwanie powiązanych zasobów 
        DELETE FROM dbo.CtiZasobDok WHERE CZS_CTNID = @id AND CZS_DokTyp = 4

        --Usuwanie powiązanych załączników
        DELETE FROM CtiZalaczniki WHERE CZAL_DokTyp = 7 AND CZAL_DokID = @id

        --Usuwanie powiązanych cech 
        DELETE FROM dbo.CtiTechnolCecha WHERE CTC_CTNID = @id AND CTC_Typ = 3

        --Usuwanie powiązanych działów
        DELETE FROM dbo.CtiTechnolDzialy WHERE CTDZ_TechnolID = @id AND CTDZ_DokTyp = 4
		
		DELETE FROM dbo.CtiZlecenieZElem WHERE  CZZE_CZZNId = @id
		DELETE FROM dbo.CtiZlecenieZNag WHERE CZZN_ID = @id
		
	END TRY
	BEGIN CATCH
        -- Sprawdzenie, czy istnieje aktywna transakcja
        IF @@TRANCOUNT > 0
        BEGIN
            IF @@TRANCOUNT = 1
            BEGIN
                ROLLBACK TRANSACTION;
            END
            ELSE
            BEGIN
                ROLLBACK TRANSACTION @SavePointName;
            END
        END

        -- Przekazanie informacji o błędzie
        DECLARE @ErrorMessage NVARCHAR(4000);
        DECLARE @ErrorSeverity INT;
        DECLARE @ErrorState INT;

        SELECT 
            @ErrorMessage = ERROR_MESSAGE(),
            @ErrorSeverity = ERROR_SEVERITY(),
            @ErrorState = ERROR_STATE();

        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH
END
ELSE
BEGIN
	RAISERROR('Status zlecenia złożonego nie pozwala na jego usunięcie.', 16, 1);
END
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZElemAdd]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieZElemAdd]
@idZleceniaNad int,
@idZleceniaPow int,
@typZleceniaPow int = 2 --2-ZP --22-ZPZ
AS
BEGIN	
	if(@idZleceniaNad <> @idZleceniaPow and @typZleceniaPow = 22)--ZPZ
	BEGIN	
			declare @ZPZglownyOld int = NULL select @ZPZglownyOld=CZZN_DokNad from dbo.CtiZlecenieZNag  where CZZN_ID = @idZleceniaPow
			if isnull(@ZPZglownyOld,0) > 0 BEGIN RETURN END--nie może być podpięte do innego ZPZ
			declare @ZPZglowny int = NULL select @ZPZglowny=CZZN_DokNad from dbo.CtiZlecenieZNag  where CZZN_ID = @idZleceniaNad
			update dbo.CtiZlecenieZNag set CZZN_DokNad=isnull(@ZPZglowny,@idZleceniaNad) where CZZN_ID = @idZleceniaPow
	END
	
	if(@idZleceniaNad = @idZleceniaPow and @typZleceniaPow = 22) RETURN
	else
	BEGIN
		insert into dbo.CtiZlecenieZElem (CZZE_CZZNID, CZZE_CZNID, CZZE_Typ, CZZE_Lp)
		SELECT CZZN_ID,@idZleceniaPow,@typZleceniaPow,isnull((select Max(isnull(CZZE_Lp,0)) from CtiZlecenieZElem where CZZE_CZZNID = @idZleceniaNad),0)+1
		FROM dbo.CtiZlecenieZNag WHERE CZZN_ID = @idZleceniaNad AND isnull((select top 1 czze_czznid from dbo.CtiZlecenieZElem where czze_cznid = CZZN_ID 
			and isnull(CZZE_Typ,2)=@typZleceniaPow),0) = 0--nie może być podpięte do innego ZPZ
	END	


END
GO
```

# [dbo].[CTIProdukcjaZlecenieZElemGeneruj]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZElemGeneruj]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZlecenieZElemGeneruj]
@idZleceniaNad int,
@LogujacyId int
AS
BEGIN
CREATE TABLE #TmpM (nrWlasny VARCHAR(50))  

		DECLARE kursor cursor FOR
			SELECT   CTZE_CTNID, CTZE_Lp,CZZN_CzasStart, CZZN_CzasEnd, cast( isnull(czzn_ilosc,1)*isnull(ctze_ilosc,1) as decimal(15,4)) as Ilosc
			FROM dbo.CtiTechnolZElem	inner join dbo.CtiZlecenieZNag on CTZE_CTZNID = CZZN_CTZNId
			where CZZN_ID =@idZleceniaNad
		DECLARE @idTechP int, @Lp int,@CzasStart datetime, @CzasEnd datetime,@ilosc decimal(15,4)OPEN kursor
		FETCH kursor INTO @idTechP, @Lp, @CzasStart, @CzasEnd, @ilosc
		while @@FETCH_STATUS = 0
		begin
		
			insert into #TmpM (nrWlasny)        
 			exec dbo.CTIProdukcjaKnfNrWlasny 2,  0
			declare @nrWlasny nvarchar(50) =''  select top 1 @nrWlasny=nrWlasny from #TmpM 
			--dodanie zlecenia prostego
				declare @idZleceniaPow int = 0
				exec dbo.CTIProdukcjaZlecenieAddEdit 0,@ilosc,@LogujacyId,0,@idTechP,'',@CzasStart,@CzasEnd,1,@nrWlasny,@idZleceniaPow OUTPUT 
			--powiązanie z zleceniem złożonym    
				insert into dbo.CtiZlecenieZElem (CZZE_CZZNID, CZZE_CZNID, CZZE_Lp, CZZE_Typ)
				Values(@idZleceniaNad,@idZleceniaPow,@Lp,2) --bez ZPZ

			
	

			FETCH kursor INTO @idTechP, @Lp, @CzasStart, @CzasEnd, @ilosc
		end

		CLOSE kursor deallocate kursor
END
GO
```

# [dbo].[CTIProdukcjaZlecenieZElemRenumeracja]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZlecenieZElemRenumeracja]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CTIProdukcjaZlecenieZElemRenumeracja]
AS
BEGIN
BEGIN TRY
	BEGIN TRAN T1
		DECLARE kursor CURSOR FOR	
				SELECT DISTINCT CZZE_CZZNID FROM dbo.CtiZlecenieZElem WHERE ISNULL(CZZE_Typ,2) = 22	
		DECLARE @idZPZ INT OPEN kursor
		FETCH NEXT FROM kursor INTO @idZPZ
		WHILE @@FETCH_STATUS = 0
		BEGIN	
			UPDATE dbo.CtiZlecenieZElem SET CZZE_Lp = lp FROM
			(	
				SELECT ROW_NUMBER() OVER (PARTITION BY CZZE_CZZNID ORDER BY  ISNULL(CZZE_Lp,0),CZZE_ID) AS Lp, CZZE_ID AS Id FROM  dbo.CtiZlecenieZElem  WHERE CZZE_CZZNID = @idZPZ	AND ISNULL(CZZE_Typ,2) = 22	
			) a 
			WHERE CZZE_ID = Id
			FETCH NEXT FROM kursor INTO @idZPZ
		END
		CLOSE kursor DEALLOCATE kursor
		
		COMMIT TRAN T1
	END TRY
	BEGIN CATCH
	   ROLLBACK TRAN T1 RETURN
	END CATCH
END;
GO
```

# [dbo].[CTIProdukcjaZPDostawy]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZPDostawy]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZPDostawy]
@idZp int
AS
BEGIN
	
	declare @prmPrc int = 0 SELECT  @prmPrc = cast(ISNULL(KNF_Wartosc,0) as int) FROM dbo.CtiKonfiguracja  where KNF_Numer = 61 and ISNULL(KNF_Wartosc,'') <> ''
	if @prmPrc in (2,3,4) begin set @prmPrc = 1 end else begin set @prmPrc = 0 end

	select 	
		TrNId as DokID, trnG.TrN_TypDokumentu as DokTyp
		, case trnG.TrN_Bufor when 1 then 0 when -1 then 4 else 1 end as Status
		, ISNULL(trnG.TrN_NumerPelny,'Dokument usuniety') as DokKod
		, trnG.trn_datawys as Data
		, case  when DokTyp in (3,5) then 'S' when DokTyp in (4,6) then 'P' else '' end as TwrTyp
		, Twr_Kod as TwrKod, replace(Twr_Nazwa,char(10), ' ') as TwrNazwa
		, ISNULL(CZO_PrcKod, '') AS PrcKod --pracownik
		, ISNULL(c.Mag_Symbol, '') AS MagDocKod 
		, ISNULL(d.Mag_Symbol, '') AS MagZrdKod
		, DstId
		, ISNULL(DokDst.TrN_NumerPelny,'') AS DstDokKod
		, DokDst.trn_datawys as DstData
		, ISNULL(a.CO_Kod, '') AS OpeWystKod, ISNULL(CZD_DataWyst, GETDATE())AS DataWyst
		, Ilosc
		, Twr_JM as JM
		--, Dst.*
	from dbo.CTIProdukcjaDostawyNaZP (@idZp) as Dst
		INNER JOIN CDN.tranag trnG on TrNId = trn_trnid
		INNER JOIN CDN.Towary on TwrId = twr_twrid
		INNER JOIN dbo.CtiZlecenieDok on CZD_TrnId = TrNId
		LEFT JOIN CDN.Magazyny AS c ON TrN_MagDocId = c.Mag_MagId 
		LEFT JOIN CDN.Magazyny AS d ON TrN_MagZrdId = d.Mag_MagId 
		LEFT JOIN dbo.CtiZleceniePrc ON  CZO_CZNId = CZD_CZNId and CZO_PrcId = CZD_PracownikId and @prmPrc = 1
		INNER JOIN cdn.TraSElem ON DstId = TrS_TrSIdDost AND TrS_Typ=1
		INNER JOIN cdn.TraElem ON TrE_TrEId = TrS_TrEId
		INNER JOIN CDN.tranag DokDst  ON DokDst.TrN_TrNId = TrE_TrNId
		LEFT JOIN CtiOperatorzy AS a ON CZD_OpeWyst = a.CO_ID 
	order by CZD_DataWyst

END
GO
```

# [dbo].[CTIProdukcjaZPZestawienia]

```sql
/****** Object:  StoredProcedure [dbo].[CTIProdukcjaZPZestawienia]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[CTIProdukcjaZPZestawienia]
	@dataOd		varchar(20),
	@dataDo		varchar(20)
AS
BEGIN

-------------ZESTAWIENIE 1 - ilość zasobów
	SELECT * INTO #zestawienia 
	FROM (
			select 1 as Dane_Nr, Dane_ID, Dane_CZNID, Dane_CZID, Dane_Txt1, Dane_Txt2, Dane_Txt3
			, 'Ilość planowana: '+convert(varchar,IloscPlan)+' Ilość zrealizowana: ' +convert(varchar,IloscRealizacja)+
				case when IloscRealizacja<IloscPlan then ' Braki w realizacji'
					 when IloscRealizacja=IloscPlan then ' Realizacja zakończona'
					 when IloscRealizacja>IloscPlan then ' '
				end as Dane_Txt4
			from
			(	
				SELECT Realizacja.ZZs_ID as Dane_ID, Realizacja.ZZs_CZNID as Dane_CZNID, Realizacja.ZZs_CZID as Dane_CZID, dbo.CTIProdukcjaZlecenieNumer (Zlecenie.CZN_ID) AS Dane_Txt1 , Zasob.CZ_kod AS Dane_Txt2 
					, case when CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then '' else CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'') , 120) end AS Dane_Txt3
					, isnull(( select 	sum(isnull(CZS_Ilosc,0)) 
						from dbo.CtiZasobDok  where ISNULL(CZS_DokTyp,0) = 2 and ISNULL(CZS_CTNID,0) = Zlecenie.CZN_id and CZS_CZID = Zasob.CZ_ID
					  ),0) as IloscPlan
					, isnull(( select 	sum(isnull(RealizacZestaw.ZZs_Ilosc,0)) 
						from dbo.CtiZlecenieZasob as RealizacZestaw where RealizacZestaw.ZZs_CZNID = Zlecenie.CZN_ID and RealizacZestaw.ZZs_CZID = Realizacja.ZZs_CZID 
					   ),0) as IloscRealizacja
		  
				from dbo.CtiZasob as Zasob
					INNER JOIN dbo.CtiZlecenieZasob as Realizacja on Realizacja.ZZs_CZID = Zasob.CZ_ID
					INNER JOIN dbo.CtiZlecenieNag	as Zlecenie on Realizacja.ZZs_CZNID = Zlecenie.CZN_ID
				where CONVERT(VARCHAR, ISNULL(Zlecenie.CZN_DataWystaw, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)
				and ZZs_DokTyp = 2
			) Zestaw1
		) AS x

-------------ZESTAWIENIE 2	- czas zasobów		
	INSERT INTO #zestawienia 
	select 2 as Dane_Nr, Dane_ID, Dane_CZNID, Dane_CZID, Dane_Txt1, Dane_Txt2, Dane_Txt3
	, 'Czas planowany: '+convert(varchar,CzasPlan)+' Czas zrealizowany: ' +convert(varchar,CzasRealizacja)+
		case when CzasRealizacja<CzasPlan then ' '
			 when CzasRealizacja=CzasPlan then ' '
			 when CzasRealizacja>CzasPlan then ' '
		end as Dane_Txt4
	from
	(	
		SELECT Realizacja.ZZs_ID as Dane_ID, Realizacja.ZZs_CZNID as Dane_CZNID, Realizacja.ZZs_CZID as Dane_CZID, dbo.CTIProdukcjaZlecenieNumer (Zlecenie.CZN_ID) AS Dane_Txt1 , Zasob.CZ_kod AS Dane_Txt2 
			, case when CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then '' else CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'') , 120) end AS Dane_Txt3
		
	
			, isnull(( select 	sum(isnull(
					(
					 case when CONVERT(VARCHAR,isnull(RealizacZestaw.ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then isnull(RealizacZestaw.ZZs_CzasMin,0)
						  when CONVERT(VARCHAR,isnull(RealizacZestaw.ZZs_DataDo,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then  CONVERT(decimal(15,4),DATEDIFF(second, RealizacZestaw.ZZs_DataOd,GETDATE())/60.0000)
						  else CONVERT(decimal(15,4),(isnull(DATEDIFF(SECOND, RealizacZestaw.ZZs_DataOd,RealizacZestaw.ZZs_DataDo) ,0))/60.0000)  end 
					),0)) 
				from dbo.CtiZlecenieZasob as RealizacZestaw where RealizacZestaw.ZZs_CZNID = Zlecenie.CZN_ID and RealizacZestaw.ZZs_CZID = Zasob.CZ_ID
			   ),0) as CzasRealizacja
			 ,
			  isnull(( select sum(isnull((convert(decimal(15,2), dbo.CTIProdukcjaZPZasobCzasCalkowity(PlanZestaw.CZS_CzasPracy, PlanZestaw.CZS_CzasUzbrojenia, PlanZestaw.CZS_CzasRozbrojenia, ZlecenieZestaw.CZN_Ilosc, COALESCE(ZasobZestaw.CZ_Dzielnik,ZlecenieZestaw.CZN_Ilosc,0), PlanZestaw.CZS_JMCzasu, CZS_ID)/60.0)),0))
				from dbo.CtiZasobDok as PlanZestaw INNER JOIN dbo.CtiZlecenieNag	as ZlecenieZestaw on ISNULL(PlanZestaw.CZS_CTNID,0) = ZlecenieZestaw.CZN_ID 
				inner join dbo.CtiZasob as ZasobZestaw on PlanZestaw.CZS_CZID = ZasobZestaw.CZ_ID
				 where ISNULL(PlanZestaw.CZS_DokTyp,0) = 2 and ISNULL(PlanZestaw.CZS_CTNID,0)=Zlecenie.CZN_ID  and  PlanZestaw.CZS_CZID = Zasob.CZ_ID
			  ),0)
		  
			   as CzasPlan
		from dbo.CtiZasob as Zasob
			INNER JOIN dbo.CtiZlecenieZasob as Realizacja on Realizacja.ZZs_CZID = Zasob.CZ_ID
			INNER JOIN dbo.CtiZlecenieNag	as Zlecenie on Realizacja.ZZs_CZNID = Zlecenie.CZN_ID
		where CONVERT(VARCHAR, ISNULL(Zlecenie.CZN_DataWystaw, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)
				and ZZs_DokTyp = 2
	) Zestaw2


-------------ZESTAWIENIE 3	- daty	
	INSERT INTO #zestawienia 
	select 3 as Dane_Nr, Dane_ID, Dane_CZNID, Dane_CZID, Dane_Txt1, Dane_Txt2, Dane_Txt3
	, 'Planowana data zakończenia: '''+convert(varchar,DataPlan, 23)+''''+
		CASE WHEN convert(varchar,isnull(DataRealizacja,''), 23) = '1900-01-01' then '' else ' Realizacja: ''' +convert(varchar,isnull(DataRealizacja,''), 23)+'''' end
	 as Dane_Txt4
	from
	(	
		SELECT Realizacja.ZZs_ID as Dane_ID, Realizacja.ZZs_CZNID as Dane_CZNID, Realizacja.ZZs_CZID as Dane_CZID, dbo.CTIProdukcjaZlecenieNumer (Zlecenie.CZN_ID) AS Dane_Txt1 , Zasob.CZ_kod AS Dane_Txt2 
			, case when CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'1900-01-01 00:00:00') , 120) = '1900-01-01 00:00:00' then '' else CONVERT(VARCHAR,isnull(Realizacja.ZZs_DataOd,'') , 120) end AS Dane_Txt3
		
	
			, ( select 	max(RealizacZestaw.ZZs_DataDo) 
				from dbo.CtiZlecenieZasob as RealizacZestaw where RealizacZestaw.ZZs_CZNID = Zlecenie.CZN_ID and RealizacZestaw.ZZs_CZID = Zasob.CZ_ID
			   ) as DataRealizacja
			 ,
			  isnull(( select max(ISNULL(PlanZestaw.CZS_CzasEnd,ZlecenieZestaw.CZN_CzasEnd))		
					from dbo.CtiZasobDok as PlanZestaw INNER JOIN dbo.CtiZlecenieNag	as ZlecenieZestaw on ISNULL(PlanZestaw.CZS_CTNID,0) = ZlecenieZestaw.CZN_ID 
				 where ISNULL(PlanZestaw.CZS_DokTyp,0) = 2 and ISNULL(PlanZestaw.CZS_CTNID,0)=Zlecenie.CZN_ID  and  PlanZestaw.CZS_CZID = Zasob.CZ_ID
			  ),Zlecenie.CZN_CzasEnd)
		  
			   as DataPlan
		from dbo.CtiZasob as Zasob
			INNER JOIN dbo.CtiZlecenieZasob as Realizacja on Realizacja.ZZs_CZID = Zasob.CZ_ID
			INNER JOIN dbo.CtiZlecenieNag	as Zlecenie on Realizacja.ZZs_CZNID = Zlecenie.CZN_ID
		where CONVERT(VARCHAR, ISNULL(Zlecenie.CZN_DataWystaw, GETDATE()), 23)  BETWEEN CONVERT(VARCHAR, @dataOd, 23) AND CONVERT(VARCHAR, @dataDo, 23)
				and ZZs_DokTyp = 2
	) Zestaw

		
			
		
		
	SELECT * FROM  #zestawienia	ORDER BY 1,2
	DROP TABLE #zestawienia 

END

GO
```

# [Towary_After_Insert_Trigger]

```sql
/****** Object:  Trigger [CDN].[Towary_After_Insert_Trigger]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[Towary_After_Insert_Trigger]
ON [CDN].[Towary]
AFTER INSERT
AS
 SET NOCOUNT ON
 declare @dokumentTypGrupuTowarowe int = 0;
 select @dokumentTypGrupuTowarowe = ID from CDN.ListaDokumentowWymiarowanych() where nazwa = 'Grupy towarowe';
 declare @dokumentTypTowary int = 0;
 select @dokumentTypTowary = ID from CDN.ListaDokumentowWymiarowanych() where nazwa = 'Towary';
  declare @ID int
  declare @GrupaGIDNUmer int
  declare @twgID int
 DECLARE Dodane CURSOR LOCAL FAST_FORWARD FOR
 SELECT TwR_TwRID, Twr_TwGGIDNumer FROM inserted
 OPEN Dodane
 WHILE 1=1
 begin
  FETCH NEXT FROM Dodane INTO @ID, @GrupaGIDNUmer
  IF @@FETCH_STATUS <> 0 BREAK
  select @twgID=TwG_TwGID from CDN.TwrGrupy where TwG_GIDNumer = @GrupaGIDNUmer and TwG_GIDTyp = -16
  if @twgID is not null
  begin
   insert into CDN.WymiaryWartosci (WyW_WyyId, WyW_Guid, WyW_Wartosc, WyW_DokumentTyp, WyW_DokumentId, WyW_Procent, WyW_Kwota, WyW_SlownikId)
   select WyW_WyyId, WyW_Guid, WyW_Wartosc, @dokumentTypTowary, @ID, WyW_Procent, WyW_Kwota, WyW_SlownikId from CDN.WymiaryWartosci
   where  WyW_DokumentTyp = @dokumentTypGrupuTowarowe and WyW_DokumentId = @twgID
   order by WyW_Guid;
  end
 end
 Close Dodane
 Deallocate Dodane
 SET NOCOUNT OFF

GO
ALTER TABLE [CDN].[Towary] ENABLE TRIGGER [Towary_After_Insert_Trigger]
GO
```

# [Towary_Delete_Trigger]

```sql
/****** Object:  Trigger [CDN].[Towary_Delete_Trigger]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
Create Trigger [CDN].[Towary_Delete_Trigger]
  On [CDN].[Towary]
  For Delete
As
Set NoCount On
-- Sprawdzenie, czy towar nie jest wzięty na urządzenie
  IF EXISTS (SELECT * FROM cdn.SrsUrzadzenia  JOIN deleted ON SrU_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34553##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest wzięty na zlecenie
  IF EXISTS (SELECT * FROM cdn.SrsZlecenia JOIN deleted ON SrZ_ZbiorczeTwrIdCzesci=deleted.Twr_TwrId WHERE SrZ_ZbiorczeFaCzesci=1)
  OR EXISTS (SELECT * FROM cdn.SrsZlecenia JOIN deleted ON SrZ_ZbiorczeTwrIdCzynnosci=deleted.Twr_TwrId WHERE SrZ_ZbiorczeFaCzynnosci=1)
    BEGIN
      RAISERROR('##OPT34554##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
  IF EXISTS (SELECT * FROM cdn.SrsCzesci  JOIN deleted ON SrC_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34554##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
  IF EXISTS (SELECT * FROM cdn.SrsCzynnosci JOIN deleted ON SrY_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34554##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest wysłany do ESklepu
  IF EXISTS (SELECT * FROM deleted where Twr_GidFirma = 1)
    BEGIN
      if exists (select * from cdn.eSklepStanowiska where ESS_Typ = 2 ) and exists (select * from cdn.eSklepStanowiska where ESS_Typ = 1)
  RAISERROR('##OPT34555##', 16, 1)
      else
  if exists (select * from cdn.eSklepStanowiska where ESS_Typ = 2 )
   RAISERROR('##OPT34556##', 16, 1)
      else
  if exists (select * from cdn.eSklepStanowiska where ESS_Typ = 1 )
   RAISERROR('##OPT34557##', 16, 1)
      else
  RAISERROR('##OPT34555##', 16, 1)
      --RAISERROR('Pozycja cennika nie może być wykasowana. Nie można usunąć towaru, który został wysłany do E-Sklepu.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
 -- Sprawdzenie, czy towar nie jest kosztem przesylki dla eCommerce
    IF EXISTS (select * from cdn.eSklepStanowiska JOIN deleted ON ESS_KosztPrzesylki=deleted.Twr_TwrId  )
    BEGIN
      RAISERROR('##OPT35175##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest wysłany do detalu
  IF EXISTS (SELECT * FROM deleted where Twr_GidFirma = 2)
    BEGIN
      RAISERROR('##OPT34558##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
  IF EXISTS (SELECT * FROM cdn.ZestawyTwrSkladniki  JOIN deleted ON ZTS_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34559##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest wykorzystany na ofercie
  IF EXISTS (SELECT * FROM cdn.OfertyElem  JOIN deleted ON OfL_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34560##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
   IF EXISTS (SELECT * FROM cdn.OfertyElemCeny  JOIN deleted ON OfC_TwrId=deleted.Twr_TwrId)
    BEGIN
      RAISERROR('##OPT34560##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest zamiennikiem
 IF EXISTS (SELECT * FROM cdn.Zamienniki  JOIN deleted ON deleted.Twr_TwrId = ZAM_TwrId) or EXISTS (SELECT * FROM cdn.Zamienniki  JOIN deleted ON deleted.Twr_TwrId = ZAM_ZamTwrId)
    BEGIN
      RAISERROR('##OPT34561##', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- obsługa kasowania rabatow dla towarów
Delete From CDN.Rabaty Where
 Rab_Typ In (4, 6, 8, 11, 12, 13) And
 Rab_TwrId In (Select Twr_TwrId From deleted)
Delete From CDN.TwrGrupy
Where TwG_GIDTyp = 16
 AND TwG_Kod In (Select Twr_Kod From deleted)
-- obsługa kaskadowego kasowania przywiązania towaru do grup
 DELETE CDN.TwrGrupy
    FROM CDN.TwrGrupy, deleted
    WHERE
   TwG_GIDTyp  = 16
  AND TwG_Kod  = deleted.Twr_Kod
-- obsługa kasowania z wzorca faktur cyklicznych
DELETE CDN.FaCyklTwrKnt FROM CDN.FaCyklTwrKnt A JOIN CDN.FaCyklElem B ON  A.FcT_FcEId=B.FCE_FcEID
WHERE FcE_TwrId In (Select Twr_TwrId From deleted)
Delete From CDN.FaCyklElem Where FcE_TwrId In (Select Twr_TwrId From deleted)
-- obsługa kasowania załączników
DELETE FROM CDN.DaneBinarneLinki WHERE DBL_TwrId IN (SELECT Twr_TwrId FROM deleted)
 -- Usuwanie wymiarów analitycznych
  DECLARE @WADokumentTyp int = 0;
  select @WADokumentTyp=id from CDN.ListaDokumentowWymiarowanych() where nazwa='Towary';
 if exists(select * from CDN.WymiaryWartosci where WyW_DokumentTyp = @WADokumentTyp and WyW_DokumentId in (select deleted.Twr_TwrId from deleted))
  delete from CDN.WymiaryWartosci where WyW_DokumentTyp = @WADokumentTyp
  and WyW_DokumentId in (select deleted.Twr_TwrId from deleted)
Set NoCount Off

GO
ALTER TABLE [CDN].[Towary] ENABLE TRIGGER [Towary_Delete_Trigger]
GO
```

# [Towary_Delete_TriggerProdukcja]

```sql
/****** Object:  Trigger [CDN].[Towary_Delete_TriggerProdukcja]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create Trigger [CDN].[Towary_Delete_TriggerProdukcja]
  On [CDN].[Towary]
  For Delete
As
Set NoCount On
-- Sprawdzenie, czy towar nie jest na technologii produkcji
  IF EXISTS ( SELECT  top 1 CTN_TwrId FROM dbo.CtiTechnolNag JOIN deleted ON  CTN_TwrId = deleted.Twr_TwrId)
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w technologii produkcji.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
	
-- Sprawdzenie, czy towar nie jest na elementach technologii produkcji
  IF EXISTS ( SELECT  top 1 CTE_TwrId FROM dbo.CtiTechnolElem JOIN deleted ON  CTE_TwrId = deleted.Twr_TwrId)
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w technologii produkcji.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
-- Sprawdzenie, czy towar nie jest na zleceniu wydania
  IF EXISTS ( SELECT top 1 isnull(CWP_PodId,0) FROM dbo.CtiProdWydaniePoz JOIN deleted ON  isnull(CWP_PodId,0) = deleted.Twr_TwrId where CWP_Typ = 33)
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w zleceniu wydania.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END

-- Sprawdzenie, czy towar nie jest na zleceniu produkcyjnym
  IF EXISTS ( SELECT top 1 isnull(CZN_TwrId,0) FROM dbo.CtiZlecenieNag JOIN deleted ON  isnull(CZN_TwrId,0) = deleted.Twr_TwrId)
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w zleceniu produkcyjnym.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END

-- Sprawdzenie, czy towar nie jest na elemencie zlecenia produkcyjnego
  IF EXISTS ( SELECT top 1 isnull(CZE_TwrId,0) FROM dbo.CtiZlecenieElem JOIN deleted ON  isnull(CZE_TwrId,0) = deleted.Twr_TwrId)
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w elemencie ZP.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END

-- Sprawdzenie, czy towar nie jest na elemencie technologii złożonej
  IF EXISTS ( SELECT TOP 1 isnull(CTZE_CTNID,0) FROM dbo.CtiTechnolZElem JOIN deleted ON  isnull(CTZE_CTNID,0) = deleted.Twr_TwrId WHERE CTZE_Typ IN (2,3,4,5,6))
    BEGIN
      RAISERROR('Pozycja cennika nie może być wykasowana. Pozycja została użyta w elemencie THZ.', 16, 1)
      ROLLBACK TRAN
      SET NOCOUNT OFF
      RETURN
    END
Set NoCount Off
GO
ALTER TABLE [CDN].[Towary] ENABLE TRIGGER [Towary_Delete_TriggerProdukcja]
GO
```

# [Towary_DeleteHandler]

```sql
/****** Object:  Trigger [CDN].[Towary_DeleteHandler]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[Towary_DeleteHandler]
  ON [CDN].[Towary]
  FOR DELETE
AS
SET NOCOUNT ON
    IF EXISTS(
   SELECT * FROM CDN.TwrJm INNER JOIN deleted
   ON deleted.Twr_GIDTyp = CDN.TwrJm.TwJ_TwrTyp
   AND deleted.Twr_GIDFirma = CDN.TwrJm.TwJ_TwrFirma
   AND deleted.Twr_GIDNumer = CDN.TwrJm.TwJ_TwrNumer
  )
  DELETE CDN.TwrJm
  FROM CDN.TwrJm, deleted
  WHERE
   deleted.Twr_GIDTyp = CDN.TwrJm.TwJ_TwrTyp
  AND deleted.Twr_GIDFirma = CDN.TwrJm.TwJ_TwrFirma
  AND deleted.Twr_GIDNumer = CDN.TwrJm.TwJ_TwrNumer
    IF EXISTS(
   SELECT * FROM CDN.TwrDost INNER JOIN deleted
   ON deleted.Twr_GIDTyp = CDN.TwrDost.TWD_TwrTyp
   AND deleted.Twr_GIDFirma = CDN.TwrDost.TWD_TwrFirma
   AND deleted.Twr_GIDNumer = CDN.TwrDost.TWD_TwrNumer
  )
  DELETE CDN.TwrDost
  FROM CDN.TwrDost, deleted
  WHERE
   deleted.Twr_GIDTyp = CDN.TwrDost.TWD_TwrTyp
  AND deleted.Twr_GIDFirma = CDN.TwrDost.TWD_TwrFirma
  AND deleted.Twr_GIDNumer = CDN.TwrDost.TWD_TwrNumer
SET NOCOUNT OFF

GO
ALTER TABLE [CDN].[Towary] ENABLE TRIGGER [Towary_DeleteHandler]
GO
```

# [Towary_Update_TriggerProdukcja]

```sql
/****** Object:  Trigger [CDN].[Towary_Update_TriggerProdukcja]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[Towary_Update_TriggerProdukcja]
ON [CDN].[Towary] FOR UPDATE
AS BEGIN
	IF (UPDATE(Twr_Nieaktywny))
	BEGIN
		IF EXISTS (
			SELECT inserted.Twr_TwrId FROM inserted
			INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
			WHERE deleted.Twr_NieAktywny = 0 and inserted.Twr_NieAktywny = 1
		)
		BEGIN
			DECLARE @Msg VARCHAR(MAX) = 'Nie można ustawić towaru na nieaktywny, występuje na aktywnych dokumentach TH w Produkcji (jako pozycja lub wyrób):';
			DECLARE @Zablokowac INT = 0;
			DECLARE @NumerTH VARCHAR(MAX) = '';
			
			-- Wyrób na TH
			DECLARE Kursor1 CURSOR FOR
			SELECT CTN_Nazwa FROM dbo.CtiTechnolNag
			WHERE CTN_TwrId IN (
				SELECT inserted.Twr_TwrId FROM inserted
				INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
				WHERE deleted.Twr_NieAktywny = 0 AND inserted.Twr_NieAktywny = 1
			)
			AND CTN_Status <> 4;

			OPEN Kursor1;
			FETCH NEXT FROM Kursor1 INTO @NumerTH;

			WHILE @@FETCH_STATUS = 0
			BEGIN
				SET @Zablokowac = 1;
				SET @Msg = @Msg + CHAR(13) + CHAR(10) + @NumerTH;

				FETCH NEXT FROM Kursor1 INTO @NumerTH;
			END

			CLOSE Kursor1;
			DEALLOCATE Kursor1;

			-- Pozycja na TH
			DECLARE Kursor2 CURSOR FOR
			SELECT DISTINCT CTN_Nazwa FROM dbo.CtiTechnolNag
			LEFT OUTER JOIN dbo.CtiTechnolElem ON CTN_ID = CTE_CTNId
			WHERE CTE_TwrId IN (
				SELECT inserted.Twr_TwrId FROM inserted
				INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
				WHERE deleted.Twr_NieAktywny = 0 AND inserted.Twr_NieAktywny = 1
			)
			AND CTN_Status <> 4

			OPEN Kursor2;
			FETCH NEXT FROM Kursor2 INTO @NumerTH;

			WHILE @@FETCH_STATUS = 0
			BEGIN
				SET @Zablokowac = 1;
				SET @Msg = @Msg + CHAR(13) + CHAR(10) + @NumerTH;

				FETCH NEXT FROM Kursor2 INTO @NumerTH;
			END

			CLOSE Kursor2;
			DEALLOCATE Kursor2;

			-- Wyrób na THZ
			DECLARE Kursor3 CURSOR FOR
			SELECT CTZN_Nazwa FROM dbo.CtiTechnolZNag
			WHERE CTZN_TwrId IN (
				SELECT inserted.Twr_TwrId FROM inserted
				INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
				WHERE deleted.Twr_NieAktywny = 0 AND inserted.Twr_NieAktywny = 1
			)
			AND CTZN_Nieaktywna = 0;

			OPEN Kursor3;
			FETCH NEXT FROM Kursor3 INTO @NumerTH;

			WHILE @@FETCH_STATUS = 0
			BEGIN
				SET @Zablokowac = 1;
				SET @Msg = @Msg + CHAR(13) + CHAR(10) + @NumerTH;

				FETCH NEXT FROM Kursor3 INTO @NumerTH;
			END

			CLOSE Kursor3;
			DEALLOCATE Kursor3;

			-- Pozycja na THZ
			DECLARE Kursor4 CURSOR FOR
			SELECT DISTINCT CTZN_Nazwa FROM dbo.CtiTechnolZNag
			LEFT OUTER JOIN dbo.CtiTechnolZElem ON CTZN_ID = CTZE_CTZNId
			WHERE CTZE_Typ IN (2,3,4,5,6)
			AND CTZE_CTNID IN (
				SELECT inserted.Twr_TwrId FROM inserted
				INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
				WHERE deleted.Twr_NieAktywny = 0 AND inserted.Twr_NieAktywny = 1
			)
			AND CTZN_Nieaktywna = 0;

			OPEN Kursor4;
			FETCH NEXT FROM Kursor4 INTO @NumerTH;

			WHILE @@FETCH_STATUS = 0
			BEGIN
				SET @Zablokowac = 1;
				SET @Msg = @Msg + CHAR(13) + CHAR(10) + @NumerTH;

				FETCH NEXT FROM Kursor4 INTO @NumerTH;
			END

			CLOSE Kursor4;
			DEALLOCATE Kursor4;



			IF (@Zablokowac = 1)
			BEGIN
				RAISERROR(@Msg, 16, 1)
				ROLLBACK TRAN
				SET NOCOUNT OFF
				RETURN
			END
		END
	END
	IF (UPDATE(Twr_JM))
	BEGIN
		IF EXISTS (
			SELECT inserted.Twr_TwrId FROM inserted
			INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
			WHERE inserted.Twr_JM <> deleted.Twr_JM 
		)
		BEGIN
			DECLARE @Msg2 VARCHAR(MAX) = 'Nie można zmienić JM towaru, występuje na dokumentach TH/THZ/ZP/ZPZ w Produkcji (jako pozycja lub wyrób)';
			DECLARE @Zablokowac2 INT = 0;
			DECLARE @tempProd VARCHAR(MAX) = '';
		
			DECLARE Kursor5 CURSOR FOR
			select distinct ID from (select distinct CTN_TwrId as ID from dbo.CtiTechnolNag 
			union all
			select distinct CTZN_TwrId as ID from dbo.CtiTechnolZNag 
			union all
			select distinct CZN_TwrId as ID from dbo.CtiZlecenieNag 
			union all
			select distinct CZZN_TwrId as ID from dbo.CtiZlecenieZNag 
			union all
			select distinct CTE_TwrId as ID from dbo.CtiTechnolElem
			union all
			select distinct CTZE_CTNID as ID from dbo.CtiTechnolZElem
			where CTZE_Typ IN(2,3,4,5,6)
			union all
			select distinct CZE_TwrId as ID from dbo.CtiZlecenieElem 
			union all
			select distinct CZZE_CZNID as ID from dbo.CtiZlecenieZElem 
			where CZZE_Typ IN(12,13,14,15,16)
			) a
			WHERE ID IN (
				SELECT inserted.Twr_TwrId FROM inserted
				INNER JOIN deleted ON inserted.Twr_TwrId = deleted.Twr_TwrId
			);
			OPEN Kursor5;
			FETCH NEXT FROM Kursor5 INTO @tempProd;

			WHILE @@FETCH_STATUS = 0
			BEGIN
				SET @Zablokowac2 = 1;
				FETCH NEXT FROM Kursor5 INTO @tempProd;
			END

			CLOSE Kursor5;
			DEALLOCATE Kursor5;

			IF (@Zablokowac2 = 1)
			BEGIN
				RAISERROR(@Msg2, 16, 1)
				ROLLBACK TRAN
				SET NOCOUNT OFF
				RETURN
			END
		END
	END
END
GO
ALTER TABLE [CDN].[Towary] ENABLE TRIGGER [Towary_Update_TriggerProdukcja]
GO
```

# [TraElem_CTI_BlokowanieZP_Delete]

```sql
/****** Object:  Trigger [CDN].[TraElem_CTI_BlokowanieZP_Delete]    Script Date: 5.01.2026 21:12:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[TraElem_CTI_BlokowanieZP_Delete]
ON [CDN].[TraElem]
FOR DELETE
AS BEGIN
	DECLARE @DokTyp int = (SELECT TOP 1 TrE_TypDokumentu FROM deleted);
	DECLARE @DokID int = (SELECT TOP 1 TrE_TrNId FROM deleted);
	DECLARE @Treid int = (SELECT TOP 1 TrE_treid FROM deleted);
	DECLARE @ZwrID int = (SELECT TOP 1 TrN_ZwrID FROM CDN.TraNag WHERE TrN_TrNID = @DokID);	
	DECLARE @TwrID int = (SELECT TOP 1 TrE_TwrId FROM inserted);
	DECLARE @ZpID int = (SELECT TOP 1 ISNULL(CZD_CZNId, -1) FROM dbo.CtiZlecenieDok WHERE (CZD_TrnId = @DokID OR CZD_TrnId = @ZwrID));

	IF (select KNF_Wartosc from dbo.CtiKonfiguracja where KNF_Numer=1139)=1 
	BEGIN
	IF @ZpID IS NOT NULL AND @ZpID > 0
		BEGIN
		IF @DokTyp = 303	
			EXEC dbo.CTIProdukcjaUpdateStatusyZP 24, @ZpID
					
		IF @DokTyp = 308 and (select KNF_Wartosc from dbo.CtiKonfiguracja where KNF_Numer=1142)=1
			EXEC dbo.CTIProdukcjaUpdateStatusyZPProdukcja @Treid, @TwrID
		END
	END
	-- Nie mamy do czynienia z PW, wychodzimy z funkcji
	IF (@DokTyp NOT IN (303, 312))
		RETURN;

	-- Dokument nie ma powiązania ze zleceniem, wychodzimy z funkcji
	IF NOT EXISTS (SELECT * FROM dbo.CtiZlecenieDok WHERE CZD_Typ IN (4, 6) AND 
		(CZD_TrnId = @DokID OR CZD_TrnId = @ZwrID))
		RETURN;

	DECLARE @IloscZlecona DECIMAL(15,4) = (SELECT TOP 1 CZN_Ilosc FROM dbo.CtiZlecenieNag WHERE CZN_ID = @ZpID);
	DECLARE @KontrolaWydania varchar(255) = ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 63), '0')
	DECLARE @IloscSprawdzana DECIMAL(15,4);

	IF (@KontrolaWydania = '1')
	BEGIN
		SET @IloscSprawdzana = (SELECT dbo.CTIProdukcjaZlecenieIloscSkontrolowana(@ZpID));
	END
	ELSE
	BEGIN
		SET @IloscSprawdzana = (SELECT dbo.CTIProdukcjaZlecenieIloscZrealizowana(@ZpID, 0, 0));
	END

	-- Blokowanie
	IF (@IloscSprawdzana >= @IloscZlecona AND ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 64), '') IN ('True', 'TRUE', '1'))
	BEGIN
		UPDATE dbo.CtiZlecenieNag SET CZN_Status = 3 WHERE CZN_ID = @ZpID;
	END

	-- Odblokowanie
	IF (@IloscSprawdzana < @IloscZlecona AND ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 1020), '') IN ('True', 'TRUE', '1'))
	BEGIN
		UPDATE dbo.CtiZlecenieNag SET CZN_Status = 2 WHERE CZN_ID = @ZpID;
	END
END
GO
ALTER TABLE [CDN].[TraElem] ENABLE TRIGGER [TraElem_CTI_BlokowanieZP_Delete]
GO
```

# [TraNag_CTI_BlokowanieZP_Delete]

```sql
/****** Object:  Trigger [CDN].[TraNag_CTI_BlokowanieZP_Delete]    Script Date: 5.01.2026 21:12:42 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[TraNag_CTI_BlokowanieZP_Delete]
ON [CDN].[TraNag]
FOR DELETE
AS BEGIN
	DECLARE @DokTyp int = (SELECT TOP 1 TrN_TypDokumentu FROM deleted);
	DECLARE @DokID int = (SELECT TOP 1 TrN_TrNId FROM deleted);
	DECLARE @ZwrID int = (SELECT TOP 1 TrN_ZwrID FROM deleted);
	DECLARE @ZpID int = (SELECT TOP 1 ISNULL(CZD_CZNId, -1) FROM dbo.CtiZlecenieDok WHERE (CZD_TrnId = @DokID OR CZD_TrnId = @ZwrID));

	-- Nie mamy do czynienia z PW, wychodzimy z funkcji
	IF (@DokTyp NOT IN (303, 312))
		RETURN;

	IF (select KNF_Wartosc from dbo.CtiKonfiguracja where KNF_Numer=1139)=1
	BEGIN
	IF @ZpID IS NOT NULL AND @ZpID > 0
		BEGIN
		IF @DokTyp = 303	
			EXEC dbo.CTIProdukcjaUpdateStatusyZP 24, @ZpID
					
		IF @DokTyp = 312	
			BEGIN
				DECLARE @DokTypProd int = ISNULL((select top 1 CZD_Typ from dbo.CtiZlecenieDok where CZD_Typ=3 and CZD_TrnId=@DokID),-1)
				IF @DokTypProd = 3
					EXEC dbo.CTIProdukcjaUpdateStatusyZP 23, @ZpID
				IF @DokTypProd = 6
					EXEC dbo.CTIProdukcjaUpdateStatusyZP 26, @ZpID
			END
		END
	END

	-- Dokument nie ma powiązania ze zleceniem, wychodzimy z funkcji
	IF NOT EXISTS (SELECT * FROM dbo.CtiZlecenieDok WHERE CZD_Typ IN (4, 6) AND 
		(CZD_TrnId = @DokID OR CZD_TrnId = @ZwrID))
		RETURN;

	DECLARE @IloscZlecona DECIMAL(15,4) = (SELECT TOP 1 CZN_Ilosc FROM dbo.CtiZlecenieNag WHERE CZN_ID = @ZpID);
	DECLARE @KontrolaWydania varchar(255) = ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 63), '0')
	DECLARE @IloscSprawdzana DECIMAL(15,4);

	IF (@KontrolaWydania = '1')
	BEGIN
		SET @IloscSprawdzana = (SELECT dbo.CTIProdukcjaZlecenieIloscSkontrolowana(@ZpID));
	END
	ELSE
	BEGIN
		SET @IloscSprawdzana = (SELECT dbo.CTIProdukcjaZlecenieIloscZrealizowana(@ZpID, 0, 0));
	END

	-- Blokowanie
	IF (@IloscSprawdzana >= @IloscZlecona AND ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 64), '') IN ('True', 'TRUE', '1'))
	BEGIN
		UPDATE dbo.CtiZlecenieNag SET CZN_Status = 3 WHERE CZN_ID = @ZpID;
	END

	-- Odblokowanie
	IF (@IloscSprawdzana < @IloscZlecona AND ISNULL((SELECT TOP 1 Knf_Wartosc FROM dbo.CtiKonfiguracja WHERE Knf_Numer = 1020), '') IN ('True', 'TRUE', '1'))
	BEGIN
		UPDATE dbo.CtiZlecenieNag SET CZN_Status = 2 WHERE CZN_ID = @ZpID;
	END
END
GO
ALTER TABLE [CDN].[TraNag] ENABLE TRIGGER [TraNag_CTI_BlokowanieZP_Delete]
GO
```

# [TraSElem_InsertUpdateDelete_Trigger]

```sql
/****** Object:  Trigger [CDN].[TraSElem_InsertUpdateDelete_Trigger]    Script Date: 5.01.2026 21:12:42 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[TraSElem_InsertUpdateDelete_Trigger]
  ON [CDN].[TraSElem]
  FOR INSERT, UPDATE, DELETE
AS
  DECLARE
    -- zmienne pomocnicze dla kursora
    @TrS_TrSId      INT,
    @TrS_TrEId      INT,
    @TrS_TrSIdDost  INT,
    @TrS_ZwrId      INT,
    @TrS_TwrId      INT,
    @TrS_MagId      INT,
    @TrS_Rodzaj     INT,
    @TrS_Typ        INT,
    @TrS_Metoda     INT,
    @TrS_DataOpe    DATETIME,
    @TrS_Ilosc      DECIMAL(15,4),
    @TrS_Wartosc    DECIMAL(15,2),
    @TrS_Cena       DECIMAL(17,4)
  SET NOCOUNT ON
 IF  UPDATE( TrS_WyborDostaw )
 AND NOT UPDATE( TrS_TrEId)
 AND NOT UPDATE( TrS_TrSIdDost)
 AND NOT UPDATE( TrS_TrEIdWydania)
 AND NOT UPDATE( TrS_ZwrId)
 AND NOT UPDATE( TrS_TwrId)
 AND NOT UPDATE( TrS_MagId)
 AND NOT UPDATE( TrS_Rodzaj)
 AND NOT UPDATE( TrS_Typ)
 AND NOT UPDATE( TrS_Metoda)
 AND NOT UPDATE( TrS_DataOpe)
 AND NOT UPDATE( TrS_Ilosc)
 AND NOT UPDATE( TrS_Wartosc)
 AND NOT UPDATE( TrS_Cena)
 return
  -- subelementy kasowane
  DECLARE Kasowane CURSOR LOCAL FAST_FORWARD FOR
    SELECT                             TrS_TrSId,  TrS_TrEId,  TrS_TrSIdDost,  TrS_ZwrId,  TrS_TwrId,  TrS_MagId,  TrS_Rodzaj,  TrS_Typ,  TrS_Metoda,  TrS_DataOpe,  TrS_Ilosc,  TrS_Wartosc,  TrS_Cena
    FROM deleted
    ORDER BY TrS_Typ DESC, TrS_TrSId DESC
    OPEN Kasowane
      WHILE 1=1
      BEGIN
        FETCH NEXT FROM Kasowane INTO @TrS_TrSId, @TrS_TrEId, @TrS_TrSIdDost, @TrS_ZwrId, @TrS_TwrId, @TrS_MagId, @TrS_Rodzaj, @TrS_Typ, @TrS_Metoda, @TrS_DataOpe, @TrS_Ilosc, @TrS_Wartosc, @TrS_Cena
        IF @@FETCH_STATUS <> 0 BREAK
          EXECUTE CDN.AktualizujTwrZasoby 2, @TrS_TrSId, @TrS_TrEId, @TrS_TrSIdDost, @TrS_ZwrId, @TrS_TwrId, @TrS_MagId, @TrS_Rodzaj, @TrS_Typ, @TrS_Metoda, @TrS_DataOpe, @TrS_Ilosc, @TrS_Wartosc, @TrS_Cena
      END
    CLOSE Kasowane
  DEALLOCATE Kasowane
  -- subelementy dodawane
  DECLARE Dodawane CURSOR LOCAL FAST_FORWARD FOR
    SELECT                             TrS_TrSId,  TrS_TrEId,  TrS_TrSIdDost,  TrS_ZwrId,  TrS_TwrId,  TrS_MagId,  TrS_Rodzaj,  TrS_Typ,  TrS_Metoda,  TrS_DataOpe,  TrS_Ilosc,  TrS_Wartosc,  TrS_Cena
    FROM inserted
    ORDER BY TrS_Typ DESC, TrS_TrSId DESC
    OPEN Dodawane
      WHILE 1=1
      BEGIN
        FETCH NEXT FROM Dodawane INTO @TrS_TrSId, @TrS_TrEId, @TrS_TrSIdDost, @TrS_ZwrId, @TrS_TwrId, @TrS_MagId, @TrS_Rodzaj, @TrS_Typ, @TrS_Metoda, @TrS_DataOpe, @TrS_Ilosc, @TrS_Wartosc, @TrS_Cena
        IF @@FETCH_STATUS <> 0 BREAK
          EXECUTE CDN.AktualizujTwrZasoby 1, @TrS_TrSId, @TrS_TrEId, @TrS_TrSIdDost, @TrS_ZwrId, @TrS_TwrId, @TrS_MagId, @TrS_Rodzaj, @TrS_Typ, @TrS_Metoda, @TrS_DataOpe, @TrS_Ilosc, @TrS_Wartosc, @TrS_Cena
          --IF @TrS_Rodzaj in(308000,308011,309000)
          --   EXECUTE CDN.UstawStatusRezerwacji @TrS_TrEId
      END
    CLOSE Dodawane
  DEALLOCATE Dodawane
  SET NOCOUNT OFF

GO
ALTER TABLE [CDN].[TraSElem] ENABLE TRIGGER [TraSElem_InsertUpdateDelete_Trigger]
GO
```

# [TwrCeny_InsertUpdate_Trigger]

```sql
/****** Object:  Trigger [CDN].[TwrCeny_InsertUpdate_Trigger]    Script Date: 5.01.2026 21:12:42 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [CDN].[TwrCeny_InsertUpdate_Trigger]
ON [CDN].[TwrCeny]
FOR INSERT, UPDATE
AS
  SET NOCOUNT ON
  If Not Exists (Select * From deleted)
    Update CDN.TwrCeny Set
    TwC_Wartosc = Twr_KosztUslugiWal, TwC_Waluta  = Twr_Waluta
    From CDN.TwrCeny
    Join inserted On inserted.TwC_TwCId = CDN.TwrCeny.TwC_TwCId
    Join CDN.Towary On inserted.TwC_TwrId = Twr_TwrId
    Where Twr_Typ = 0 And Twr_KosztUslugiTyp = 1
    And (inserted.TwC_Wartosc <> Twr_KosztUslugiWal Or inserted.TwC_Waluta <> Twr_Waluta)
    And inserted.TwC_TwCNumer = 1
-- Uuzpełnienie tabeli przecen
  Declare @TwrID Int, @TwCNumer Int, @Wartosc Decimal(17,4), @Waluta NVarchar(3), @Data DateTime, @WartoscP Decimal(17,4), @WalutaP NVarchar(3), @DataP DateTime
  Declare Dodane Cursor Local Fast_Forward For
   Select TwC_TwrID, TwC_TwCNumer, TwC_Wartosc, TwC_Waluta From inserted
  Open Dodane
  While 1=1
   Begin
    Fetch Next From Dodane InTo @TwrID, @TwCNumer, @Wartosc, @Waluta
    If @@Fetch_Status <> 0 Break
    If Exists (Select * From CDN.TwrPrzeceny Where TPC_TwrId=@TwrId And TPC_TwCNumer=@TwCNumer)
     Begin
      Set @Data = Convert (DateTime, Convert(Char(10), GetDate(), 120), 120)
      Select Top 1 @WartoscP = TPC_Wartosc, @WalutaP = TPC_Waluta, @DataP = TPC_Data
       From CDN.TwrPrzeceny
       Where TPC_TwrId = @TwrId And TPC_TwCNumer = @TwCNumer And TPC_Data <= @Data
       Order By TPC_TwrId, TPC_TwCNumer, TPC_Data DESC
      If @Wartosc = @WartoscP And @Waluta = @WalutaP Continue
      If @DataP = @Data
       Update CDN.TwrPrzeceny Set TPC_Wartosc = @Wartosc, TPC_Waluta = @Waluta
         Where TPC_TwrId = @TwrId And TPC_TwCNumer = @TwCNumer And TPC_Data = @Data
      Else
       Insert InTo CDN.TwrPrzeceny (TPC_TwrID, TPC_TwCNumer, TPC_Data, TPC_Wartosc, TPC_Waluta)
                            Values (@TwrId,    @TwCNumer,    @Data,    @Wartosc,    @Waluta   )
     End
    Else
     Insert InTo CDN.TwrPrzeceny (
      TPC_TwrID,
      TPC_TwCNumer,
      TPC_Data,
      TPC_Wartosc,
      TPC_Waluta
     )
     Values (
      @TwrId,
      @TwCNumer,
      Convert(DateTime, '1900-01-01', 120),
      @Wartosc,
      @Waluta
     )
   End
  Close Dodane
  Deallocate Dodane
  update cdn.towary set twr_ts_mod = getdate() where twr_twrid in ( select distinct twc_twrid From inserted)
    -- modyfikujemy zestawy zeby poszły do esklep jak sie ceny zmienia
  update CDN.ZestawyTwr set ZTw_TS_Mod = getdate() where ZTw_TwrId in ( select distinct twc_twrid From inserted)
  update Zestawy set ZTw_TS_Mod = getdate()from CDN.ZestawyTwr Zestawy join cdn.ZestawyTwrSkladniki on ZTw_ZTwId = zts_ztwid where ZTs_TwrId in ( select distinct twc_twrid From inserted)
  SET NOCOUNT OFF

GO
ALTER TABLE [CDN].[TwrCeny] ENABLE TRIGGER [TwrCeny_InsertUpdate_Trigger]
GO
```

# [CTI_RealizacjaWieleOsob]

```sql
/****** Object:  Trigger [dbo].[CTI_RealizacjaWieleOsob]    Script Date: 5.01.2026 21:12:43 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [dbo].[CTI_RealizacjaWieleOsob]
ON [dbo].[CtiZlecenieZasob] FOR INSERT, UPDATE
AS BEGIN
	SET NOCOUNT ON;

	DECLARE @ZlecenieID int;
	DECLARE @ZasobID int;
	DECLARE @PracownikID int;
	DECLARE @DataDo DATETIME;
	DECLARE @MozeWieleOsob int = 0;
	DECLARE @Zablokowac int = 0;
	DECLARE @DokTyp int = 0;
	
	DECLARE insert_kursorZZS CURSOR FOR
	SELECT ZZs_CZNID, ZZs_CZID, ZZs_PrcID, ZZs_DataDo, ZZs_DokTyp FROM inserted;

	OPEN insert_kursorZZS;
	FETCH NEXT FROM insert_kursorZZS INTO @ZlecenieID, @ZasobID, @PracownikID, @DataDo, @DokTyp;

	WHILE @@FETCH_STATUS = 0  
	BEGIN  
		IF (@DataDo IS NOT NULL)
		BEGIN
			-- zasób został zakończony więc nie blokujemy
			FETCH NEXT FROM insert_kursorZZS INTO @ZlecenieID, @ZasobID, @PracownikID, @DataDo, @DokTyp;
		END

		SET @MozeWieleOsob = (
			SELECT TOP 1 CZS_WieleOsob
			FROM dbo.CtiZasobDok
			WHERE CZS_DokTyp = 2
			AND CZS_CTNID = @ZlecenieID
			AND CZS_CZID = @ZasobID
		);

		IF (@MozeWieleOsob IS NULL)
		BEGIN
			SET @MozeWieleOsob = (
				SELECT TOP 1 CZ_WieleOsob
				FROM dbo.CtiZasob
				WHERE CZ_ID = @ZasobID
			);
		END

		IF (@MozeWieleOsob = 0)
		BEGIN
			-- zasób ma blokadę wielu osób
			-- sprawdzamy czy ktoś jeszcze w tym czasie nie pracuje
			IF EXISTS (
				SELECT * FROM dbo.CtiZlecenieZasob
				WHERE ZZs_CZNID = @ZlecenieID
				AND ZZs_CZID = @ZasobID
				AND ZZs_PrcId <> @PracownikID
				AND ZZs_DataDo IS NULL
				AND ZZs_DokTyp = @DokTyp
			)
			BEGIN
				SET @Zablokowac = 1;
			END
		END

		IF (@Zablokowac <> 1)
		BEGIN
			-- nie blokujemy więc sprawdzamy kolejny zasób
			FETCH NEXT FROM insert_kursorZZS INTO @ZlecenieID, @ZasobID, @PracownikID, @DataDo, @DokTyp;
		END
	END

	CLOSE insert_kursorZZS;
	DEALLOCATE insert_kursorZZS; 

	IF (@Zablokowac = 1)
	BEGIN
		RAISERROR('Inny pracownik rozpoczął już pracę nad tym zasobem.', 16, 1);
		ROLLBACK TRAN
	END

	SET NOCOUNT OFF;
END
GO
ALTER TABLE [dbo].[CtiZlecenieZasob] ENABLE TRIGGER [CTI_RealizacjaWieleOsob]
GO
```

# [CtiProdukcjaCofniecieZakonczeniaZasobu]

```sql
/****** Object:  Trigger [dbo].[CtiProdukcjaCofniecieZakonczeniaZasobu]    Script Date: 5.01.2026 21:12:43 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER
[dbo].[CtiProdukcjaCofniecieZakonczeniaZasobu]
ON [dbo].[CtiZlecenieZasob]
AFTER DELETE
AS BEGIN

	DECLARE @ZasobID INT;
	DECLARE @ZlecenieID INT;
	DECLARE @DokTyp INT;

	DECLARE del_kursor CURSOR FOR
	SELECT ZZs_CZID, ZZs_CZNID, ZZs_DokTyp FROM deleted;

	OPEN del_kursor;
	FETCH NEXT FROM del_kursor INTO @ZasobID, @ZlecenieID, @DokTyp;

	WHILE @@FETCH_STATUS = 0
	BEGIN
		IF NOT EXISTS (
			SELECT * FROM dbo.CtiZlecenieZasob
			WHERE ZZs_CZNID = @ZlecenieID
			AND ZZs_CZID = @ZasobID
			AND ZZs_DokTyp = @DokTyp
		)
		BEGIN
			UPDATE dbo.CtiZasobDok
			SET CZS_Zakonczono = 0
			WHERE CZS_CTNID = @ZlecenieID
			AND CZS_CZID = @ZasobID
			AND CZS_DokTyp = @DokTyp;
		END

		FETCH NEXT FROM del_kursor INTO @ZasobID, @ZlecenieID, @DokTyp;
	END

	CLOSE del_kursor DEALLOCATE del_kursor
END
GO
ALTER TABLE [dbo].[CtiZlecenieZasob] ENABLE TRIGGER [CtiProdukcjaCofniecieZakonczeniaZasobu]
GO
```

# [CtiZlecenieZNag_Update_Trigger]

```sql
/****** Object:  Trigger [dbo].[CtiZlecenieZNag_Update_Trigger]    Script Date: 5.01.2026 21:12:43 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create Trigger [dbo].[CtiZlecenieZNag_Update_Trigger]
  On [dbo].[CtiZlecenieZNag]
   FOR UPDATE
As

IF (UPDATE (CZZN_Ilosc))
 BEGIN
    UPDATE dbo.CtiZlecenieZElem
    SET CZZE_Ilosc = (select top 1 CZZN_Ilosc from inserted)
    WHERE CZZE_CZNID = (select top 1 CZZN_ID from inserted)
    AND CZZE_Typ = 22
 END

 IF (UPDATE (CZZN_DokNad))--Modyfikacja ZPZ głównego - Kontrola zapętlenia
 begin
	declare @ZPZglownyNew int=0, @ZPZglownyOld int=0, @idZPZ int=0
	select top 1 @ZPZglownyNew =isnull(inserted.CZZN_DokNad,0), @idZPZ=CZZN_ID from inserted
	set @ZPZglownyOld = isnull((select top 1 deleted.CZZN_DokNad from deleted),0)

	IF (@ZPZglownyNew <> @ZPZglownyOld)
	BEGIN
		IF(@idZPZ=@ZPZglownyNew) 
		 BEGIN
			RAISERROR('Kontrola zapętlenia. Podpinane ZPZ jest głównym ZPZ', 16, 1)
			ROLLBACK TRAN
			SET NOCOUNT OFF
			RETURN
		END
		 ELSE IF ( @ZPZglownyOld = 0 and @ZPZglownyNew > 0)
		 BEGIN--Zmiana ZPZ głównego gdy podpinane do ZPZ
						update dbo.CtiZlecenieZNag  set CZZN_DokNad=@ZPZglownyNew
			from dbo.CtiZlecenieZNag inner join	dbo.CtiZlecenieZElem on CZZE_CZNID =czzn_id
			inner join inserted on inserted.CZZN_ID = CZZE_CZZNID
			where isnull(CZZE_Typ,2) = 22--ZPZ
		 END
		ELSE IF (@ZPZglownyOld > 0 and @ZPZglownyNew = 0)
		BEGIN--usuwamy powiązanie to obecny staję się głónym
			update dbo.CtiZlecenieZNag  set CZZN_DokNad=inserted.CZZN_ID
			from dbo.CtiZlecenieZNag inner join	dbo.CtiZlecenieZElem on CZZE_CZNID =czzn_id
			inner join inserted on inserted.CZZN_ID = CZZE_CZZNID
			where isnull(CZZE_Typ,2) = 22--ZPZ
		END
	END
END
GO
ALTER TABLE [dbo].[CtiZlecenieZNag] ENABLE TRIGGER [CtiZlecenieZNag_Update_Trigger]
GO
```
